### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ì œí’ˆ(Product) í´ë˜ìŠ¤

````python
class Product:
    """ì œí’ˆ ì •ë³´ë¥¼ ê´€ë¦¬í•˜ëŠ” í´ë˜ìŠ¤"""
    
    def __init__(self, product_id, name, price, stock):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.stock = stock
    
    def get_total_value(self):
        """ì¬ê³  ì´ ê°€ì¹˜ ê³„ì‚°"""
        return self.price * self.stock
    
    def is_in_stock(self):
        """ì¬ê³  ìˆëŠ”ì§€ í™•ì¸"""
        return self.stock > 0
    
    def sell(self, quantity):
        """íŒë§¤ ì²˜ë¦¬"""
        if self.stock >= quantity:
            self.stock -= quantity
            return f"{self.name} {quantity}ê°œ íŒë§¤ ì™„ë£Œ. ë‚¨ì€ ì¬ê³ : {self.stock}"
        return f"ì¬ê³  ë¶€ì¡±! í˜„ì¬ ì¬ê³ : {self.stock}"

# ì‹¤ì œ ì‚¬ìš©
laptop = Product("IT001", "ë§¥ë¶ í”„ë¡œ", 2500000, 10)
print(f"ì´ ì¬ê³  ê°€ì¹˜: {laptop.get_total_value():,}ì›")
print(laptop.sell(3))
print(laptop.sell(10))  # ì¬ê³  ë¶€ì¡±
````

````python
class UserSession:
    """ì‚¬ìš©ì ì„¸ì…˜ ê´€ë¦¬ í´ë˜ìŠ¤"""
    
    # í´ë˜ìŠ¤ ë³€ìˆ˜ - ëª¨ë“  ì„¸ì…˜ì˜ ì „ì²´ ë¡œê·¸ì¸ ì¹´ìš´íŠ¸ (ê³µìœ )
    total_logins = 0
    
    def __init__(self, username):
        self.username = username
        self.login_history = []  # âœ… ê° ì‚¬ìš©ìë³„ ë…ë¦½ì ì¸ íˆìŠ¤í† ë¦¬
        self.preferences = {}    # âœ… ê° ì‚¬ìš©ìë³„ ë…ë¦½ì ì¸ ì„¤ì •
        UserSession.total_logins += 1
    
    def add_login(self, timestamp):
        """ë¡œê·¸ì¸ ê¸°ë¡ ì¶”ê°€"""
        self.login_history.append(timestamp)
    
    def set_preference(self, key, value):
        """ì‚¬ìš©ì ì„¤ì •"""
        self.preferences[key] = value

# ì‚¬ìš© ì˜ˆì‹œ
user1 = UserSession("alice")
user2 = UserSession("bob")

user1.add_login("2024-01-15 09:00")
user1.set_preference("theme", "dark")

user2.add_login("2024-01-15 09:30")
user2.set_preference("theme", "light")

# ê° ì‚¬ìš©ìëŠ” ë…ë¦½ì ì¸ ë°ì´í„°ë¥¼ ê°€ì§
print(f"{user1.username}ì˜ ë¡œê·¸ì¸: {user1.login_history}")
print(f"{user2.username}ì˜ ë¡œê·¸ì¸: {user2.login_history}")
print(f"\nì „ì²´ ë¡œê·¸ì¸ ìˆ˜: {UserSession.total_logins}íšŒ")  # ê³µìœ ë˜ëŠ” ê°’
````

### 1) ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ (Instance Method)

ê°€ì¥ ì¼ë°˜ì ì¸ ë©”ì„œë“œ. ì¸ìŠ¤í„´ìŠ¤ì˜ ì†ì„±ì— ì ‘ê·¼í•˜ê³  ìˆ˜ì •í•©ë‹ˆë‹¤.

````python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    
    # ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ - selfë¥¼ í†µí•´ ì¸ìŠ¤í„´ìŠ¤ ì†ì„± ì ‘ê·¼
    def deposit(self, amount):
        self.balance += amount
        return f"{amount:,}ì› ì…ê¸ˆ. ì”ì•¡: {self.balance:,}ì›"
    
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            return f"{amount:,}ì› ì¶œê¸ˆ. ì”ì•¡: {self.balance:,}ì›"
        return "ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤."

account = BankAccount("ê¹€ì² ìˆ˜", 10000)
print(account.deposit(5000))
print(account.withdraw(3000))
````

### 2) í´ë˜ìŠ¤ ë©”ì„œë“œ (@classmethod)

í´ë˜ìŠ¤ ë³€ìˆ˜ì— ì ‘ê·¼í•˜ê±°ë‚˜, **ëŒ€ì²´ ìƒì„±ì(alternative constructor)**ë¥¼ ë§Œë“¤ ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.

````python
class Student:
    school = "ë‹¨í…Œ ë°ì´í„° ê³ ë“±í•™êµ"  # í´ë˜ìŠ¤ ë³€ìˆ˜
    student_count = 0
    
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
        Student.student_count += 1
    
    # ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ
    def introduce(self):
        return f"ì €ëŠ” {self.name}ì´ê³ , {self.grade}í•™ë…„ì…ë‹ˆë‹¤."
    
    # í´ë˜ìŠ¤ ë©”ì„œë“œ - clsë¥¼ í†µí•´ í´ë˜ìŠ¤ ë³€ìˆ˜ ì ‘ê·¼
    @classmethod
    def get_school_info(cls):
        return f"{cls.school} (ì´ {cls.student_count}ëª… ì¬í•™)"
    
    # ëŒ€ì²´ ìƒì„±ì: ë¬¸ìì—´ì—ì„œ í•™ìƒ ê°ì²´ ìƒì„±
    @classmethod
    def from_string(cls, student_str):
        """'ì´ë¦„-í•™ë…„' í˜•ì‹ì˜ ë¬¸ìì—´ì—ì„œ Student ìƒì„±"""
        name, grade = student_str.split('-')
        return cls(name, int(grade))

# ì¼ë°˜ ìƒì„±ì
student1 = Student("ê¹€ë¯¼ìˆ˜", 1)
student2 = Student("ì´ì˜í¬", 2)

# ëŒ€ì²´ ìƒì„±ì ì‚¬ìš©
student3 = Student.from_string("ë°•ì² ìˆ˜-3")

print(student1.introduce())
print(student3.introduce())
print(Student.get_school_info())  # í´ë˜ìŠ¤ ë©”ì„œë“œëŠ” í´ë˜ìŠ¤ë¡œ í˜¸ì¶œ
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ë‚ ì§œ íŒŒì‹±ì„ ìœ„í•œ í´ë˜ìŠ¤ ë©”ì„œë“œ

````python
from datetime import datetime

class DateParser:
    """ë‹¤ì–‘í•œ í˜•ì‹ì˜ ë‚ ì§œë¥¼ íŒŒì‹±í•˜ëŠ” í´ë˜ìŠ¤"""
    
    def __init__(self, date):
        self.date = date
    
    # ëŒ€ì²´ ìƒì„±ìë“¤ - ë‹¤ì–‘í•œ í˜•ì‹ ì§€ì›
    @classmethod
    def from_iso(cls, date_str):
        """ISO í˜•ì‹: 2024-01-15"""
        date = datetime.strptime(date_str, "%Y-%m-%d")
        return cls(date)
    
    @classmethod
    def from_korean(cls, date_str):
        """í•œêµ­ í˜•ì‹: 2024ë…„ 1ì›” 15ì¼"""
        date = datetime.strptime(date_str, "%Yë…„ %mì›” %dì¼")
        return cls(date)
    
    @classmethod
    def from_slash(cls, date_str):
        """ìŠ¬ë˜ì‹œ í˜•ì‹: 01/15/2024"""
        date = datetime.strptime(date_str, "%m/%d/%Y")
        return cls(date)
    
    def format(self):
        return self.date.strftime("%Yë…„ %mì›” %dì¼")

# ë‹¤ì–‘í•œ í˜•ì‹ì˜ ë‚ ì§œë¥¼ ì¼ê´€ëœ ë°©ì‹ìœ¼ë¡œ ì²˜ë¦¬
date1 = DateParser.from_iso("2024-01-15")
date2 = DateParser.from_korean("2024ë…„ 3ì›” 20ì¼")
date3 = DateParser.from_slash("12/31/2024")

print(date1.format())
print(date2.format())
print(date3.format())
````

````python
import statistics

class BaseAnalyzer:
    """ê¸°ë³¸ ë°ì´í„° ë¶„ì„ê¸°"""
    
    def __init__(self, data):
        self.data = data
    
    def mean(self):
        return statistics.mean(self.data)
    
    def median(self):
        return statistics.median(self.data)
    
    def summary(self):
        return {
            'count': len(self.data),
            'mean': self.mean(),
            'median': self.median()
        }

class SalesAnalyzer(BaseAnalyzer):
    """ë§¤ì¶œ ë°ì´í„° ì „ìš© ë¶„ì„ê¸°"""
    
    def __init__(self, data, product_name):
        super().__init__(data)
        self.product_name = product_name
    
    def total_revenue(self):
        """ì´ ë§¤ì¶œ"""
        return sum(self.data)
    
    def summary(self):
        # ë¶€ëª¨ ë©”ì„œë“œ í™•ì¥
        base_summary = super().summary()
        base_summary['product'] = self.product_name
        base_summary['total'] = self.total_revenue()
        return base_summary

# ì‚¬ìš©
sales = SalesAnalyzer([100, 150, 200, 180, 220], "ë…¸íŠ¸ë¶")
print(sales.summary())
````

## 2.2 @property: ìº¡ìŠí™”ì™€ ë°ì´í„° ê²€ì¦

### ê¸°ë³¸ ì‚¬ìš©ë²•

````python
# property ì—†ì´ (ë¬¸ì œì )
class ProductBad:
    def __init__(self, name, price):
        self.name = name
        self.price = price  # ìŒìˆ˜ ê°€ê²©ë„ í—ˆìš©ë¨! (ë¬¸ì œ)

# ìŒìˆ˜ ê°€ê²©ì´ ë“¤ì–´ê°ˆ ìˆ˜ ìˆìŒ
bad_product = ProductBad("ë…¸íŠ¸ë¶", -1000)  # ê²€ì¦ ì•ˆ ë¨!
print(f"ê°€ê²©: {bad_product.price}ì›")  # -1000ì› (ë¬¸ì œ!)
````

````python
# property ì‚¬ìš© (í•´ê²°)
class ProductGood:
    def __init__(self, name, price):
        self.name = name
        self._price = 0  # private ë³€ìˆ˜ (ê´€ë¡€: _ prefix)
        self.price = price  # setterë¥¼ í†µí•´ ê²€ì¦
    
    @property
    def price(self):
        """ê°€ê²© ì¡°íšŒ (getter)"""
        return self._price
    
    @price.setter
    def price(self, value):
        """ê°€ê²© ì„¤ì • (setter) - ê²€ì¦ í¬í•¨"""
        if value < 0:
            raise ValueError("ê°€ê²©ì€ 0 ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤.")
        self._price = value
    
    @property
    def formatted_price(self):
        """í¬ë§·ëœ ê°€ê²© (ì½ê¸° ì „ìš©)"""
        return f"{self._price:,}ì›"

# ì‚¬ìš©
product = ProductGood("ë…¸íŠ¸ë¶", 1500000)
print(f"ê°€ê²©: {product.price}")  # getter í˜¸ì¶œ
print(f"í¬ë§·: {product.formatted_price}")

# ê°€ê²© ë³€ê²½ ì‹œ ìë™ ê²€ì¦
product.price = 1800000  # setter í˜¸ì¶œ
print(f"ë³€ê²½ëœ ê°€ê²©: {product.formatted_price}")

# ìŒìˆ˜ ê°€ê²© ì„¤ì • ì‹œë„
try:
    product.price = -1000  # ValueError ë°œìƒ!
except ValueError as e:
    print(f"ì—ëŸ¬: {e}")
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ì˜¨ë„ ë³€í™˜ í´ë˜ìŠ¤

````python
class Temperature:
    """ì˜¨ë„ ê´€ë¦¬ í´ë˜ìŠ¤ (ì„­ì”¨ â†” í™”ì”¨ ìë™ ë³€í™˜)"""
    
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        """ì„­ì”¨ ì˜¨ë„"""
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        if value < -273.15:  # ì ˆëŒ€ì˜ë„ ê²€ì¦
            raise ValueError("ì˜¨ë„ëŠ” ì ˆëŒ€ì˜ë„(-273.15Â°C) ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤.")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        """í™”ì”¨ ì˜¨ë„ (ìë™ ë³€í™˜)"""
        return (self._celsius * 9/5) + 32
    
    @fahrenheit.setter
    def fahrenheit(self, value):
        """í™”ì”¨ë¡œ ì„¤ì •í•˜ë©´ ì„­ì”¨ë¡œ ìë™ ë³€í™˜"""
        self.celsius = (value - 32) * 5/9

# ì‚¬ìš©
temp = Temperature(25)
print(f"{temp.celsius}Â°C = {temp.fahrenheit}Â°F")

# í™”ì”¨ë¡œ ì„¤ì •
temp.fahrenheit = 86
print(f"{temp.celsius}Â°C = {temp.fahrenheit}Â°F")
````

## 2.4 ë§¤ì§ ë©”ì„œë“œ (Magic Methods)

**ë§¤ì§ ë©”ì„œë“œ**ëŠ” `__method__` í˜•ì‹ìœ¼ë¡œ, Pythonì˜ íŠ¹ìˆ˜í•œ ë™ì‘ì„ ì •ì˜í•©ë‹ˆë‹¤.  
ê°ì²´ì˜ ì¶œë ¥ í˜•ì‹, ì—°ì‚°ì ë™ì‘ ë“±ì„ ì»¤ìŠ¤í„°ë§ˆì´ì§•í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### ìì£¼ ì‚¬ìš©í•˜ëŠ” ë§¤ì§ ë©”ì„œë“œ

| ë©”ì„œë“œ | ì„¤ëª… | í˜¸ì¶œ ì‹œì  |
|--------|------|----------|
| `__init__` | ìƒì„±ì | ì¸ìŠ¤í„´ìŠ¤ ìƒì„± ì‹œ |
| `__str__` | ë¬¸ìì—´ í‘œí˜„ (ì‚¬ìš©ììš©) | `str()`, `print()` |
| `__repr__` | ë¬¸ìì—´ í‘œí˜„ (ê°œë°œììš©) | `repr()`, ì½˜ì†” ì¶œë ¥ |
| `__len__` | ê¸¸ì´ ë°˜í™˜ | `len()` |
| `__getitem__` | ì¸ë±ì‹± | `obj[key]` |
| `__add__` | ë§ì…ˆ | `obj1 + obj2` |

````python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    # ì‚¬ìš©ì ì¹œí™”ì ì¸ ì¶œë ¥
    def __str__(self):
        return f"'{self.title}' by {self.author}"
    
    # ê°œë°œììš© ìƒì„¸ ì¶œë ¥
    def __repr__(self):
        return f"Book(title='{self.title}', author='{self.author}', pages={self.pages})"
    
    # len() ì§€ì›
    def __len__(self):
        return self.pages
    
    # + ì—°ì‚°ì ì˜¤ë²„ë¡œë”© (ì±… í•©ì¹˜ê¸°)
    def __add__(self, other):
        total_pages = self.pages + other.pages
        return Book(f"{self.title} & {other.title}", "í•©ë³¸", total_pages)

# ì‚¬ìš©
book1 = Book("íŒŒì´ì¬ ê¸°ì´ˆ", "ë‹¨í…Œ", 300)
book2 = Book("ë°ì´í„° ë¶„ì„", "ë‹¨í…Œ", 250)

print(book1)  # __str__ í˜¸ì¶œ
print(repr(book1))  # __repr__ í˜¸ì¶œ
print(f"í˜ì´ì§€ ìˆ˜: {len(book1)}ìª½")  # __len__ í˜¸ì¶œ

# ë‘ ì±… í•©ì¹˜ê¸°
combined = book1 + book2  # __add__ í˜¸ì¶œ
print(combined)
print(f"ì´ í˜ì´ì§€: {len(combined)}ìª½")
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ë²¡í„° í´ë˜ìŠ¤

````python
import math

class Vector2D:
    """2D ë²¡í„° (ë°ì´í„° ì‹œê°í™”ì—ì„œ ì¢Œí‘œ ê³„ì‚° ì‹œ ìœ ìš©)"""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"
    
    def __repr__(self):
        return f"Vector2D(x={self.x}, y={self.y})"
    
    # ë²¡í„° ë§ì…ˆ
    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)
    
    # ë²¡í„° ëº„ì…ˆ
    def __sub__(self, other):
        return Vector2D(self.x - other.x, self.y - other.y)
    
    # ìŠ¤ì¹¼ë¼ ê³±ì…ˆ
    def __mul__(self, scalar):
        return Vector2D(self.x * scalar, self.y * scalar)
    
    # ì ˆëŒ“ê°’ (ë²¡í„° í¬ê¸°)
    def __abs__(self):
        return math.sqrt(self.x**2 + self.y**2)
    
    # ë™ë“±ì„± ë¹„êµ
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

# ì‚¬ìš©
v1 = Vector2D(3, 4)
v2 = Vector2D(1, 2)

print(v1)  # Vector(3, 4)
print(f"ë²¡í„° í¬ê¸°: {abs(v1)}")  # 5.0

v3 = v1 + v2  # ë²¡í„° ë§ì…ˆ
print(f"v1 + v2 = {v3}")

v4 = v1 * 2  # ìŠ¤ì¹¼ë¼ ê³±
print(f"v1 * 2 = {v4}")

print(f"v1 == v2? {v1 == v2}")
````
