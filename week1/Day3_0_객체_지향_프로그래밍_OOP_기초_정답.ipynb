## Q4. @propertyë¡œ ë°ì´í„° ê²€ì¦ â­â­â­â­

**ë¬¸ì œ**: `BankAccount` í´ë˜ìŠ¤ë¥¼ ë§Œë“œì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:

- `@balance.setter`ë¡œ setter êµ¬í˜„ (ìŒìˆ˜ ê¸ˆì•¡ ê²€ì¦)
- ë©”ì„œë“œ: `deposit(amount)`, `withdraw(amount)`

```python
# ì •ë‹µ
class BankAccount:
    def __init__(self, owner, balance=0):
        """ê³„ì¢Œ ì´ˆê¸°í™”"""
        self.owner = owner
        self._balance = 0  # private ë³€ìˆ˜ (ê´€ë¡€: _ prefix)
        self.balance = balance  # setterë¥¼ í†µí•´ ê²€ì¦
    
    @property
    def balance(self):
        """ì”ì•¡ ì¡°íšŒ (getter)"""
        return self._balance
    
    @balance.setter
    def balance(self, value):
        """ì”ì•¡ ì„¤ì • (setter) - ìŒìˆ˜ ê²€ì¦"""
        if value < 0:
            raise ValueError("ì”ì•¡ì€ 0 ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤.")
        self._balance = value
    
    def deposit(self, amount):
        """ì…ê¸ˆ"""
        if amount <= 0:
            raise ValueError("ì…ê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤.")
        self._balance += amount
        return f"{amount:,}ì› ì…ê¸ˆ ì™„ë£Œ. í˜„ì¬ ì”ì•¡: {self._balance:,}ì›"
    
    def withdraw(self, amount):
        """ì¶œê¸ˆ"""
        if amount <= 0:
            raise ValueError("ì¶œê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤.")
        if amount > self._balance:
            raise ValueError("ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤.")
        self._balance -= amount
        return f"{amount:,}ì› ì¶œê¸ˆ ì™„ë£Œ. í˜„ì¬ ì”ì•¡: {self._balance:,}ì›"

# í…ŒìŠ¤íŠ¸
account = BankAccount("ê¹€ì² ìˆ˜")
print(account.deposit(10000))
print(f"ì”ì•¡: {account.balance:,}ì›")
print(account.withdraw(3000))
print(f"ì”ì•¡: {account.balance:,}ì›")

# ìŒìˆ˜ ì”ì•¡ ì„¤ì • ì‹œë„
try:
    account.balance = -1000  # ì—ëŸ¬ ë°œìƒ
except ValueError as e:
    print(f"ì—ëŸ¬: {e}")
```

## Q5. ë‹¤í˜•ì„± í™œìš© â­â­â­â­

**ë¬¸ì œ**: ë„í˜• í´ë˜ìŠ¤ ê³„ì¸µì„ ë§Œë“œì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:

**Shape í´ë˜ìŠ¤** (ë¶€ëª¨):
- ë©”ì„œë“œ: `area()` - "ë©´ì ì„ ê³„ì‚°í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤" ë°˜í™˜

**Rectangle, Circle í´ë˜ìŠ¤** (Shape ìƒì†):
- Rectangle: `width`, `height` ì†ì„±, `area()` ì˜¤ë²„ë¼ì´ë”©
- Circle: `radius` ì†ì„±, `area()` ì˜¤ë²„ë¼ì´ë”©

**í•¨ìˆ˜**: `print_area(shape)` - ë‹¤í˜•ì„±ìœ¼ë¡œ ëª¨ë“  ë„í˜• ì²˜ë¦¬

```python
import math

# ì •ë‹µ
class Shape:
    """ë„í˜• ê¸°ë³¸ í´ë˜ìŠ¤"""
    
    def area(self):
        """ë©´ì  ê³„ì‚° - ìì‹ í´ë˜ìŠ¤ì—ì„œ ì˜¤ë²„ë¼ì´ë”© í•„ìš”"""
        return "ë©´ì ì„ ê³„ì‚°í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤"

class Rectangle(Shape):
    """ì§ì‚¬ê°í˜•"""
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        """ì§ì‚¬ê°í˜• ë©´ì  = ê°€ë¡œ Ã— ì„¸ë¡œ"""
        return self.width * self.height

class Circle(Shape):
    """ì›"""
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        """ì› ë©´ì  = Ï€ Ã— ë°˜ì§€ë¦„Â²"""
        return math.pi * self.radius ** 2

def print_area(shape):
    """ë‹¤í˜•ì„±: ì–´ë–¤ ë„í˜•ì´ë“  ê°™ì€ ë°©ì‹ìœ¼ë¡œ ì²˜ë¦¬"""
    area = shape.area()  # ê° ë„í˜•ì— ë§ëŠ” area() ë©”ì„œë“œ í˜¸ì¶œ
    
    if isinstance(area, (int, float)):
        print(f"{shape.__class__.__name__} ë©´ì : {area:.2f}")
    else:
        print(f"{shape.__class__.__name__}: {area}")

# í…ŒìŠ¤íŠ¸
shapes = [
    Rectangle(5, 10),
    Circle(7),
    Shape()
]

for shape in shapes:
    print_area(shape)
```

### ğŸ’¡ í•´ì„¤

- **ë‹¤í˜•ì„± (Polymorphism)**:
  - "ê°™ì€ ì¸í„°í˜ì´ìŠ¤, ë‹¤ë¥¸ ë™ì‘"
  - ëª¨ë“  ë„í˜•ì´ `area()` ë©”ì„œë“œë¥¼ ê°€ì§€ì§€ë§Œ, ê°ì ë‹¤ë¥´ê²Œ êµ¬í˜„

## Q6. ë§¤ì§ ë©”ì„œë“œ ì¢…í•© â­â­â­â­â­

**ë¬¸ì œ**: `Money` í´ë˜ìŠ¤ë¥¼ ë§Œë“œì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
- ì†ì„±: `amount`, `currency`
- `__str__`: "ê¸ˆì•¡ í†µí™”" í˜•ì‹ (ì˜ˆ: "1,000 KRW")
- `__repr__`: "Money(amount=..., currency='...')" í˜•ì‹
- `__add__`: ê°™ì€ í†µí™”ë¼ë¦¬ë§Œ ë§ì…ˆ ê°€ëŠ¥ (ë‹¤ë¥´ë©´ ValueError)
- `__eq__`: ê¸ˆì•¡ê³¼ í†µí™”ê°€ ëª¨ë‘ ê°™ì•„ì•¼ True

```python
# ì •ë‹µ
class Money:
    """ê¸ˆì•¡ê³¼ í†µí™”ë¥¼ ê´€ë¦¬í•˜ëŠ” í´ë˜ìŠ¤"""
    
    def __init__(self, amount, currency):
        self.amount = amount
        self.currency = currency
    
    def __str__(self):
        """ì‚¬ìš©ì ì¹œí™”ì  ì¶œë ¥ (print, str)"""
        return f"{self.amount:,} {self.currency}"
    
    def __repr__(self):
        """ê°œë°œììš© ìƒì„¸ ì¶œë ¥ (repr, ì½˜ì†”)"""
        return f"Money(amount={self.amount}, currency='{self.currency}')"
    
    def __add__(self, other):
        """ë§ì…ˆ ì—°ì‚°ì (+) ì˜¤ë²„ë¡œë”©"""
        if not isinstance(other, Money):
            raise TypeError("Money ê°ì²´ë¼ë¦¬ë§Œ ë§ì…ˆí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
        
        if self.currency != other.currency:
            raise ValueError(f"í†µí™”ê°€ ë‹¤ë¦…ë‹ˆë‹¤: {self.currency} vs {other.currency}")
        
        return Money(self.amount + other.amount, self.currency)
    
    def __eq__(self, other):
        """ë™ë“±ì„± ë¹„êµ (==)"""
        if not isinstance(other, Money):
            return False
        return self.amount == other.amount and self.currency == other.currency
    
    def __lt__(self, other):
        """ì‘ìŒ ë¹„êµ (<)"""
        if not isinstance(other, Money):
            raise TypeError("Money ê°ì²´ë¼ë¦¬ë§Œ ë¹„êµí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
        
        if self.currency != other.currency:
            raise ValueError(f"í†µí™”ê°€ ë‹¤ë¦…ë‹ˆë‹¤: {self.currency} vs {other.currency}")
        
        return self.amount < other.amount
    
    def __le__(self, other):
        """ì‘ê±°ë‚˜ ê°™ìŒ ë¹„êµ (<=)"""
        return self == other or self < other
    
    def __gt__(self, other):
        """í¼ ë¹„êµ (>)"""
        return not self <= other
    
    def __ge__(self, other):
        """í¬ê±°ë‚˜ ê°™ìŒ ë¹„êµ (>=)"""
        return not self < other

# í…ŒìŠ¤íŠ¸
krw1 = Money(10000, "KRW")
krw2 = Money(5000, "KRW")
usd = Money(100, "USD")

print(krw1)  # __str__ í˜¸ì¶œ
print(repr(krw2))  # __repr__ í˜¸ì¶œ

# ë§ì…ˆ (__add__)
total = krw1 + krw2
print(f"í•©ê³„: {total}")

# ë¹„êµ (__gt__, __eq__)
print(f"krw1 > krw2? {krw1 > krw2}")
print(f"krw1 == krw2? {krw1 == krw2}")

# ë‹¤ë¥¸ í†µí™” ë§ì…ˆ (ì—ëŸ¬)
try:
    wrong = krw1 + usd
except ValueError as e:
    print(f"ì—ëŸ¬: {e}")
```

### ğŸ’¡ í•´ì„¤

- **`__str__` vs `__repr__`**:
  - `__str__`: ì‚¬ìš©ì ì¹œí™”ì  ì¶œë ¥ (`print()`, `str()`)
  - `__repr__`: ê°œë°œììš© ìƒì„¸ ì¶œë ¥ (ì½˜ì†”, `repr()`)
