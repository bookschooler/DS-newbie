### 3.2 Counter: ë¹ˆë„ ë¶„ì„ì˜ ê°•ë ¥í•œ ë„êµ¬

````python
from collections import Counter

# [ê¸°ë³¸] ë¦¬ìŠ¤íŠ¸ì˜ ìš”ì†Œ ê°œìˆ˜ ì„¸ê¸°
fruits = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
counter = Counter(fruits)

print(counter)
print(f"apple ê°œìˆ˜: {counter['apple']}")
print(f"grape ê°œìˆ˜: {counter['grape']}")  # ì—†ìœ¼ë©´ 0 ë°˜í™˜
````

````python
# [ì‹¤ë¬´ ì˜ˆì œ] í…ìŠ¤íŠ¸ ë‹¨ì–´ ë¹ˆë„ ë¶„ì„
text = "data science is amazing data analysis is powerful data is everywhere"
words = text.split()

word_count = Counter(words)
print("ë‹¨ì–´ ë¹ˆë„:")
print(word_count)

# ê°€ì¥ ë§ì´ ë‚˜ì˜¨ ë‹¨ì–´ ìƒìœ„ 3ê°œ
print("\nê°€ì¥ ë§ì´ ë‚˜ì˜¨ ë‹¨ì–´:")
for word, count in word_count.most_common(3):
    print(f"  {word}: {count}íšŒ")
````

````python
# [ì‹¤ìŠµ] ë¬¸ì ë¹ˆë„ ë¶„ì„
sentence = "hello world"
char_counter = Counter(sentence)

print("ë¬¸ì ë¹ˆë„:")
for char, count in char_counter.most_common():
    if char != ' ':  # ê³µë°± ì œì™¸
        print(f"  '{char}': {count}íšŒ")
````

````python
# Counterì˜ ì‚°ìˆ  ì—°ì‚°
c1 = Counter(['a', 'b', 'c', 'a'])
c2 = Counter(['a', 'c', 'd'])

print("c1:", c1)
print("c2:", c2)
print("c1 + c2:", c1 + c2)  # í•©ì§‘í•©
print("c1 - c2:", c1 - c2)  # ì°¨ì§‘í•©
print("c1 & c2:", c1 & c2)  # êµì§‘í•© (ìµœì†Œê°’)
print("c1 | c2:", c1 | c2)  # í•©ì§‘í•© (ìµœëŒ€ê°’)
````

````python
# [ì‹¤ë¬´ ì˜ˆì œ] ê·¸ë£¹í•‘ ì‘ì—…
students = [
    ('Alice', 'Math'),
    ('Bob', 'Science'),
    ('Charlie', 'Math'),
    ('David', 'Science'),
    ('Eve', 'Math')
]

# ê³¼ëª©ë³„ë¡œ í•™ìƒì„ ê·¸ë£¹í•‘
subject_groups = defaultdict(list)  # ê¸°ë³¸ê°’: ë¹ˆ ë¦¬ìŠ¤íŠ¸

for name, subject in students:
    subject_groups[subject].append(name)

print("ê³¼ëª©ë³„ í•™ìƒ:")
for subject, names in subject_groups.items():
    print(f"  {subject}: {names}")
````

````python
# [ì‹¤ìŠµ] ë‹¨ì–´ ê¸¸ì´ë³„ ê·¸ë£¹í•‘
words = ['cat', 'dog', 'elephant', 'ant', 'bear', 'lion']

length_groups = defaultdict(list)
for word in words:
    length_groups[len(word)].append(word)

print("ê¸¸ì´ë³„ ë‹¨ì–´:")
for length, word_list in sorted(length_groups.items()):
    print(f"  {length}ê¸€ì: {word_list}")
````

### 3.4 deque: ì–‘ë°©í–¥ í

`deque` (Double-Ended Queue): ì–‘ìª½ ëì—ì„œ ë¹ ë¥´ê²Œ ì¶”ê°€/ì œê±°í•  ìˆ˜ ìˆëŠ” ìë£Œêµ¬ì¡°

**ë¦¬ìŠ¤íŠ¸ vs deque**:
* ë¦¬ìŠ¤íŠ¸: ëì— ì¶”ê°€/ì œê±°ëŠ” ë¹ ë¥´ì§€ë§Œ **ì•**ì— ì¶”ê°€/ì œê±°ëŠ” ëŠë¦¼ (O(n))
* deque: ì–‘ìª½ ë ëª¨ë‘ ë¹ ë¦„ (O(1))

````python
from collections import deque

# [ê¸°ë³¸] deque ìƒì„±
dq = deque([1, 2, 3])
print("ì´ˆê¸° deque:", dq)

# ì˜¤ë¥¸ìª½ì— ì¶”ê°€/ì œê±°
dq.append(4)
print("append(4):", dq)

removed = dq.pop()
print(f"pop() â†’ {removed}, ê²°ê³¼:", dq)
````

````python
# ì™¼ìª½ì— ì¶”ê°€/ì œê±° (ë¦¬ìŠ¤íŠ¸ë³´ë‹¤ í›¨ì”¬ ë¹ ë¦„!)
dq.appendleft(0)
print("appendleft(0):", dq)

removed = dq.popleft()
print(f"popleft() â†’ {removed}, ê²°ê³¼:", dq)
````

````python
# [ì‹¤ë¬´ ì˜ˆì œ] ìµœê·¼ Nê°œ í•­ëª© ìœ ì§€ (ìŠ¬ë¼ì´ë”© ìœˆë„ìš°)
recent_logs = deque(maxlen=5)  # ìµœëŒ€ 5ê°œë§Œ ìœ ì§€

for i in range(10):
    recent_logs.append(f"ë¡œê·¸ {i}")
    print(f"ì¶”ê°€ í›„: {list(recent_logs)}")

print(f"\nìµœì¢… ìµœê·¼ 5ê°œ: {list(recent_logs)}")
````

````python
# [ì‹¤ìŠµ] rotate: íšŒì „ ê¸°ëŠ¥
dq = deque([1, 2, 3, 4, 5])
print("ì›ë³¸:", dq)

dq.rotate(2)  # ì˜¤ë¥¸ìª½ìœ¼ë¡œ 2ì¹¸ íšŒì „
print("rotate(2):", dq)

dq.rotate(-2)  # ì™¼ìª½ìœ¼ë¡œ 2ì¹¸ íšŒì „ (ì›ìƒë³µêµ¬)
print("rotate(-2):", dq)
````

````python
# [ì‹¤ë¬´ ì˜ˆì œ] BFS íë¡œ ì‚¬ìš©
# ë„ˆë¹„ ìš°ì„  íƒìƒ‰(Breadth-First Search)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []
    
    while queue:
        node = queue.popleft()  # íì—ì„œ êº¼ë‚´ê¸°
        if node not in visited:
            visited.add(node)
            result.append(node)
            queue.extend(graph[node])  # ì´ì›ƒ ë…¸ë“œ ì¶”ê°€
    
    return result

print("BFS ìˆœíšŒ:", bfs(graph, 'A'))
````

````python
# [Q2 ì •ë‹µ]
from collections import Counter

statistics_calc_code = '''
"""í†µê³„ ê³„ì‚° ëª¨ë“ˆ"""
from collections import Counter

def mean(numbers):
    """í‰ê·  ê³„ì‚°"""
    return sum(numbers) / len(numbers)

def median(numbers):
    """ì¤‘ì•™ê°’ ê³„ì‚°"""
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    mid = n // 2
    
    if n % 2 == 0:
        return (sorted_nums[mid - 1] + sorted_nums[mid]) / 2
    else:
        return sorted_nums[mid]

def mode(numbers):
    """ìµœë¹ˆê°’ ê³„ì‚° (ê°€ì¥ ë§ì´ ë‚˜ì˜¨ ê°’)"""
    counter = Counter(numbers)
    return counter.most_common(1)[0][0]

# í…ŒìŠ¤íŠ¸ ì½”ë“œ
if __name__ == "__main__":
    print("í†µê³„ ê³„ì‚° ëª¨ë“ˆ í…ŒìŠ¤íŠ¸")
    
    data = [1, 2, 2, 3, 4, 4, 4, 5]
    print(f"ë°ì´í„°: {data}")
    print(f"í‰ê· (mean): {mean(data)}")
    print(f"ì¤‘ì•™ê°’(median): {median(data)}")
    print(f"ìµœë¹ˆê°’(mode): {mode(data)}")
'''

with open('statistics_calc.py', 'w', encoding='utf-8') as f:
    f.write(statistics_calc_code)

print("í…ŒìŠ¤íŠ¸ 1: ì§ì ‘ ì‹¤í–‰")
!python statistics_calc.py

print("\ní…ŒìŠ¤íŠ¸ 2: import (í…ŒìŠ¤íŠ¸ ì½”ë“œê°€ ì‹¤í–‰ë˜ì§€ ì•ŠìŒ)")
import statistics_calc
print(statistics_calc.mean([10, 20, 30]))
````

### Q4. [defaultdict] í•™ìƒ ì„±ì  ê´€ë¦¬ (ë‚œì´ë„: â­â­â­)

````python
# [Q4 ì •ë‹µ]
from collections import defaultdict

grades = [
    ('Alice', 'Math', 90),
    ('Bob', 'Math', 85),
    ('Alice', 'Science', 95),
    ('Charlie', 'Math', 88),
    ('Bob', 'Science', 92),
    ('Alice', 'English', 87)
]

# 1. í•™ìƒë³„ ê·¸ë£¹í•‘
student_grades = defaultdict(dict)
for student, subject, score in grades:
    student_grades[student][subject] = score

print("í•™ìƒë³„ ì„±ì :")
for student, subjects in student_grades.items():
    print(f"  {student}: {subjects}")

# 2. í•™ìƒë³„ í‰ê· 
print("\ní•™ìƒë³„ í‰ê·  ì ìˆ˜:")
for student, subjects in student_grades.items():
    avg = sum(subjects.values()) / len(subjects)
    print(f"  {student}: {avg:.2f}ì ")

# 3. ê³¼ëª©ë³„ í‰ê· 
subject_scores = defaultdict(list)
for student, subject, score in grades:
    subject_scores[subject].append(score)

print("\nê³¼ëª©ë³„ í‰ê·  ì ìˆ˜:")
for subject, scores in subject_scores.items():
    avg = sum(scores) / len(scores)
    print(f"  {subject}: {avg:.2f}ì ")
````

### Q5. [deque] ìŠ¬ë¼ì´ë”© ìœˆë„ìš° í‰ê·  (ë‚œì´ë„: â­â­â­)

````python
# [Q5 ì •ë‹µ]
from collections import deque

def moving_average(prices, window_size):
    """
    ì´ë™ í‰ê·  ê³„ì‚°
    
    Args:
        prices (list): ê°€ê²© ë°ì´í„°
        window_size (int): ìœˆë„ìš° í¬ê¸°
    
    Returns:
        list: ê° ì‹œì ì˜ ì´ë™ í‰ê· 
    """
    window = deque(maxlen=window_size)
    result = []
    
    for price in prices:
        window.append(price)
        # í˜„ì¬ ìœˆë„ìš°ì˜ í‰ê·  ê³„ì‚°
        avg = sum(window) / len(window)
        result.append(round(avg, 2))
    
    return result

# í…ŒìŠ¤íŠ¸
prices = [100, 102, 101, 105, 103, 107, 110]
result = moving_average(prices, 3)
print("ì›ë³¸ ê°€ê²©:", prices)
print("ì´ë™ í‰ê· (ìœˆë„ìš°=3):", result)
````

### Q6. [ì¢…í•©] í…ìŠ¤íŠ¸ ë¶„ì„ ëª¨ë“ˆ ë§Œë“¤ê¸° (ë‚œì´ë„: â­â­â­â­)

````python
# [Q6 ì •ë‹µ]
text_analyzer_code = '''
"""í…ìŠ¤íŠ¸ ë¶„ì„ ëª¨ë“ˆ"""
from collections import Counter, defaultdict

def word_frequency(text, top_n=10):
    """ê°€ì¥ ë§ì´ ë‚˜ì˜¨ ë‹¨ì–´ top_nê°œë¥¼ ë°˜í™˜"""
    words = text.lower().split()
    counter = Counter(words)
    return counter.most_common(top_n)

def char_frequency(text):
    """ë¬¸ì ë¹ˆë„ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ ë°˜í™˜ (ê³µë°± ì œì™¸)"""
    chars = [char for char in text if char != ' ']
    return dict(Counter(chars))

def word_length_distribution(text):
    """ë‹¨ì–´ ê¸¸ì´ë³„ ë¶„í¬ ë°˜í™˜"""
    words = text.split()
    length_dist = defaultdict(list)
    
    for word in words:
        length_dist[len(word)].append(word)
    
    return dict(length_dist)

if __name__ == "__main__":
    sample = """data science is amazing
data analysis is powerful
python makes data science easy"""
    
    print("=== í…ìŠ¤íŠ¸ ë¶„ì„ ê²°ê³¼ ===")
    print(f"\nì›ë³¸ í…ìŠ¤íŠ¸:\n{sample}\n")
    
    # 1. ë‹¨ì–´ ë¹ˆë„
    print("1. ìƒìœ„ 5ê°œ ë‹¨ì–´:")
    for word, count in word_frequency(sample, 5):
        print(f"   {word}: {count}íšŒ")
    
    # 2. ë¬¸ì ë¹ˆë„
    print("\n2. ë¬¸ì ë¹ˆë„ (ìƒìœ„ 5ê°œ):")
    char_freq = char_frequency(sample)
    for char, count in sorted(char_freq.items(), key=lambda x: x[1], reverse=True)[:5]:
        print(f"   '{char}': {count}íšŒ")
    
    # 3. ë‹¨ì–´ ê¸¸ì´ ë¶„í¬
    print("\n3. ë‹¨ì–´ ê¸¸ì´ë³„ ë¶„í¬:")
    length_dist = word_length_distribution(sample)
    for length in sorted(length_dist.keys()):
        words = length_dist[length]
        print(f"   {length}ê¸€ì ({len(words)}ê°œ): {words[:3]}...")  # ì²˜ìŒ 3ê°œë§Œ
'''

with open('text_analyzer.py', 'w', encoding='utf-8') as f:
    f.write(text_analyzer_code)

print("text_analyzer.py ìƒì„± ì™„ë£Œ!\n")
!python text_analyzer.py
````

### Q7. [ì¢…í•©/ì‹¬í™”] ì¬ì‚¬ìš© ê°€ëŠ¥í•œ ë°ì´í„° ì²˜ë¦¬ ë¼ì´ë¸ŒëŸ¬ë¦¬ (ë‚œì´ë„: â­â­â­â­â­)

````python
# [Q7 ì •ë‹µ]
import os

# 1. ë””ë ‰í† ë¦¬ ìƒì„±
os.makedirs('data_utils', exist_ok=True)

# 2. cleaners.py ìƒì„±
cleaners_code = '''
"""ë°ì´í„° ì •ë¦¬ í•¨ìˆ˜ë“¤"""

def remove_duplicates(data):
    """ë¦¬ìŠ¤íŠ¸ì—ì„œ ì¤‘ë³µ ì œê±° (ìˆœì„œ ìœ ì§€)"""
    seen = set()
    result = []
    for item in data:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

def fill_missing(data, fill_value=0):
    """Noneì„ fill_valueë¡œ ëŒ€ì²´"""
    return [fill_value if x is None else x for x in data]
'''

with open('data_utils/cleaners.py', 'w', encoding='utf-8') as f:
    f.write(cleaners_code)

# 3. validators.py ìƒì„±
validators_code = '''
"""ë°ì´í„° ê²€ì¦ í•¨ìˆ˜ë“¤"""

def is_valid_email(email):
    """ì´ë©”ì¼ í˜•ì‹ ê²€ì¦ (ê°„ë‹¨í•˜ê²Œ @ í¬í•¨ ì—¬ë¶€)"""
    return '@' in email and '.' in email.split('@')[1]

def is_in_range(value, min_val, max_val):
    """ê°’ì´ ë²”ìœ„ ë‚´ì— ìˆëŠ”ì§€ í™•ì¸"""
    return min_val <= value <= max_val
'''

with open('data_utils/validators.py', 'w', encoding='utf-8') as f:
    f.write(validators_code)

# 4. aggregators.py ìƒì„±
aggregators_code = '''
"""ë°ì´í„° ì§‘ê³„ í•¨ìˆ˜ë“¤"""
from collections import defaultdict, Counter

def group_by_key(items, key_func):
    """key_func ê¸°ì¤€ìœ¼ë¡œ ê·¸ë£¹í•‘"""
    groups = defaultdict(list)
    for item in items:
        key = key_func(item)
        groups[key].append(item)
    return dict(groups)

def count_by_category(items, category_key):
    """ì¹´í…Œê³ ë¦¬ë³„ ê°œìˆ˜ ì„¸ê¸°"""
    if callable(category_key):
        # í•¨ìˆ˜ê°€ ì£¼ì–´ì§„ ê²½ìš°
        categories = [category_key(item) for item in items]
    else:
        # ë”•ì…”ë„ˆë¦¬ í‚¤ê°€ ì£¼ì–´ì§„ ê²½ìš°
        categories = [item[category_key] for item in items]
    return dict(Counter(categories))
'''

with open('data_utils/aggregators.py', 'w', encoding='utf-8') as f:
    f.write(aggregators_code)

# 5. __init__.py ìƒì„±
init_code = '''
"""ë°ì´í„° ìœ í‹¸ë¦¬í‹° íŒ¨í‚¤ì§€"""

# ì£¼ìš” í•¨ìˆ˜ë“¤ì„ íŒ¨í‚¤ì§€ ë ˆë²¨ì—ì„œ ë°”ë¡œ ì‚¬ìš© ê°€ëŠ¥í•˜ë„ë¡
from .cleaners import remove_duplicates, fill_missing
from .validators import is_valid_email, is_in_range
from .aggregators import group_by_key, count_by_category

__all__ = [
    'remove_duplicates',
    'fill_missing',
    'is_valid_email',
    'is_in_range',
    'group_by_key',
    'count_by_category'
]
'''

with open('data_utils/__init__.py', 'w', encoding='utf-8') as f:
    f.write(init_code)

print("data_utils íŒ¨í‚¤ì§€ ìƒì„± ì™„ë£Œ!")
print("\nìƒì„±ëœ íŒŒì¼:")
for root, dirs, files in os.walk('data_utils'):
    for file in files:
        print(f"  {os.path.join(root, file)}")
````
