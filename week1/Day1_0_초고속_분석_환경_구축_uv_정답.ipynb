## Q1. uv 버전 확인하기 ⭐


````python
%%bash
# 배시 셀에서 실행
uv --version
````

```
uv 0.5.1
```

## Q2. 프로젝트 초기화하기 ⭐

**문제**: `my_first_project`라는 이름의 새 프로젝트를 만드세요.

### 💡 정답

````python
%%bash
# /tmp 폴더로 이동
cd /tmp

# 디렉토리 생성
mkdir -p my_first_project
cd my_first_project

# uv로 프로젝트 초기화
uv init

# 생성된 파일 확인
echo "=== 생성된 파일 ==="
ls -lh
````

**예상 출력**:
```
Initialized project `my-first-project`
=== 생성된 파일 ===
total 16K
-rw-r--r-- 1 user user   12 Dec 30 10:00 .python-version
-rw-r--r-- 1 user user  156 Dec 30 10:00 pyproject.toml
-rw-r--r-- 1 user user   45 Dec 30 10:00 README.md
-rw-r--r-- 1 user user   89 Dec 30 10:00 hello.py
```

### 📝 해설

**생성된 파일 설명**:

1. **`.python-version`**: 사용할 Python 버전 지정
   ```
   3.12
   ```

2. **`pyproject.toml`**: 프로젝트 설정 파일
   ```toml
   [project]
   name = "my-first-project"
   version = "0.1.0"
   dependencies = []
   ```

3. **`README.md`**: 프로젝트 설명 파일

4. **`hello.py`**: 샘플 Python 스크립트

**명령어 설명**:
- `mkdir -p`: 디렉토리 생성 (`-p`는 부모 디렉토리도 함께 생성)
- `cd`: 디렉토리 이동
- `uv init`: 현재 디렉토리를 uv 프로젝트로 초기화
- `ls -lh`: 파일 목록을 사람이 읽기 쉬운 형태로 표시

**💡 실무 팁**:
- 프로젝트명에는 하이픈(`-`)이 언더스코어(`_`)보다 권장됨
- `pyproject.toml`과 `.python-version`은 Git에 커밋 필수
- `.venv` 폴더는 `.gitignore`에 추가

## Q3. 패키지 추가하기 ⭐⭐

**문제**: 프로젝트에 `pandas` 패키지를 추가하세요.

### 💡 정답

````python
%%bash
cd /tmp/my_first_project

# pandas 추가
uv add pandas

# pyproject.toml 확인
echo "\n=== pyproject.toml 내용 ==="
cat pyproject.toml
````

**예상 출력**:
```
Resolved 6 packages in 245ms
Installed 6 packages in 1.2s
 + numpy==1.26.4
 + pandas==2.2.0
 + python-dateutil==2.8.2
 + pytz==2024.1
 + six==1.16.0
 + tzdata==2024.1

=== pyproject.toml 내용 ===
[project]
name = "my-first-project"
version = "0.1.0"
dependencies = [
    "pandas>=2.2.0",
]
```

### 💡 추가 팁: 여러 패키지 한 번에 추가

````python
%%bash
cd /tmp/my_first_project

# 여러 패키지를 한 번에 추가 (더 빠름!)
uv add pandas numpy matplotlib
````

### 📝 해설

**uv add의 특징**:

1. **자동 의존성 해결**: pandas를 추가하면 numpy도 자동으로 설치됨
2. **pyproject.toml 자동 업데이트**: 수동 편집 불필요
3. **빠른 설치 속도**: pip보다 10-100배 빠름

**명령어 설명**:
- `uv add pandas`: pandas 패키지를 프로젝트에 추가
- `cat pyproject.toml`: 파일 내용 출력

**설치된 패키지들**:
- `pandas`: 데이터 분석 라이브러리 (직접 설치)
- `numpy`: 수치 계산 (pandas의 의존성)
- `python-dateutil`: 날짜 처리 (pandas의 의존성)
- `pytz`, `tzdata`: 시간대 처리 (pandas의 의존성)

**💡 실무 팁**:
- 필요한 패키지만 직접 추가하면, 의존성은 자동으로 설치됨
- `pyproject.toml`에는 직접 추가한 패키지만 기록됨
- 모든 패키지의 정확한 버전은 `uv.lock` 파일에 기록됨

## Q5. 패키지 제거하기 ⭐⭐

**문제**: 설치했던 `pandas` 패키지를 제거하세요.

### 💡 정답

````python
%%bash
cd /tmp/my_first_project

# pandas 제거
uv remove pandas

echo "\n=== 제거 후 패키지 목록 ==="
uv pip list
````

**예상 출력**:
```
Resolved 0 packages in 12ms
Uninstalled 6 packages in 45ms
 - numpy==1.26.4
 - pandas==2.2.0
 - python-dateutil==2.8.2
 - pytz==2024.1
 - six==1.16.0
 - tzdata==2024.1

=== 제거 후 패키지 목록 ===
Package    Version
---------- -------
(빈 목록)
```

### 📝 해설

**uv remove의 동작**:

1. **주 패키지 제거**: pandas를 pyproject.toml에서 삭제
2. **불필요한 의존성 제거**: numpy, pytz 등도 함께 제거
3. **자동 정리**: 더 이상 필요 없는 패키지는 모두 삭제

**명령어 설명**:
- `uv remove pandas`: pandas 패키지 제거
- pandas만 제거했는데 6개가 삭제된 이유: 의존성 패키지들이 더 이상 필요 없어짐

**제거 전후 비교**:

| 상태 | 패키지 수 | 설명 |
|------|-----------|------|
| 제거 전 | 6개 | pandas + 의존성 5개 |
| 제거 후 | 0개 | 모두 정리됨 |

**💡 실무 팁**:
- `uv remove`는 불필요한 의존성도 자동으로 정리함
- pip에서는 `pip uninstall pandas`만 실행되고, numpy 등은 남아있음
- 깔끔한 환경 유지: uv가 자동으로 처리

**pip와의 차이**:
```bash
# pip (문제)
pip uninstall pandas  # pandas만 제거, numpy 등은 남음

# uv (자동 정리)
uv remove pandas  # 불필요한 의존성까지 모두 제거
```

## Q6. 개발 의존성 추가하기 ⭐⭐⭐

**문제**: 테스트용 패키지 `pytest`를 **개발 의존성**으로 추가하세요.

### 💡 정답

````python
%%bash
cd /tmp/my_first_project

# pytest를 개발 의존성으로 추가
uv add --dev pytest

echo "\n=== pyproject.toml 확인 ==="
cat pyproject.toml
````

**예상 출력**:
```
Resolved 4 packages in 123ms
Installed 4 packages in 567ms
 + iniconfig==2.0.0
 + packaging==23.2
 + pluggy==1.4.0
 + pytest==8.0.0

=== pyproject.toml 확인 ===
[project]
name = "my-first-project"
version = "0.1.0"
dependencies = []

[project.optional-dependencies]
dev = [
    "pytest>=8.0.0",
]
```

### 📝 해설

**개발 의존성이란?**

프로그램 실행에는 필요 없지만, 개발/테스트할 때만 필요한 패키지:

| 구분 | 예시 | 용도 |
|------|------|------|
| **일반 의존성** | pandas, numpy | 실제 프로그램 실행에 필수 |
| **개발 의존성** | pytest, black | 테스트, 코드 포맷팅 등 |

**pyproject.toml 구조**:

```toml
[project]
dependencies = [           # 프로덕션 필수
    "pandas>=2.0.0",
]

[project.optional-dependencies]
dev = [                    # 개발 전용
    "pytest>=8.0.0",
    "black>=23.0.0",
]
```

**명령어 설명**:
- `uv add --dev pytest`: pytest를 개발 의존성으로 추가
- `--dev` 플래그가 중요: 이게 없으면 일반 의존성으로 추가됨

**왜 분리할까?**

1. **배포 크기 최소화**: 프로덕션에는 실행에 필요한 패키지만
2. **명확한 구분**: 어떤 도구가 개발용인지 한눈에 파악
3. **보안**: 개발 도구는 보안 위험 증가 가능

**💡 실무 팁**:
- 테스트 도구: `pytest`, `pytest-cov` → 개발 의존성
- 코드 포맷팅: `black`, `isort` → 개발 의존성
- 타입 체킹: `mypy` → 개발 의존성
- 실제 분석 라이브러리: `pandas`, `numpy` → 일반 의존성

**일반 의존성 vs 개발 의존성**:
```bash
# 일반 의존성 추가
uv add pandas          # 실행에 필수

# 개발 의존성 추가
uv add --dev pytest    # 개발/테스트용
```

## Q7. 간단한 스크립트 실행하기 ⭐⭐⭐

**문제**: 간단한 Python 스크립트를 만들고 `uv run`으로 실행하세요.

### 💡 정답

````python
%%bash
cd /tmp/my_first_project

# hello.py 파일 생성
cat > hello.py << 'EOF'
print("=== uv 환경 구축 성공! ===")
print("Hello from my first uv project!")
print("이제 데이터 분석을 시작할 준비가 되었습니다!")
EOF

# 스크립트 실행
echo "\n=== 스크립트 실행 ==="
uv run python hello.py
````

**예상 출력**:
```
=== 스크립트 실행 ===
=== uv 환경 구축 성공! ===
Hello from my first uv project!
이제 데이터 분석을 시작할 준비가 되었습니다!
```

### 💡 추가: 생성된 파일 확인

````python
%%bash
cd /tmp/my_first_project

# hello.py 내용 확인
echo "=== hello.py 내용 ==="
cat hello.py
````

### 📝 해설

**uv run의 장점**:

| 비교 | 전통적인 방법 (pip + venv) | uv 방법 |
|------|----------------------------|----------|
| 환경 활성화 | `source .venv/bin/activate` 필요 | **불필요!** |
| 패키지 설치 | `pip install ...` | 자동 설치 |
| 스크립트 실행 | `python script.py` | `uv run python script.py` |

**명령어 설명**:

1. `cat > hello.py << 'EOF'`: hello.py 파일을 생성하고 내용 작성
   - `cat >`: 파일 생성
   - `<< 'EOF'`: 여러 줄 입력 (EOF까지)

2. `uv run python hello.py`: uv 환경에서 스크립트 실행
   - 가상환경 자동 활성화
   - 필요한 패키지 자동 설치
   - 스크립트 실행

**실행 방법 비교**:

```bash
# 전통적인 방법 (pip + venv)
python -m venv .venv
source .venv/bin/activate
pip install pandas
python hello.py
deactivate

# uv 방법
uv run python hello.py  # 끝!
```

**uv run 활용 예시**:

```bash
# Python 스크립트 실행
uv run python script.py

# Jupyter Lab 실행
uv run jupyter lab

# pytest 실행
uv run pytest tests/

# 모듈 직접 실행
uv run python -m http.server
```

**💡 실무 팁**:
- `uv run`은 항상 프로젝트 루트에서 실행 권장
- 가상환경 activate 없이 바로 실행 가능
- CI/CD 파이프라인에서도 `uv run`으로 테스트 실행

**주의사항**:
- 스크립트 내 상대 경로는 실행 위치 기준
- 항상 프로젝트 폴더에서 실행하는 것이 안전

## 🎓 종합 정리

### 학습한 uv 명령어

| 명령어 | 기능 | 사용 예시 |
|--------|------|----------|
| `uv --version` | 버전 확인 | `uv --version` |
| `uv init` | 프로젝트 초기화 | `uv init` |
| `uv add` | 패키지 추가 | `uv add pandas` |
| `uv add --dev` | 개발 패키지 추가 | `uv add --dev pytest` |
| `uv remove` | 패키지 제거 | `uv remove pandas` |
| `uv pip list` | 패키지 목록 | `uv pip list` |
| `uv run` | 스크립트 실행 | `uv run python script.py` |

### pip vs uv 비교

| 작업 | pip (기존 방식) | uv (새로운 방식) |
|------|----------------|------------------|
| 가상환경 생성 | `python -m venv .venv` | 자동 생성 |
| 환경 활성화 | `source .venv/bin/activate` | **불필요!** |
| 패키지 설치 | `pip install pandas` | `uv add pandas` |
| 패키지 제거 | `pip uninstall pandas` | `uv remove pandas` |
| 스크립트 실행 | `python script.py` | `uv run python script.py` |
| 속도 | 느림 | **10-100배 빠름** |

### 💡 실무 활용 팁

1. **새 프로젝트 시작**: 항상 `uv init`으로 시작
2. **패키지 관리**: `uv add`로 추가, 자동으로 `pyproject.toml`에 기록
3. **개발 도구**: 테스트/린팅 도구는 `--dev` 플래그로 추가
4. **팀 협업**: `pyproject.toml`과 `uv.lock`을 Git에 커밋
5. **빠른 실행**: `uv run`으로 activate 없이 바로 실행

### 생성되는 중요 파일

| 파일 | 역할 | Git 커밋 |
|------|------|----------|
| `.python-version` | 사용할 Python 버전 | ✅ 필수 |
| `pyproject.toml` | 프로젝트 설정 및 패키지 목록 | ✅ 필수 |
| `uv.lock` | 정확한 패키지 버전 기록 | ✅ 필수 |
| `.venv/` | 가상환경 폴더 | ❌ .gitignore |

## ✅ 학습 완료 체크리스트

다음 항목을 모두 이해했다면 Day1_0 완료!

- [ ] `uv --version`으로 버전 확인할 수 있다
- [ ] `uv init`으로 새 프로젝트를 만들 수 있다
- [ ] `uv add`로 패키지를 추가할 수 있다
- [ ] `uv pip list`로 설치된 패키지를 확인할 수 있다
- [ ] `uv remove`로 패키지를 제거할 수 있다
- [ ] `uv add --dev`로 개발 의존성을 추가할 수 있다
- [ ] `uv run`으로 스크립트를 실행할 수 있다
- [ ] `pyproject.toml`과 `uv.lock`의 차이를 이해했다
- [ ] uv와 pip의 차이를 설명할 수 있다
