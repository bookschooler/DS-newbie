````python
# [ì‹¤ë¬´ ì˜ˆì œ] ì´ì ê³„ì‚°
from decimal import Decimal, ROUND_HALF_UP

principal = Decimal('1000000')  # ì›ê¸ˆ 100ë§Œì›
rate = Decimal('0.035')         # ì—° ì´ìœ¨ 3.5%
months = 12

# ì›” ì´ì ê³„ì‚°
monthly_rate = rate / Decimal('12')
monthly_interest = principal * monthly_rate

print(f"ì›ê¸ˆ: {principal:,}ì›")
print(f"ì—° ì´ìœ¨: {rate * 100}%")
print(f"ì›” ì´ì: {monthly_interest.quantize(Decimal('0.01'), rounding=ROUND_HALF_UP):,}ì›")
````

### 3.3 ì •ë°€ë„ ì œì–´: quantize()

`quantize()` ë©”ì„œë“œë¡œ ì†Œìˆ˜ì  ìë¦¿ìˆ˜ì™€ ë°˜ì˜¬ë¦¼ ë°©ì‹ì„ ì œì–´í•©ë‹ˆë‹¤.

### 3.4 ì‹¤ë¬´ í™œìš©: ê°€ê²© ê³„ì‚° ì‹œìŠ¤í…œ

````python
# [ì‹¤ë¬´ ì˜ˆì œ] VAT í¬í•¨ ê°€ê²© ê³„ì‚°
from decimal import Decimal, ROUND_HALF_UP

def calculate_total_price(unit_price, quantity, vat_rate=Decimal('0.1')):
    """
    VAT í¬í•¨ ì´ ê°€ê²© ê³„ì‚°
    
    Args:
        unit_price: ë‹¨ê°€ (ë¬¸ìì—´ ë˜ëŠ” Decimal)
        quantity: ìˆ˜ëŸ‰ (ì •ìˆ˜)
        vat_rate: ë¶€ê°€ì„¸ìœ¨ (ê¸°ë³¸ 10%)
    
    Returns:
        dict: ìƒì„¸ ê°€ê²© ì •ë³´
    """
    price = Decimal(str(unit_price))
    qty = Decimal(str(quantity))
    
    subtotal = price * qty
    vat = (subtotal * vat_rate).quantize(Decimal('0.01'), rounding=ROUND_HALF_UP)
    total = subtotal + vat
    
    return {
        'subtotal': subtotal,
        'vat': vat,
        'total': total
    }

# í…ŒìŠ¤íŠ¸
result = calculate_total_price('15000.50', 3)
print(f"ê³µê¸‰ê°€ì•¡: {result['subtotal']:,}ì›")
print(f"ë¶€ê°€ì„¸: {result['vat']:,}ì›")
print(f"ì´ì•¡: {result['total']:,}ì›")
````

### Q3. ë¶€ë™ ì†Œìˆ˜ì  ì •ë°€ë„ ì´ìŠˆ â­â­â­

````python
# [Q3 ì •ë‹µ]
import math

def safe_float_sum(numbers):
    """
    ë¶€ë™ ì†Œìˆ˜ì  ë¦¬ìŠ¤íŠ¸ì˜ í•©ê³„ë¥¼ ê³„ì‚°í•˜ê³  ì˜¤ì°¨ë¥¼ ê²€ì¦
    
    Args:
        numbers (list): ë¶€ë™ ì†Œìˆ˜ì  ìˆ«ì ë¦¬ìŠ¤íŠ¸
    
    Returns:
        tuple: (ì‹¤ì œ í•©ê³„, ì˜ˆìƒ í•©ê³„, ì¼ì¹˜ ì—¬ë¶€)
    """
    actual_sum = sum(numbers)
    expected_sum = len(numbers) * numbers[0]  # ëª¨ë“  ê°’ì´ ê°™ë‹¤ê³  ê°€ì •
    is_close = math.isclose(actual_sum, expected_sum)
    
    print(f"ì¼ë°˜ í•©ê³„: {actual_sum}")
    print(f"ì˜ˆìƒ ê²°ê³¼: {expected_sum}")
    print(f"ì •ë°€ í‘œì‹œ: {actual_sum:.20f}")
    print(f"ì¼ì¹˜ ì—¬ë¶€: {is_close} (ì˜¤ì°¨ ë²”ìœ„ ë‚´)")
    
    return (actual_sum, expected_sum, is_close)

# í…ŒìŠ¤íŠ¸
numbers = [0.1] * 10
result = safe_float_sum(numbers)
````

### Q5. ê¸ˆìœµ ê³„ì‚°: ë³µë¦¬ ì´ì â­â­â­â­

````python
# [Q5 ì •ë‹µ]
from decimal import Decimal, ROUND_HALF_UP

def calculate_compound_interest(principal, rate, years):
    """
    ë³µë¦¬ ì´ì ê³„ì‚°
    
    Args:
        principal (str): ì›ê¸ˆ
        rate (str): ì—° ì´ìœ¨ (ì˜ˆ: '0.05' = 5%)
        years (int): ê¸°ê°„ (ë…„)
    
    Returns:
        dict: {'principal': ì›ê¸ˆ, 'interest': ì´ì, 'total': ì´ì•¡}
    """
    p = Decimal(principal)
    r = Decimal(rate)
    
    # ë³µë¦¬ ê³µì‹: P Ã— (1 + r)^n
    total = p * ((Decimal('1') + r) ** years)
    
    # ì› ë‹¨ìœ„ë¡œ ë°˜ì˜¬ë¦¼
    total = total.quantize(Decimal('1'), rounding=ROUND_HALF_UP)
    interest = total - p
    
    print(f"ì›ê¸ˆ: {p:,}ì›")
    print(f"ì´ì: {interest:,}ì›")
    print(f"ì´ì•¡: {total:,}ì›")
    
    return {
        'principal': p,
        'interest': interest,
        'total': total
    }

# í…ŒìŠ¤íŠ¸
result = calculate_compound_interest('1000000', '0.05', 3)
````

### Q6. ë¬¸ìì—´ í¬ë§·íŒ…: ì¬ë¬´ ë³´ê³ ì„œ ìƒì„± â­â­â­â­

````python
# [Q6 ì •ë‹µ]
quarterly_data = [
    {"quarter": "Q1", "revenue": 15000000, "expense": 12000000, "growth": 0.15},
    {"quarter": "Q2", "revenue": 18000000, "expense": 13500000, "growth": 0.20},
    {"quarter": "Q3", "revenue": 16500000, "expense": 12800000, "growth": -0.08},
    {"quarter": "Q4", "revenue": 21000000, "expense": 15000000, "growth": 0.27}
]

def generate_financial_report(data):
    """
    ë¶„ê¸°ë³„ ì¬ë¬´ ë³´ê³ ì„œ ìƒì„±
    
    Args:
        data (list): ë¶„ê¸°ë³„ ì¬ë¬´ ë°ì´í„°
    """
    # í—¤ë” ì¶œë ¥
    print("=" * 65)
    print(f"{'ë¶„ê¸°':^6} {'ë§¤ì¶œ':^16} {'ë¹„ìš©':^16} {'ì´ìµ':^16} {'ì„±ì¥ë¥ ':^8}")
    print("=" * 65)
    
    # ì§‘ê³„ ë³€ìˆ˜
    total_revenue = 0
    total_expense = 0
    total_profit = 0
    
    # ê° ë¶„ê¸° ë°ì´í„° ì¶œë ¥
    for item in data:
        quarter = item['quarter']
        revenue = item['revenue']
        expense = item['expense']
        profit = revenue - expense
        growth = item['growth']
        
        print(f"{quarter:^6} {revenue:>13,}ì› {expense:>13,}ì› {profit:>13,}ì› {growth:>7.1%}")
        
        # í•©ê³„ ëˆ„ì 
        total_revenue += revenue
        total_expense += expense
        total_profit += profit
    
    # ì—°ê°„ í•©ê³„ ì¶œë ¥
    print("=" * 65)
    print(f"{'ì—°ê°„':^6} {total_revenue:>13,}ì› {total_expense:>13,}ì› {total_profit:>13,}ì›")
    print("=" * 65)

# í…ŒìŠ¤íŠ¸
generate_financial_report(quarterly_data)
````

````python
# [Q7 ì •ë‹µ]
from decimal import Decimal, ROUND_HALF_UP

order_items = [
    {"name": "ë…¸íŠ¸ë¶", "price": "1200000", "quantity": 1, "discount": "0.1"},
    {"name": "ë§ˆìš°ìŠ¤", "price": "35000", "quantity": 2, "discount": "0.05"},
    {"name": "í‚¤ë³´ë“œ", "price": "89000", "quantity": 1, "discount": "0.15"}
]

def calculate_order_total(items, shipping_threshold=50000, shipping_fee=3000):
    """
    ì£¼ë¬¸ ì´ì•¡ ê³„ì‚° ë° ì˜ìˆ˜ì¦ ì¶œë ¥
    
    Args:
        items (list): ì£¼ë¬¸ ìƒí’ˆ ëª©ë¡
        shipping_threshold (int): ë¬´ë£Œ ë°°ì†¡ ê¸°ì¤€ ê¸ˆì•¡
        shipping_fee (int): ë°°ì†¡ë¹„
    
    Returns:
        Decimal: ìµœì¢… ê²°ì œ ê¸ˆì•¡
    """
    print("=" * 45)
    print(f"{'ì£¼ë¬¸ ì˜ìˆ˜ì¦':^45}")
    print("=" * 45)
    
    subtotal = Decimal('0')
    
    # ê° ìƒí’ˆ ì²˜ë¦¬
    for item in items:
        name = item['name']
        price = Decimal(item['price'])
        quantity = item['quantity']
        discount_rate = Decimal(item['discount'])
        
        # ìƒí’ˆ ê¸ˆì•¡ ê³„ì‚°
        item_total = price * quantity
        discount_amount = (item_total * discount_rate).quantize(Decimal('1'), ROUND_HALF_UP)
        
        # ìƒí’ˆ ì •ë³´ ì¶œë ¥
        print(f"{name:<8} x{quantity}      {item_total:>12,}ì›")
        print(f"  í• ì¸({discount_rate*100:.0f}%)        {-discount_amount:>12,}ì›")
        
        # í•©ê³„ ëˆ„ì 
        subtotal += item_total - discount_amount
    
    # ë°°ì†¡ë¹„ ê³„ì‚°
    shipping = Decimal('0') if subtotal >= shipping_threshold else Decimal(str(shipping_fee))
    shipping_text = "(ë¬´ë£Œ)" if shipping == 0 else ""
    
    print("-" * 45)
    print(f"ìƒí’ˆ í•©ê³„:         {subtotal:>12,}ì›")
    print(f"ë°°ì†¡ë¹„:            {shipping:>12,}ì› {shipping_text}")
    
    # ì†Œê³„ ë° ë¶€ê°€ì„¸
    subtotal_with_shipping = subtotal + shipping
    vat = (subtotal_with_shipping * Decimal('0.1')).quantize(Decimal('1'), ROUND_HALF_UP)
    
    print("-" * 45)
    print(f"ì†Œê³„:              {subtotal_with_shipping:>12,}ì›")
    print(f"ë¶€ê°€ì„¸(10%):       {vat:>12,}ì›")
    
    # ìµœì¢… ê¸ˆì•¡
    total = subtotal_with_shipping + vat
    
    print("=" * 45)
    print(f"ìµœì¢… ê¸ˆì•¡:         {total:>12,}ì›")
    print("=" * 45)
    
    return total

# í…ŒìŠ¤íŠ¸
total = calculate_order_total(order_items)
````
