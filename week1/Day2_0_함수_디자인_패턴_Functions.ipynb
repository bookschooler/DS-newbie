````python
# [ì˜ˆì œ 13] ìƒì„¸í•œ Docstring (Google Style)
def process_sales_data(data, threshold, include_tax=True):
    """
    ë§¤ì¶œ ë°ì´í„°ë¥¼ ì²˜ë¦¬í•˜ì—¬ ë¶„ì„ ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    
    Args:
        data (list): ë§¤ì¶œ ë°ì´í„° ë¦¬ìŠ¤íŠ¸
        threshold (float): í•„í„°ë§ ê¸°ì¤€ê°’
        include_tax (bool, optional): ì„¸ê¸ˆ í¬í•¨ ì—¬ë¶€. ê¸°ë³¸ê°’ì€ True.
    
    Returns:
        dict: ë¶„ì„ ê²°ê³¼ ë”•ì…”ë„ˆë¦¬
            - 'total': ì´ ë§¤ì¶œ
            - 'count': ê±°ë˜ ê±´ìˆ˜
            - 'average': í‰ê·  ë§¤ì¶œ
    
    Examples:
        >>> process_sales_data([100, 200, 300], 150)
        {'total': 500, 'count': 2, 'average': 250.0}
    """
    filtered = [x for x in data if x >= threshold]
    if include_tax:
        filtered = [x * 1.1 for x in filtered]
    
    return {
        'total': sum(filtered),
        'count': len(filtered),
        'average': sum(filtered) / len(filtered) if filtered else 0
    }

# ë„ì›€ë§ í™•ì¸
help(process_sales_data)
````

````python
# [ì˜ˆì œ 19] í‚¤ì›Œë“œ ì „ìš© ë§¤ê°œë³€ìˆ˜
def filter_data(data, *, min_value=None, max_value=None, ascending=True):
    """
    ë°ì´í„°ë¥¼ í•„í„°ë§í•˜ê³  ì •ë ¬í•©ë‹ˆë‹¤.
    
    Args:
        data (list): ë°ì´í„° ë¦¬ìŠ¤íŠ¸
        min_value: ìµœì†Œê°’ (í‚¤ì›Œë“œ ì „ìš©)
        max_value: ìµœëŒ€ê°’ (í‚¤ì›Œë“œ ì „ìš©)
        ascending: ì˜¤ë¦„ì°¨ìˆœ ì—¬ë¶€ (í‚¤ì›Œë“œ ì „ìš©)
    """
    result = data.copy()
    
    if min_value is not None:
        result = [x for x in result if x >= min_value]
    if max_value is not None:
        result = [x for x in result if x <= max_value]
    
    result.sort(reverse=not ascending)
    return result

# ì˜¬ë°”ë¥¸ ì‚¬ìš© - ì˜µì…˜ì„ í‚¤ì›Œë“œë¡œ
data = [5, 2, 8, 1, 9, 3]
print(filter_data(data, min_value=3, max_value=7))
print(filter_data(data, min_value=5, ascending=False))

# ì˜¤ë¥˜ - ìœ„ì¹˜ë¡œ ì „ë‹¬ ë¶ˆê°€
# filter_data(data, 3, 7)  # TypeError!
````

````python
# [ì˜ˆì œ 24] ê°„ë‹¨í•œ ë°ì½”ë ˆì´í„°
def log_function_call(func):
    """í•¨ìˆ˜ í˜¸ì¶œì„ ë¡œê¹…í•˜ëŠ” ë°ì½”ë ˆì´í„°"""
    def wrapper(*args, **kwargs):
        print(f"[LOG] {func.__name__}() í˜¸ì¶œë¨")
        result = func(*args, **kwargs)
        print(f"[LOG] {func.__name__}() ì™„ë£Œ")
        return result
    return wrapper

@log_function_call
def greet(name):
    print(f"ì•ˆë…•í•˜ì„¸ìš”, {name}ë‹˜!")

greet("í™ê¸¸ë™")
````

````python
# [ì˜ˆì œ 25] ì‹¤í–‰ ì‹œê°„ ì¸¡ì • ë°ì½”ë ˆì´í„°
import time
import functools

def timer(func):
    """í•¨ìˆ˜ ì‹¤í–‰ ì‹œê°„ì„ ì¸¡ì •í•˜ëŠ” ë°ì½”ë ˆì´í„°"""
    @functools.wraps(func)  # ì›ë³¸ í•¨ìˆ˜ ë©”íƒ€ë°ì´í„° ë³´ì¡´
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        elapsed = (end - start) * 1000  # ë°€ë¦¬ì´ˆ
        print(f"[TIMER] {func.__name__}() ì‹¤í–‰ ì‹œê°„: {elapsed:.2f}ms")
        return result
    return wrapper

@timer
def slow_function():
    """ëŠë¦° í•¨ìˆ˜ ì‹œë®¬ë ˆì´ì…˜"""
    time.sleep(0.5)
    return "ì™„ë£Œ"

result = slow_function()
print(f"ê²°ê³¼: {result}")
````

````python
# [ì˜ˆì œ 26] ì—¬ëŸ¬ ë°ì½”ë ˆì´í„° ì¡°í•©
@timer
@log_function_call
def process_data(data):
    """ë°ì´í„° ì²˜ë¦¬ (ë¡œê¹… + íƒ€ì´ë¨¸)"""
    time.sleep(0.1)
    return [x * 2 for x in data]

result = process_data([1, 2, 3, 4, 5])
print(f"ê²°ê³¼: {result}")
````
