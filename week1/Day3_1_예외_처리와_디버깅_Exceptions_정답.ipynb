````python
# ë°©ë²• 1: ê¸°ë³¸ try-except (ê¶Œì¥)
def safe_int(value):
    """
    ì•ˆì „í•˜ê²Œ ì •ìˆ˜ë¡œ ë³€í™˜
    
    Args:
        value: ë³€í™˜í•  ê°’
    
    Returns:
        int or None: ë³€í™˜ëœ ì •ìˆ˜ ë˜ëŠ” None
    """
    try:
        return int(value)
    except ValueError:
        return None

# í…ŒìŠ¤íŠ¸
print(safe_int("123"))    # 123
print(safe_int("abc"))    # None
print(safe_int("45.6"))   # None
print(safe_int(45.6))     # 45 (floatëŠ” ë³€í™˜ ê°€ëŠ¥)
````

````python
# ë°©ë²• 1: ìˆœì°¨ì  except (ê¶Œì¥)
def safe_get(data, index):
    """
    ì•ˆì „í•˜ê²Œ ë¦¬ìŠ¤íŠ¸ì—ì„œ ê°’ ê°€ì ¸ì˜¤ê¸°
    
    Args:
        data: ë¦¬ìŠ¤íŠ¸
        index: ì¸ë±ìŠ¤
    
    Returns:
        ê°’ ë˜ëŠ” None (ì—ëŸ¬ ë©”ì‹œì§€ ì¶œë ¥)
    """
    try:
        return data[index]
    except IndexError:
        print("ì¸ë±ìŠ¤ ë²”ìœ„ ì´ˆê³¼")
        return None
    except TypeError:
        print("ì˜ëª»ëœ íƒ€ì…")
        return None
    except Exception as e:
        print(f"ì•Œ ìˆ˜ ì—†ëŠ” ì—ëŸ¬: {e}")
        return None

# í…ŒìŠ¤íŠ¸
data = [10, 20, 30]
print(safe_get(data, 1))      # 20
print(safe_get(data, 10))     # None (ì¸ë±ìŠ¤ ë²”ìœ„ ì´ˆê³¼)
print(safe_get(data, "a"))    # None (ì˜ëª»ëœ íƒ€ì…)
````

## Q3. try-except-else-finally ì™„ì „ ì •ë³µ (ë‚œì´ë„: â­â­â­)

**ë¬¸ì œ**: íŒŒì¼ì„ ì½ê³  ì²˜ë¦¬í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
- `try`: íŒŒì¼ ì½ê¸° ì‹œë„
- `except FileNotFoundError`: "íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤" ì¶œë ¥ í›„ 0 ë°˜í™˜
- `except PermissionError`: "íŒŒì¼ ì½ê¸° ê¶Œí•œì´ ì—†ìŠµë‹ˆë‹¤" ì¶œë ¥ í›„ -1 ë°˜í™˜
- `else`: ì—ëŸ¬ ì—†ìœ¼ë©´ "íŒŒì¼ ì²˜ë¦¬ ì™„ë£Œ" ì¶œë ¥

````python
# ë°©ë²• 1: ëª…ì‹œì  íŒŒì¼ ê´€ë¦¬ (í•™ìŠµìš©)
def count_file_lines(filepath):
    """
    íŒŒì¼ì˜ ì¤„ ìˆ˜ë¥¼ ì„¸ê¸°
    
    Args:
        filepath: íŒŒì¼ ê²½ë¡œ
    
    Returns:
        int: ì¤„ ìˆ˜ (ì—ëŸ¬ ì‹œ 0 ë˜ëŠ” -1)
    """
    file_handle = None
    line_count = 0
    
    try:
        # íŒŒì¼ ì—´ê¸°
        file_handle = open(filepath, 'r', encoding='utf-8')
        # ì¤„ ìˆ˜ ì„¸ê¸°
        line_count = len(file_handle.readlines())
        
    except FileNotFoundError:
        print("íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
        return 0
    
    except PermissionError:
        print("íŒŒì¼ ì½ê¸° ê¶Œí•œì´ ì—†ìŠµë‹ˆë‹¤")
        return -1
    
    else:
        # ì—ëŸ¬ê°€ ì—†ì„ ë•Œë§Œ ì‹¤í–‰
        print("íŒŒì¼ ì²˜ë¦¬ ì™„ë£Œ")
    
    finally:
        # í•­ìƒ ì‹¤í–‰ (ì—ëŸ¬ ì—¬ë¶€ ë¬´ê´€)
        print("íŒŒì¼ ì²˜ë¦¬ ì¢…ë£Œ")
        # íŒŒì¼ì´ ì—´ë ¸ë‹¤ë©´ ë‹«ê¸°
        if file_handle is not None:
            file_handle.close()
    
    return line_count

# í…ŒìŠ¤íŠ¸ íŒŒì¼ ìƒì„±
with open('test_file.txt', 'w') as f:
    f.write('line1\nline2\nline3\n')

# í…ŒìŠ¤íŠ¸
print("=== ì •ìƒ ì¼€ì´ìŠ¤ ===")
result = count_file_lines('test_file.txt')
print(f"ì¤„ ìˆ˜: {result}\n")

print("=== íŒŒì¼ ì—†ìŒ ===")
result = count_file_lines('nonexistent.txt')
print(f"ì¤„ ìˆ˜: {result}\n")
````

````python
# ë°©ë²• 2: with ë¬¸ ì‚¬ìš© (ì‹¤ë¬´ ê¶Œì¥)
def count_file_lines_v2(filepath):
    """
    íŒŒì¼ì˜ ì¤„ ìˆ˜ë¥¼ ì„¸ê¸° (with ë¬¸ ì‚¬ìš©)
    """
    try:
        # with ë¬¸ì€ ìë™ìœ¼ë¡œ íŒŒì¼ì„ ë‹«ì•„ì¤Œ
        with open(filepath, 'r', encoding='utf-8') as f:
            line_count = len(f.readlines())
    
    except FileNotFoundError:
        print("íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
        return 0
    
    except PermissionError:
        print("íŒŒì¼ ì½ê¸° ê¶Œí•œì´ ì—†ìŠµë‹ˆë‹¤")
        return -1
    
    except UnicodeDecodeError:
        print("íŒŒì¼ ì¸ì½”ë”© ì˜¤ë¥˜")
        return -1
    
    else:
        print("íŒŒì¼ ì²˜ë¦¬ ì™„ë£Œ")
    
    finally:
        print("íŒŒì¼ ì²˜ë¦¬ ì¢…ë£Œ")
    
    return line_count

# í…ŒìŠ¤íŠ¸
print("=== with ë¬¸ ë²„ì „ ===")
result = count_file_lines_v2('test_file.txt')
print(f"ì¤„ ìˆ˜: {result}")
````

````python
# ë°©ë²• 2: í•„ë“œ ì •ë³´ë¥¼ í¬í•¨í•˜ëŠ” ì˜ˆì™¸ (ê³ ê¸‰)
class ValidationError(Exception):
    """ë°ì´í„° ê²€ì¦ ì‹¤íŒ¨ ì˜ˆì™¸ (í•„ë“œ ì •ë³´ í¬í•¨)"""
    
    def __init__(self, message, field=None, value=None):
        super().__init__(message)
        self.field = field
        self.value = value
    
    def __str__(self):
        if self.field:
            return f"[{self.field}] {super().__str__()}"
        return super().__str__()

def validate_user_v2(name, age, email):
    """
    ì‚¬ìš©ì ë°ì´í„° ê²€ì¦ (í•„ë“œ ì •ë³´ í¬í•¨)
    """
    # ì´ë¦„ ê²€ì¦
    if not name or not name.strip():
        raise ValidationError(
            "ì´ë¦„ì€ ë¹ˆ ë¬¸ìì—´ì¼ ìˆ˜ ì—†ìŠµë‹ˆë‹¤",
            field='name',
            value=name
        )
    
    # ë‚˜ì´ ê²€ì¦
    if not isinstance(age, int):
        raise ValidationError(
            "ë‚˜ì´ëŠ” ì •ìˆ˜ì—¬ì•¼ í•©ë‹ˆë‹¤",
            field='age',
            value=age
        )
    
    if age < 0 or age > 150:
        raise ValidationError(
            f"ë‚˜ì´ëŠ” 0-150 ë²”ìœ„ì—¬ì•¼ í•©ë‹ˆë‹¤ (ì…ë ¥: {age})",
            field='age',
            value=age
        )
    
    # ì´ë©”ì¼ ê²€ì¦
    if '@' not in email:
        raise ValidationError(
            f"ìœ íš¨í•˜ì§€ ì•Šì€ ì´ë©”ì¼ í˜•ì‹",
            field='email',
            value=email
        )
    
    return {
        'name': name.strip(),
        'age': age,
        'email': email.lower()
    }

# í…ŒìŠ¤íŠ¸
print("=== í•„ë“œ ì •ë³´ í¬í•¨ ë²„ì „ ===")
try:
    user = validate_user_v2("", 25, "kim@example.com")
except ValidationError as e:
    print(f"ê²€ì¦ ì‹¤íŒ¨: {e}")
    print(f"ë¬¸ì œ í•„ë“œ: {e.field}")
    print(f"ì…ë ¥ê°’: {e.value}")
````

````python
# ë°©ë²• 3: ì—¬ëŸ¬ ê²€ì¦ ì—ëŸ¬ë¥¼ ëª¨ìœ¼ê¸° (ì‹¤ë¬´ ë²„ì „)
class ValidationError(Exception):
    """ë°ì´í„° ê²€ì¦ ì‹¤íŒ¨ ì˜ˆì™¸"""
    
    def __init__(self, errors):
        """
        Args:
            errors: ì—ëŸ¬ ë©”ì‹œì§€ ë¦¬ìŠ¤íŠ¸ ë˜ëŠ” ë”•ì…”ë„ˆë¦¬
        """
        self.errors = errors
        if isinstance(errors, dict):
            message = ", ".join(f"{k}: {v}" for k, v in errors.items())
        else:
            message = ", ".join(errors)
        super().__init__(message)

def validate_user_v3(name, age, email):
    """
    ì‚¬ìš©ì ë°ì´í„° ê²€ì¦ (ëª¨ë“  ì—ëŸ¬ ìˆ˜ì§‘)
    """
    errors = {}
    
    # ì´ë¦„ ê²€ì¦
    if not name or not name.strip():
        errors['name'] = "ì´ë¦„ì€ ë¹ˆ ë¬¸ìì—´ì¼ ìˆ˜ ì—†ìŠµë‹ˆë‹¤"
    
    # ë‚˜ì´ ê²€ì¦
    if not isinstance(age, int):
        errors['age'] = "ë‚˜ì´ëŠ” ì •ìˆ˜ì—¬ì•¼ í•©ë‹ˆë‹¤"
    elif age < 0 or age > 150:
        errors['age'] = f"ë‚˜ì´ëŠ” 0-150 ë²”ìœ„ì—¬ì•¼ í•©ë‹ˆë‹¤ (ì…ë ¥: {age})"
    
    # ì´ë©”ì¼ ê²€ì¦
    if '@' not in email:
        errors['email'] = f"ìœ íš¨í•˜ì§€ ì•Šì€ ì´ë©”ì¼ í˜•ì‹: {email}"
    
    # ì—ëŸ¬ê°€ ìˆìœ¼ë©´ ì˜ˆì™¸ ë°œìƒ
    if errors:
        raise ValidationError(errors)
    
    return {
        'name': name.strip(),
        'age': age,
        'email': email.lower()
    }

# í…ŒìŠ¤íŠ¸
print("=== ëª¨ë“  ì—ëŸ¬ ìˆ˜ì§‘ ë²„ì „ ===")
try:
    user = validate_user_v3("", 200, "invalid")
except ValidationError as e:
    print(f"ê²€ì¦ ì‹¤íŒ¨: {e}")
    print("\nê°œë³„ ì—ëŸ¬:")
    for field, error in e.errors.items():
        print(f"  - {field}: {error}")
````

## Q5. ì˜ˆì™¸ ì²´ì´ë‹ (raise from) (ë‚œì´ë„: â­â­â­â­)

````python
import json

# ë°©ë²• 1: ê¸°ë³¸ ì˜ˆì™¸ ì²´ì´ë‹ (ê¶Œì¥)
class ConfigError(Exception):
    """ì„¤ì • ê´€ë ¨ ì—ëŸ¬"""
    pass

def load_config(filepath):
    """
    JSON ì„¤ì • íŒŒì¼ ë¡œë“œ
    
    Args:
        filepath: ì„¤ì • íŒŒì¼ ê²½ë¡œ
    
    Returns:
        dict: ì„¤ì • ë”•ì…”ë„ˆë¦¬
    
    Raises:
        ConfigError: ì„¤ì • íŒŒì¼ ë¬¸ì œ ì‹œ (ì›ì¸ í¬í•¨)
    """
    try:
        # íŒŒì¼ ì½ê¸°
        with open(filepath, 'r', encoding='utf-8') as f:
            content = f.read()
    except FileNotFoundError as e:
        # ì›ì¸ì„ ë³´ì¡´í•˜ë©° ìƒˆ ì˜ˆì™¸ ë°œìƒ
        raise ConfigError(f"ì„¤ì • íŒŒì¼ ë¡œë“œ ì‹¤íŒ¨: {filepath}") from e
    except PermissionError as e:
        raise ConfigError(f"ì„¤ì • íŒŒì¼ ì ‘ê·¼ ê¶Œí•œ ì—†ìŒ: {filepath}") from e
    
    try:
        # JSON íŒŒì‹±
        config = json.loads(content)
    except json.JSONDecodeError as e:
        # JSON íŒŒì‹± ì—ëŸ¬ ì²´ì´ë‹
        raise ConfigError(f"ì„¤ì • íŒŒì¼ í˜•ì‹ ì˜¤ë¥˜: {filepath}") from e
    
    return config

# í…ŒìŠ¤íŠ¸ íŒŒì¼ ìƒì„±
with open('valid_config.json', 'w') as f:
    json.dump({'debug': True, 'port': 8080}, f)

with open('invalid_config.json', 'w') as f:
    f.write('{invalid json}')

# ì •ìƒ ì¼€ì´ìŠ¤
print("=== ì •ìƒ ì¼€ì´ìŠ¤ ===")
config = load_config('valid_config.json')
print(config)

# ì—ëŸ¬ ì¼€ì´ìŠ¤ 1: íŒŒì¼ ì—†ìŒ
print("\n=== íŒŒì¼ ì—†ìŒ ===")
try:
    config = load_config('nonexistent.json')
except ConfigError as e:
    print(f"ì—ëŸ¬: {e}")
    print(f"ì›ì¸ íƒ€ì…: {type(e.__cause__).__name__}")
    print(f"ì›ì¸ ë©”ì‹œì§€: {e.__cause__}")

# ì—ëŸ¬ ì¼€ì´ìŠ¤ 2: ì˜ëª»ëœ JSON
print("\n=== ì˜ëª»ëœ JSON ===")
try:
    config = load_config('invalid_config.json')
except ConfigError as e:
    print(f"ì—ëŸ¬: {e}")
    print(f"ì›ì¸ íƒ€ì…: {type(e.__cause__).__name__}")
    print(f"ì›ì¸ ë©”ì‹œì§€: {e.__cause__}")
````

````python
# ë°©ë²• 2: ë¡œê¹… + ì˜ˆì™¸ ì²´ì´ë‹ (ì‹¤ë¬´ ë²„ì „)
import logging

logging.basicConfig(level=logging.ERROR)

class ConfigError(Exception):
    """ì„¤ì • ê´€ë ¨ ì—ëŸ¬"""
    pass

def load_config_v2(filepath):
    """
    JSON ì„¤ì • íŒŒì¼ ë¡œë“œ (ë¡œê¹… í¬í•¨)
    """
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            content = f.read()
    except FileNotFoundError as e:
        logging.error(f"ì„¤ì • íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŒ: {filepath}")
        raise ConfigError(f"ì„¤ì • íŒŒì¼ ë¡œë“œ ì‹¤íŒ¨: {filepath}") from e
    except PermissionError as e:
        logging.error(f"ì„¤ì • íŒŒì¼ ì ‘ê·¼ ê¶Œí•œ ì—†ìŒ: {filepath}")
        raise ConfigError(f"ì„¤ì • íŒŒì¼ ì ‘ê·¼ ê¶Œí•œ ì—†ìŒ: {filepath}") from e
    
    try:
        config = json.loads(content)
    except json.JSONDecodeError as e:
        logging.error(f"ì„¤ì • íŒŒì¼ JSON íŒŒì‹± ì‹¤íŒ¨: {filepath}, ìœ„ì¹˜: {e.lineno}:{e.colno}")
        raise ConfigError(f"ì„¤ì • íŒŒì¼ í˜•ì‹ ì˜¤ë¥˜: {filepath}") from e
    
    return config

# í…ŒìŠ¤íŠ¸
print("=== ë¡œê¹… ë²„ì „ ===")
try:
    config = load_config_v2('invalid_config.json')
except ConfigError as e:
    print(f"\nìµœì¢… ì—ëŸ¬: {e}")
    print(f"ì›ì¸: {e.__cause__}")
````

````python
# ë°©ë²• 3: ì „ì²´ ì²´ì¸ ì¶œë ¥ (ë””ë²„ê¹…ìš©)
import traceback

def load_config_v3(filepath):
    """
    JSON ì„¤ì • íŒŒì¼ ë¡œë“œ
    """
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            content = f.read()
    except (FileNotFoundError, PermissionError) as e:
        raise ConfigError(f"ì„¤ì • íŒŒì¼ ë¡œë“œ ì‹¤íŒ¨: {filepath}") from e
    
    try:
        config = json.loads(content)
    except json.JSONDecodeError as e:
        raise ConfigError(f"ì„¤ì • íŒŒì¼ í˜•ì‹ ì˜¤ë¥˜: {filepath}") from e
    
    return config

def print_exception_chain(exc):
    """ì˜ˆì™¸ ì²´ì¸ ì „ì²´ ì¶œë ¥"""
    print("=== ì˜ˆì™¸ ì²´ì¸ ===")
    current = exc
    level = 0
    
    while current is not None:
        indent = "  " * level
        print(f"{indent}[{level}] {type(current).__name__}: {current}")
        current = current.__cause__
        level += 1

# í…ŒìŠ¤íŠ¸
print("=== ì˜ˆì™¸ ì²´ì¸ ì¶œë ¥ ===")
try:
    config = load_config_v3('invalid_config.json')
except ConfigError as e:
    print_exception_chain(e)
    
    print("\n=== ì „ì²´ íŠ¸ë ˆì´ìŠ¤ë°± ===")
    traceback.print_exc()
````

## Q6. ë¡œê¹… ëª¨ë“ˆ ê¸°ì´ˆ (ë‚œì´ë„: â­â­â­)

**ë¬¸ì œ**: `print` ëŒ€ì‹  `logging` ëª¨ë“ˆì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„° ì²˜ë¦¬ í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
1. ë¡œê¹… ì„¤ì •: INFO ë ˆë²¨, ì‹œê°„-ë ˆë²¨-ë©”ì‹œì§€ í¬ë§·

````python
import logging

# ë°©ë²• 1: ê¸°ë³¸ ë¡œê¹… ì„¤ì •
def setup_logging():
    """ë¡œê¹… ì„¤ì •"""
    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(levelname)s - %(message)s',
        handlers=[
            logging.FileHandler('process.log', encoding='utf-8'),
            logging.StreamHandler()  # ì½˜ì†” ì¶œë ¥
        ]
    )

def process_file(filepath):
    """íŒŒì¼ ì²˜ë¦¬ (ì‹œë®¬ë ˆì´ì…˜)"""
    with open(filepath, encoding='utf-8') as f:
        return len(f.readlines())

def process_files(filepaths):
    """
    ì—¬ëŸ¬ íŒŒì¼ ì²˜ë¦¬
    
    Args:
        filepaths: íŒŒì¼ ê²½ë¡œ ë¦¬ìŠ¤íŠ¸
    
    Returns:
        dict: {'ì„±ê³µ': ì„±ê³µ_ê°œìˆ˜, 'ì‹¤íŒ¨': ì‹¤íŒ¨_ê°œìˆ˜}
    """
    logging.info(f"íŒŒì¼ ì²˜ë¦¬ ì‹œì‘: ì´ {len(filepaths)}ê°œ")
    
    success_count = 0
    fail_count = 0
    
    for filepath in filepaths:
        logging.info(f"ì²˜ë¦¬ ì¤‘: {filepath}")
        
        try:
            line_count = process_file(filepath)
            logging.info(f"ì„±ê³µ: {filepath} ({line_count}ì¤„)")
            success_count += 1
            
        except FileNotFoundError:
            logging.warning(f"íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŒ: {filepath}")
            fail_count += 1
            
        except Exception as e:
            logging.error(f"ì²˜ë¦¬ ì‹¤íŒ¨: {filepath}, ì—ëŸ¬: {e}")
            fail_count += 1
    
    logging.info(f"íŒŒì¼ ì²˜ë¦¬ ì™„ë£Œ: ì„±ê³µ {success_count}ê°œ, ì‹¤íŒ¨ {fail_count}ê°œ")
    
    return {'ì„±ê³µ': success_count, 'ì‹¤íŒ¨': fail_count}

# í…ŒìŠ¤íŠ¸ íŒŒì¼ ìƒì„±
with open('test1.txt', 'w') as f:
    f.write('line1\nline2\nline3')

with open('test2.txt', 'w') as f:
    f.write('data')

# ì‹¤í–‰
setup_logging()
files = ['test1.txt', 'test2.txt', 'nonexistent.txt']
result = process_files(files)
print(f"\nìµœì¢… ê²°ê³¼: {result}")
````

## Q7. ì‹¤ì „: ë°ì´í„° íŒŒì´í”„ë¼ì¸ ì—ëŸ¬ ì²˜ë¦¬ (ë‚œì´ë„: â­â­â­â­â­)

**ë¬¸ì œ**: ì‹¤ë¬´ ìˆ˜ì¤€ì˜ ë°ì´í„° íŒŒì´í”„ë¼ì¸ì„ ë§Œë“œì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:

2. ì¬ì‹œë„ ë¡œì§
3. ë¡œê¹…
4. ì—ëŸ¬ ì²´ì´ë‹

````python
import logging
import time
import random
import json

# ë¡œê¹… ì„¤ì •
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger('pipeline')

# ì»¤ìŠ¤í…€ ì˜ˆì™¸ ê³„ì¸µ
class DataPipelineError(Exception):
    """ë°ì´í„° íŒŒì´í”„ë¼ì¸ ê¸°ë³¸ ì—ëŸ¬"""
    pass

class DataFetchError(DataPipelineError):
    """ë°ì´í„° ìˆ˜ì§‘ ì—ëŸ¬"""
    pass

class DataValidationError(DataPipelineError):
    """ë°ì´í„° ê²€ì¦ ì—ëŸ¬"""
    pass

class DataSaveError(DataPipelineError):
    """ë°ì´í„° ì €ì¥ ì—ëŸ¬"""
    pass

# ì‹œë®¬ë ˆì´ì…˜ í•¨ìˆ˜ë“¤
def fetch_data_from_api(source_id):
    """APIì—ì„œ ë°ì´í„° ê°€ì ¸ì˜¤ê¸° (ì‹œë®¬ë ˆì´ì…˜)"""
    # 30% í™•ë¥ ë¡œ ì‹¤íŒ¨
    if random.random() < 0.3:
        raise ConnectionError(f"API ì—°ê²° ì‹¤íŒ¨: {source_id}")
    
    return {
        'source_id': source_id,
        'value': random.randint(1, 100),
        'timestamp': time.time()
    }

def validate_data(data):
    """ë°ì´í„° ê²€ì¦"""
    if 'value' not in data:
        raise ValueError("value í•„ë“œ ì—†ìŒ")
    
    if not isinstance(data['value'], int):
        raise TypeError("valueëŠ” ì •ìˆ˜ì—¬ì•¼ í•¨")
    
    if data['value'] < 0 or data['value'] > 100:
        raise ValueError("valueëŠ” 0-100 ë²”ìœ„ì—¬ì•¼ í•¨")
    
    return True

def save_data(data, filepath):
    """ë°ì´í„° ì €ì¥"""
    with open(filepath, 'a', encoding='utf-8') as f:
        json.dump(data, f)
        f.write('\n')

# ë©”ì¸ íŒŒì´í”„ë¼ì¸ í•¨ìˆ˜ë“¤
def fetch_with_retry(source_id, max_attempts=3):
    """
    ì¬ì‹œë„ ë¡œì§ì´ ìˆëŠ” ë°ì´í„° ìˆ˜ì§‘
    
    Args:
        source_id: ì†ŒìŠ¤ ID
        max_attempts: ìµœëŒ€ ì‹œë„ íšŸìˆ˜
    
    Returns:
        dict: ìˆ˜ì§‘ëœ ë°ì´í„°
    
    Raises:
        DataFetchError: ëª¨ë“  ì‹œë„ ì‹¤íŒ¨ ì‹œ
    """
    for attempt in range(1, max_attempts + 1):
        try:
            logger.info(f"ë°ì´í„° ìˆ˜ì§‘ ì‹œë„ [{attempt}/{max_attempts}]: {source_id}")
            data = fetch_data_from_api(source_id)
            logger.info(f"ë°ì´í„° ìˆ˜ì§‘ ì„±ê³µ: {source_id}")
            return data
            
        except ConnectionError as e:
            if attempt == max_attempts:
                # ë§ˆì§€ë§‰ ì‹œë„ ì‹¤íŒ¨ â†’ ì˜ˆì™¸ ì²´ì´ë‹
                logger.error(f"ë°ì´í„° ìˆ˜ì§‘ ìµœì¢… ì‹¤íŒ¨: {source_id}")
                raise DataFetchError(f"ë°ì´í„° ìˆ˜ì§‘ ì‹¤íŒ¨ (ëª¨ë“  ì¬ì‹œë„ ì‹¤íŒ¨): {source_id}") from e
            else:
                # ì¬ì‹œë„
                logger.warning(f"ì¬ì‹œë„ ëŒ€ê¸° ì¤‘: {source_id} ({e})")
                time.sleep(1)  # 1ì´ˆ ëŒ€ê¸°

def process_source(source_id, output_file):
    """
    í•˜ë‚˜ì˜ ì†ŒìŠ¤ì—ì„œ ë°ì´í„° ìˆ˜ì§‘ â†’ ê²€ì¦ â†’ ì €ì¥
    
    Args:
        source_id: ì†ŒìŠ¤ ID
        output_file: ì¶œë ¥ íŒŒì¼ ê²½ë¡œ
    
    Returns:
        bool: ì„±ê³µ ì—¬ë¶€
    """
    try:
        # 1. ë°ì´í„° ìˆ˜ì§‘ (ì¬ì‹œë„ ë¡œì§ í¬í•¨)
        data = fetch_with_retry(source_id)
        
        # 2. ë°ì´í„° ê²€ì¦
        try:
            validate_data(data)
            logger.info(f"ë°ì´í„° ê²€ì¦ ì„±ê³µ: {source_id}")
        except (ValueError, TypeError) as e:
            raise DataValidationError(f"ë°ì´í„° ê²€ì¦ ì‹¤íŒ¨: {source_id}") from e
        
        # 3. ë°ì´í„° ì €ì¥
        try:
            save_data(data, output_file)
            logger.info(f"ë°ì´í„° ì €ì¥ ì„±ê³µ: {source_id}")
        except IOError as e:
            raise DataSaveError(f"ë°ì´í„° ì €ì¥ ì‹¤íŒ¨: {source_id}") from e
        
        return True
        
    except DataPipelineError as e:
        # íŒŒì´í”„ë¼ì¸ ì—ëŸ¬ ë¡œê¹…
        logger.error(f"íŒŒì´í”„ë¼ì¸ ì—ëŸ¬: {e}")
        if e.__cause__:
            logger.error(f"  ì›ì¸: {e.__cause__}")
        return False

def run_pipeline(source_ids, output_file='pipeline_output.jsonl'):
    """
    ë°ì´í„° íŒŒì´í”„ë¼ì¸ ì‹¤í–‰
    
    Args:
        source_ids: ì†ŒìŠ¤ ID ë¦¬ìŠ¤íŠ¸
        output_file: ì¶œë ¥ íŒŒì¼ ê²½ë¡œ
    
    Returns:
        dict: {'ì„±ê³µ': int, 'ì‹¤íŒ¨': int, 'ì‹¤íŒ¨_ëª©ë¡': list}
    """
    logger.info("="*50)
    logger.info(f"íŒŒì´í”„ë¼ì¸ ì‹œì‘: {len(source_ids)}ê°œ ì†ŒìŠ¤")
    logger.info("="*50)
    
    start_time = time.time()
    success_count = 0
    fail_count = 0
    failed_sources = []
    
    for i, source_id in enumerate(source_ids, 1):
        logger.info(f"\n[{i}/{len(source_ids)}] ì²˜ë¦¬ ì‹œì‘: {source_id}")
        
        success = process_source(source_id, output_file)
        
        if success:
            success_count += 1
        else:
            fail_count += 1
            failed_sources.append(source_id)
    
    elapsed_time = time.time() - start_time
    
    logger.info("\n" + "="*50)
    logger.info("íŒŒì´í”„ë¼ì¸ ì™„ë£Œ")
    logger.info(f"ì†Œìš” ì‹œê°„: {elapsed_time:.2f}ì´ˆ")
    logger.info(f"ì„±ê³µ: {success_count}ê°œ")
    logger.info(f"ì‹¤íŒ¨: {fail_count}ê°œ")
    if failed_sources:
        logger.info(f"ì‹¤íŒ¨ ì†ŒìŠ¤: {failed_sources}")
    logger.info("="*50)
    
    return {
        'ì„±ê³µ': success_count,
        'ì‹¤íŒ¨': fail_count,
        'ì‹¤íŒ¨_ëª©ë¡': failed_sources
    }

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
sources = ['source_1', 'source_2', 'source_3', 'source_4', 'source_5']
result = run_pipeline(sources)

print("\n" + "="*50)
print("=== íŒŒì´í”„ë¼ì¸ ê²°ê³¼ ===")
print(f"ì„±ê³µ: {result['ì„±ê³µ']}")
print(f"ì‹¤íŒ¨: {result['ì‹¤íŒ¨']}")
if result['ì‹¤íŒ¨_ëª©ë¡']:
    print(f"ì‹¤íŒ¨ ì†ŒìŠ¤: {result['ì‹¤íŒ¨_ëª©ë¡']}")
print("="*50)
````

````python
# ë³´ë„ˆìŠ¤ 1: ì‹¤íŒ¨í•œ ì†ŒìŠ¤ë§Œ ì¬ì²˜ë¦¬
def retry_failed(failed_sources, output_file='pipeline_output.jsonl'):
    """
    ì‹¤íŒ¨í•œ ì†ŒìŠ¤ë§Œ ì¬ì²˜ë¦¬
    
    Args:
        failed_sources: ì‹¤íŒ¨í•œ ì†ŒìŠ¤ ID ë¦¬ìŠ¤íŠ¸
        output_file: ì¶œë ¥ íŒŒì¼ ê²½ë¡œ
    
    Returns:
        dict: ì¬ì²˜ë¦¬ ê²°ê³¼
    """
    logger.info("\n" + "="*50)
    logger.info(f"ì‹¤íŒ¨ ì†ŒìŠ¤ ì¬ì²˜ë¦¬: {len(failed_sources)}ê°œ")
    logger.info("="*50)
    
    return run_pipeline(failed_sources, output_file)

# í…ŒìŠ¤íŠ¸
if result['ì‹¤íŒ¨_ëª©ë¡']:
    print("\nì‹¤íŒ¨í•œ ì†ŒìŠ¤ë¥¼ ì¬ì²˜ë¦¬í•©ë‹ˆë‹¤...")
    retry_result = retry_failed(result['ì‹¤íŒ¨_ëª©ë¡'])
    print(f"\nì¬ì²˜ë¦¬ ê²°ê³¼: {retry_result}")
````

````python
# ë³´ë„ˆìŠ¤ 2: ì²˜ë¦¬ ì‹œê°„ ì¸¡ì •
import functools

def timer(func):
    """í•¨ìˆ˜ ì‹¤í–‰ ì‹œê°„ ì¸¡ì • ë°ì½”ë ˆì´í„°"""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        elapsed = time.time() - start
        logger.info(f"{func.__name__} ì‹¤í–‰ ì‹œê°„: {elapsed:.2f}ì´ˆ")
        return result
    return wrapper

@timer
def run_pipeline_with_timer(source_ids, output_file='pipeline_output.jsonl'):
    """ì‹œê°„ ì¸¡ì •ì´ í¬í•¨ëœ íŒŒì´í”„ë¼ì¸"""
    return run_pipeline(source_ids, output_file)

# í…ŒìŠ¤íŠ¸
print("\n=== ì‹œê°„ ì¸¡ì • ë²„ì „ ===")
result = run_pipeline_with_timer(['source_A', 'source_B'])
````

````python
# ë³´ë„ˆìŠ¤ 3: ì—ëŸ¬ í†µê³„
from collections import defaultdict

class PipelineStats:
    """íŒŒì´í”„ë¼ì¸ í†µê³„ ìˆ˜ì§‘ê¸°"""
    
    def __init__(self):
        self.error_counts = defaultdict(int)
        self.total_attempts = 0
        self.total_successes = 0
    
    def record_error(self, error_type):
        """ì—ëŸ¬ ê¸°ë¡"""
        self.error_counts[error_type.__name__] += 1
    
    def record_attempt(self):
        """ì‹œë„ ê¸°ë¡"""
        self.total_attempts += 1
    
    def record_success(self):
        """ì„±ê³µ ê¸°ë¡"""
        self.total_successes += 1
    
    def get_stats(self):
        """í†µê³„ ë°˜í™˜"""
        return {
            'ì´_ì‹œë„': self.total_attempts,
            'ì´_ì„±ê³µ': self.total_successes,
            'ì„±ê³µë¥ ': f"{self.total_successes / self.total_attempts * 100:.1f}%" if self.total_attempts > 0 else '0%',
            'ì—ëŸ¬_íƒ€ì…ë³„_ì¹´ìš´íŠ¸': dict(self.error_counts)
        }

def run_pipeline_with_stats(source_ids, output_file='pipeline_output.jsonl'):
    """í†µê³„ ìˆ˜ì§‘ì´ í¬í•¨ëœ íŒŒì´í”„ë¼ì¸"""
    stats = PipelineStats()
    
    for source_id in source_ids:
        stats.record_attempt()
        
        try:
            success = process_source(source_id, output_file)
            if success:
                stats.record_success()
        except DataPipelineError as e:
            stats.record_error(type(e))
    
    return stats.get_stats()

# í…ŒìŠ¤íŠ¸
print("\n=== í†µê³„ ìˆ˜ì§‘ ë²„ì „ ===")
stats = run_pipeline_with_stats(['S1', 'S2', 'S3'])
print("\ní†µê³„:")
for key, value in stats.items():
    print(f"  {key}: {value}")
````
