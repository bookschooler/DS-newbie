### 3.2 Counter: ë¹ˆë„ ë¶„ì„ì˜ ê°•ë ¥í•œ ë„êµ¬

````python
from collections import Counter

# [ê¸°ë³¸] ë¦¬ìŠ¤íŠ¸ì˜ ìš”ì†Œ ê°œìˆ˜ ì„¸ê¸°
fruits = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
counter = Counter(fruits)

print(counter)
print(f"apple ê°œìˆ˜: {counter['apple']}")
print(f"grape ê°œìˆ˜: {counter['grape']}")  # ì—†ìœ¼ë©´ 0 ë°˜í™˜
````

````python
# [ì‹¤ë¬´ ì˜ˆì œ] í…ìŠ¤íŠ¸ ë‹¨ì–´ ë¹ˆë„ ë¶„ì„
text = "data science is amazing data analysis is powerful data is everywhere"
words = text.split()

word_count = Counter(words)
print("ë‹¨ì–´ ë¹ˆë„:")
print(word_count)

# ê°€ì¥ ë§ì´ ë‚˜ì˜¨ ë‹¨ì–´ ìƒìœ„ 3ê°œ
print("\nê°€ì¥ ë§ì´ ë‚˜ì˜¨ ë‹¨ì–´:")
for word, count in word_count.most_common(3):
    print(f"  {word}: {count}íšŒ")
````

````python
# [ì‹¤ìŠµ] ë¬¸ì ë¹ˆë„ ë¶„ì„
sentence = "hello world"
char_counter = Counter(sentence)

print("ë¬¸ì ë¹ˆë„:")
for char, count in char_counter.most_common():
    if char != ' ':  # ê³µë°± ì œì™¸
        print(f"  '{char}': {count}íšŒ")
````

````python
# Counterì˜ ì‚°ìˆ  ì—°ì‚°
c1 = Counter(['a', 'b', 'c', 'a'])
c2 = Counter(['a', 'c', 'd'])

print("c1:", c1)
print("c2:", c2)
print("c1 + c2:", c1 + c2)  # í•©ì§‘í•©
print("c1 - c2:", c1 - c2)  # ì°¨ì§‘í•©
print("c1 & c2:", c1 & c2)  # êµì§‘í•© (ìµœì†Œê°’)
print("c1 | c2:", c1 | c2)  # í•©ì§‘í•© (ìµœëŒ€ê°’)
````

`defaultdict`ëŠ” í‚¤ê°€ ì—†ì„ ë•Œ ê¸°ë³¸ê°’ì„ ìë™ìœ¼ë¡œ ìƒì„±í•©ë‹ˆë‹¤.

````python
# [ì‹¤ë¬´ ì˜ˆì œ] ê·¸ë£¹í•‘ ì‘ì—…
students = [
    ('Alice', 'Math'),
    ('Bob', 'Science'),
    ('Charlie', 'Math'),
    ('David', 'Science'),
    ('Eve', 'Math')
]

# ê³¼ëª©ë³„ë¡œ í•™ìƒì„ ê·¸ë£¹í•‘
subject_groups = defaultdict(list)  # ê¸°ë³¸ê°’: ë¹ˆ ë¦¬ìŠ¤íŠ¸

for name, subject in students:
    subject_groups[subject].append(name)

print("ê³¼ëª©ë³„ í•™ìƒ:")
for subject, names in subject_groups.items():
    print(f"  {subject}: {names}")
````

````python
# [ì‹¤ìŠµ] ë‹¨ì–´ ê¸¸ì´ë³„ ê·¸ë£¹í•‘
words = ['cat', 'dog', 'elephant', 'ant', 'bear', 'lion']

length_groups = defaultdict(list)
for word in words:
    length_groups[len(word)].append(word)

print("ê¸¸ì´ë³„ ë‹¨ì–´:")
for length, word_list in sorted(length_groups.items()):
    print(f"  {length}ê¸€ì: {word_list}")
````

### 3.4 deque: ì–‘ë°©í–¥ í

`deque` (Double-Ended Queue): ì–‘ìª½ ëì—ì„œ ë¹ ë¥´ê²Œ ì¶”ê°€/ì œê±°í•  ìˆ˜ ìˆëŠ” ìë£Œêµ¬ì¡°

**ë¦¬ìŠ¤íŠ¸ vs deque**:
* ë¦¬ìŠ¤íŠ¸: ëì— ì¶”ê°€/ì œê±°ëŠ” ë¹ ë¥´ì§€ë§Œ **ì•**ì— ì¶”ê°€/ì œê±°ëŠ” ëŠë¦¼ (O(n))
* deque: ì–‘ìª½ ë ëª¨ë‘ ë¹ ë¦„ (O(1))

````python
from collections import deque

# [ê¸°ë³¸] deque ìƒì„±
dq = deque([1, 2, 3])
print("ì´ˆê¸° deque:", dq)

# ì˜¤ë¥¸ìª½ì— ì¶”ê°€/ì œê±°
dq.append(4)
print("append(4):", dq)

removed = dq.pop()
print(f"pop() â†’ {removed}, ê²°ê³¼:", dq)
````

````python
# ì™¼ìª½ì— ì¶”ê°€/ì œê±° (ë¦¬ìŠ¤íŠ¸ë³´ë‹¤ í›¨ì”¬ ë¹ ë¦„!)
dq.appendleft(0)
print("appendleft(0):", dq)

removed = dq.popleft()
print(f"popleft() â†’ {removed}, ê²°ê³¼:", dq)
````

````python
# [ì‹¤ë¬´ ì˜ˆì œ] ìµœê·¼ Nê°œ í•­ëª© ìœ ì§€ (ìŠ¬ë¼ì´ë”© ìœˆë„ìš°)
recent_logs = deque(maxlen=5)  # ìµœëŒ€ 5ê°œë§Œ ìœ ì§€

for i in range(10):
    recent_logs.append(f"ë¡œê·¸ {i}")
    print(f"ì¶”ê°€ í›„: {list(recent_logs)}")

print(f"\nìµœì¢… ìµœê·¼ 5ê°œ: {list(recent_logs)}")
````

````python
# [ì‹¤ìŠµ] rotate: íšŒì „ ê¸°ëŠ¥
dq = deque([1, 2, 3, 4, 5])
print("ì›ë³¸:", dq)

dq.rotate(2)  # ì˜¤ë¥¸ìª½ìœ¼ë¡œ 2ì¹¸ íšŒì „
print("rotate(2):", dq)

dq.rotate(-2)  # ì™¼ìª½ìœ¼ë¡œ 2ì¹¸ íšŒì „ (ì›ìƒë³µêµ¬)
print("rotate(-2):", dq)
````

````python
# [ì‹¤ë¬´ ì˜ˆì œ] BFS íë¡œ ì‚¬ìš©
# ë„ˆë¹„ ìš°ì„  íƒìƒ‰(Breadth-First Search)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []
    
    while queue:
        node = queue.popleft()  # íì—ì„œ êº¼ë‚´ê¸°
        if node not in visited:
            visited.add(node)
            result.append(node)
            queue.extend(graph[node])  # ì´ì›ƒ ë…¸ë“œ ì¶”ê°€
    
    return result

print("BFS ìˆœíšŒ:", bfs(graph, 'A'))
````
