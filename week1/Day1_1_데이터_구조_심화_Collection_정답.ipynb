````python
# ê±°ë˜ ë‚´ì—­
transactions = [
    {"product": "ë…¸íŠ¸ë¶", "amount": 1000},
    {"product": "ë§ˆìš°ìŠ¤", "amount": 30},
    {"product": "ë…¸íŠ¸ë¶", "amount": 1200},
    {"product": "í‚¤ë³´ë“œ", "amount": 80},
    {"product": "ë§ˆìš°ìŠ¤", "amount": 25}
]

# ì œí’ˆë³„ ì´ ë§¤ì¶œ ê³„ì‚°
sales_summary = {}
for tx in transactions:
    product = tx["product"]
    amount = tx["amount"]
    sales_summary[product] = sales_summary.get(product, 0) + amount

print("ì œí’ˆë³„ ë§¤ì¶œ:")
for product, total in sales_summary.items():
    print(f"  {product}: {total:,}ë§Œì›")
````

````python
# ê³ ê° ë§ˆìŠ¤í„° ë°ì´í„°
customers = [
    {"id": 1, "name": "ê¹€ì² ìˆ˜", "grade": "VIP"},
    {"id": 2, "name": "ì´ì˜í¬", "grade": "ì¼ë°˜"},
    {"id": 3, "name": "ë°•ë¯¼ìˆ˜", "grade": "VIP"}
]

# ì£¼ë¬¸ ë°ì´í„°
orders = [
    {"order_id": "A001", "customer_id": 1, "amount": 5000},
    {"order_id": "A002", "customer_id": 3, "amount": 8000},
    {"order_id": "A003", "customer_id": 2, "amount": 1000},
    {"order_id": "A004", "customer_id": 999, "amount": 500}  # ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ê³ ê°
]
````
