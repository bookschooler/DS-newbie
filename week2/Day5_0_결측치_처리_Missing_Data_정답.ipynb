## Q6. ê²°ì¸¡ì¹˜ ë¹„ìœ¨ ê³„ì‚° â­â­â­

**ë¬¸ì œ**: ê° ì»¬ëŸ¼ì˜ ê²°ì¸¡ì¹˜ ë¹„ìœ¨(%)ì„ ê³„ì‚°í•˜ê³ , 20% ì´ìƒì¸ ì»¬ëŸ¼ë§Œ ì¶œë ¥í•˜ì„¸ìš”.

````python
# ì •ë‹µ ì½”ë“œ
df = pd.DataFrame({
    'A': [1, np.nan, 3, np.nan, 5, np.nan, 7, 8, 9, 10],
    'B': [np.nan, 2, 3, 4, 5, 6, 7, 8, 9, 10],
    'C': [1, 2, 3, np.nan, 5, 6, np.nan, 8, np.nan, 10]
})

# ì»¬ëŸ¼ë³„ ê²°ì¸¡ì¹˜ ë¹„ìœ¨ ê³„ì‚°
missing_ratio = (df.isna().sum() / len(df)) * 100

# 20% ì´ìƒì¸ ì»¬ëŸ¼ë§Œ í•„í„°ë§
high_missing = missing_ratio[missing_ratio >= 20]

print("ê²°ì¸¡ì¹˜ ë¹„ìœ¨ 20% ì´ìƒì¸ ì»¬ëŸ¼:")
print(high_missing)
````

## Q7. Forward Fill ì ìš© â­â­â­

**ë¬¸ì œ**: ë‹¤ìŒ ì¬ê³  ë°ì´í„°ì˜ ê²°ì¸¡ì¹˜ë¥¼ forward fillë¡œ ì±„ìš°ì„¸ìš”.

````python
# ì •ë‹µ ì½”ë“œ
inventory = pd.Series(
    [100, np.nan, np.nan, 85, np.nan, 80], 
    index=pd.date_range('2024-01-01', periods=6)
)

# ffill()ë¡œ ì´ì „ ê°’ìœ¼ë¡œ ì±„ìš°ê¸°
inventory_filled = inventory.ffill()

print("Forward Fill ê²°ê³¼:")
print(inventory_filled)
````

````python
# ê²€ì¦
expected = [100.0, 100.0, 100.0, 85.0, 85.0, 80.0]
assert list(inventory_filled.values) == expected, f"ì˜ˆìƒ: {expected}"
assert inventory_filled.isna().sum() == 0, "ê²°ì¸¡ì¹˜ê°€ ë‚¨ì•„ìˆìŠµë‹ˆë‹¤"
print("âœ… ì •ë‹µì…ë‹ˆë‹¤!")
````

### ğŸ’¡ í’€ì´ ì„¤ëª…

**ì ‘ê·¼ ë°©ë²•**:
- `ffill()` (forward fill)ì€ ê²°ì¸¡ì¹˜ë¥¼ **ë°”ë¡œ ì´ì „ ê°’**ìœ¼ë¡œ ì±„ìš°ê¸°

````python
# ì •ë‹µ ì½”ë“œ
np.random.seed(42)
survey_data = pd.DataFrame({
    'ì‘ë‹µìID': range(1, 21),
    'ë‚˜ì´': np.random.choice([25, 30, 35, np.nan], 20),
    'ë§Œì¡±ë„': np.random.choice([1, 2, 3, 4, 5, np.nan], 20),
    'ì¬êµ¬ë§¤ì˜í–¥': np.random.choice(['ì˜ˆ', 'ì•„ë‹ˆì˜¤', np.nan], 20),
    'ì¶”ì²œì ìˆ˜': np.random.choice([7, 8, 9, 10, np.nan], 20)
})

print("ì›ë³¸ ë°ì´í„°:")
print(survey_data.head())
print(f"\nê²°ì¸¡ì¹˜ ê°œìˆ˜:\n{survey_data.isna().sum()}")
````

````python
# 1. ê²°ì¸¡ì¹˜ íˆíŠ¸ë§µ
missing_matrix = survey_data.drop('ì‘ë‹µìID', axis=1).isna().astype(int)

fig1 = px.imshow(
    missing_matrix.T,
    labels=dict(x="í–‰ ë²ˆí˜¸", y="ì»¬ëŸ¼ëª…", color="ê²°ì¸¡ ì—¬ë¶€"),
    title="ğŸ“Š ê²°ì¸¡ì¹˜ íŒ¨í„´ íˆíŠ¸ë§µ",
    color_continuous_scale=['#2ecc71', '#e74c3c'],
    aspect='auto'
)
fig1.update_layout(height=400)
fig1.show()
````

````python
# 3. ì ì ˆí•œ ì „ëµìœ¼ë¡œ ì²˜ë¦¬
survey_filled = survey_data.copy()

# ë‚˜ì´: ì¤‘ì•™ê°’
survey_filled['ë‚˜ì´'] = survey_filled['ë‚˜ì´'].fillna(survey_filled['ë‚˜ì´'].median())

# ë§Œì¡±ë„, ì¶”ì²œì ìˆ˜: í‰ê·  (ë°˜ì˜¬ë¦¼)
survey_filled['ë§Œì¡±ë„'] = survey_filled['ë§Œì¡±ë„'].fillna(survey_filled['ë§Œì¡±ë„'].mean().round())
survey_filled['ì¶”ì²œì ìˆ˜'] = survey_filled['ì¶”ì²œì ìˆ˜'].fillna(survey_filled['ì¶”ì²œì ìˆ˜'].mean().round())

# ì¬êµ¬ë§¤ì˜í–¥: 'ì‘ë‹µ ì•ˆ í•¨'
survey_filled['ì¬êµ¬ë§¤ì˜í–¥'] = survey_filled['ì¬êµ¬ë§¤ì˜í–¥'].fillna('ì‘ë‹µ ì•ˆ í•¨')

print("\nì²˜ë¦¬ í›„ ê²°ì¸¡ì¹˜ ê°œìˆ˜:")
print(survey_filled.isna().sum())
````
