## 1.2 Series: 1차원 데이터

### Series란?

- pandas의 **1차원 배열** 자료구조
- 인덱스(index)와 값(value)으로 구성
- 엑셀의 한 열(Column)과 유사

````python
# 리스트로 Series 생성
sales = pd.Series([120, 150, 180, 200, 160])
print("매출 데이터:")
print(sales)
print(f"\n타입: {type(sales)}")
````

````python
# 인덱스 지정하여 Series 생성
sales = pd.Series(
    [120, 150, 180, 200, 160],
    index=["월", "화", "수", "목", "금"]
)
print("요일별 매출:")
print(sales)
````

````python
# 딕셔너리로 Series 생성
product_prices = pd.Series({
    "노트북": 1200,
    "마우스": 30,
    "키보드": 80,
    "모니터": 350
})
print("제품 가격:")
print(product_prices)
````

````python
# Series 접근
print(f"노트북 가격: {product_prices['노트북']:,}만원")
print(f"\n첫 번째 제품 가격: {product_prices.iloc[0]:,}만원")
print(f"\n가격 평균: {product_prices.mean():.1f}만원")
````

### 💡 실무 예시: 월별 매출 추이

월별 매출 데이터를 Series로 관리하면 통계 계산이 간편합니다.

````python
monthly_sales = pd.Series({
    "1월": 850, "2월": 920, "3월": 1100,
    "4월": 980, "5월": 1200, "6월": 1350
})

print("월별 매출 통계:")
print(f"평균: {monthly_sales.mean():.1f}만원")
print(f"최대: {monthly_sales.max():,}만원")
print(f"최소: {monthly_sales.min():,}만원")
print(f"표준편차: {monthly_sales.std():.1f}만원")
````

## 1.3 DataFrame: 2차원 데이터

### DataFrame이란?

- pandas의 **2차원 표(table)** 자료구조
- 여러 Series를 모아놓은 것
- 엑셀 시트와 가장 유사
- **행(row)**과 **열(column)**로 구성

````python
# 딕셔너리로 DataFrame 생성
data = {
    "제품명": ["노트북", "마우스", "키보드", "모니터"],
    "가격": [1200, 30, 80, 350],
    "재고": [5, 20, 15, 8]
}

df = pd.DataFrame(data)
print("제품 데이터:")
print(df)
````

````python
# DataFrame 기본 정보
print(f"행 x 열: {df.shape}")
print(f"전체 행 수: {len(df)}")
print(f"\n컬럼 목록:")
print(df.columns.tolist())
print(f"\n인덱스:")
print(df.index.tolist())
````

### 리스트의 리스트로 DataFrame 생성

````python
# 행 중심 데이터
sales_data = [
    ["김철수", "서울", 1200],
    ["이영희", "부산", 980],
    ["박민수", "대구", 1100],
    ["최지영", "서울", 1350]
]

df_sales = pd.DataFrame(
    sales_data,
    columns=["이름", "지역", "매출"]
)
print("영업사원 매출:")
print(df_sales)
````

## 1.6 인덱싱과 슬라이싱

### 컬럼 선택

````python
# 한 개 컬럼 선택 (Series 반환)
print("제품 컬럼:")
print(large_data["제품"])
print(f"\n타입: {type(large_data['제품'])}")
````

````python
# 여러 컬럼 선택 (DataFrame 반환)
print("제품과 판매량:")
print(large_data[["제품", "판매량"]])
````

### loc: 라벨 기반 인덱싱

````python
# 행 번호로 선택 (인덱스 라벨)
print("첫 번째 행:")
print(large_data.loc[0])
````

````python
# 범위 선택 (0~2행)
print("0~2행:")
print(large_data.loc[0:2])
````

````python
# 행과 컬럼 동시 선택
print("0~2행의 제품, 판매량 컬럼:")
print(large_data.loc[0:2, ["제품", "판매량"]])
````

### iloc: 위치 기반 인덱싱

````python
# 위치로 선택 (0부터 시작)
print("첫 번째 행:")
print(large_data.iloc[0])
````

````python
# 범위 선택 (0~2, 마지막 미포함)
print("0~2행 (2 미포함):")
print(large_data.iloc[0:2])
````

````python
# 행과 컬럼 위치로 선택
print("0~2행, 1~3 컬럼:")
print(large_data.iloc[0:3, 1:3])
````

### 💡 실무 예시: loc vs iloc 비교

````python
# loc: 라벨 기반 (마지막 포함)
print("loc[0:2] - 0, 1, 2행 모두 포함:")
print(large_data.loc[0:2, "제품"])

# iloc: 위치 기반 (마지막 미포함)
print("\niloc[0:2] - 0, 1행만 포함 (2 미포함):")
print(large_data.iloc[0:2, 1])
````

## 1.7 조건 필터링

### 단일 조건

````python
# 판매량이 50 이상인 행만
high_sales = large_data[large_data["판매량"] >= 50]
print("판매량 50 이상:")
print(high_sales)
````

````python
# 제품A만
product_a = large_data[large_data["제품"] == "제품A"]
print("제품A:")
print(product_a)
````

### 복합 조건

````python
# AND 조건: & (괄호 필수!)
result = large_data[(large_data["제품"] == "제품A") & (large_data["판매량"] >= 50)]
print("제품A이면서 판매량 50 이상:")
print(result)
````

````python
# OR 조건: |
result = large_data[(large_data["판매량"] >= 80) | (large_data["매출"] >= 800)]
print("판매량 80 이상 또는 매출 800 이상:")
print(result)
````

### isin(): 여러 값 중 하나

````python
# 제품A 또는 제품B만
result = large_data[large_data["제품"].isin(["제품A", "제품B"])]
print("제품A 또는 제품B:")
print(result.head())
````

### 💡 실무 예시: 고액 거래 추출

````python
# 매출 상위 30% 추출
threshold = large_data["매출"].quantile(0.7)
top_sales = large_data[large_data["매출"] >= threshold]

print(f"매출 상위 30% 기준: {threshold:.0f}만원 이상")
print(f"\n해당 거래 수: {len(top_sales)}건")
print(top_sales)
````

## 2.1 컬럼 추가/삭제/변경

### 새 컬럼 추가

````python
# 기존 컬럼으로 계산
df_calc = pd.DataFrame({
    "제품": ["A", "B", "C"],
    "단가": [1000, 500, 800],
    "수량": [5, 10, 7]
})

# 총액 컬럼 추가
df_calc["총액"] = df_calc["단가"] * df_calc["수량"]
print(df_calc)
````

````python
# 고정값 컬럼 추가
df_calc["구분"] = "국내"
print(df_calc)
````

### 컬럼 삭제

````python
# drop() 메소드 (원본 유지)
df_dropped = df_calc.drop("구분", axis=1)
print("구분 컬럼 제거:")
print(df_dropped)

print("\n원본은 그대로:")
print(df_calc)
````

````python
# inplace=True (원본 수정)
df_calc.drop("구분", axis=1, inplace=True)
print("원본이 변경됨:")
print(df_calc)
````

````python
# del 키워드 (원본 수정)
df_temp = df_calc.copy()
del df_temp["총액"]
print("del로 삭제:")
print(df_temp)
````

### 컬럼명 변경

````python
# rename() 메소드
df_renamed = df_calc.rename(columns={
    "제품": "상품명",
    "단가": "가격"
})
print("컬럼명 변경:")
print(df_renamed)
````

````python
# 전체 컬럼명 변경
df_temp = df_calc.copy()
df_temp.columns = ["Product", "Price", "Quantity", "Total"]
print("영문 컬럼명:")
print(df_temp)
````

### 💡 실무 예시: 파생 변수 생성

````python
# 고객 등급 계산
customers = pd.DataFrame({
    "이름": ["김철수", "이영희", "박민수", "최지영"],
    "구매금액": [5000, 1200, 8000, 3500]
})

# 등급 컬럼 추가
def get_grade(amount):
    if amount >= 5000:
        return "VIP"
    elif amount >= 3000:
        return "골드"
    else:
        return "일반"

customers["등급"] = customers["구매금액"].apply(get_grade)
print(customers)
````

## 2.2 데이터 정렬

### sort_values(): 값 기준 정렬

````python
# 샘플 데이터
df_sort = pd.DataFrame({
    "이름": ["김철수", "이영희", "박민수", "최지영"],
    "나이": [35, 28, 42, 31],
    "매출": [1200, 980, 1500, 1100]
})

print("원본:")
print(df_sort)
````

````python
# 나이 오름차순
print("나이 오름차순:")
print(df_sort.sort_values("나이"))
````

````python
# 매출 내림차순
print("매출 내림차순:")
print(df_sort.sort_values("매출", ascending=False))
````

````python
# 다중 컬럼 정렬
df_multi = pd.DataFrame({
    "부서": ["영업", "개발", "영업", "개발", "영업"],
    "등급": ["A", "B", "B", "A", "A"],
    "매출": [1200, 980, 1100, 1500, 1300]
})

# 부서 → 등급 → 매출 순 정렬
print("다중 정렬:")
print(df_multi.sort_values(["부서", "등급", "매출"], ascending=[True, True, False]))
````

### sort_index(): 인덱스 기준 정렬

````python
# 인덱스가 섞인 데이터
df_idx = pd.DataFrame(
    {"값": [10, 20, 30]},
    index=[2, 0, 1]
)

print("섞인 인덱스:")
print(df_idx)

print("\n인덱스 정렬:")
print(df_idx.sort_index())
````

### 💡 실무 예시: Top N 추출

````python
# 매출 상위 3명
top3 = df_sort.sort_values("매출", ascending=False).head(3)
print("매출 Top 3:")
print(top3)
````

## 2.3 중복 데이터 처리

### duplicated(): 중복 확인

````python
# 중복 포함 데이터
df_dup = pd.DataFrame({
    "제품": ["A", "B", "A", "C", "B"],
    "수량": [10, 20, 10, 30, 20]
})

print("원본:")
print(df_dup)

print("\n중복 여부:")
print(df_dup.duplicated())
````

````python
# 특정 컬럼 기준 중복
print("제품 기준 중복:")
print(df_dup.duplicated(subset=["제품"]))
````

### drop_duplicates(): 중복 제거

````python
# 전체 중복 제거 (모든 컬럼이 동일한 행)
print("전체 중복 제거:")
print(df_dup.drop_duplicates())
````

````python
# 제품 기준 중복 제거 (첫 번째 행 유지)
print("제품 기준 중복 제거:")
print(df_dup.drop_duplicates(subset=["제품"], keep="first"))
````

````python
# 마지막 행 유지
print("마지막 행 유지:")
print(df_dup.drop_duplicates(subset=["제품"], keep="last"))
````

### 💡 실무 예시: 고객 마스터 정제

### Q5. loc vs iloc ⭐⭐

**문제**: 다음 DataFrame에서 2번째 행의 "이름"과 "나이" 컬럼을 loc과 iloc 두 가지 방법으로 추출하세요.

```python
people = pd.DataFrame({
    "이름": ["김철수", "이영희", "박민수"],
    "나이": [35, 28, 42],
    "직업": ["개발자", "디자이너", "기획자"]
})
```

````python
# Q5 정답

people = pd.DataFrame({
    "이름": ["김철수", "이영희", "박민수"],
    "나이": [35, 28, 42],
    "직업": ["개발자", "디자이너", "기획자"]
})

# 방법 1: loc (라벨 기반)
result_loc = people.loc[1, ["이름", "나이"]]
print("loc 결과:")
print(result_loc)

# 방법 2: iloc (위치 기반)
result_iloc = people.iloc[1, [0, 1]]  # 0번째, 1번째 컬럼
print("\niloc 결과:")
print(result_iloc)
````

````python
# 테스트 케이스로 검증
assert result_loc["이름"] == "이영희", "이름이 이영희여야 합니다"
assert result_loc["나이"] == 28, "나이가 28이어야 합니다"
assert result_iloc["이름"] == "이영희", "iloc 결과도 동일해야 합니다"
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- loc: 라벨(인덱스 번호, 컬럼명) 사용
- iloc: 위치(0부터 시작하는 정수) 사용

**핵심 개념**:
- `loc[행라벨, 컬럼라벨]`: 라벨 기반 인덱싱
- `iloc[행위치, 컬럼위치]`: 위치 기반 인덱싱
- 2번째 행 = 인덱스 1 (0부터 시작)

**대안 솔루션**:
```python
# 컬럼만 먼저 선택 후 행 선택
result = people[["이름", "나이"]].iloc[1]
```

**흔한 실수**:
- ❌ `loc[2, ...]`: 2번째 행은 인덱스 1 (0부터 시작)
- ❌ `iloc[1, ["이름", "나이"]]`: iloc은 컬럼명 사용 불가

**실무 팁**:
컬럼명을 아는 경우 loc이 명확하고, 위치만 아는 경우 iloc을 사용합니다.

````python
# Q8 정답

visits = pd.DataFrame({
    "고객명": ["김철수", "이영희", "김철수", "박민수", "이영희"],
    "방문일": ["2024-01-01", "2024-01-02", "2024-01-05", "2024-01-03", "2024-01-06"]
})

# 1. 고객명 기준 중복 제거 (마지막 행 유지)
result = visits.drop_duplicates(subset=["고객명"], keep="last")

# 결과 출력
print(result)
````

### Q9. apply 활용 ⭐⭐⭐⭐

**문제**: 다음 DataFrame의 "가격" 컬럼에 apply를 사용하여 "가격대" 컬럼을 추가하세요.

- 1000 이상: "고가"
- 100 이상 1000 미만: "중가"
- 100 미만: "저가"

```python
products = pd.DataFrame({
    "제품": ["노트북", "마우스", "키보드", "모니터"],
    "가격": [1200, 30, 80, 350]
})
```

````python
# Q9 정답

products = pd.DataFrame({
    "제품": ["노트북", "마우스", "키보드", "모니터"],
    "가격": [1200, 30, 80, 350]
})

# 1. 가격대 판별 함수 정의
def categorize_price(price):
    if price >= 1000:
        return "고가"
    elif price >= 100:
        return "중가"
    else:
        return "저가"

# 2. apply로 함수 적용
products["가격대"] = products["가격"].apply(categorize_price)

# 결과 출력
print(products)
````

````python
# 테스트 케이스로 검증
assert products.loc[0, "가격대"] == "고가", "노트북은 고가여야 합니다"
assert products.loc[1, "가격대"] == "저가", "마우스는 저가여야 합니다"
assert products.loc[3, "가격대"] == "중가", "모니터는 중가여야 합니다"
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
가격대를 판별하는 함수를 정의하고, apply()로 각 행에 적용합니다.

**핵심 개념**:
- `Series.apply(함수)`: 각 값에 함수 적용
- if-elif-else로 조건 분기
- 함수는 단일 값을 받아 단일 값 반환

**대안 솔루션**:
```python
# lambda 함수 사용 (한 줄)
products["가격대"] = products["가격"].apply(
    lambda x: "고가" if x >= 1000 else "중가" if x >= 100 else "저가"
)

# np.select() 사용 (더 빠름)
conditions = [products["가격"] >= 1000, products["가격"] >= 100]
choices = ["고가", "중가"]
products["가격대"] = np.select(conditions, choices, default="저가")
```

**흔한 실수**:
- ❌ 조건 순서 잘못: `>= 100`을 먼저 체크하면 1000 이상도 "중가"로 분류됨
- ❌ `apply(categorize_price())`: 괄호 있으면 함수 호출 결과 전달 (함수 자체 전달해야 함)

**실무 팁**:
복잡한 로직은 함수로 분리하면 가독성과 재사용성이 높아집니다.

### Q10. 복합 조건 필터링 ⭐⭐⭐⭐⭐

**문제**: 다음 DataFrame에서 아래 조건을 모두 만족하는 행을 추출하세요.

1. 지역이 "서울" 또는 "부산"
2. 매출이 1000 이상

```python
sales_report = pd.DataFrame({
    "지역": ["서울", "부산", "대구", "서울", "부산", "인천"],
    "매출": [1200, 800, 950, 1500, 1100, 700],
    "직원수": [5, 3, 4, 6, 4, 3]
})
```

````python
# Q10 정답

sales_report = pd.DataFrame({
    "지역": ["서울", "부산", "대구", "서울", "부산", "인천"],
    "매출": [1200, 800, 950, 1500, 1100, 700],
    "직원수": [5, 3, 4, 6, 4, 3]
})

# 1. 조건 1: 지역이 서울 또는 부산
condition1 = sales_report["지역"].isin(["서울", "부산"])

# 2. 조건 2: 매출이 1000 이상
condition2 = sales_report["매출"] >= 1000

# 3. 두 조건을 AND로 결합
result = sales_report[condition1 & condition2]

# 결과 출력
print(result)
````

````python
# 테스트 케이스로 검증
assert len(result) == 3, "결과가 3행이어야 합니다"
assert all(result["지역"].isin(["서울", "부산"])), "모든 지역이 서울 또는 부산이어야 합니다"
assert all(result["매출"] >= 1000), "모든 매출이 1000 이상이어야 합니다"
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
1. isin()으로 지역 조건 생성
2. 비교 연산자로 매출 조건 생성
3. & 연산자로 두 조건을 AND 결합

**핵심 개념**:
- `isin([값1, 값2])`: 여러 값 중 하나인지 확인
- `&`: AND 연산 (괄호 필수!)
- `|`: OR 연산
- Boolean Series 결합 가능

**대안 솔루션**:
```python
# 한 줄로 축약
result = sales_report[
    (sales_report["지역"].isin(["서울", "부산"])) & 
    (sales_report["매출"] >= 1000)
]

# query() 메소드 (더 읽기 쉬움)
result = sales_report.query("지역 in ['서울', '부산'] and 매출 >= 1000")
```

**흔한 실수**:
- ❌ `condition1 and condition2`: Python의 and는 사용 불가, `&` 사용
- ❌ 괄호 누락: `a & b | c`는 `a & (b | c)`로 해석됨
- ❌ `지역 == "서울" or "부산"`: isin() 사용해야 함

**실무 팁**:
복잡한 조건은 변수로 분리하거나 query() 사용이 가독성을 높입니다.
