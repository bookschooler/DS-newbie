### 1.2 px.scatter() - 산점도 시각화

**산점도 (Scatter Plot)**: 두 연속형 변수의 관계를 점으로 표현

**기본 문법**:
```python
px.scatter(
    data_frame=df,      # pandas DataFrame
    x='컬럼명',          # x축 변수
    y='컬럼명',          # y축 변수
    title='차트 제목'    # 제목
)
```

````python
# 간단한 데이터셋 생성
sales_data = pd.DataFrame({
    '광고비': [50, 80, 120, 150, 200, 250, 300, 350],  # 만원
    '매출': [200, 280, 350, 420, 500, 580, 650, 720],  # 만원
    '지역': ['서울', '부산', '서울', '대구', '부산', '서울', '대구', '부산']
})

# 기본 산점도
fig = px.scatter(
    sales_data,
    x='광고비',
    y='매출',
    title='광고비 vs 매출 분석'
)

fig.show()
````

````python
# 색상으로 카테고리 구분
fig = px.scatter(
    sales_data,
    x='광고비',
    y='매출',
    color='지역',  # 지역별 다른 색상
    title='지역별 광고비-매출 관계'
)

fig.show()
````

💡 **실무 예시**: 마케팅 부서에서 광고비 대비 매출 효율을 지역별로 비교할 때, 색상으로 지역을 구분하면 한눈에 패턴 파악이 가능합니다.

### 1.3 px.line() - 선 그래프 (트렌드)

**선 그래프**: 시간에 따른 변화, 트렌드를 선으로 연결하여 표현

**기본 문법**:
```python
px.line(
    data_frame=df,
    x='날짜',
    y='값',
    title='트렌드 차트'
)
```

````python
# 월별 매출 데이터
monthly_revenue = pd.DataFrame({
    '월': ['1월', '2월', '3월', '4월', '5월', '6월'],
    '매출': [1200, 1350, 1280, 1500, 1650, 1800],  # 만원
    '비용': [800, 850, 820, 900, 950, 1000]  # 만원
})

# 단일 선 그래프
fig = px.line(
    monthly_revenue,
    x='월',
    y='매출',
    title='월별 매출 트렌드',
    markers=True  # 데이터 포인트 표시
)

fig.show()
````

````python
# 다중 선 그래프 (와이드 포맷)
fig = px.line(
    monthly_revenue,
    x='월',
    y=['매출', '비용'],  # 여러 컬럼
    title='월별 매출 vs 비용',
    markers=True
)

fig.show()
````

💡 **실무 예시**: 경영진에게 월별 실적을 보고할 때, 매출과 비용을 한 차트에 겹쳐 그리면 수익성 추이를 즉시 파악할 수 있습니다.

### 1.4 px.bar() - 막대 그래프

**막대 그래프**: 카테고리별 값을 막대 높이로 비교

**기본 문법**:
```python
px.bar(
    data_frame=df,
    x='카테고리',
    y='값',
    title='비교 차트'
)
```

````python
# 제품별 판매량 데이터
product_sales = pd.DataFrame({
    '제품': ['노트북', '모니터', '키보드', '마우스', '헤드셋'],
    '판매량': [120, 200, 350, 280, 150],
    '카테고리': ['PC', 'PC', '주변기기', '주변기기', '주변기기']
})

# 기본 막대 그래프
fig = px.bar(
    product_sales,
    x='제품',
    y='판매량',
    title='제품별 판매량 비교'
)

fig.show()
````

````python
# 카테고리별 색상 구분
fig = px.bar(
    product_sales,
    x='제품',
    y='판매량',
    color='카테고리',  # 카테고리별 색상
    title='카테고리별 제품 판매량'
)

fig.show()
````

````python
# 가로 막대 그래프 (orientation='h')
fig = px.bar(
    product_sales,
    x='판매량',
    y='제품',
    orientation='h',  # horizontal
    title='제품별 판매량 (가로)',
    color='카테고리'
)

fig.show()
````

💡 **실무 예시**: 제품명이 긴 경우 가로 막대 그래프(`orientation='h'`)를 사용하면 라벨이 겹치지 않아 가독성이 높아집니다.

### 1.5 px.histogram() - 히스토그램

**히스토그램**: 연속형 데이터의 분포를 구간(bin)으로 나누어 표현

**기본 문법**:
```python
px.histogram(
    data_frame=df,
    x='값',
    nbins=10,  # 구간 개수
    title='분포 차트'
)
```

````python
# 고객 나이 데이터 생성 (정규분포)
np.random.seed(42)
customer_data = pd.DataFrame({
    '나이': np.random.normal(35, 10, 500).astype(int),  # 평균 35, 표준편차 10
    '성별': np.random.choice(['남성', '여성'], 500)
})

# 기본 히스토그램
fig = px.histogram(
    customer_data,
    x='나이',
    nbins=20,
    title='고객 나이 분포'
)

fig.show()
````

````python
# 그룹별 히스토그램 (겹쳐 그리기)
fig = px.histogram(
    customer_data,
    x='나이',
    color='성별',
    nbins=20,
    title='성별 고객 나이 분포',
    barmode='overlay',  # 겹쳐서 표시
    opacity=0.7  # 투명도 조절
)

fig.show()
````

### 1.6 px.box() - 박스플롯

**박스플롯**: 최소값, 1사분위수, 중앙값, 3사분위수, 최대값을 박스로 표현

**기본 문법**:
```python
px.box(
    data_frame=df,
    x='카테고리',
    y='값',
    title='분포 비교'
)
```

````python
# 부서별 연봉 데이터
salary_data = pd.DataFrame({
    '부서': ['영업']*50 + ['개발']*50 + ['마케팅']*50,
    '연봉': np.concatenate([
        np.random.normal(4500, 800, 50),  # 영업: 평균 4500만원
        np.random.normal(5500, 1000, 50),  # 개발: 평균 5500만원
        np.random.normal(4000, 600, 50)   # 마케팅: 평균 4000만원
    ])
})

# 박스플롯
fig = px.box(
    salary_data,
    x='부서',
    y='연봉',
    title='부서별 연봉 분포',
    points='all'  # 모든 데이터 포인트 표시
)

fig.show()
````

💡 **실무 예시**: 박스플롯은 이상치(outlier)를 점으로 표시하므로, 급여 데이터에서 비정상적으로 높거나 낮은 값을 한눈에 발견할 수 있습니다.

## Part 2: 인터랙티브 기능

### 2.1 hover_data, color, size - 다차원 데이터 표현

**인터랙티브 요소**:
- `hover_data`: 마우스 오버 시 추가 정보 표시
- `color`: 카테고리 또는 연속값으로 색상 매핑
- `size`: 값의 크기를 점 크기로 표현

이 세 가지를 조합하면 **4차원 이상의 데이터**를 2D 차트에 표현 가능!

````python
# 다차원 매장 데이터
store_data = pd.DataFrame({
    '매장명': ['강남점', '홍대점', '신촌점', '잠실점', '건대점', '수원점'],
    '월_매출': [5000, 4200, 3800, 4500, 3500, 3000],  # 만원
    '방문객': [1200, 980, 850, 1050, 800, 650],
    '직원수': [12, 10, 8, 11, 7, 6],
    '지역': ['서울', '서울', '서울', '서울', '서울', '경기'],
    '평점': [4.5, 4.3, 4.1, 4.4, 4.0, 3.9]
})

# 다차원 산점도
fig = px.scatter(
    store_data,
    x='방문객',
    y='월_매출',
    size='직원수',  # 점 크기 = 직원수
    color='평점',  # 색상 = 평점 (연속값)
    hover_data=['매장명', '지역'],  # 마우스 오버 시 추가 정보
    title='매장 성과 분석 (4차원)',
    labels={'방문객': '월간 방문객 수', '월_매출': '월 매출 (만원)'},
    size_max=30  # 최대 점 크기
)

fig.show()
````

💡 **실무 예시**: 위 차트는 x축(방문객), y축(매출), 점 크기(직원수), 색상(평점) 총 4개 변수를 동시에 표현합니다. 마우스를 올리면 매장명과 지역도 확인 가능!

### 2.2 facet_row, facet_col - 서브플롯

**서브플롯 (Facets)**: 카테고리별로 여러 차트를 자동 생성

**기본 문법**:
```python
px.scatter(
    df,
    x='값1',
    y='값2',
    facet_col='카테고리1',  # 열 방향 분할
    facet_row='카테고리2'   # 행 방향 분할
)
```

````python
# 제품별, 분기별 데이터
quarterly_sales = pd.DataFrame({
    '분기': ['Q1', 'Q2', 'Q3', 'Q4'] * 3,
    '제품': ['A']*4 + ['B']*4 + ['C']*4,
    '매출': [120, 150, 180, 200, 90, 110, 130, 150, 200, 220, 250, 280],
    '광고비': [30, 35, 40, 45, 20, 25, 30, 35, 50, 55, 60, 65]
})

# 제품별로 열 분할 (facet_col)
fig = px.scatter(
    quarterly_sales,
    x='광고비',
    y='매출',
    facet_col='제품',  # 제품별로 열 분할
    title='제품별 광고비-매출 관계',
    trendline='ols'  # 추세선 추가 (선형회귀)
)

fig.show()
````

💡 **실무 예시**: `facet_col`을 사용하면 각 제품의 광고비-매출 관계를 별도 차트로 자동 생성하여, 제품별 광고 효율을 한눈에 비교할 수 있습니다.

### 2.3 animation_frame - 애니메이션

**애니메이션**: 시간 흐름에 따른 변화를 동적으로 표현

**기본 문법**:
```python
px.scatter(
    df,
    x='값1',
    y='값2',
    animation_frame='시간',  # 애니메이션 기준 (년도, 월 등)
    range_x=[최소, 최대],  # x축 고정 (애니메이션 시 축 변동 방지)
    range_y=[최소, 최대]   # y축 고정
)
```

````python
# 월별 매출 추이 (애니메이션용 데이터)
monthly_trend = pd.DataFrame({
    '월': ['1월']*3 + ['2월']*3 + ['3월']*3 + ['4월']*3,
    '지역': ['서울', '부산', '대구'] * 4,
    '매출': [100, 80, 60, 120, 95, 70, 140, 110, 85, 160, 125, 95],
    '방문객': [500, 400, 300, 600, 480, 350, 700, 550, 425, 800, 625, 475]
})

# 애니메이션 산점도
fig = px.scatter(
    monthly_trend,
    x='방문객',
    y='매출',
    animation_frame='월',  # 월별 애니메이션
    color='지역',
    size='매출',
    title='월별 매출 변화 (애니메이션)',
    range_x=[0, 900],  # x축 범위 고정
    range_y=[0, 200],  # y축 범위 고정
    size_max=30
)

fig.show()
````

💡 **실무 예시**: 경영진 보고에서 애니메이션 차트를 재생하면, 시간에 따른 지역별 성장 추이를 영화처럼 생생하게 전달할 수 있습니다.

### 2.4 테마와 템플릿

**템플릿**: Plotly에서 제공하는 사전 정의된 디자인 세트

**주요 템플릿**:
- `plotly` (기본)
- `plotly_white` (깔끔한 흰색 배경)
- `plotly_dark` (다크 모드)
- `simple_white` (미니멀)
- `ggplot2`, `seaborn` (다른 라이브러리 스타일)

**적용 방법**:
```python
fig = px.scatter(..., template='plotly_white')
```

````python
# 동일한 데이터, 다른 템플릿
sample_data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [10, 15, 13, 17, 20]
})

# plotly_white 템플릿
fig1 = px.scatter(
    sample_data,
    x='x',
    y='y',
    title='plotly_white 템플릿',
    template='plotly_white'
)

fig1.show()
````

````python
# plotly_dark 템플릿
fig2 = px.scatter(
    sample_data,
    x='x',
    y='y',
    title='plotly_dark 템플릿',
    template='plotly_dark'
)

fig2.show()
````

💡 **실무 팁**: 회사 브랜드 가이드에 맞춰 템플릿을 선택하면 시각적 일관성을 유지할 수 있습니다. `plotly_white`는 보고서용, `plotly_dark`는 발표용으로 적합합니다.

### Part 1: 기초 핵심 요약

| 차트 유형 | 함수 | 주요 매개변수 | 실무 활용 |
|----------|------|--------------|----------|
| 산점도 | `px.scatter()` | `x, y, color, size` | 두 변수 간 상관관계 분석 |
| 선 그래프 | `px.line()` | `x, y, markers` | 시간에 따른 트렌드 파악 |
| 막대 그래프 | `px.bar()` | `x, y, color, orientation` | 카테고리별 값 비교 |
| 히스토그램 | `px.histogram()` | `x, nbins, barmode` | 데이터 분포 확인 |
| 박스플롯 | `px.box()` | `x, y, points` | 분포 요약 통계량 + 이상치 탐지 |

### Part 2: 인터랙티브 기능 핵심 요약

| 기능 | 매개변수 | 효과 | 언제 쓰나? |
|------|---------|------|----------|
| 다차원 표현 | `size, color, hover_data` | 3개 이상 변수를 2D에 표현 | 복합 지표 대시보드 |
| 서브플롯 | `facet_col, facet_row` | 카테고리별 차트 자동 생성 | 그룹 간 패턴 비교 |
| 애니메이션 | `animation_frame` | 시간 흐름 시각화 | 변화 추이 강조 발표 |
| 템플릿 | `template` | 시각적 일관성 유지 | 브랜드 가이드 준수 |

### 💡 실무 팁

1. **한 줄로 복잡한 차트**: `px.scatter(df, x='A', y='B', color='C', size='D')`만으로 4차원 시각화 가능
2. **HTML 저장**: `fig.write_html('report.html')`로 저장 후 이메일 첨부, 누구나 브라우저에서 확인
3. **축 범위 고정**: 애니메이션 시 `range_x, range_y`로 축을 고정하면 비교가 쉬움
4. **템플릿 통일**: 프로젝트 시작 시 `template='plotly_white'` 같은 템플릿을 정해두면 모든 차트가 일관된 디자인
5. **hover_data 활용**: 차트에 모든 정보를 표시하지 말고, 중요한 것만 표시하고 나머지는 hover_data로 숨기면 깔끔함
