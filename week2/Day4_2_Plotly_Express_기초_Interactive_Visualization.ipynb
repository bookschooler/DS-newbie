````python
# ë¶€ì„œë³„ ì—°ë´‰ ë°ì´í„°
salary_data = pd.DataFrame({
    'ë¶€ì„œ': ['ì˜ì—…']*50 + ['ê°œë°œ']*50 + ['ë§ˆì¼€íŒ…']*50,
    'ì—°ë´‰': np.concatenate([
        np.random.normal(4500, 800, 50),  # ì˜ì—…: í‰ê·  4500ë§Œì›
        np.random.normal(5500, 1000, 50),  # ê°œë°œ: í‰ê·  5500ë§Œì›
        np.random.normal(4000, 600, 50)   # ë§ˆì¼€íŒ…: í‰ê·  4000ë§Œì›
    ])
})

# ë°•ìŠ¤í”Œë¡¯
fig = px.box(
    salary_data,
    x='ë¶€ì„œ',
    y='ì—°ë´‰',
    title='ë¶€ì„œë³„ ì—°ë´‰ ë¶„í¬',
    points='all'  # ëª¨ë“  ë°ì´í„° í¬ì¸íŠ¸ í‘œì‹œ
)

fig.show()
````

- `hover_data`: ë§ˆìš°ìŠ¤ ì˜¤ë²„ ì‹œ ì¶”ê°€ ì •ë³´ í‘œì‹œ
- `color`: ì¹´í…Œê³ ë¦¬ ë˜ëŠ” ì—°ì†ê°’ìœ¼ë¡œ ìƒ‰ìƒ ë§¤í•‘
