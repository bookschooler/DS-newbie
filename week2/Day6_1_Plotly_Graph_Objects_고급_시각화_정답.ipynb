## Q1. ê¸°ë³¸ Scatter ì°¨íŠ¸ ë§Œë“¤ê¸° â­

## Q4. 1í–‰ 2ì—´ Subplot ë§Œë“¤ê¸° â­â­â­

### ğŸ“‹ ë¬¸ì œ
make_subplots()ë¥¼ ì‚¬ìš©í•˜ì—¬ ìˆ˜ìµê³¼ ë¹„ìš©ì„ ë‚˜ë€íˆ í‘œì‹œí•˜ì„¸ìš”.

```python

```

**ê¸°ëŒ€ ê²°ê³¼**: 1í–‰ 2ì—´ subplot (ìˆ˜ìµ ì™¼ìª½, ë¹„ìš© ì˜¤ë¥¸ìª½)

````python
# ë°ì´í„°
months = ['1ì›”', '2ì›”', '3ì›”']
revenue = [200, 250, 300]
cost = [150, 180, 200]

# ì •ë‹µ ì½”ë“œ
# 1í–‰ 2ì—´ Subplot ìƒì„±
fig = make_subplots(
    rows=1,                             # í–‰ ê°œìˆ˜
    cols=2,                             # ì—´ ê°œìˆ˜
    subplot_titles=('ìˆ˜ìµ', 'ë¹„ìš©')     # ê° subplot ì œëª©
)

# ìˆ˜ìµ ì°¨íŠ¸ (1í–‰ 1ì—´)
fig.add_trace(
    go.Scatter(x=months, y=revenue, mode='lines+markers', name='ìˆ˜ìµ'),
    row=1,  # ì²« ë²ˆì§¸ í–‰
    col=1   # ì²« ë²ˆì§¸ ì—´
)

# ë¹„ìš© ì°¨íŠ¸ (1í–‰ 2ì—´)
fig.add_trace(
    go.Scatter(x=months, y=cost, mode='lines+markers', name='ë¹„ìš©'),
    row=1,  # ì²« ë²ˆì§¸ í–‰
    col=2   # ë‘ ë²ˆì§¸ ì—´
)

# ì „ì²´ ì œëª© ì„¤ì •
fig.update_layout(title_text='ìˆ˜ìµ vs ë¹„ìš©', height=400, width=900)

fig.show()
````

## Q5. ì´ì¤‘ Yì¶• ì°¨íŠ¸ ë§Œë“¤ê¸° â­â­â­

### ğŸ“‹ ë¬¸ì œ
íŒë§¤ëŸ‰(ê°œ)ê³¼ ë§Œì¡±ë„(ì )ë¥¼ ì´ì¤‘ Yì¶•ìœ¼ë¡œ í‘œì‹œí•˜ì„¸ìš”.

```python
months = ['1ì›”', '2ì›”', '3ì›”', '4ì›”']
quantity = [1000, 1200, 1150, 1300]
satisfaction = [4.2, 4.5, 4.3, 4.7]
```

**ìš”êµ¬ì‚¬í•­**:
- Primary Yì¶•: íŒë§¤ëŸ‰

````python
# ë°ì´í„°
months = ['1ì›”', '2ì›”', '3ì›”', '4ì›”']
quantity = [1000, 1200, 1150, 1300]
satisfaction = [4.2, 4.5, 4.3, 4.7]

# ì •ë‹µ ì½”ë“œ
# Secondary Yì¶•ì„ ê°€ì§„ Subplot ìƒì„±
fig = make_subplots(
    specs=[[{"secondary_y": True}]]  # Secondary Yì¶• í™œì„±í™”
)

# íŒë§¤ëŸ‰ (Primary Yì¶•)
fig.add_trace(
    go.Scatter(x=months, y=quantity, mode='lines+markers', name='íŒë§¤ëŸ‰'),
    secondary_y=False  # Primary Yì¶• ì‚¬ìš©
)

# ë§Œì¡±ë„ (Secondary Yì¶•)
fig.add_trace(
    go.Scatter(x=months, y=satisfaction, mode='lines+markers', name='ë§Œì¡±ë„', 
               line=dict(color='red')),
    secondary_y=True   # Secondary Yì¶• ì‚¬ìš©
)

# ì¶• ì œëª© ì„¤ì •
fig.update_xaxes(title_text='ì›”')
fig.update_yaxes(title_text='íŒë§¤ëŸ‰ (ê°œ)', secondary_y=False)    # Primary Yì¶•
fig.update_yaxes(title_text='ë§Œì¡±ë„ (ì )', secondary_y=True)     # Secondary Yì¶•

# ì „ì²´ ì œëª©
fig.update_layout(title_text='íŒë§¤ëŸ‰ vs ë§Œì¡±ë„', hovermode='x unified')

fig.show()
````

## Q10. ì¢…í•© ì¸í„°ë™í‹°ë¸Œ ëŒ€ì‹œë³´ë“œ â­â­â­â­â­

### ğŸ“‹ ë¬¸ì œ
3ê°œ ì§€ì—­ì˜ ë§¤ì¶œì„ Dropdownìœ¼ë¡œ ì „í™˜í•˜ê³ , ëª©í‘œì„ ê³¼ ì£¼ì„ì„ í¬í•¨í•œ ëŒ€ì‹œë³´ë“œë¥¼ ë§Œë“œì„¸ìš”.

```python
months = ['1ì›”', '2ì›”', '3ì›”', '4ì›”']
seoul = [300, 320, 350, 380]
busan = [150, 160, 170, 180]
daegu = [100, 110, 120, 130]
target_line = 200
```

**ìš”êµ¬ì‚¬í•­**:
- Dropdown ë©”ë‰´ë¡œ ì„œìš¸/ë¶€ì‚°/ëŒ€êµ¬ ì „í™˜
- ëª©í‘œì„  (200) ì¶”ê°€
- ê° ì§€ì—­ ìµœê³ ì ì— ì£¼ì„
- ì´ˆê¸° í™”ë©´: ì„œìš¸

````python
# ë°ì´í„°
months = ['1ì›”', '2ì›”', '3ì›”', '4ì›”']
seoul = [300, 320, 350, 380]
busan = [150, 160, 170, 180]
daegu = [100, 110, 120, 130]
target_line = 200

# ì§€ì—­ë³„ ë°ì´í„°ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ êµ¬ì„±
regions_data = {
    'ì„œìš¸': seoul,
    'ë¶€ì‚°': busan,
    'ëŒ€êµ¬': daegu
}

# ì •ë‹µ ì½”ë“œ
# Figure ìƒì„±
fig = go.Figure()

# ê° ì§€ì—­ trace ì¶”ê°€
for idx, (region, data) in enumerate(regions_data.items()):
    fig.add_trace(go.Scatter(
        x=months,
        y=data,
        mode='lines+markers',
        name=region,
        visible=(idx == 0)  # ì²« ë²ˆì§¸(ì„œìš¸)ë§Œ ì´ˆê¸° í‘œì‹œ
    ))

# ëª©í‘œì„  ì¶”ê°€
fig.add_hline(
    y=target_line,
    line_dash='dash',
    line_color='red',
    annotation_text=f'ëª©í‘œ ë§¤ì¶œ ({target_line})',
    annotation_position='right'
)

# ê° ì§€ì—­ì˜ ìµœê³ ì ì— ì£¼ì„ ì¶”ê°€
for region, data in regions_data.items():
    max_value = max(data)
    max_idx = data.index(max_value)
    max_month = months[max_idx]
    
    fig.add_annotation(
        x=max_month,
        y=max_value,
        text=f'{region} ìµœê³ <br>{max_value}ë§Œì›',
        showarrow=True,
        arrowhead=2,
        ax=30,
        ay=-30,
        bgcolor='lightyellow',
        bordercolor='orange',
        visible=(region == 'ì„œìš¸')  # ì´ˆê¸°ì—ëŠ” ì„œìš¸ ì£¼ì„ë§Œ í‘œì‹œ
    )

# Dropdown ë©”ë‰´ ìƒì„±
dropdown_buttons = []
for idx, region in enumerate(regions_data.keys()):
    # trace visible ì„¤ì • (ì„ íƒëœ ì§€ì—­ë§Œ True)
    visible_array = [False] * len(regions_data)
    visible_array[idx] = True
    
    # annotation visible ì„¤ì • (ì„ íƒëœ ì§€ì—­ì˜ ì£¼ì„ë§Œ True)
    # hline annotation(1ê°œ) + ê° ì§€ì—­ ì£¼ì„(3ê°œ) = ì´ 4ê°œ annotation
    annotation_visible = [True]  # ëª©í‘œì„  annotationì€ í•­ìƒ í‘œì‹œ
    for ann_idx in range(len(regions_data)):
        annotation_visible.append(ann_idx == idx)
    
    dropdown_buttons.append(
        dict(
            label=region,
            method='update',
            args=[
                {'visible': visible_array},  # trace í‘œì‹œ ì„¤ì •
                {
                    'title': f'{region} ë§¤ì¶œ ì¶”ì´',
                    'annotations': [
                        ann for ann_idx, ann in enumerate(fig.layout.annotations)
                        if annotation_visible[ann_idx]
                    ]
                }
            ]
        )
    )

# ë ˆì´ì•„ì›ƒ ì„¤ì •
fig.update_layout(
    updatemenus=[
        dict(
            buttons=dropdown_buttons,
            direction='down',    # ë“œë¡­ë‹¤ìš´ ë°©í–¥
            x=0.1,
            y=1.15,
            showactive=True      # í˜„ì¬ ì„ íƒëœ ì˜µì…˜ ê°•ì¡°
        )
    ],
    title='ì„œìš¸ ë§¤ì¶œ ì¶”ì´',  # ì´ˆê¸° ì œëª©
    yaxis_range=[0, 400],    # Yì¶• ë²”ìœ„ ê³ ì • (ì§€ì—­ ì „í™˜ ì‹œ ì¼ê´€ì„±)
    height=500,
    width=800
)

fig.show()
````
