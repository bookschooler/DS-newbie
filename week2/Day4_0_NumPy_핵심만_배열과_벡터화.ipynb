## 1.1 NumPy ë°°ì—´ì´ë€?

### Python ë¦¬ìŠ¤íŠ¸ vs NumPy ë°°ì—´

````python
import numpy as np

# Python ë¦¬ìŠ¤íŠ¸
python_list = [10, 20, 30, 40, 50]
print("Python ë¦¬ìŠ¤íŠ¸:", python_list)
print("íƒ€ì…:", type(python_list))

# NumPy ë°°ì—´
numpy_array = np.array([10, 20, 30, 40, 50])
print("\nNumPy ë°°ì—´:", numpy_array)
print("íƒ€ì…:", type(numpy_array))
print("ë°ì´í„° íƒ€ì…:", numpy_array.dtype)
````

## 2.1 ë¸Œë¡œë“œìºìŠ¤íŒ… (Broadcasting)

### ë¸Œë¡œë“œìºìŠ¤íŒ…ì´ë€?

**í¬ê¸°ê°€ ë‹¤ë¥¸ ë°°ì—´ë¼ë¦¬ ì—°ì‚°í•  ë•Œ ìë™ìœ¼ë¡œ í™•ì¥**í•˜ëŠ” NumPyì˜ ê°•ë ¥í•œ ê¸°ëŠ¥ì…ë‹ˆë‹¤.

````python
# ë°°ì—´ê³¼ ìŠ¤ì¹¼ë¼ ì—°ì‚°
prices = np.array([100, 200, 150, 300])

# ëª¨ë“  ê°€ê²©ì— 10% í• ì¸ (ìŠ¤ì¹¼ë¼ 0.9ê°€ ìë™ í™•ì¥)
discounted = prices * 0.9
print("ì›ê°€:", prices)
print("í• ì¸ê°€:", discounted)
````

````python
# 1D + 1D ë¸Œë¡œë“œìºìŠ¤íŒ…
base_price = np.array([1000, 2000, 1500])
markup = np.array([100, 200, 150])  # ë§ˆì§„

final_price = base_price + markup
print("ê¸°ë³¸ê°€:", base_price)
print("ë§ˆì§„:", markup)
print("ìµœì¢…ê°€:", final_price)
````

### 2D ë¸Œë¡œë“œìºìŠ¤íŒ…

````python
# ì§€ì ë³„ íŒë§¤ëŸ‰ (2D)
sales = np.array([
    [100, 200, 150],  # ê°•ë‚¨ì 
    [120, 180, 160]   # ë¶„ë‹¹ì 
])

# ìƒí’ˆë³„ ê°€ê²© (1D)
prices = np.array([10, 20, 15])  # ë§Œì›

# ë¸Œë¡œë“œìºìŠ¤íŒ…: ê° ìƒí’ˆ ê°€ê²©ì´ ëª¨ë“  ì§€ì ì— ì ìš©ë¨
revenue = sales * prices
print("íŒë§¤ëŸ‰:")
print(sales)
print("\nê°€ê²©:", prices)
print("\në§¤ì¶œ:")
print(revenue)
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ì§€ì—­ë³„ í• ì¸ìœ¨ ì ìš©

````python
# 3ê°œ ì§€ì , 4ê°œ ìƒí’ˆ íŒë§¤ê°€
prices_matrix = np.array([
    [10000, 20000, 15000, 30000],  # ì„œìš¸
    [10000, 20000, 15000, 30000],  # ë¶€ì‚°
    [10000, 20000, 15000, 30000]   # ëŒ€êµ¬
])

# ì§€ì—­ë³„ í• ì¸ìœ¨ (1D ë°°ì—´, ì—´ ë°©í–¥ í™•ì¥)
regional_discount = np.array([0.9, 0.85, 0.88]).reshape(3, 1)  # 10%, 15%, 12%

# ë¸Œë¡œë“œìºìŠ¤íŒ…ìœ¼ë¡œ ì§€ì—­ë³„ í• ì¸ê°€ ê³„ì‚°
discounted_prices = prices_matrix * regional_discount

print("ì›ê°€:")
print(prices_matrix)
print("\ní• ì¸ìœ¨:", regional_discount.flatten())
print("\ní• ì¸ê°€:")
print(discounted_prices.astype(int))
````

## 2.2 ë²¡í„°í™” ì—°ì‚°ì˜ ì„±ëŠ¥ ì´ì 

### ë°˜ë³µë¬¸ vs ë²¡í„°í™” ì†ë„ ë¹„êµ

````python
import time

# 100ë§Œ ê°œ ë°ì´í„°
data = np.random.randint(1, 100, size=1_000_000)

# ë°©ë²• 1: Python ë°˜ë³µë¬¸
start = time.time()
result_loop = []
for d in data:
    result_loop.append(d * 2 + 10)
time_loop = time.time() - start

# ë°©ë²• 2: NumPy ë²¡í„°í™”
start = time.time()
result_numpy = data * 2 + 10
time_numpy = time.time() - start

print(f"ë°˜ë³µë¬¸ ì‹œê°„: {time_loop:.4f}ì´ˆ")
print(f"ë²¡í„°í™” ì‹œê°„: {time_numpy:.4f}ì´ˆ")
print(f"\nì„±ëŠ¥ í–¥ìƒ: {time_loop / time_numpy:.1f}ë°° ë¹ ë¦„!")
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ëŒ€ìš©ëŸ‰ ë§¤ì¶œ ë°ì´í„° ì²˜ë¦¬

````python
# 10ë§Œ ê±´ ê±°ë˜ ë°ì´í„°
transaction_amounts = np.random.randint(10, 1000, size=100_000)

# ë³µì¡í•œ ê³„ì‚°: ë¶€ê°€ì„¸ + ìˆ˜ìˆ˜ë£Œ + í¬ì¸íŠ¸ ì ë¦½
start = time.time()
vat = transaction_amounts * 0.1
fee = transaction_amounts * 0.02
points = transaction_amounts * 0.01
final_amount = transaction_amounts + vat + fee - points
time_taken = time.time() - start

print(f"ì²˜ë¦¬ ê±´ìˆ˜: {len(transaction_amounts):,}ê±´")
print(f"ì²˜ë¦¬ ì‹œê°„: {time_taken * 1000:.2f}ms")
print(f"í‰ê·  ê±°ë˜ì•¡: {transaction_amounts.mean():.0f}ì›")
print(f"ì´ ë¶€ê°€ì„¸: {vat.sum():,.0f}ì›")
print(f"ì´ ìˆ˜ìˆ˜ë£Œ: {fee.sum():,.0f}ì›")
````

### ë°°ì—´ ê²°í•© (concatenate, vstack, hstack)

````python
# 1ì›”, 2ì›” ë§¤ì¶œ
jan_sales = np.array([100, 200, 150])
feb_sales = np.array([120, 180, 160])

# ìˆ˜í‰ ê²°í•© (ì´ì–´ë¶™ì´ê¸°)
combined = np.concatenate([jan_sales, feb_sales])
print("ê²°í•© (1D):", combined)

# ìˆ˜ì§ ìŒ“ê¸° (í–‰ ì¶”ê°€)
stacked = np.vstack([jan_sales, feb_sales])
print("\nìˆ˜ì§ ìŒ“ê¸° (2D):")
print(stacked)
````
