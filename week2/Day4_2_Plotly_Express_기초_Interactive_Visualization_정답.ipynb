### Q1. 기본 산점도 그리기 ⭐

**문제**: 아래 데이터로 `공부시간(x축)` vs `시험점수(y축)` 산점도를 그리세요.

````python
# Q1 정답
import plotly.express as px
import pandas as pd

study_data = pd.DataFrame({
    '공부시간': [2, 4, 6, 8, 10],
    '시험점수': [60, 70, 80, 85, 95]
})

# 산점도 생성
fig = px.scatter(
    study_data,
    x='공부시간',
    y='시험점수',
    title='공부시간 vs 시험점수'
)

fig.show()
````

````python
# 테스트: 차트 객체가 생성되었는지 확인
assert fig is not None, "차트 객체가 생성되지 않았습니다."
assert len(fig.data) == 1, "산점도는 하나의 trace를 가져야 합니다."
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- `px.scatter()`의 기본 형태: `데이터프레임`, `x`, `y`, `title`
- DataFrame을 첫 번째 인자로, 컬럼명을 문자열로 전달

**핵심 개념**:
- Plotly Express는 pandas DataFrame과 완벽 통합
- `x='컬럼명'` 형태로 간단하게 축 지정

**대안 솔루션**:
```python
# Graph Objects를 사용한 방법 (더 복잡)
# import plotly.graph_objects as go
# fig = go.Figure(data=go.Scatter(x=study_data['공부시간'], y=study_data['시험점수']))
```

**흔한 실수**:
- ❌ `x=study_data['공부시간']` (시리즈 전달) → 동작하지만 권장하지 않음
- ❌ `fig.plot()` → Plotly는 `.show()` 사용

**실무 팁**:
px.scatter()는 matplotlib의 plt.scatter()보다 훨씬 간결하며, 자동으로 인터랙티브 기능(줌, 팬, hover)이 추가됩니다.

### Q2. 색상으로 카테고리 구분 ⭐⭐

**문제**: 아래 데이터로 산점도를 그리되, `과목`별로 다른 색상을 적용하세요.

````python
# Q2 정답
grade_data = pd.DataFrame({
    '공부시간': [3, 5, 7, 4, 6, 8],
    '점수': [65, 75, 85, 70, 80, 90],
    '과목': ['수학', '수학', '수학', '영어', '영어', '영어']
})

# 과목별 색상 구분
fig = px.scatter(
    grade_data,
    x='공부시간',
    y='점수',
    color='과목',  # 핵심: 카테고리 컬럼을 color에 지정
    title='과목별 공부시간-점수 관계'
)

fig.show()
````

````python
# 테스트: 두 개의 trace가 생성되었는지 확인 (과목 2개)
assert len(fig.data) == 2, "과목이 2개이므로 trace도 2개여야 합니다."
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- `color` 매개변수에 카테고리 컬럼명을 전달하면 자동으로 색상 매핑
- Plotly가 고유값(수학, 영어)을 인식하고 각각 다른 색상 할당

**핵심 개념**:
- `color`: 카테고리형 → 이산 색상, 연속형 → 그라디언트 색상
- 범례(legend)가 자동 생성됨

**대안 솔루션**:
```python
# 색상을 직접 지정하는 방법
# fig = px.scatter(..., color='과목', color_discrete_map={'수학': 'red', '영어': 'blue'})
```

**흔한 실수**:
- ❌ `color=['red', 'red', 'red', 'blue', 'blue', 'blue']` → 리스트가 아닌 컬럼명 전달

**실무 팁**:

### Q3. 선 그래프로 트렌드 표현 ⭐⭐

**문제**: 아래 데이터로 `주차`별 `판매량` 선 그래프를 그리세요. 데이터 포인트도 표시해야 합니다.

````python
# Q3 정답
weekly_sales = pd.DataFrame({
    '주차': ['1주', '2주', '3주', '4주', '5주'],
    '판매량': [120, 135, 150, 145, 170]
})

# 선 그래프 + 마커
fig = px.line(
    weekly_sales,
    x='주차',
    y='판매량',
    title='주차별 판매량 트렌드',
    markers=True  # 핵심: 데이터 포인트 표시
)

fig.show()
````

````python
# 테스트: 마커가 표시되었는지 확인
assert fig.data[0].mode == 'lines+markers', "markers=True 설정이 필요합니다."
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- `px.line()`의 기본 형태는 선만 표시
- `markers=True`를 추가하면 각 데이터 포인트에 점이 표시됨

**핵심 개념**:
- 선 그래프는 시간/순서가 있는 데이터에 적합
- `markers=True`: 데이터 포인트 강조, 특히 데이터가 적을 때 유용

**대안 솔루션**:
```python
# 마커 스타일 직접 지정
# fig.update_traces(marker=dict(size=10, symbol='circle'))
```

**흔한 실수**:
- ❌ `marker=True` → 매개변수명은 `markers` (복수형)
- ❌ `points=True` → 이건 px.box()에서 사용

**실무 팁**:
데이터 포인트가 5개 이하일 때는 markers=True를 권장합니다. 100개 이상이면 선만 표시하는 것이 깔끔합니다.

### Q5. 히스토그램으로 분포 파악 ⭐⭐⭐

**문제**: 아래 나이 데이터로 히스토그램을 그리되, 구간(nbins)을 15개로 설정하세요.

````python
# Q5 정답
ages = pd.DataFrame({
    '나이': [23, 25, 28, 30, 32, 35, 38, 40, 42, 45, 48, 50, 52, 55, 58,
             22, 26, 29, 31, 34, 36, 39, 41, 44, 47, 49, 51, 54, 56, 59]
})

# 히스토그램 (구간 15개)
fig = px.histogram(
    ages,
    x='나이',
    nbins=15,  # 핵심: 구간 개수 지정
    title='나이 분포 (15개 구간)'
)

fig.show()
````

````python
# 테스트: nbins 설정 확인
# (실제 bin 개수는 데이터 범위에 따라 약간 달라질 수 있음)
assert fig.data[0].nbinsx == 15, "nbins는 15로 설정되어야 합니다."
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- 히스토그램은 연속형 데이터를 구간으로 나누어 빈도 표시
- `nbins`로 구간 개수를 조절 (기본값: 자동 계산)

**핵심 개념**:
- `nbins`가 클수록 세밀한 분포, 작을수록 전체 패턴 파악
- 데이터가 30개일 때 적정 bin 개수는 5~20개

**대안 솔루션**:
```python
# 구간 경계를 직접 지정하는 방법
# import numpy as np
# bins = np.arange(20, 65, 3)  # 20부터 65까지 3씩 증가
# fig = px.histogram(ages, x='나이', nbins=len(bins))
```

**흔한 실수**:
- ❌ `bins=15` → Plotly Express는 `nbins` 사용 (matplotlib과 다름)
- ❌ 너무 많은 bin (nbins=100) → 데이터가 적으면 빈 구간 많아짐

**실무 팁**:
스타지스 공식 `nbins = sqrt(n)`을 참고하되, 시각적으로 적절한 값을 실험적으로 찾는 것이 좋습니다 (여기서는 sqrt(30) ≈ 5.5이지만 15로 설정).

### Q6. 박스플롯으로 분포 비교 ⭐⭐⭐

**문제**: 아래 데이터로 `부서`별 `만족도` 박스플롯을 그리되, 모든 데이터 포인트를 표시하세요.

````python
# Q6 정답
satisfaction = pd.DataFrame({
    '부서': ['영업']*10 + ['개발']*10,
    '만족도': [7, 8, 6, 9, 7, 8, 7, 6, 8, 9,  # 영업
              8, 9, 9, 10, 8, 9, 8, 9, 10, 9]  # 개발
})

# 박스플롯 + 모든 포인트 표시
fig = px.box(
    satisfaction,
    x='부서',
    y='만족도',
    title='부서별 만족도 분포',
    points='all'  # 핵심: 모든 데이터 포인트 표시
)

fig.show()
````

````python
# 테스트: points 설정 확인
assert fig.data[0].boxpoints == 'all', "points='all' 설정이 필요합니다."
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- 박스플롯은 5가지 요약 통계량(최소, Q1, 중앙값, Q3, 최대)을 표시
- `points='all'`로 원본 데이터 포인트까지 표시하면 분포를 더 정확히 파악

**핵심 개념**:
- `points` 옵션: `'all'` (모든 점), `'outliers'` (이상치만), `False` (안 보임)
- 박스 내부 선 = 중앙값, 박스 = IQR (사분위 범위)

**대안 솔루션**:
```python
# 이상치만 표시
# fig = px.box(satisfaction, x='부서', y='만족도', points='outliers')
```

**흔한 실수**:
- ❌ `show_points=True` → 매개변수명은 `points`
- ❌ `points=True` → 'all', 'outliers', False 중 하나 선택

**실무 팁**:
박스플롯은 이상치 탐지에 유용합니다. IQR 범위를 벗어난 점들이 자동으로 강조되므로, 급여 데이터나 성과 지표에서 비정상 값을 빠르게 발견할 수 있습니다.

### Q7. 다차원 산점도 (size + color) ⭐⭐⭐⭐

**문제**: 아래 데이터로 산점도를 그리되, `직원수`를 점 크기로, `지역`을 색상으로 표현하세요.

````python
# Q7 정답
branch_data = pd.DataFrame({
    '방문객': [800, 650, 900, 750],
    '매출': [3500, 2800, 4000, 3200],
    '직원수': [10, 8, 12, 9],
    '지역': ['서울', '경기', '서울', '경기']
})

# 다차원 산점도 (x, y, size, color)
fig = px.scatter(
    branch_data,
    x='방문객',
    y='매출',
    size='직원수',  # 핵심1: 점 크기로 직원수 표현
    color='지역',   # 핵심2: 색상으로 지역 구분
    title='지점 성과 분석 (4차원)',
    size_max=30  # 최대 점 크기 제한
)

fig.show()
````

````python
# 테스트: size와 color가 모두 설정되었는지 확인
assert len(fig.data) == 2, "지역이 2개이므로 trace도 2개여야 합니다."
# size 매핑은 marker.size에 반영됨
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- x축(방문객) + y축(매출) + size(직원수) + color(지역) = 4차원 데이터를 2D에 표현
- `size`는 연속형 값을 받아 점 크기를 자동 조정

**핵심 개념**:
- `size`: 연속형 변수를 점 크기로 매핑 (값이 클수록 큰 점)
- `color`: 카테고리 또는 연속형 변수를 색상으로 매핑
- `size_max`: 점이 너무 커지지 않도록 제한

**대안 솔루션**:
```python
# hover_data로 추가 정보 표시
# fig = px.scatter(..., hover_data=['지점명', '평점'])
```

**흔한 실수**:
- ❌ `size=[10, 8, 12, 9]` → 리스트가 아닌 컬럼명 전달
- ❌ size 없이 color만 사용 → 3차원 표현 (4차원 아님)

**실무 팁**:
다차원 산점도는 대시보드에서 매우 효과적입니다. 한 차트로 여러 KPI를 동시에 모니터링하여 이상 패턴을 빠르게 발견할 수 있습니다.

### Q8. 서브플롯으로 카테고리별 분석 ⭐⭐⭐⭐

**문제**: 아래 데이터로 `제품`별로 열을 나눠(facet_col) 산점도를 그리세요. x축은 `광고비`, y축은 `판매량`입니다.

````python
# Q8 정답
product_ad = pd.DataFrame({
    '광고비': [10, 20, 30, 40, 15, 25, 35, 45],
    '판매량': [50, 70, 90, 110, 60, 80, 100, 120],
    '제품': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B']
})

# 서브플롯 (제품별 열 분할)
fig = px.scatter(
    product_ad,
    x='광고비',
    y='판매량',
    facet_col='제품',  # 핵심: 제품별로 열 방향 분할
    title='제품별 광고비-판매량 관계'
)

fig.show()
````

````python
# 테스트: subplot 개수 확인
# facet_col로 제품 2개이므로 subplot 2개
assert len(fig.data) == 2, "제품이 2개이므로 subplot도 2개여야 합니다."
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- `facet_col`을 사용하면 카테고리 값마다 별도 차트를 자동 생성
- 제품 A와 B의 광고비-판매량 관계를 나란히 비교 가능

**핵심 개념**:
- `facet_col`: 열 방향 분할 (가로로 나열)
- `facet_row`: 행 방향 분할 (세로로 나열)
- 각 subplot은 동일한 x, y 스케일 유지 (비교 용이)

**대안 솔루션**:
```python
# 행 방향 분할로 변경
# fig = px.scatter(..., facet_row='제품')
```

**흔한 실수**:
- ❌ `subplot='제품'` → 매개변수명은 `facet_col` 또는 `facet_row`
- ❌ facet과 color를 동일 컬럼에 동시 적용 → 중복 표현

**실무 팁**:
facet은 A/B 테스트 결과 비교, 지역별 성과 비교, 시간대별 패턴 비교 등 "그룹 간 비교"가 필요할 때 매우 강력한 도구입니다.

### Q9. 애니메이션 산점도 ⭐⭐⭐⭐⭐

**문제**: 아래 데이터로 `년도`별 애니메이션 산점도를 그리세요. x축은 `GDP`, y축은 `기대수명`, 색상은 `대륙`입니다.

````python
# Q9 정답
world_data = pd.DataFrame({
    '년도': [2020, 2020, 2021, 2021, 2022, 2022],
    '국가': ['한국', '미국', '한국', '미국', '한국', '미국'],
    'GDP': [30000, 60000, 32000, 62000, 34000, 64000],
    '기대수명': [82, 78, 83, 79, 84, 80],
    '대륙': ['아시아', '북미', '아시아', '북미', '아시아', '북미']
})

# 애니메이션 산점도
fig = px.scatter(
    world_data,
    x='GDP',
    y='기대수명',
    color='대륙',
    animation_frame='년도',  # 핵심: 년도별 애니메이션
    title='년도별 GDP-기대수명 변화',
    range_x=[25000, 70000],  # x축 범위 고정
    range_y=[75, 90],        # y축 범위 고정
    hover_data=['국가']      # 마우스 오버 시 국가명 표시
)

fig.show()
````

````python
# 테스트: animation_frame 설정 확인
assert 'frames' in fig.to_dict(), "animation_frame이 설정되어야 합니다."
assert len(fig.frames) == 3, "년도가 3개(2020, 2021, 2022)이므로 프레임도 3개여야 합니다."
print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- `animation_frame`에 시간 변수(년도)를 지정하면 각 시점별 프레임 자동 생성
- `range_x`, `range_y`로 축 범위를 고정하면 애니메이션 시 비교가 쉬움

**핵심 개념**:
- `animation_frame`: 시간/순서 변수로 애니메이션 생성
- 축 고정: 프레임마다 축이 바뀌면 변화량 파악 어려움
- hover_data: 애니메이션 중 추가 정보 표시

**대안 솔루션**:
```python
# size로 추가 차원 표현
# fig = px.scatter(..., size='인구', animation_frame='년도')
```

**흔한 실수**:
- ❌ `animation='년도'` → 매개변수명은 `animation_frame`
- ❌ 축 범위 미고정 → 프레임마다 축이 바뀌어 변화 파악 어려움

**실무 팁**:
애니메이션은 경영진 발표에서 "스토리텔링"에 매우 효과적입니다. 시간에 따른 성장, 경쟁사 비교, 지역별 변화를 영화처럼 전달하여 임팩트를 극대화할 수 있습니다.

### Q10. 템플릿 적용 + 다중 변수 시각화 ⭐⭐⭐⭐⭐

**문제**: 아래 데이터로 산점도를 그리되, 다음 조건을 모두 만족하세요:
1. x축: `경험년수`, y축: `연봉`
2. 점 크기: `프로젝트수`
3. 색상: `직급`
4. 템플릿: `plotly_white`
5. hover_data에 `이름` 포함

````python
# Q10 정답
employee_data = pd.DataFrame({
    '이름': ['김철수', '이영희', '박민수', '정수진'],
    '경험년수': [3, 5, 8, 10],
    '연봉': [4000, 5500, 7000, 9000],
    '프로젝트수': [5, 8, 12, 15],
    '직급': ['사원', '대리', '과장', '차장']
})

# 종합 산점도 (5가지 조건 모두 충족)
fig = px.scatter(
    employee_data,
    x='경험년수',          # 조건1
    y='연봉',              # 조건1
    size='프로젝트수',     # 조건2
    color='직급',          # 조건3
    template='plotly_white',  # 조건4
    hover_data=['이름'],   # 조건5
    title='직원 성과 분석 (경험-연봉-프로젝트)',
    labels={'경험년수': '경험 (년)', '연봉': '연봉 (만원)'},
    size_max=40
)

fig.show()
````

````python
# 테스트: 모든 조건 충족 확인
# 1. 축 확인
assert fig.data[0].x.tolist() == [3, 5, 8, 10], "x축이 경험년수여야 합니다."

# 2. 템플릿 확인
assert fig.layout.template.layout.plot_bgcolor == 'white', "템플릿이 plotly_white여야 합니다."

print("✅ 테스트 통과!")
````

### 💡 풀이 설명

**접근 방법**:
- 5가지 조건을 하나씩 매개변수로 추가
- `template`, `hover_data`, `labels`까지 활용한 종합 문제

**핵심 개념**:
- **다차원 표현**: x(경험) + y(연봉) + size(프로젝트) + color(직급) + hover(이름) = 5차원
- **template**: 시각적 일관성 유지 (plotly_white = 깔끔한 흰색 배경)
- **hover_data**: 차트에 모든 정보를 표시하지 않고 마우스 오버 시 추가 정보 제공
- **labels**: 축 이름 커스터마이징

**대안 솔루션**:
```python
# 템플릿을 plotly_dark로 변경
# fig = px.scatter(..., template='plotly_dark')

# 직급별 색상 직접 지정
# color_map = {'사원': 'blue', '대리': 'green', '과장': 'orange', '차장': 'red'}
# fig = px.scatter(..., color_discrete_map=color_map)
```

**흔한 실수**:
- ❌ `hover=['이름']` → 매개변수명은 `hover_data`
- ❌ `theme='plotly_white'` → 매개변수명은 `template`
- ❌ 너무 많은 변수 표현 → 6차원 이상은 오히려 혼란 가중

**실무 팁**:
이러한 종합 차트는 대시보드의 핵심 KPI 모니터링에 활용됩니다. 예를 들어 영업 대시보드에서 "영업사원(x: 경험, y: 매출, size: 고객수, color: 지역, hover: 이름)"을 한눈에 파악하여 성과가 낮은 영업사원을 즉시 식별할 수 있습니다.
