### Q5. explicit_wait ì‚¬ìš© â­â­â­

**ë¬¸ì œ**: `WebDriverWait`ì™€ `EC.element_to_be_clickable`ì„ ì‚¬ìš©í•˜ì—¬ ê²€ìƒ‰ì°½ì´ í´ë¦­ ê°€ëŠ¥í•  ë•Œê¹Œì§€ ê¸°ë‹¤ë¦° í›„ "numpy"ë¥¼ ì…ë ¥í•˜ì„¸ìš”.

```python
# ì •ë‹µ ì½”ë“œ
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
driver.get("https://www.python.org")

# 1. WebDriverWait ê°ì²´ ìƒì„± (ìµœëŒ€ 10ì´ˆ)
wait = WebDriverWait(driver, 10)

# 2. ê²€ìƒ‰ì°½ì´ í´ë¦­ ê°€ëŠ¥í•  ë•Œê¹Œì§€ ëŒ€ê¸°
search_box = wait.until(
    EC.element_to_be_clickable((By.ID, "id-search-field"))
)
print("âœ… ê²€ìƒ‰ì°½ì´ í´ë¦­ ê°€ëŠ¥ ìƒíƒœ")

# 3. í…ìŠ¤íŠ¸ ì…ë ¥
search_box.send_keys("numpy")
print("âœ… 'numpy' ì…ë ¥ ì™„ë£Œ")

driver.quit()
```

### Q9. ìƒˆ ì°½ ì „í™˜ â­â­â­â­

```python
# ì •ë‹µ ì½”ë“œ
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.python.org")

# 1. ì›ë˜ ì°½ í•¸ë“¤ ì €ì¥
main_window = driver.current_window_handle
print(f"ë©”ì¸ ì°½ í•¸ë“¤: {main_window}")

# 2. Downloads ë§í¬ í´ë¦­
downloads_link = driver.find_element(By.LINK_TEXT, "Downloads")
downloads_link.click()
time.sleep(2)  # í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°

# 3. ëª¨ë“  ì°½ í•¸ë“¤ ê°€ì ¸ì˜¤ê¸°
all_windows = driver.window_handles
print(f"ì „ì²´ ì°½ ê°œìˆ˜: {len(all_windows)}ê°œ")

# 4. ìƒˆ ì°½ìœ¼ë¡œ ì „í™˜
for window in all_windows:
    if window != main_window:
        driver.switch_to.window(window)
        print(f"ìƒˆ ì°½ ì œëª©: {driver.title}")
        break

# 5. ì›ë˜ ì°½ìœ¼ë¡œ ë³µê·€
driver.switch_to.window(main_window)
print(f"ë³µê·€ í›„ ì œëª©: {driver.title}")
print("âœ… ë©”ì¸ ì°½ìœ¼ë¡œ ë³µê·€ ì™„ë£Œ")

driver.quit()
```

### Q10. ì¢…í•©: ë‰´ìŠ¤ í—¤ë“œë¼ì¸ ìˆ˜ì§‘ íŒŒì´í”„ë¼ì¸ â­â­â­â­â­

```python
# ì •ë‹µ ì½”ë“œ
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from datetime import datetime
import time

# 1. headless ëª¨ë“œ ì„¤ì •
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-gpu')

driver = webdriver.Chrome(options=options)

try:
    # 2. ë„¤ì´ë²„ ë‰´ìŠ¤ ì ‘ì†
    driver.get("https://news.naver.com/")
    print("âœ… ë„¤ì´ë²„ ë‰´ìŠ¤ ì ‘ì† ì™„ë£Œ")
    
    # 3. í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°
    wait = WebDriverWait(driver, 10)
    wait.until(
        EC.presence_of_element_located((By.CSS_SELECTOR, ".cjs_news_bl"))
    )
    time.sleep(2)  # ì¶”ê°€ ì•ˆì •í™”
    
    # 4. í—¤ë“œë¼ì¸ ìˆ˜ì§‘
    headlines_elements = driver.find_elements(By.CSS_SELECTOR, ".cjs_news_bl .cjs_t")
    print(f"âœ… {len(headlines_elements)}ê°œ í—¤ë“œë¼ì¸ ë°œê²¬")
    
    # 5. ìƒìœ„ 10ê°œ ì¶”ì¶œ
    data = []
    for i, element in enumerate(headlines_elements[:10], 1):
        data.append({
            "ìˆœë²ˆ": i,
            "í—¤ë“œë¼ì¸": element.text,
            "ìˆ˜ì§‘ì‹œê°„": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        })
    
    # 6. DataFrame ìƒì„±
    df = pd.DataFrame(data)
    print("\nğŸ“° ìˆ˜ì§‘ëœ í—¤ë“œë¼ì¸:")
    print(df)
    
    # 7. CSV ì €ì¥
    output_file = "news_headlines.csv"
    df.to_csv(output_file, index=False, encoding='utf-8-sig')
    print(f"\nâœ… {output_file}ë¡œ ì €ì¥ ì™„ë£Œ")
    
finally:
    # 8. ë¸Œë¼ìš°ì € ì¢…ë£Œ (ì˜¤ë¥˜ ë°œìƒí•´ë„ ë°˜ë“œì‹œ ì‹¤í–‰)
    driver.quit()
    print("âœ… ë¸Œë¼ìš°ì € ì¢…ë£Œ")
```
