| **Î™®Ïàò (Parameter)** | Î™®ÏßëÎã®Ïùò ÌäπÏÑ±Í∞í (ÏïåÍ≥† Ïã∂ÏùÄ Í∞í) | 20ÎåÄ Ï†ÑÏ≤¥Ïùò ÌèâÍ∑† ÏõîÍ∏â |
| **ÌÜµÍ≥ÑÎüâ (Statistic)** | ÌëúÎ≥∏ÏúºÎ°ú Í≥ÑÏÇ∞Ìïú Í∞í | ÌëúÎ≥∏ 1,000Î™ÖÏùò ÌèâÍ∑† ÏõîÍ∏â |

````python
# ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏
import numpy as np
import pandas as pd
from scipy import stats
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ÎÇúÏàò ÏãúÎìú Í≥†Ï†ï (Ïû¨ÌòÑÏÑ±)
np.random.seed(2025)

print("ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏ ÏôÑÎ£å!")
````

````python
# Î™®ÏßëÎã®Í≥º ÌëúÎ≥∏ ÏòàÏãú: 20ÎåÄ ÏõîÍ∏â Î∂ÑÏÑù
print("Î™®ÏßëÎã®Í≥º ÌëúÎ≥∏ Í∞úÎÖê ÏòàÏãú")
print("=" * 50)

# Í∞ÄÏÉÅÏùò Î™®ÏßëÎã® ÏÉùÏÑ± (Ïã§Ï†úÎ°úÎäî Ïïå Ïàò ÏóÜÎäî Í∞í!)
# 20ÎåÄ Ï†ÑÏ≤¥ ÏõîÍ∏â Î∂ÑÌè¨: ÌèâÍ∑† 280ÎßåÏõê, ÌëúÏ§ÄÌé∏Ï∞® 50ÎßåÏõê Í∞ÄÏ†ï
population_size = 5_000_000  # 500ÎßåÎ™Ö (20ÎåÄ Ïù∏Íµ¨ Í∞ÄÏ†ï)
population_mean = 280        # Î™®ÌèâÍ∑† (Ïö∞Î¶¨Í∞Ä ÏïåÍ≥† Ïã∂ÏùÄ Í∞í)
population_std = 50          # Î™®ÌëúÏ§ÄÌé∏Ï∞®

# Î™®ÏßëÎã® Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (ÏãúÎÆ¨Î†àÏù¥ÏÖò Î™©Ï†Å)
population = np.random.normal(population_mean, population_std, population_size)

print(f"Î™®ÏßëÎã® ÌÅ¨Í∏∞: {population_size:,}Î™Ö")
print(f"Î™®ÌèâÍ∑† (mu): {population_mean}ÎßåÏõê (ÏïåÍ≥† Ïã∂ÏùÄ Í∞í)")
print(f"Î™®ÌëúÏ§ÄÌé∏Ï∞® (sigma): {population_std}ÎßåÏõê")
````

````python
# ÌëúÎ≥∏ Ï∂îÏ∂ú Î∞è ÌÜµÍ≥ÑÎüâ Í≥ÑÏÇ∞
sample_size = 1000  # ÌëúÎ≥∏ ÌÅ¨Í∏∞
sample = np.random.choice(population, size=sample_size, replace=False)

# ÌÜµÍ≥ÑÎüâ Í≥ÑÏÇ∞
sample_mean = np.mean(sample)    # ÌëúÎ≥∏ÌèâÍ∑†
sample_std = np.std(sample, ddof=1)  # ÌëúÎ≥∏ÌëúÏ§ÄÌé∏Ï∞®

print(f"\nÌëúÎ≥∏ ÌÅ¨Í∏∞: {sample_size}Î™Ö")
print(f"ÌëúÎ≥∏ÌèâÍ∑† (xÃÑ): {sample_mean:.2f}ÎßåÏõê")
print(f"ÌëúÎ≥∏ÌëúÏ§ÄÌé∏Ï∞® (s): {sample_std:.2f}ÎßåÏõê")
print(f"\nÏ∂îÏ†ï Ïò§Ï∞®: {abs(sample_mean - population_mean):.2f}ÎßåÏõê")
print(f"\nÍ≤∞Î°†: ÌëúÎ≥∏ÌèâÍ∑† {sample_mean:.2f}ÎßåÏõêÏúºÎ°ú Î™®ÌèâÍ∑† {population_mean}ÎßåÏõêÏùÑ Ï∂îÏ†ï!")
````

## 1.2 ÎåÄÏàòÏùò Î≤ïÏπô (Law of Large Numbers)

### ÎåÄÏàòÏùò Î≤ïÏπôÏù¥ÎûÄ?

**ÌëúÎ≥∏ ÌÅ¨Í∏∞Í∞Ä Ïª§ÏßàÏàòÎ°ù, ÌëúÎ≥∏ÌèâÍ∑†ÏùÄ Î™®ÌèâÍ∑†Ïóê Í∞ÄÍπåÏõåÏßÑÎã§!**

- ÎßéÏù¥ Í¥ÄÏ∞∞Ìï†ÏàòÎ°ù "ÏßÑÏßú ÌèâÍ∑†"Ïóê ÏàòÎ†¥
- Î≥¥Ìóò, ÎèÑÎ∞ï, ÌíàÏßàÍ¥ÄÎ¶¨ Îì± Ïû•Í∏∞ ÏòàÏ∏°Ïùò Í∑ºÍ±∞
- "Ïö¥"Ïùò ÏòÅÌñ•Ïù¥ Ï§ÑÏñ¥Îì§Í≥† "Ïã§Î†•"Ïù¥ ÎìúÎü¨ÎÇ®

### Ïôú Ï§ëÏöîÌïúÍ∞Ä?

ÎåÄÏàòÏùò Î≤ïÏπôÏù¥ ÏóÜÎã§Î©¥, ÏïÑÎ¨¥Î¶¨ ÎßéÏù¥ Ï°∞ÏÇ¨Ìï¥ÎèÑ Ï†ïÌôïÌïú Í≤∞Î°†ÏùÑ ÎÇ¥Î¶¥ Ïàò ÏóÜÏäµÎãàÎã§!

````python
# ÎåÄÏàòÏùò Î≤ïÏπô ÏãúÎÆ¨Î†àÏù¥ÏÖò: ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞
np.random.seed(2025)

# Í≥µÏ†ïÌïú ÎèôÏ†Ñ: ÏïûÎ©¥(1) ÌôïÎ•† = 0.5
true_prob = 0.5
max_flips = 5000

# ÎàÑÏ†Å ÌèâÍ∑† Í≥ÑÏÇ∞
flips = np.random.binomial(1, true_prob, max_flips)  # 0 ÎòêÎäî 1
cumulative_means = np.cumsum(flips) / np.arange(1, max_flips + 1)

print("ÎåÄÏàòÏùò Î≤ïÏπô: ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞ ÏãúÎÆ¨Î†àÏù¥ÏÖò")
print("=" * 50)
print(f"Ïã§Ï†ú ÏïûÎ©¥ ÌôïÎ•† (Î™®ÌèâÍ∑†): {true_prob}")
print(f"\nÎçòÏßÑ ÌöüÏàòÎ≥Ñ ÏïûÎ©¥ ÎπÑÏú® (ÌëúÎ≥∏ÌèâÍ∑†):")
for n in [10, 100, 500, 1000, 5000]:
    print(f"  {n:>5}Î≤à ÎçòÏ°åÏùÑ Îïå: {cumulative_means[n-1]:.4f}")
````

````python
# ÎåÄÏàòÏùò Î≤ïÏπô ÏãúÍ∞ÅÌôî
df_lln = pd.DataFrame({
    'ÎçòÏßÑ ÌöüÏàò': range(1, max_flips + 1),
    'ÏïûÎ©¥ ÎπÑÏú® (ÎàÑÏ†ÅÌèâÍ∑†)': cumulative_means
})

fig = px.line(df_lln, x='ÎçòÏßÑ ÌöüÏàò', y='ÏïûÎ©¥ ÎπÑÏú® (ÎàÑÏ†ÅÌèâÍ∑†)',
              title='ÎåÄÏàòÏùò Î≤ïÏπô: ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞ (Î™®ÌèâÍ∑† = 0.5)')

# Î™®ÌèâÍ∑† Í∏∞Ï§ÄÏÑ†
fig.add_hline(y=true_prob, line_dash="dash", line_color="red",
              annotation_text=f"Î™®ÌèâÍ∑† = {true_prob}",
              annotation_position="bottom right")

fig.update_layout(height=450, width=800,
                  yaxis_range=[0.3, 0.7])
fig.show()
````

### üí° Ïã§Î¨¥ ÏòàÏãú: Ï£ºÏÇ¨ÏúÑ ÎçòÏßÄÍ∏∞Î°ú ÌôïÏù∏ÌïòÎäî ÎåÄÏàòÏùò Î≤ïÏπô

````python
# Ï£ºÏÇ¨ÏúÑ ÎçòÏßÄÍ∏∞ ÏãúÎÆ¨Î†àÏù¥ÏÖò
np.random.seed(42)

# Í≥µÏ†ïÌïú Ï£ºÏÇ¨ÏúÑ: Í∏∞ÎåìÍ∞í = (1+2+3+4+5+6)/6 = 3.5
true_mean = 3.5
max_rolls = 3000

# Ï£ºÏÇ¨ÏúÑ ÎçòÏßÄÍ∏∞ (1~6)
rolls = np.random.randint(1, 7, max_rolls)
cumulative_means_dice = np.cumsum(rolls) / np.arange(1, max_rolls + 1)

print("ÎåÄÏàòÏùò Î≤ïÏπô: Ï£ºÏÇ¨ÏúÑ ÎçòÏßÄÍ∏∞")
print("=" * 50)
print(f"Ï£ºÏÇ¨ÏúÑ Í∏∞ÎåìÍ∞í (Ïù¥Î°†Í∞í): {true_mean}")
print(f"\nÎçòÏßÑ ÌöüÏàòÎ≥Ñ ÌèâÍ∑†:")
for n in [10, 50, 100, 500, 1000, 3000]:
    mean_val = cumulative_means_dice[n-1]
    error = abs(mean_val - true_mean)
    print(f"  {n:>4}Î≤à: ÌèâÍ∑† = {mean_val:.3f}, Ïò§Ï∞® = {error:.3f}")
````

````python
# Ï£ºÏÇ¨ÏúÑ ÎåÄÏàòÏùò Î≤ïÏπô ÏãúÍ∞ÅÌôî
df_dice = pd.DataFrame({
    'ÎçòÏßÑ ÌöüÏàò': range(1, max_rolls + 1),
    'ÌèâÍ∑†Í∞í': cumulative_means_dice
})

fig = px.line(df_dice, x='ÎçòÏßÑ ÌöüÏàò', y='ÌèâÍ∑†Í∞í',
              title='ÎåÄÏàòÏùò Î≤ïÏπô: Ï£ºÏÇ¨ÏúÑ ÎçòÏßÄÍ∏∞ (Í∏∞ÎåìÍ∞í = 3.5)')

fig.add_hline(y=true_mean, line_dash="dash", line_color="red",
              annotation_text=f"Í∏∞ÎåìÍ∞í = {true_mean}",
              annotation_position="bottom right")

fig.update_layout(height=450, width=800,
                  yaxis_range=[2.5, 4.5])
fig.show()

print("\nÍ≤∞Î°†: ÏãúÌñâ ÌöüÏàòÍ∞Ä ÎßéÏïÑÏßàÏàòÎ°ù ÌëúÎ≥∏ÌèâÍ∑†Ïù¥ Î™®ÌèâÍ∑†(3.5)Ïóê ÏàòÎ†¥!")
````

## 1.3 Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨ (Central Limit Theorem)

### Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨ÎûÄ?

**Î™®ÏßëÎã®Ïùò Î∂ÑÌè¨Í∞Ä Ïñ¥ÎñªÎì†, ÌëúÎ≥∏ ÌÅ¨Í∏∞Í∞Ä Ï∂©Î∂ÑÌûà ÌÅ¨Î©¥(n >= 30) ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨Îäî Ï†ïÍ∑úÎ∂ÑÌè¨Î•º Îî∞Î•∏Îã§!**

### ÌïµÏã¨ Í≥µÏãù

ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨:
- **ÌèâÍ∑†**: E(XÃÑ) = Œº (Î™®ÌèâÍ∑†)
- **ÌëúÏ§ÄÌé∏Ï∞® (ÌëúÏ§ÄÏò§Ï∞®)**: SE = œÉ / ‚àön

### Ïôú Ï§ëÏöîÌïúÍ∞Ä?

CLT ÎçïÎ∂ÑÏóê Î™®ÏßëÎã® Î∂ÑÌè¨Î•º Î™∞ÎùºÎèÑ ÌëúÎ≥∏ÌèâÍ∑†ÏùÑ Ïù¥Ïö©Ìïú Ï∂îÎ°†Ïù¥ Í∞ÄÎä•Ìï©ÎãàÎã§!

````python
# Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨ ÏãúÎÆ¨Î†àÏù¥ÏÖò: Í∑†ÏùºÎ∂ÑÌè¨ Î™®ÏßëÎã®
np.random.seed(2025)

# Î™®ÏßëÎã®: 0~1 ÏÇ¨Ïù¥ Í∑†ÏùºÎ∂ÑÌè¨ (Ï†ïÍ∑úÎ∂ÑÌè¨ ÏïÑÎãò!)
population_uniform = np.random.uniform(0, 1, 100000)

# Î™®ÌèâÍ∑†, Î™®ÌëúÏ§ÄÌé∏Ï∞®
mu = np.mean(population_uniform)
sigma = np.std(population_uniform)

print("Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨ ÏãúÎÆ¨Î†àÏù¥ÏÖò")
print("=" * 50)
print(f"Î™®ÏßëÎã®: Í∑†ÏùºÎ∂ÑÌè¨ (0~1)")
print(f"Î™®ÌèâÍ∑† (Œº): {mu:.4f}")
print(f"Î™®ÌëúÏ§ÄÌé∏Ï∞® (œÉ): {sigma:.4f}")
````

````python
# Î™®ÏßëÎã® Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî (Í∑†ÏùºÎ∂ÑÌè¨ - ÌèâÌèâÌïú Î™®Ïñë)
fig = px.histogram(x=population_uniform, nbins=50,
                   title='Î™®ÏßëÎã® Î∂ÑÌè¨ (Í∑†ÏùºÎ∂ÑÌè¨ - Ï†ïÍ∑úÎ∂ÑÌè¨ ÏïÑÎãò!)',
                   labels={'x': 'Í∞í', 'y': 'ÎπàÎèÑ'})
fig.update_layout(height=350, width=700)
fig.show()
````

````python
# Îã§ÏñëÌïú ÌëúÎ≥∏ ÌÅ¨Í∏∞Î°ú ÌëúÎ≥∏ÌèâÍ∑† Î∂ÑÌè¨ ÏÉùÏÑ±
sample_sizes = [2, 10, 30, 100]
n_samples = 1000  # Í∞Å ÌÅ¨Í∏∞Î≥ÑÎ°ú 1000Í∞úÏùò ÌëúÎ≥∏ÌèâÍ∑† ÏÉùÏÑ±

results = {}

for n in sample_sizes:
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.choice(population_uniform, size=n, replace=True)
        sample_means.append(np.mean(sample))
    results[n] = sample_means

print("ÌëúÎ≥∏ ÌÅ¨Í∏∞Î≥Ñ ÌëúÎ≥∏ÌèâÍ∑† Î∂ÑÌè¨ ÌÜµÍ≥Ñ")
print("=" * 60)
print(f"{'ÌëúÎ≥∏ÌÅ¨Í∏∞':<10} {'ÌèâÍ∑†':>10} {'ÌëúÏ§ÄÌé∏Ï∞®':>12} {'Ïù¥Î°†Ï†Å SE':>12}")
print("-" * 60)
for n in sample_sizes:
    mean_val = np.mean(results[n])
    std_val = np.std(results[n])
    theoretical_se = sigma / np.sqrt(n)
    print(f"n = {n:<6} {mean_val:>10.4f} {std_val:>12.4f} {theoretical_se:>12.4f}")
````

## 2.2 Í≤ÄÏ†ïÎ†• (Statistical Power)

### Í≤ÄÏ†ïÎ†•Ïù¥ÎûÄ?

**Power = 1 - Œ≤ = Ïã§Ï†ú Ìö®Í≥ºÍ∞Ä ÏûàÏùÑ Îïå Ïù¥Î•º Î∞úÍ≤¨Ìï† ÌôïÎ•†**

- Œ≤ = Type II Ïò§Î•ò ÌôïÎ•† (Ìö®Í≥º ÏûàÎäîÎç∞ ÏóÜÎã§Í≥† ÌåêÎã®)
- Power = Ìö®Í≥ºÎ•º Ï†úÎåÄÎ°ú Î∞úÍ≤¨Ìï† ÌôïÎ•†
- ÏùºÎ∞òÏ†ÅÏúºÎ°ú 80% Ïù¥ÏÉÅÏùò Í≤ÄÏ†ïÎ†• Í∂åÏû•

### Í≤ÄÏ†ïÎ†•Ïóê ÏòÅÌñ•ÏùÑ Ï£ºÎäî ÏöîÏÜå

| ÏöîÏÜå | Í≤ÄÏ†ïÎ†• ÏòÅÌñ• |
|------|------------|
| ÌëúÎ≥∏ ÌÅ¨Í∏∞ (n) ‚Üë | Power ‚Üë |
| Ìö®Í≥º ÌÅ¨Í∏∞ ‚Üë | Power ‚Üë |
| Ïú†ÏùòÏàòÏ§Ä Œ± ‚Üë | Power ‚Üë (Îã®, Type I Ïò§Î•ò ‚Üë) |
| Î∂ÑÏÇ∞ (œÉ¬≤) ‚Üì | Power ‚Üë |

````python
# Í≤ÄÏ†ïÎ†• ÏãúÍ∞ÅÌôî: Îëê Î∂ÑÌè¨Ïùò Í¥ÄÍ≥Ñ
import numpy as np
from scipy import stats
import plotly.graph_objects as go

# ÏÑ§Ï†ï
mu_null = 0      # Í∑ÄÎ¨¥Í∞ÄÏÑ§ ÌïòÏùò ÌèâÍ∑†
mu_alt = 0.5     # ÎåÄÎ¶ΩÍ∞ÄÏÑ§ ÌïòÏùò ÌèâÍ∑† (Ìö®Í≥º ÏûàÏùå)
sigma = 1        # ÌëúÏ§ÄÌé∏Ï∞®
n = 30           # ÌëúÎ≥∏ ÌÅ¨Í∏∞
alpha = 0.05     # Ïú†ÏùòÏàòÏ§Ä

# ÌëúÏ§ÄÏò§Ï∞®
se = sigma / np.sqrt(n)

# ÏûÑÍ≥ÑÍ∞í (Îã®Ï∏°Í≤ÄÏ†ï, Ïò§Î•∏Ï™Ω)
critical_value = stats.norm.ppf(1 - alpha, mu_null, se)

# Í≤ÄÏ†ïÎ†• Í≥ÑÏÇ∞
power = 1 - stats.norm.cdf(critical_value, mu_alt, se)
beta = 1 - power

print(f"Í≤ÄÏ†ïÎ†• Î∂ÑÏÑù")
print("=" * 50)
print(f"Í∑ÄÎ¨¥Í∞ÄÏÑ§ ÌèâÍ∑† (H0): {mu_null}")
print(f"ÎåÄÎ¶ΩÍ∞ÄÏÑ§ ÌèâÍ∑† (H1): {mu_alt}")
print(f"ÌëúÎ≥∏ ÌÅ¨Í∏∞: {n}")
print(f"Ïú†ÏùòÏàòÏ§Ä (Œ±): {alpha}")
print(f"ÏûÑÍ≥ÑÍ∞í: {critical_value:.4f}")
print(f"\nType II Ïò§Î•ò (Œ≤): {beta:.4f}")
print(f"Í≤ÄÏ†ïÎ†• (1-Œ≤): {power:.4f} = {power*100:.1f}%")
````

````python
# Í≤ÄÏ†ïÎ†• ÏãúÍ∞ÅÌôî
x = np.linspace(-1, 1.5, 500)

fig = go.Figure()

# H0 Î∂ÑÌè¨
y_null = stats.norm.pdf(x, mu_null, se)
fig.add_trace(go.Scatter(x=x, y=y_null, mode='lines', name='H0 (Ìö®Í≥º ÏóÜÏùå)',
                         line=dict(color='blue', width=2)))

# H1 Î∂ÑÌè¨
y_alt = stats.norm.pdf(x, mu_alt, se)
fig.add_trace(go.Scatter(x=x, y=y_alt, mode='lines', name='H1 (Ìö®Í≥º ÏûàÏùå)',
                         line=dict(color='red', width=2)))

# ÏûÑÍ≥ÑÍ∞í ÏÑ†
fig.add_vline(x=critical_value, line_dash="dash", line_color="green",
              annotation_text=f"ÏûÑÍ≥ÑÍ∞í={critical_value:.3f}")

# Type I Ïò§Î•ò ÏòÅÏó≠ (Œ±)
x_alpha = np.linspace(critical_value, 1.5, 100)
y_alpha = stats.norm.pdf(x_alpha, mu_null, se)
fig.add_trace(go.Scatter(x=x_alpha, y=y_alpha, fill='tozeroy', name='Œ± (Type I)',
                         fillcolor='rgba(0,0,255,0.3)', line=dict(width=0)))

# Type II Ïò§Î•ò ÏòÅÏó≠ (Œ≤)
x_beta = np.linspace(-1, critical_value, 100)
y_beta = stats.norm.pdf(x_beta, mu_alt, se)
fig.add_trace(go.Scatter(x=x_beta, y=y_beta, fill='tozeroy', name='Œ≤ (Type II)',
                         fillcolor='rgba(255,0,0,0.3)', line=dict(width=0)))

fig.update_layout(title=f'Í≤ÄÏ†ïÎ†• ÏãúÍ∞ÅÌôî (Power = {power:.1%})',
                  xaxis_title='ÌëúÎ≥∏ÌèâÍ∑†',
                  yaxis_title='ÌôïÎ•†Î∞ÄÎèÑ',
                  height=500, width=800)
fig.show()
````

## 2.3 Ìö®Í≥º ÌÅ¨Í∏∞ (Effect Size)

### Ìö®Í≥º ÌÅ¨Í∏∞ÎûÄ?

**Ìö®Í≥ºÍ∞Ä "ÏñºÎßàÎÇò" ÌÅ∞ÏßÄÎ•º ÎÇòÌÉÄÎÇ¥Îäî ÌëúÏ§ÄÌôîÎêú ÏßÄÌëú**

- p-valueÎäî Ìö®Í≥ºÏùò "Ïú†Î¨¥"Îßå ÏïåÎ†§Ï§å
- Ìö®Í≥º ÌÅ¨Í∏∞Îäî Ìö®Í≥ºÏùò "ÌÅ¨Í∏∞"Î•º ÏïåÎ†§Ï§å

### Cohen's d

**d = (Œº‚ÇÅ - Œº‚ÇÇ) / œÉ**

| Cohen's d | Ìï¥ÏÑù |
|-----------|------|
| 0.2 | ÏûëÏùÄ Ìö®Í≥º (Small) |
| 0.5 | Ï§ëÍ∞Ñ Ìö®Í≥º (Medium) |
| 0.8 | ÌÅ∞ Ìö®Í≥º (Large) |

````python
# Cohen's d Í≥ÑÏÇ∞ Ìï®Ïàò
def cohens_d(group1, group2):
    """Îëê Í∑∏Î£π Í∞Ñ Cohen's d Í≥ÑÏÇ∞"""
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    
    # pooled standard deviation
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))
    
    return (np.mean(group1) - np.mean(group2)) / pooled_std

# ÏòàÏãú: A/B ÌÖåÏä§Ìä∏ Ï†ÑÌôòÏú®
np.random.seed(42)

# Í∑∏Î£π A (Í∏∞Ï°¥): Ï†ÑÌôòÏú® 5%
group_a = np.random.binomial(1, 0.05, 1000)

# Í∑∏Î£π B (Ïã†Í∑ú): Ï†ÑÌôòÏú® 6%
group_b = np.random.binomial(1, 0.06, 1000)

d = cohens_d(group_b, group_a)

print("A/B ÌÖåÏä§Ìä∏ Ìö®Í≥º ÌÅ¨Í∏∞ Î∂ÑÏÑù")
print("=" * 50)
print(f"Í∑∏Î£π A (Í∏∞Ï°¥) Ï†ÑÌôòÏú®: {np.mean(group_a):.2%}")
print(f"Í∑∏Î£π B (Ïã†Í∑ú) Ï†ÑÌôòÏú®: {np.mean(group_b):.2%}")
print(f"\nCohen's d: {d:.3f}")

if abs(d) < 0.2:
    effect_size = "Îß§Ïö∞ ÏûëÏùå"
elif abs(d) < 0.5:
    effect_size = "ÏûëÏùå (Small)"
elif abs(d) < 0.8:
    effect_size = "Ï§ëÍ∞Ñ (Medium)"
else:
    effect_size = "ÌÅº (Large)"
    
print(f"Ìö®Í≥º ÌÅ¨Í∏∞ Ìï¥ÏÑù: {effect_size}")
````

````python
# Ìö®Í≥º ÌÅ¨Í∏∞Î≥Ñ Î∂ÑÌè¨ ÎπÑÍµê ÏãúÍ∞ÅÌôî
effect_sizes = [0.2, 0.5, 0.8]
labels = ['ÏûëÏùÄ Ìö®Í≥º (d=0.2)', 'Ï§ëÍ∞Ñ Ìö®Í≥º (d=0.5)', 'ÌÅ∞ Ìö®Í≥º (d=0.8)']

fig = make_subplots(rows=1, cols=3, subplot_titles=labels)

x = np.linspace(-4, 4, 200)

for i, d in enumerate(effect_sizes):
    # Í∑∏Î£π 1 (ÌèâÍ∑† 0)
    y1 = stats.norm.pdf(x, 0, 1)
    # Í∑∏Î£π 2 (ÌèâÍ∑† d)
    y2 = stats.norm.pdf(x, d, 1)
    
    fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='Í∑∏Î£π1',
                             line=dict(color='blue')), row=1, col=i+1)
    fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Í∑∏Î£π2',
                             line=dict(color='red')), row=1, col=i+1)

fig.update_layout(height=350, width=900, showlegend=False,
                  title_text='Ìö®Í≥º ÌÅ¨Í∏∞Ïóê Îî∞Î•∏ Îëê Î∂ÑÌè¨Ïùò Ï∞®Ïù¥')
fig.show()

print("Ìö®Í≥º ÌÅ¨Í∏∞Í∞Ä ÌÅ¥ÏàòÎ°ù Îëê Î∂ÑÌè¨Í∞Ä Îçî ÎßéÏù¥ Î∂ÑÎ¶¨Îê®!")
````

## 2.4 ÌëúÎ≥∏ ÌÅ¨Í∏∞ Í≤∞Ï†ï

### ÌëúÎ≥∏ ÌÅ¨Í∏∞Í∞Ä Ï§ëÏöîÌïú Ïù¥Ïú†

- **ÎÑàÎ¨¥ ÏûëÏùÄ ÌëúÎ≥∏**: Í≤ÄÏ†ïÎ†• Î∂ÄÏ°± ‚Üí Type II Ïò§Î•ò Ï¶ùÍ∞Ä
- **ÎÑàÎ¨¥ ÌÅ∞ ÌëúÎ≥∏**: ÎπÑÏö©/ÏãúÍ∞Ñ ÎÇ≠ÎπÑ

### ÌëúÎ≥∏ ÌÅ¨Í∏∞ Í≤∞Ï†ï ÏöîÏÜå

1. **Ïú†ÏùòÏàòÏ§Ä (Œ±)**: Î≥¥ÌÜµ 0.05
2. **Í≤ÄÏ†ïÎ†• (1-Œ≤)**: Î≥¥ÌÜµ 0.80
3. **Ìö®Í≥º ÌÅ¨Í∏∞ (d)**: ÌÉêÏßÄÌïòÍ≥† Ïã∂ÏùÄ ÏµúÏÜå Ìö®Í≥º
4. **Í≤ÄÏ†ï Ïú†Ìòï**: Îã®Ï∏° vs ÏñëÏ∏°

````python
# statsmodelsÎ°ú ÌëúÎ≥∏ ÌÅ¨Í∏∞ Í≥ÑÏÇ∞
from statsmodels.stats.power import TTestIndPower

# Í≤ÄÏ†ïÎ†• Î∂ÑÏÑù Í∞ùÏ≤¥ ÏÉùÏÑ±
power_analysis = TTestIndPower()

# ÏÑ§Ï†ï
effect_size = 0.5   # Ï§ëÍ∞Ñ Ìö®Í≥º
alpha = 0.05        # Ïú†ÏùòÏàòÏ§Ä
power = 0.80        # Î™©Ìëú Í≤ÄÏ†ïÎ†•

# ÌïÑÏöî ÌëúÎ≥∏ ÌÅ¨Í∏∞ Í≥ÑÏÇ∞ (Í∞Å Í∑∏Î£πÎãπ)
sample_size = power_analysis.solve_power(
    effect_size=effect_size,
    alpha=alpha,
    power=power,
    alternative='two-sided'  # ÏñëÏ∏°Í≤ÄÏ†ï
)

print("ÌëúÎ≥∏ ÌÅ¨Í∏∞ Í≥ÑÏÇ∞ (ÎèÖÎ¶Ω t-Í≤ÄÏ†ï)")
print("=" * 50)
print(f"Ìö®Í≥º ÌÅ¨Í∏∞ (Cohen's d): {effect_size}")
print(f"Ïú†ÏùòÏàòÏ§Ä (Œ±): {alpha}")
print(f"Î™©Ìëú Í≤ÄÏ†ïÎ†•: {power}")
print(f"\nÌïÑÏöî ÌëúÎ≥∏ ÌÅ¨Í∏∞: Í∞Å Í∑∏Î£π {int(np.ceil(sample_size))}Î™Ö")
print(f"Ï¥ù ÌïÑÏöî Ïù∏Ïõê: {int(np.ceil(sample_size)) * 2}Î™Ö")
````

````python
# Ìö®Í≥º ÌÅ¨Í∏∞Î≥Ñ ÌïÑÏöî ÌëúÎ≥∏ ÌÅ¨Í∏∞
effect_sizes = [0.2, 0.3, 0.5, 0.8]

print("Ìö®Í≥º ÌÅ¨Í∏∞Î≥Ñ ÌïÑÏöî ÌëúÎ≥∏ ÌÅ¨Í∏∞ (Œ±=0.05, Power=0.80)")
print("=" * 60)
print(f"{'Ìö®Í≥º ÌÅ¨Í∏∞':<12} {'Ìï¥ÏÑù':<12} {'Í∞Å Í∑∏Î£π':<10} {'Ï¥ù Ïù∏Ïõê'}")
print("-" * 60)

for d in effect_sizes:
    n = power_analysis.solve_power(effect_size=d, alpha=0.05, power=0.80)
    n = int(np.ceil(n))
    
    if d <= 0.2:
        interpret = "ÏûëÏùÄ Ìö®Í≥º"
    elif d <= 0.5:
        interpret = "Ï§ëÍ∞Ñ Ìö®Í≥º"
    else:
        interpret = "ÌÅ∞ Ìö®Í≥º"
    
    print(f"d = {d:<8} {interpret:<12} {n:<10} {n*2}")
````

### üí° Ïã§Î¨¥ ÏòàÏãú: A/B ÌÖåÏä§Ìä∏ ÏÑ§Í≥Ñ

````python
# A/B ÌÖåÏä§Ìä∏ ÌëúÎ≥∏ ÌÅ¨Í∏∞ Í≥ÑÏÇ∞
from statsmodels.stats.power import NormalIndPower

# ÌòÑÏû¨ Ï†ÑÌôòÏú®Í≥º Î™©Ìëú Í∞úÏÑ†
baseline_rate = 0.05    # ÌòÑÏû¨ Ï†ÑÌôòÏú® 5%
target_rate = 0.06      # Î™©Ìëú Ï†ÑÌôòÏú® 6%
relative_lift = (target_rate - baseline_rate) / baseline_rate

# Ìö®Í≥º ÌÅ¨Í∏∞ Í≥ÑÏÇ∞ (ÎπÑÏú®Ïùò Cohen's h)
effect_size_h = 2 * (np.arcsin(np.sqrt(target_rate)) - np.arcsin(np.sqrt(baseline_rate)))

# ÌëúÎ≥∏ ÌÅ¨Í∏∞ Í≥ÑÏÇ∞
power_analysis_prop = NormalIndPower()
sample_size_ab = power_analysis_prop.solve_power(
    effect_size=effect_size_h,
    alpha=0.05,
    power=0.80,
    alternative='two-sided'
)

print("A/B ÌÖåÏä§Ìä∏ ÌëúÎ≥∏ ÌÅ¨Í∏∞ ÏÑ§Í≥Ñ")
print("=" * 50)
print(f"ÌòÑÏû¨ Ï†ÑÌôòÏú®: {baseline_rate:.1%}")
print(f"Î™©Ìëú Ï†ÑÌôòÏú®: {target_rate:.1%}")
print(f"ÏÉÅÎåÄÏ†Å Í∞úÏÑ†Ïú®: {relative_lift:.1%}")
print(f"\nÌö®Í≥º ÌÅ¨Í∏∞ (Cohen's h): {effect_size_h:.4f}")
print(f"\nÌïÑÏöî ÌëúÎ≥∏ ÌÅ¨Í∏∞: Í∞Å Í∑∏Î£π {int(np.ceil(sample_size_ab)):,}Î™Ö")
print(f"Ï¥ù ÌïÑÏöî Ìä∏ÎûòÌîΩ: {int(np.ceil(sample_size_ab)) * 2:,}Î™Ö")

# ÌÖåÏä§Ìä∏ Í∏∞Í∞Ñ Ï∂îÏ†ï
daily_traffic = 5000  # ÏùºÏùº Ìä∏ÎûòÌîΩ
test_days = int(np.ceil(sample_size_ab * 2 / daily_traffic))
print(f"\nÏòàÏÉÅ ÌÖåÏä§Ìä∏ Í∏∞Í∞Ñ: {test_days}Ïùº (ÏùºÏùº {daily_traffic:,}Î™Ö Í∏∞Ï§Ä)")
````

````python
# ÌëúÎ≥∏ ÌÅ¨Í∏∞ÏôÄ Í≤ÄÏ†ïÎ†• Í¥ÄÍ≥Ñ ÏãúÍ∞ÅÌôî
n_values = np.arange(10, 501, 10)
powers = []

for n in n_values:
    p = power_analysis.solve_power(effect_size=0.5, alpha=0.05, nobs1=n)
    powers.append(p)

fig = px.line(x=n_values, y=powers,
              title='ÌëúÎ≥∏ ÌÅ¨Í∏∞ÏôÄ Í≤ÄÏ†ïÎ†• Í¥ÄÍ≥Ñ (d=0.5, Œ±=0.05)',
              labels={'x': 'Í∞Å Í∑∏Î£π ÌëúÎ≥∏ ÌÅ¨Í∏∞', 'y': 'Í≤ÄÏ†ïÎ†• (Power)'})

# 80% Í≤ÄÏ†ïÎ†• Í∏∞Ï§ÄÏÑ†
fig.add_hline(y=0.80, line_dash="dash", line_color="red",
              annotation_text="Î™©Ìëú Í≤ÄÏ†ïÎ†• 80%")

# ÌïÑÏöî ÌëúÎ≥∏ ÌÅ¨Í∏∞ ÌëúÏãú
required_n = power_analysis.solve_power(effect_size=0.5, alpha=0.05, power=0.80)
fig.add_vline(x=required_n, line_dash="dash", line_color="green",
              annotation_text=f"n={int(required_n)}")

fig.update_layout(height=450, width=800)
fig.show()
````
