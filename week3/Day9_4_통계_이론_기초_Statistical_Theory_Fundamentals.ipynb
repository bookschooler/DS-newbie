## 1.1 ëª¨ì§‘ë‹¨, í‘œë³¸, ëª¨ìˆ˜, í†µê³„ëŸ‰

| **í‘œë³¸ (Sample)** | ëª¨ì§‘ë‹¨ì—ì„œ ì¶”ì¶œí•œ ì¼ë¶€ | ì„¤ë¬¸ì¡°ì‚¬ì— ì°¸ì—¬í•œ 1,000ëª… |
| **ëª¨ìˆ˜ (Parameter)** | ëª¨ì§‘ë‹¨ì˜ íŠ¹ì„±ê°’ (ì•Œê³  ì‹¶ì€ ê°’) | 20ëŒ€ ì „ì²´ì˜ í‰ê·  ì›”ê¸‰ |
| **í†µê³„ëŸ‰ (Statistic)** | í‘œë³¸ìœ¼ë¡œ ê³„ì‚°í•œ ê°’ | í‘œë³¸ 1,000ëª…ì˜ í‰ê·  ì›”ê¸‰ |

````python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
import numpy as np
import pandas as pd
from scipy import stats
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ë‚œìˆ˜ ì‹œë“œ ê³ ì • (ì¬í˜„ì„±)
np.random.seed(2025)

print("ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸ ì™„ë£Œ!")
````

````python
# ëª¨ì§‘ë‹¨ê³¼ í‘œë³¸ ì˜ˆì‹œ: 20ëŒ€ ì›”ê¸‰ ë¶„ì„
print("ëª¨ì§‘ë‹¨ê³¼ í‘œë³¸ ê°œë… ì˜ˆì‹œ")
print("=" * 50)

# ê°€ìƒì˜ ëª¨ì§‘ë‹¨ ìƒì„± (ì‹¤ì œë¡œëŠ” ì•Œ ìˆ˜ ì—†ëŠ” ê°’!)
# 20ëŒ€ ì „ì²´ ì›”ê¸‰ ë¶„í¬: í‰ê·  280ë§Œì›, í‘œì¤€í¸ì°¨ 50ë§Œì› ê°€ì •
population_size = 5_000_000  # 500ë§Œëª… (20ëŒ€ ì¸êµ¬ ê°€ì •)
population_mean = 280        # ëª¨í‰ê·  (ìš°ë¦¬ê°€ ì•Œê³  ì‹¶ì€ ê°’)
population_std = 50          # ëª¨í‘œì¤€í¸ì°¨

# ëª¨ì§‘ë‹¨ ë°ì´í„° ìƒì„± (ì‹œë®¬ë ˆì´ì…˜ ëª©ì )
population = np.random.normal(population_mean, population_std, population_size)

print(f"ëª¨ì§‘ë‹¨ í¬ê¸°: {population_size:,}ëª…")
print(f"ëª¨í‰ê·  (mu): {population_mean}ë§Œì› (ì•Œê³  ì‹¶ì€ ê°’)")
print(f"ëª¨í‘œì¤€í¸ì°¨ (sigma): {population_std}ë§Œì›")
````

````python
# í‘œë³¸ ì¶”ì¶œ ë° í†µê³„ëŸ‰ ê³„ì‚°
sample_size = 1000  # í‘œë³¸ í¬ê¸°
sample = np.random.choice(population, size=sample_size, replace=False)

# í†µê³„ëŸ‰ ê³„ì‚°
sample_mean = np.mean(sample)    # í‘œë³¸í‰ê· 
sample_std = np.std(sample, ddof=1)  # í‘œë³¸í‘œì¤€í¸ì°¨

print(f"\ní‘œë³¸ í¬ê¸°: {sample_size}ëª…")
print(f"í‘œë³¸í‰ê·  (xÌ„): {sample_mean:.2f}ë§Œì›")
print(f"í‘œë³¸í‘œì¤€í¸ì°¨ (s): {sample_std:.2f}ë§Œì›")
print(f"\nì¶”ì • ì˜¤ì°¨: {abs(sample_mean - population_mean):.2f}ë§Œì›")
print(f"\nê²°ë¡ : í‘œë³¸í‰ê·  {sample_mean:.2f}ë§Œì›ìœ¼ë¡œ ëª¨í‰ê·  {population_mean}ë§Œì›ì„ ì¶”ì •!")
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ê³ ê° ë§Œì¡±ë„ ì¡°ì‚¬

**ìƒí™©**: ì „ì²´ ê³ ê° 10ë§Œëª…ì˜ ë§Œì¡±ë„ë¥¼ ì•Œê³  ì‹¶ì§€ë§Œ, ëª¨ë‘ì—ê²Œ ì„¤ë¬¸í•˜ê¸° ì–´ë µìŠµë‹ˆë‹¤.

````python
# ê³ ê° ë§Œì¡±ë„ ì¡°ì‚¬ ì‹œë‚˜ë¦¬ì˜¤
print("ê³ ê° ë§Œì¡±ë„ ì¡°ì‚¬ ì‹œë‚˜ë¦¬ì˜¤")
print("=" * 50)

# ëª¨ì§‘ë‹¨: ì „ì²´ ê³ ê° (1-5ì  ì²™ë„)
total_customers = 100_000
true_satisfaction = 3.8  # ì‹¤ì œ í‰ê·  ë§Œì¡±ë„ (ëª¨ìˆ˜)

# ì‹¤ì œ ë°ì´í„° (ì‹œë®¬ë ˆì´ì…˜)
all_scores = np.random.normal(true_satisfaction, 0.8, total_customers).clip(1, 5)

# í‘œë³¸ ì¶”ì¶œ: 500ëª…ì—ê²Œ ì„¤ë¬¸
survey_sample = np.random.choice(all_scores, size=500, replace=False)

print(f"ì „ì²´ ê³ ê° ìˆ˜ (ëª¨ì§‘ë‹¨): {total_customers:,}ëª…")
print(f"ì„¤ë¬¸ ì‘ë‹µì ìˆ˜ (í‘œë³¸): {len(survey_sample)}ëª…")
print(f"\nì‹¤ì œ í‰ê·  ë§Œì¡±ë„ (ëª¨ìˆ˜): {true_satisfaction:.2f}ì ")
print(f"ì¡°ì‚¬ëœ í‰ê·  ë§Œì¡±ë„ (í†µê³„ëŸ‰): {np.mean(survey_sample):.2f}ì ")
print(f"\ní‘œë³¸ìœ¼ë¡œ ëª¨ì§‘ë‹¨ì„ ì¶”ì •í•˜ëŠ” ê²ƒì´ í†µê³„ì˜ í•µì‹¬!")
````

## 1.2 ëŒ€ìˆ˜ì˜ ë²•ì¹™ (Law of Large Numbers)

### ëŒ€ìˆ˜ì˜ ë²•ì¹™ì´ë€?

**í‘œë³¸ í¬ê¸°ê°€ ì»¤ì§ˆìˆ˜ë¡, í‘œë³¸í‰ê· ì€ ëª¨í‰ê· ì— ê°€ê¹Œì›Œì§„ë‹¤!**

- ë§ì´ ê´€ì°°í• ìˆ˜ë¡ "ì§„ì§œ í‰ê· "ì— ìˆ˜ë ´
- ë³´í—˜, ë„ë°•, í’ˆì§ˆê´€ë¦¬ ë“± ì¥ê¸° ì˜ˆì¸¡ì˜ ê·¼ê±°
- "ìš´"ì˜ ì˜í–¥ì´ ì¤„ì–´ë“¤ê³  "ì‹¤ë ¥"ì´ ë“œëŸ¬ë‚¨

### ì™œ ì¤‘ìš”í•œê°€?

ëŒ€ìˆ˜ì˜ ë²•ì¹™ì´ ì—†ë‹¤ë©´, ì•„ë¬´ë¦¬ ë§ì´ ì¡°ì‚¬í•´ë„ ì •í™•í•œ ê²°ë¡ ì„ ë‚´ë¦´ ìˆ˜ ì—†ìŠµë‹ˆë‹¤!

````python
# ëŒ€ìˆ˜ì˜ ë²•ì¹™ ì‹œë®¬ë ˆì´ì…˜: ë™ì „ ë˜ì§€ê¸°
np.random.seed(2025)

# ê³µì •í•œ ë™ì „: ì•ë©´(1) í™•ë¥  = 0.5
true_prob = 0.5
max_flips = 5000

# ëˆ„ì  í‰ê·  ê³„ì‚°
flips = np.random.binomial(1, true_prob, max_flips)  # 0 ë˜ëŠ” 1
cumulative_means = np.cumsum(flips) / np.arange(1, max_flips + 1)

print("ëŒ€ìˆ˜ì˜ ë²•ì¹™: ë™ì „ ë˜ì§€ê¸° ì‹œë®¬ë ˆì´ì…˜")
print("=" * 50)
print(f"ì‹¤ì œ ì•ë©´ í™•ë¥  (ëª¨í‰ê· ): {true_prob}")
print(f"\në˜ì§„ íšŸìˆ˜ë³„ ì•ë©´ ë¹„ìœ¨ (í‘œë³¸í‰ê· ):")
for n in [10, 100, 500, 1000, 5000]:
    print(f"  {n:>5}ë²ˆ ë˜ì¡Œì„ ë•Œ: {cumulative_means[n-1]:.4f}")
````

````python
# ëŒ€ìˆ˜ì˜ ë²•ì¹™ ì‹œê°í™”
df_lln = pd.DataFrame({
    'ë˜ì§„ íšŸìˆ˜': range(1, max_flips + 1),
    'ì•ë©´ ë¹„ìœ¨ (ëˆ„ì í‰ê· )': cumulative_means
})

fig = px.line(df_lln, x='ë˜ì§„ íšŸìˆ˜', y='ì•ë©´ ë¹„ìœ¨ (ëˆ„ì í‰ê· )',
              title='ëŒ€ìˆ˜ì˜ ë²•ì¹™: ë™ì „ ë˜ì§€ê¸° (ëª¨í‰ê·  = 0.5)')

# ëª¨í‰ê·  ê¸°ì¤€ì„ 
fig.add_hline(y=true_prob, line_dash="dash", line_color="red",
              annotation_text=f"ëª¨í‰ê·  = {true_prob}",
              annotation_position="bottom right")

fig.update_layout(height=450, width=800,
                  yaxis_range=[0.3, 0.7])
fig.show()
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ì£¼ì‚¬ìœ„ ë˜ì§€ê¸°ë¡œ í™•ì¸í•˜ëŠ” ëŒ€ìˆ˜ì˜ ë²•ì¹™

````python
# ì£¼ì‚¬ìœ„ ë˜ì§€ê¸° ì‹œë®¬ë ˆì´ì…˜
np.random.seed(42)

# ê³µì •í•œ ì£¼ì‚¬ìœ„: ê¸°ëŒ“ê°’ = (1+2+3+4+5+6)/6 = 3.5
true_mean = 3.5
max_rolls = 3000

# ì£¼ì‚¬ìœ„ ë˜ì§€ê¸° (1~6)
rolls = np.random.randint(1, 7, max_rolls)
cumulative_means_dice = np.cumsum(rolls) / np.arange(1, max_rolls + 1)

print("ëŒ€ìˆ˜ì˜ ë²•ì¹™: ì£¼ì‚¬ìœ„ ë˜ì§€ê¸°")
print("=" * 50)
print(f"ì£¼ì‚¬ìœ„ ê¸°ëŒ“ê°’ (ì´ë¡ ê°’): {true_mean}")
print(f"\në˜ì§„ íšŸìˆ˜ë³„ í‰ê· :")
for n in [10, 50, 100, 500, 1000, 3000]:
    mean_val = cumulative_means_dice[n-1]
    error = abs(mean_val - true_mean)
    print(f"  {n:>4}ë²ˆ: í‰ê·  = {mean_val:.3f}, ì˜¤ì°¨ = {error:.3f}")
````

````python
# ì£¼ì‚¬ìœ„ ëŒ€ìˆ˜ì˜ ë²•ì¹™ ì‹œê°í™”
df_dice = pd.DataFrame({
    'ë˜ì§„ íšŸìˆ˜': range(1, max_rolls + 1),
    'í‰ê· ê°’': cumulative_means_dice
})

fig = px.line(df_dice, x='ë˜ì§„ íšŸìˆ˜', y='í‰ê· ê°’',
              title='ëŒ€ìˆ˜ì˜ ë²•ì¹™: ì£¼ì‚¬ìœ„ ë˜ì§€ê¸° (ê¸°ëŒ“ê°’ = 3.5)')

fig.add_hline(y=true_mean, line_dash="dash", line_color="red",
              annotation_text=f"ê¸°ëŒ“ê°’ = {true_mean}",
              annotation_position="bottom right")

fig.update_layout(height=450, width=800,
                  yaxis_range=[2.5, 4.5])
fig.show()

print("\nê²°ë¡ : ì‹œí–‰ íšŸìˆ˜ê°€ ë§ì•„ì§ˆìˆ˜ë¡ í‘œë³¸í‰ê· ì´ ëª¨í‰ê· (3.5)ì— ìˆ˜ë ´!")
````

## 1.3 ì¤‘ì‹¬ê·¹í•œì •ë¦¬ (Central Limit Theorem)

### ì¤‘ì‹¬ê·¹í•œì •ë¦¬ë€?

**ëª¨ì§‘ë‹¨ì˜ ë¶„í¬ê°€ ì–´ë–»ë“ , í‘œë³¸ í¬ê¸°ê°€ ì¶©ë¶„íˆ í¬ë©´(n >= 30) í‘œë³¸í‰ê· ì˜ ë¶„í¬ëŠ” ì •ê·œë¶„í¬ë¥¼ ë”°ë¥¸ë‹¤!**

### í•µì‹¬ ê³µì‹

í‘œë³¸í‰ê· ì˜ ë¶„í¬:
- **í‰ê· **: E(XÌ„) = Î¼ (ëª¨í‰ê· )
- **í‘œì¤€í¸ì°¨ (í‘œì¤€ì˜¤ì°¨)**: SE = Ïƒ / âˆšn

### ì™œ ì¤‘ìš”í•œê°€?

CLT ë•ë¶„ì— ëª¨ì§‘ë‹¨ ë¶„í¬ë¥¼ ëª°ë¼ë„ í‘œë³¸í‰ê· ì„ ì´ìš©í•œ ì¶”ë¡ ì´ ê°€ëŠ¥í•©ë‹ˆë‹¤!

````python
# ì¤‘ì‹¬ê·¹í•œì •ë¦¬ ì‹œë®¬ë ˆì´ì…˜: ê· ì¼ë¶„í¬ ëª¨ì§‘ë‹¨
np.random.seed(2025)

# ëª¨ì§‘ë‹¨: 0~1 ì‚¬ì´ ê· ì¼ë¶„í¬ (ì •ê·œë¶„í¬ ì•„ë‹˜!)
population_uniform = np.random.uniform(0, 1, 100000)

# ëª¨í‰ê· , ëª¨í‘œì¤€í¸ì°¨
mu = np.mean(population_uniform)
sigma = np.std(population_uniform)

print("ì¤‘ì‹¬ê·¹í•œì •ë¦¬ ì‹œë®¬ë ˆì´ì…˜")
print("=" * 50)
print(f"ëª¨ì§‘ë‹¨: ê· ì¼ë¶„í¬ (0~1)")
print(f"ëª¨í‰ê·  (Î¼): {mu:.4f}")
print(f"ëª¨í‘œì¤€í¸ì°¨ (Ïƒ): {sigma:.4f}")
````

````python
# ëª¨ì§‘ë‹¨ ë¶„í¬ ì‹œê°í™” (ê· ì¼ë¶„í¬ - í‰í‰í•œ ëª¨ì–‘)
fig = px.histogram(x=population_uniform, nbins=50,
                   title='ëª¨ì§‘ë‹¨ ë¶„í¬ (ê· ì¼ë¶„í¬ - ì •ê·œë¶„í¬ ì•„ë‹˜!)',
                   labels={'x': 'ê°’', 'y': 'ë¹ˆë„'})
fig.update_layout(height=350, width=700)
fig.show()
````

````python
# ë‹¤ì–‘í•œ í‘œë³¸ í¬ê¸°ë¡œ í‘œë³¸í‰ê·  ë¶„í¬ ìƒì„±
sample_sizes = [2, 10, 30, 100]
n_samples = 1000  # ê° í¬ê¸°ë³„ë¡œ 1000ê°œì˜ í‘œë³¸í‰ê·  ìƒì„±

results = {}

for n in sample_sizes:
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.choice(population_uniform, size=n, replace=True)
        sample_means.append(np.mean(sample))
    results[n] = sample_means

print("í‘œë³¸ í¬ê¸°ë³„ í‘œë³¸í‰ê·  ë¶„í¬ í†µê³„")
print("=" * 60)
print(f"{'í‘œë³¸í¬ê¸°':<10} {'í‰ê· ':>10} {'í‘œì¤€í¸ì°¨':>12} {'ì´ë¡ ì  SE':>12}")
print("-" * 60)
for n in sample_sizes:
    mean_val = np.mean(results[n])
    std_val = np.std(results[n])
    theoretical_se = sigma / np.sqrt(n)
    print(f"n = {n:<6} {mean_val:>10.4f} {std_val:>12.4f} {theoretical_se:>12.4f}")
````

````python
# í‘œë³¸ í¬ê¸°ë³„ í‘œë³¸í‰ê·  ë¶„í¬ ì‹œê°í™”
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=[f'n = {n}' for n in sample_sizes])

positions = [(1, 1), (1, 2), (2, 1), (2, 2)]

for n, pos in zip(sample_sizes, positions):
    fig.add_trace(
        go.Histogram(x=results[n], nbinsx=30, name=f'n={n}',
                     marker_color='steelblue', opacity=0.7),
        row=pos[0], col=pos[1]
    )

fig.update_layout(height=600, width=900,
                  title_text='ì¤‘ì‹¬ê·¹í•œì •ë¦¬: í‘œë³¸ í¬ê¸°ê°€ ì»¤ì§ˆìˆ˜ë¡ ì •ê·œë¶„í¬ì— ê°€ê¹Œì›Œì§',
                  showlegend=False)
fig.show()

print("\nê´€ì°° í¬ì¸íŠ¸:")
print("1. n=2ì¼ ë•Œ: ì‚¼ê°í˜• ëª¨ì–‘ (ì •ê·œë¶„í¬ ì•„ë‹˜)")
print("2. n=10ì¼ ë•Œ: ì¢… ëª¨ì–‘ì— ê°€ê¹Œì›Œì§")
print("3. n=30ì¼ ë•Œ: ê±°ì˜ ì •ê·œë¶„í¬!")
print("4. n=100ì¼ ë•Œ: ì™„ì „í•œ ì •ê·œë¶„í¬")
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ë°°ì†¡ ì‹œê°„ ë¶„ì„

## 2.2 ê²€ì •ë ¥ (Statistical Power)

### ê²€ì •ë ¥ì´ë€?

**Power = 1 - Î² = ì‹¤ì œ íš¨ê³¼ê°€ ìˆì„ ë•Œ ì´ë¥¼ ë°œê²¬í•  í™•ë¥ **

- Î² = Type II ì˜¤ë¥˜ í™•ë¥  (íš¨ê³¼ ìˆëŠ”ë° ì—†ë‹¤ê³  íŒë‹¨)
- Power = íš¨ê³¼ë¥¼ ì œëŒ€ë¡œ ë°œê²¬í•  í™•ë¥ 
- ì¼ë°˜ì ìœ¼ë¡œ 80% ì´ìƒì˜ ê²€ì •ë ¥ ê¶Œì¥

### ê²€ì •ë ¥ì— ì˜í–¥ì„ ì£¼ëŠ” ìš”ì†Œ

| ìš”ì†Œ | ê²€ì •ë ¥ ì˜í–¥ |
|------|------------|
| í‘œë³¸ í¬ê¸° (n) â†‘ | Power â†‘ |
| íš¨ê³¼ í¬ê¸° â†‘ | Power â†‘ |
| ìœ ì˜ìˆ˜ì¤€ Î± â†‘ | Power â†‘ (ë‹¨, Type I ì˜¤ë¥˜ â†‘) |
| ë¶„ì‚° (ÏƒÂ²) â†“ | Power â†‘ |

````python
# ê²€ì •ë ¥ ì‹œê°í™”: ë‘ ë¶„í¬ì˜ ê´€ê³„
import numpy as np
from scipy import stats
import plotly.graph_objects as go

# ì„¤ì •
mu_null = 0      # ê·€ë¬´ê°€ì„¤ í•˜ì˜ í‰ê· 
mu_alt = 0.5     # ëŒ€ë¦½ê°€ì„¤ í•˜ì˜ í‰ê·  (íš¨ê³¼ ìˆìŒ)
sigma = 1        # í‘œì¤€í¸ì°¨
n = 30           # í‘œë³¸ í¬ê¸°
alpha = 0.05     # ìœ ì˜ìˆ˜ì¤€

# í‘œì¤€ì˜¤ì°¨
se = sigma / np.sqrt(n)

# ì„ê³„ê°’ (ë‹¨ì¸¡ê²€ì •, ì˜¤ë¥¸ìª½)
critical_value = stats.norm.ppf(1 - alpha, mu_null, se)

# ê²€ì •ë ¥ ê³„ì‚°
power = 1 - stats.norm.cdf(critical_value, mu_alt, se)
beta = 1 - power

print(f"ê²€ì •ë ¥ ë¶„ì„")
print("=" * 50)
print(f"ê·€ë¬´ê°€ì„¤ í‰ê·  (H0): {mu_null}")
print(f"ëŒ€ë¦½ê°€ì„¤ í‰ê·  (H1): {mu_alt}")
print(f"í‘œë³¸ í¬ê¸°: {n}")
print(f"ìœ ì˜ìˆ˜ì¤€ (Î±): {alpha}")
print(f"ì„ê³„ê°’: {critical_value:.4f}")
print(f"\nType II ì˜¤ë¥˜ (Î²): {beta:.4f}")
print(f"ê²€ì •ë ¥ (1-Î²): {power:.4f} = {power*100:.1f}%")
````

````python
# ê²€ì •ë ¥ ì‹œê°í™”
x = np.linspace(-1, 1.5, 500)

fig = go.Figure()

# H0 ë¶„í¬
y_null = stats.norm.pdf(x, mu_null, se)
fig.add_trace(go.Scatter(x=x, y=y_null, mode='lines', name='H0 (íš¨ê³¼ ì—†ìŒ)',
                         line=dict(color='blue', width=2)))

# H1 ë¶„í¬
y_alt = stats.norm.pdf(x, mu_alt, se)
fig.add_trace(go.Scatter(x=x, y=y_alt, mode='lines', name='H1 (íš¨ê³¼ ìˆìŒ)',
                         line=dict(color='red', width=2)))

# ì„ê³„ê°’ ì„ 
fig.add_vline(x=critical_value, line_dash="dash", line_color="green",
              annotation_text=f"ì„ê³„ê°’={critical_value:.3f}")

# Type I ì˜¤ë¥˜ ì˜ì—­ (Î±)
x_alpha = np.linspace(critical_value, 1.5, 100)
y_alpha = stats.norm.pdf(x_alpha, mu_null, se)
fig.add_trace(go.Scatter(x=x_alpha, y=y_alpha, fill='tozeroy', name='Î± (Type I)',
                         fillcolor='rgba(0,0,255,0.3)', line=dict(width=0)))

# Type II ì˜¤ë¥˜ ì˜ì—­ (Î²)
x_beta = np.linspace(-1, critical_value, 100)
y_beta = stats.norm.pdf(x_beta, mu_alt, se)
fig.add_trace(go.Scatter(x=x_beta, y=y_beta, fill='tozeroy', name='Î² (Type II)',
                         fillcolor='rgba(255,0,0,0.3)', line=dict(width=0)))

fig.update_layout(title=f'ê²€ì •ë ¥ ì‹œê°í™” (Power = {power:.1%})',
                  xaxis_title='í‘œë³¸í‰ê· ',
                  yaxis_title='í™•ë¥ ë°€ë„',
                  height=500, width=800)
fig.show()
````

## 2.3 íš¨ê³¼ í¬ê¸° (Effect Size)

### íš¨ê³¼ í¬ê¸°ë€?

**íš¨ê³¼ê°€ "ì–¼ë§ˆë‚˜" í°ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” í‘œì¤€í™”ëœ ì§€í‘œ**

- p-valueëŠ” íš¨ê³¼ì˜ "ìœ ë¬´"ë§Œ ì•Œë ¤ì¤Œ
- íš¨ê³¼ í¬ê¸°ëŠ” íš¨ê³¼ì˜ "í¬ê¸°"ë¥¼ ì•Œë ¤ì¤Œ

### Cohen's d

**d = (Î¼â‚ - Î¼â‚‚) / Ïƒ**

| Cohen's d | í•´ì„ |
|-----------|------|
| 0.2 | ì‘ì€ íš¨ê³¼ (Small) |
| 0.5 | ì¤‘ê°„ íš¨ê³¼ (Medium) |
| 0.8 | í° íš¨ê³¼ (Large) |

````python
# Cohen's d ê³„ì‚° í•¨ìˆ˜
def cohens_d(group1, group2):
    """ë‘ ê·¸ë£¹ ê°„ Cohen's d ê³„ì‚°"""
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    
    # pooled standard deviation
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))
    
    return (np.mean(group1) - np.mean(group2)) / pooled_std

# ì˜ˆì‹œ: A/B í…ŒìŠ¤íŠ¸ ì „í™˜ìœ¨
np.random.seed(42)

# ê·¸ë£¹ A (ê¸°ì¡´): ì „í™˜ìœ¨ 5%
group_a = np.random.binomial(1, 0.05, 1000)

# ê·¸ë£¹ B (ì‹ ê·œ): ì „í™˜ìœ¨ 6%
group_b = np.random.binomial(1, 0.06, 1000)

d = cohens_d(group_b, group_a)

print("A/B í…ŒìŠ¤íŠ¸ íš¨ê³¼ í¬ê¸° ë¶„ì„")
print("=" * 50)
print(f"ê·¸ë£¹ A (ê¸°ì¡´) ì „í™˜ìœ¨: {np.mean(group_a):.2%}")
print(f"ê·¸ë£¹ B (ì‹ ê·œ) ì „í™˜ìœ¨: {np.mean(group_b):.2%}")
print(f"\nCohen's d: {d:.3f}")

if abs(d) < 0.2:
    effect_size = "ë§¤ìš° ì‘ìŒ"
elif abs(d) < 0.5:
    effect_size = "ì‘ìŒ (Small)"
elif abs(d) < 0.8:
    effect_size = "ì¤‘ê°„ (Medium)"
else:
    effect_size = "í¼ (Large)"
    
print(f"íš¨ê³¼ í¬ê¸° í•´ì„: {effect_size}")
````

````python
# íš¨ê³¼ í¬ê¸°ë³„ ë¶„í¬ ë¹„êµ ì‹œê°í™”
effect_sizes = [0.2, 0.5, 0.8]
labels = ['ì‘ì€ íš¨ê³¼ (d=0.2)', 'ì¤‘ê°„ íš¨ê³¼ (d=0.5)', 'í° íš¨ê³¼ (d=0.8)']

fig = make_subplots(rows=1, cols=3, subplot_titles=labels)

x = np.linspace(-4, 4, 200)

for i, d in enumerate(effect_sizes):
    # ê·¸ë£¹ 1 (í‰ê·  0)
    y1 = stats.norm.pdf(x, 0, 1)
    # ê·¸ë£¹ 2 (í‰ê·  d)
    y2 = stats.norm.pdf(x, d, 1)
    
    fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='ê·¸ë£¹1',
                             line=dict(color='blue')), row=1, col=i+1)
    fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='ê·¸ë£¹2',
                             line=dict(color='red')), row=1, col=i+1)

fig.update_layout(height=350, width=900, showlegend=False,
                  title_text='íš¨ê³¼ í¬ê¸°ì— ë”°ë¥¸ ë‘ ë¶„í¬ì˜ ì°¨ì´')
fig.show()

print("íš¨ê³¼ í¬ê¸°ê°€ í´ìˆ˜ë¡ ë‘ ë¶„í¬ê°€ ë” ë§ì´ ë¶„ë¦¬ë¨!")
````

## 2.4 í‘œë³¸ í¬ê¸° ê²°ì •

### í‘œë³¸ í¬ê¸°ê°€ ì¤‘ìš”í•œ ì´ìœ 

- **ë„ˆë¬´ ì‘ì€ í‘œë³¸**: ê²€ì •ë ¥ ë¶€ì¡± â†’ Type II ì˜¤ë¥˜ ì¦ê°€
- **ë„ˆë¬´ í° í‘œë³¸**: ë¹„ìš©/ì‹œê°„ ë‚­ë¹„

### í‘œë³¸ í¬ê¸° ê²°ì • ìš”ì†Œ

2. **ê²€ì •ë ¥ (1-Î²)**: ë³´í†µ 0.80
3. **íš¨ê³¼ í¬ê¸° (d)**: íƒì§€í•˜ê³  ì‹¶ì€ ìµœì†Œ íš¨ê³¼
4. **ê²€ì • ìœ í˜•**: ë‹¨ì¸¡ vs ì–‘ì¸¡

````python
# statsmodelsë¡œ í‘œë³¸ í¬ê¸° ê³„ì‚°
from statsmodels.stats.power import TTestIndPower

# ê²€ì •ë ¥ ë¶„ì„ ê°ì²´ ìƒì„±
power_analysis = TTestIndPower()

# ì„¤ì •
effect_size = 0.5   # ì¤‘ê°„ íš¨ê³¼
alpha = 0.05        # ìœ ì˜ìˆ˜ì¤€
power = 0.80        # ëª©í‘œ ê²€ì •ë ¥

# í•„ìš” í‘œë³¸ í¬ê¸° ê³„ì‚° (ê° ê·¸ë£¹ë‹¹)
sample_size = power_analysis.solve_power(
    effect_size=effect_size,
    alpha=alpha,
    power=power,
    alternative='two-sided'  # ì–‘ì¸¡ê²€ì •
)

print("í‘œë³¸ í¬ê¸° ê³„ì‚° (ë…ë¦½ t-ê²€ì •)")
print("=" * 50)
print(f"íš¨ê³¼ í¬ê¸° (Cohen's d): {effect_size}")
print(f"ìœ ì˜ìˆ˜ì¤€ (Î±): {alpha}")
print(f"ëª©í‘œ ê²€ì •ë ¥: {power}")
print(f"\ní•„ìš” í‘œë³¸ í¬ê¸°: ê° ê·¸ë£¹ {int(np.ceil(sample_size))}ëª…")
print(f"ì´ í•„ìš” ì¸ì›: {int(np.ceil(sample_size)) * 2}ëª…")
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: A/B í…ŒìŠ¤íŠ¸ ì„¤ê³„

````python
# A/B í…ŒìŠ¤íŠ¸ í‘œë³¸ í¬ê¸° ê³„ì‚°
from statsmodels.stats.power import NormalIndPower

# í˜„ì¬ ì „í™˜ìœ¨ê³¼ ëª©í‘œ ê°œì„ 
baseline_rate = 0.05    # í˜„ì¬ ì „í™˜ìœ¨ 5%
target_rate = 0.06      # ëª©í‘œ ì „í™˜ìœ¨ 6%
relative_lift = (target_rate - baseline_rate) / baseline_rate

# íš¨ê³¼ í¬ê¸° ê³„ì‚° (ë¹„ìœ¨ì˜ Cohen's h)
effect_size_h = 2 * (np.arcsin(np.sqrt(target_rate)) - np.arcsin(np.sqrt(baseline_rate)))

# í‘œë³¸ í¬ê¸° ê³„ì‚°
power_analysis_prop = NormalIndPower()
sample_size_ab = power_analysis_prop.solve_power(
    effect_size=effect_size_h,
    alpha=0.05,
    power=0.80,
    alternative='two-sided'
)

print("A/B í…ŒìŠ¤íŠ¸ í‘œë³¸ í¬ê¸° ì„¤ê³„")
print("=" * 50)
print(f"í˜„ì¬ ì „í™˜ìœ¨: {baseline_rate:.1%}")
print(f"ëª©í‘œ ì „í™˜ìœ¨: {target_rate:.1%}")
print(f"ìƒëŒ€ì  ê°œì„ ìœ¨: {relative_lift:.1%}")
print(f"\níš¨ê³¼ í¬ê¸° (Cohen's h): {effect_size_h:.4f}")
print(f"\ní•„ìš” í‘œë³¸ í¬ê¸°: ê° ê·¸ë£¹ {int(np.ceil(sample_size_ab)):,}ëª…")
print(f"ì´ í•„ìš” íŠ¸ë˜í”½: {int(np.ceil(sample_size_ab)) * 2:,}ëª…")

# í…ŒìŠ¤íŠ¸ ê¸°ê°„ ì¶”ì •
daily_traffic = 5000  # ì¼ì¼ íŠ¸ë˜í”½
test_days = int(np.ceil(sample_size_ab * 2 / daily_traffic))
print(f"\nì˜ˆìƒ í…ŒìŠ¤íŠ¸ ê¸°ê°„: {test_days}ì¼ (ì¼ì¼ {daily_traffic:,}ëª… ê¸°ì¤€)")
````
