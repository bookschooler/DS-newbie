### Q1. ì •ê·œì„± ê²€ì • â­

**ë¬¸ì œ**: ë‹¤ìŒ ë°ì´í„°ê°€ ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ëŠ”ì§€ Shapiro-Wilk ê²€ì •ìœ¼ë¡œ í™•ì¸í•˜ì„¸ìš”.

```python
daily_revenue = [152, 148, 155, 160, 145, 158, 162, 150, 153, 157,
                 149, 156, 151, 154, 159, 147, 161, 150, 155, 152]
```

**ê¸°ëŒ€ ê²°ê³¼**: ê²€ì • í†µê³„ëŸ‰, p-value, ì •ê·œë¶„í¬ ì—¬ë¶€ íŒì •

````python
# Q1 ì •ë‹µ

daily_revenue = [152, 148, 155, 160, 145, 158, 162, 150, 153, 157,
                 149, 156, 151, 154, 159, 147, 161, 150, 155, 152]

# Shapiro-Wilk ê²€ì • ìˆ˜í–‰
stat, p_value = stats.shapiro(daily_revenue)

# ê²°ê³¼ ì¶œë ¥
print("=" * 50)
print("ì •ê·œì„± ê²€ì • ê²°ê³¼ (Shapiro-Wilk)")
print("=" * 50)
print(f"\nê²€ì • í†µê³„ëŸ‰: {stat:.4f}")
print(f"p-value: {p_value:.4f}")

# íŒì • (ìœ ì˜ìˆ˜ì¤€ 0.05 ê¸°ì¤€)
alpha = 0.05
print(f"\nìœ ì˜ìˆ˜ì¤€: {alpha}")

if p_value >= alpha:
    print(f"\nâœ… ê²°ë¡ : p-value ({p_value:.4f}) >= {alpha}")
    print("   â†’ ì •ê·œë¶„í¬ë¥¼ ë”°ë¥¸ë‹¤ê³  í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
else:
    print(f"\nâŒ ê²°ë¡ : p-value ({p_value:.4f}) < {alpha}")
    print("   â†’ ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ì§€ ì•ŠìŠµë‹ˆë‹¤.")
````

````python
# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¡œ ê²€ì¦
assert stat > 0.9, "ê²€ì • í†µê³„ëŸ‰ì´ 0.9 ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤"
assert p_value > 0.05, "ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ëŠ” ë°ì´í„°ì…ë‹ˆë‹¤ (p > 0.05)"
print("âœ… í…ŒìŠ¤íŠ¸ í†µê³¼!")
````

### Q2. ë…ë¦½ í‘œë³¸ t-test â­â­

**ë¬¸ì œ**: ì˜¨ë¼ì¸ ì±„ë„ê³¼ ì˜¤í”„ë¼ì¸ ì±„ë„ì˜ í‰ê·  ì£¼ë¬¸ ê¸ˆì•¡ì— ìœ ì˜í•œ ì°¨ì´ê°€ ìˆëŠ”ì§€ ê²€ì •í•˜ì„¸ìš”.

```python
online_orders = [85000, 92000, 78000, 95000, 88000, 91000, 82000, 89000, 94000, 87000]
offline_orders = [72000, 68000, 75000, 70000, 73000, 69000, 74000, 71000, 76000, 67000]
```

````python
# Q2 ì •ë‹µ

online_orders = [85000, 92000, 78000, 95000, 88000, 91000, 82000, 89000, 94000, 87000]
offline_orders = [72000, 68000, 75000, 70000, 73000, 69000, 74000, 71000, 76000, 67000]

# ê¸°ì´ˆ í†µê³„ëŸ‰ ê³„ì‚°
print("=" * 50)
print("ì±„ë„ë³„ ì£¼ë¬¸ ê¸ˆì•¡ ë¹„êµ")
print("=" * 50)

print(f"\nì˜¨ë¼ì¸ ì±„ë„ (n={len(online_orders)}):")
print(f"  í‰ê· : {np.mean(online_orders):,.0f}ì›")
print(f"  í‘œì¤€í¸ì°¨: {np.std(online_orders, ddof=1):,.0f}ì›")

print(f"\nì˜¤í”„ë¼ì¸ ì±„ë„ (n={len(offline_orders)}):")
print(f"  í‰ê· : {np.mean(offline_orders):,.0f}ì›")
print(f"  í‘œì¤€í¸ì°¨: {np.std(offline_orders, ddof=1):,.0f}ì›")

# ì°¨ì´
diff = np.mean(online_orders) - np.mean(offline_orders)
print(f"\ní‰ê·  ì°¨ì´: {diff:,.0f}ì›")

# ë…ë¦½ í‘œë³¸ t-test
t_stat, p_value = stats.ttest_ind(online_orders, offline_orders)

print(f"\ní†µê³„ ê²€ì • (ë…ë¦½ í‘œë³¸ t-test):")
print(f"  t-statistic: {t_stat:.4f}")
print(f"  p-value: {p_value:.6f}")

# ê²°ë¡ 
print("\n" + "=" * 50)
if p_value < 0.05:
    print("âœ… ê²°ë¡ : ì˜¨ë¼ì¸ê³¼ ì˜¤í”„ë¼ì¸ ì±„ë„ì˜ í‰ê·  ì£¼ë¬¸ ê¸ˆì•¡ì—")
    print("   í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•œ ì°¨ì´ê°€ ìˆìŠµë‹ˆë‹¤!")
    print(f"   ì˜¨ë¼ì¸ì´ {diff:,.0f}ì› ë” ë†’ìŠµë‹ˆë‹¤.")
else:
    print("âŒ ê²°ë¡ : ì±„ë„ ê°„ í‰ê·  ì£¼ë¬¸ ê¸ˆì•¡ì— ìœ ì˜í•œ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤.")
print("=" * 50)
````

````python
# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¡œ ê²€ì¦
assert np.mean(online_orders) > np.mean(offline_orders), "ì˜¨ë¼ì¸ í‰ê· ì´ ë” ë†’ì•„ì•¼ í•©ë‹ˆë‹¤"
assert p_value < 0.05, "ìœ ì˜í•œ ì°¨ì´ê°€ ìˆì–´ì•¼ í•©ë‹ˆë‹¤ (p < 0.05)"
print("âœ… í…ŒìŠ¤íŠ¸ í†µê³¼!")
````

### Q3. ëŒ€ì‘ í‘œë³¸ t-test â­â­

**ë¬¸ì œ**: í”„ë¡œëª¨ì…˜ ì „í›„ ê³ ê° ë§Œì¡±ë„ ì ìˆ˜ì— ìœ ì˜í•œ ë³€í™”ê°€ ìˆëŠ”ì§€ ê²€ì •í•˜ì„¸ìš”.

````python
# Q3 ì •ë‹µ

before_promo = [3.5, 3.8, 3.2, 4.0, 3.6, 3.4, 3.9, 3.3, 3.7, 3.5]
after_promo = [4.2, 4.5, 3.9, 4.6, 4.3, 4.1, 4.7, 4.0, 4.4, 4.2]

# ë³€í™”ëŸ‰ ê³„ì‚°
changes = [after - before for before, after in zip(before_promo, after_promo)]

print("=" * 50)
print("í”„ë¡œëª¨ì…˜ ì „í›„ ë§Œì¡±ë„ ë³€í™” ë¶„ì„")
print("=" * 50)

print(f"\ní”„ë¡œëª¨ì…˜ ì „ í‰ê· : {np.mean(before_promo):.2f}ì ")
print(f"í”„ë¡œëª¨ì…˜ í›„ í‰ê· : {np.mean(after_promo):.2f}ì ")
print(f"í‰ê·  ë³€í™”ëŸ‰: +{np.mean(changes):.2f}ì ")

# ëŒ€ì‘ í‘œë³¸ t-test
t_stat, p_value = stats.ttest_rel(before_promo, after_promo)

print(f"\ní†µê³„ ê²€ì • (ëŒ€ì‘ í‘œë³¸ t-test):")
print(f"  t-statistic: {t_stat:.4f}")
print(f"  p-value: {p_value:.6f}")

# ê²°ë¡ 
print("\n" + "=" * 50)
if p_value < 0.05:
    print("âœ… ê²°ë¡ : í”„ë¡œëª¨ì…˜ ì „í›„ ê³ ê° ë§Œì¡±ë„ì—")
    print("   í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•œ ë³€í™”ê°€ ìˆìŠµë‹ˆë‹¤!")
    print(f"   ë§Œì¡±ë„ê°€ í‰ê·  {np.mean(changes):.2f}ì  ìƒìŠ¹í–ˆìŠµë‹ˆë‹¤.")
else:
    print("âŒ ê²°ë¡ : í”„ë¡œëª¨ì…˜ íš¨ê³¼ê°€ í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
print("=" * 50)
````

````python
# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¡œ ê²€ì¦
assert np.mean(changes) > 0, "ë§Œì¡±ë„ê°€ ìƒìŠ¹í•´ì•¼ í•©ë‹ˆë‹¤"
assert p_value < 0.05, "ìœ ì˜í•œ ë³€í™”ê°€ ìˆì–´ì•¼ í•©ë‹ˆë‹¤"
print("âœ… í…ŒìŠ¤íŠ¸ í†µê³¼!")
````

### Q4. ANOVA ë¶„ì„ â­â­â­

**ë¬¸ì œ**: 3ê°œ ê´‘ê³  í”Œë«í¼(ìœ íŠœë¸Œ, ì¸ìŠ¤íƒ€ê·¸ë¨, í˜ì´ìŠ¤ë¶)ì˜ í´ë¦­ë¥ ì— ì°¨ì´ê°€ ìˆëŠ”ì§€ ANOVAë¡œ ê²€ì •í•˜ì„¸ìš”.

````python
# Q4 ì •ë‹µ

youtube_ctr = [3.2, 3.5, 3.8, 3.1, 3.6, 3.4]
instagram_ctr = [4.5, 4.8, 4.2, 4.6, 4.9, 4.4]
facebook_ctr = [2.8, 2.5, 3.0, 2.7, 2.9, 2.6]

print("=" * 60)
print("ê´‘ê³  í”Œë«í¼ë³„ í´ë¦­ë¥ (CTR) ë¹„êµ")
print("=" * 60)

# í”Œë«í¼ë³„ í†µê³„
platforms = {
    'ìœ íŠœë¸Œ': youtube_ctr,
    'ì¸ìŠ¤íƒ€ê·¸ë¨': instagram_ctr,
    'í˜ì´ìŠ¤ë¶': facebook_ctr
}

for name, data in platforms.items():
    print(f"\n{name}:")
    print(f"  í‰ê·  CTR: {np.mean(data):.2f}%")
    print(f"  í‘œì¤€í¸ì°¨: {np.std(data, ddof=1):.2f}%")

# One-Way ANOVA
f_stat, p_value = stats.f_oneway(youtube_ctr, instagram_ctr, facebook_ctr)

print(f"\ní†µê³„ ê²€ì • (One-Way ANOVA):")
print(f"  F-statistic: {f_stat:.4f}")
print(f"  p-value: {p_value:.8f}")

# ê²°ë¡ 
print("\n" + "=" * 60)
if p_value < 0.05:
    print("âœ… ê²°ë¡ : í”Œë«í¼ ê°„ í´ë¦­ë¥ ì— í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•œ ì°¨ì´ê°€ ìˆìŠµë‹ˆë‹¤!")
    # ìµœê³ /ìµœì € ì„±ê³¼ í”Œë«í¼
    means = {name: np.mean(data) for name, data in platforms.items()}
    best = max(means, key=means.get)
    worst = min(means, key=means.get)
    print(f"   ìµœê³  ì„±ê³¼: {best} ({means[best]:.2f}%)")
    print(f"   ìµœì € ì„±ê³¼: {worst} ({means[worst]:.2f}%)")
else:
    print("âŒ ê²°ë¡ : í”Œë«í¼ ê°„ í´ë¦­ë¥ ì— ìœ ì˜í•œ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤.")
print("=" * 60)
````

````python
# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¡œ ê²€ì¦
assert p_value < 0.05, "í”Œë«í¼ ê°„ ìœ ì˜í•œ ì°¨ì´ê°€ ìˆì–´ì•¼ í•©ë‹ˆë‹¤"
assert np.mean(instagram_ctr) > np.mean(youtube_ctr) > np.mean(facebook_ctr), "ì¸ìŠ¤íƒ€ > ìœ íŠœë¸Œ > í˜ì´ìŠ¤ë¶ ìˆœì´ì–´ì•¼ í•©ë‹ˆë‹¤"
print("âœ… í…ŒìŠ¤íŠ¸ í†µê³¼!")
````

### Q5. Chi-Square ê²€ì • â­â­â­

**ë¬¸ì œ**: ë””ë°”ì´ìŠ¤ ìœ í˜•(ëª¨ë°”ì¼, PC, íƒœë¸”ë¦¿)ê³¼ êµ¬ë§¤ ì™„ë£Œ ì—¬ë¶€ê°€ ë…ë¦½ì ì¸ì§€ ê²€ì •í•˜ì„¸ìš”.

````python
# Q5 ì •ë‹µ

device_purchase = np.array([
    [180, 320],  # ëª¨ë°”ì¼: êµ¬ë§¤O, êµ¬ë§¤X
    [120, 280],  # PC
    [60, 140]    # íƒœë¸”ë¦¿
])

print("=" * 60)
print("ë””ë°”ì´ìŠ¤ ìœ í˜• vs êµ¬ë§¤ ì™„ë£Œ ì—¬ë¶€ ë¶„ì„")
print("=" * 60)

# êµì°¨í‘œ ìƒì„±
devices = ['ëª¨ë°”ì¼', 'PC', 'íƒœë¸”ë¦¿']
contingency_df = pd.DataFrame(device_purchase, 
                               index=devices,
                               columns=['êµ¬ë§¤', 'ë¯¸êµ¬ë§¤'])
print("\nêµì°¨í‘œ:")
print(contingency_df)

# ë””ë°”ì´ìŠ¤ë³„ êµ¬ë§¤ìœ¨ ê³„ì‚°
print("\në””ë°”ì´ìŠ¤ë³„ êµ¬ë§¤ìœ¨:")
for i, device in enumerate(devices):
    total = device_purchase[i].sum()
    purchased = device_purchase[i][0]
    rate = purchased / total * 100
    print(f"  {device}: {purchased}/{total} = {rate:.1f}%")

# Chi-square ê²€ì •
chi2, p_value, dof, expected = stats.chi2_contingency(device_purchase)

print(f"\nChi-square ê²€ì •:")
print(f"  Chi-square í†µê³„ëŸ‰: {chi2:.4f}")
print(f"  p-value: {p_value:.4f}")
print(f"  ììœ ë„: {dof}")

# ê²°ë¡ 
print("\n" + "=" * 60)
if p_value < 0.05:
    print("âœ… ê²°ë¡ : ë””ë°”ì´ìŠ¤ ìœ í˜•ê³¼ êµ¬ë§¤ ì—¬ë¶€ëŠ” ìœ ì˜í•œ ê´€ë ¨ì´ ìˆìŠµë‹ˆë‹¤!")
else:
    print("âŒ ê²°ë¡ : ë””ë°”ì´ìŠ¤ ìœ í˜•ê³¼ êµ¬ë§¤ ì—¬ë¶€ëŠ” ë…ë¦½ì ì…ë‹ˆë‹¤.")
    print("   (ë””ë°”ì´ìŠ¤ì— ê´€ê³„ì—†ì´ êµ¬ë§¤ìœ¨ì´ ë¹„ìŠ·í•¨)")
print("=" * 60)
````

````python
# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë¡œ ê²€ì¦
assert chi2 > 0, "Chi-square í†µê³„ëŸ‰ì€ ì–‘ìˆ˜ì—¬ì•¼ í•©ë‹ˆë‹¤"
assert dof == 2, "ììœ ë„ëŠ” (3-1)*(2-1) = 2ì—¬ì•¼ í•©ë‹ˆë‹¤"
print("âœ… í…ŒìŠ¤íŠ¸ í†µê³¼!")
````

### Q7. ë‹¨ìˆœ íšŒê·€ë¶„ì„ â­â­â­â­

### Q8. ë‹¤ì¤‘ íšŒê·€ë¶„ì„ â­â­â­â­

**ë¬¸ì œ**: ì›¹ì‚¬ì´íŠ¸ ì§€í‘œë¡œ ì „í™˜ìœ¨ì„ ì˜ˆì¸¡í•˜ëŠ” ë‹¤ì¤‘ íšŒê·€ ëª¨ë¸ì„ êµ¬ì¶•í•˜ì„¸ìš”.

````python
# Q8 ì •ë‹µ

web_data = pd.DataFrame({
    'page_views': [1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500],
    'avg_time': [2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0],
    'bounce_rate': [65, 60, 55, 50, 45, 40, 35, 30, 25, 20],
    'conversion_rate': [1.2, 1.8, 2.3, 2.9, 3.4, 4.0, 4.5, 5.1, 5.6, 6.2]
})

# ë‹¤ì¤‘ íšŒê·€ë¶„ì„
model = smf.ols('conversion_rate ~ page_views + avg_time + bounce_rate', data=web_data).fit()

print("=" * 70)
print("ë‹¤ì¤‘ íšŒê·€ë¶„ì„ ê²°ê³¼ ìš”ì•½")
print("=" * 70)
print(model.summary())
````

### Q9. ì¢…í•© ë¶„ì„ - ì œí’ˆ íŒë§¤ ë°ì´í„° â­â­â­â­â­

**ë¬¸ì œ**: ì œí’ˆ íŒë§¤ ë°ì´í„°ë¥¼ ì¢…í•©ì ìœ¼ë¡œ ë¶„ì„í•˜ì„¸ìš”.

````python
# Q9 ì •ë‹µ

np.random.seed(42)
products = pd.DataFrame({
    'product_id': range(1, 51),
    'category': np.random.choice(['ì „ìì œí’ˆ', 'ì˜ë¥˜', 'ì‹í’ˆ'], 50),
    'price': np.random.randint(10000, 100000, 50),
    'ad_cost': np.random.randint(5000, 30000, 50),
    'rating': np.random.uniform(3.0, 5.0, 50).round(1),
    'sales_qty': np.random.randint(50, 500, 50)
})

print("=" * 70)
print("ğŸ“Š ì œí’ˆ íŒë§¤ ë°ì´í„° ì¢…í•© ë¶„ì„")
print("=" * 70)

# 1. ì¹´í…Œê³ ë¦¬ë³„ í‰ê·  íŒë§¤ëŸ‰ ë¹„êµ (ANOVA)
print("\n[1] ì¹´í…Œê³ ë¦¬ë³„ í‰ê·  íŒë§¤ëŸ‰ ë¹„êµ (ANOVA)")
print("-" * 50)

categories = products.groupby('category')['sales_qty']
for name, group in categories:
    print(f"{name}: í‰ê·  {group.mean():.1f}ê°œ, í‘œì¤€í¸ì°¨ {group.std():.1f}")

# ANOVA
cat_groups = [group.values for name, group in categories]
f_stat, p_anova = stats.f_oneway(*cat_groups)
print(f"\nF-statistic: {f_stat:.4f}, p-value: {p_anova:.4f}")
if p_anova < 0.05:
    print("âœ… ì¹´í…Œê³ ë¦¬ë³„ íŒë§¤ëŸ‰ì— ìœ ì˜í•œ ì°¨ì´ê°€ ìˆìŠµë‹ˆë‹¤.")
else:
    print("âŒ ì¹´í…Œê³ ë¦¬ë³„ íŒë§¤ëŸ‰ì— ìœ ì˜í•œ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤.")
````

````python
# 2. ê°€ê²©ê³¼ íŒë§¤ëŸ‰ì˜ ìƒê´€ê´€ê³„
print("\n[2] ê°€ê²©ê³¼ íŒë§¤ëŸ‰ì˜ ìƒê´€ê´€ê³„")
print("-" * 50)

r_price, p_price = stats.pearsonr(products['price'], products['sales_qty'])
print(f"Pearson ìƒê´€ê³„ìˆ˜: {r_price:.4f}")
print(f"p-value: {p_price:.4f}")

if abs(r_price) > 0.5:
    strength = "ê°•í•œ"
elif abs(r_price) > 0.3:
    strength = "ì¤‘ê°„"
else:
    strength = "ì•½í•œ"
direction = "ì–‘ì˜" if r_price > 0 else "ìŒì˜"
print(f"í•´ì„: {strength} {direction} ìƒê´€ê´€ê³„")
````

````python
# Q10 ì •ë‹µ

np.random.seed(42)
customer_segments = pd.DataFrame({
    'customer_id': range(1, 101),
    'segment': np.random.choice(['ì‹ ê·œ', 'ì¼ë°˜', 'ì¶©ì„±', 'VIP'], 100, p=[0.3, 0.35, 0.25, 0.1]),
    'age': np.random.randint(20, 60, 100),
    'tenure_months': np.random.randint(1, 60, 100),
    'monthly_spend': np.random.randint(30000, 300000, 100),
    'visit_frequency': np.random.randint(1, 20, 100),
    'satisfaction': np.random.uniform(3.0, 5.0, 100).round(1)
})

print("=" * 80)
print("ğŸ“Š ê³ ê° ì„¸ê·¸ë¨¼íŠ¸ ì¢…í•© ë¶„ì„ ë¦¬í¬íŠ¸")
print("=" * 80)

# 1. ì„¸ê·¸ë¨¼íŠ¸ë³„ ê¸°ì´ˆ í†µê³„
print("\n[1] ì„¸ê·¸ë¨¼íŠ¸ë³„ ê¸°ì´ˆ í†µê³„")
print("-" * 60)

segment_stats = customer_segments.groupby('segment').agg({
    'customer_id': 'count',
    'monthly_spend': ['mean', 'std'],
    'visit_frequency': 'mean',
    'satisfaction': 'mean'
}).round(2)

segment_stats.columns = ['ê³ ê°ìˆ˜', 'í‰ê· ì§€ì¶œ', 'ì§€ì¶œí‘œì¤€í¸ì°¨', 'ë°©ë¬¸ë¹ˆë„', 'ë§Œì¡±ë„']
print(segment_stats)
````

````python
# 2. ì„¸ê·¸ë¨¼íŠ¸ë³„ ì›” ì§€ì¶œì•¡ ì°¨ì´ ê²€ì • (ANOVA)
print("\n[2] ì„¸ê·¸ë¨¼íŠ¸ë³„ ì›” ì§€ì¶œì•¡ ì°¨ì´ (ANOVA)")
print("-" * 60)

segments = customer_segments.groupby('segment')['monthly_spend']
segment_groups = [group.values for name, group in segments]
f_stat, p_anova = stats.f_oneway(*segment_groups)

print(f"F-statistic: {f_stat:.4f}")
print(f"p-value: {p_anova:.4f}")

if p_anova < 0.05:
    print("\nâœ… ì„¸ê·¸ë¨¼íŠ¸ë³„ ì›” ì§€ì¶œì•¡ì— ìœ ì˜í•œ ì°¨ì´ê°€ ìˆìŠµë‹ˆë‹¤!")
    means = segments.mean().sort_values(ascending=False)
    print(f"   ìµœê³ : {means.index[0]} ({means.iloc[0]:,.0f}ì›)")
    print(f"   ìµœì €: {means.index[-1]} ({means.iloc[-1]:,.0f}ì›)")
else:
    print("\nâŒ ì„¸ê·¸ë¨¼íŠ¸ë³„ ì›” ì§€ì¶œì•¡ì— ìœ ì˜í•œ ì°¨ì´ê°€ ì—†ìŠµë‹ˆë‹¤.")
````

````python
# 3. ë°©ë¬¸ ë¹ˆë„ì™€ ë§Œì¡±ë„ì˜ ìƒê´€ê´€ê³„
print("\n[3] ë°©ë¬¸ ë¹ˆë„ vs ë§Œì¡±ë„ ìƒê´€ê´€ê³„")
print("-" * 60)

r, p_corr = stats.pearsonr(customer_segments['visit_frequency'], 
                           customer_segments['satisfaction'])

print(f"Pearson ìƒê´€ê³„ìˆ˜: {r:.4f}")
print(f"p-value: {p_corr:.4f}")

if abs(r) > 0.5:
    strength = "ê°•í•œ"
elif abs(r) > 0.3:
    strength = "ì¤‘ê°„"
else:
    strength = "ì•½í•œ"
direction = "ì–‘ì˜" if r > 0 else "ìŒì˜"

print(f"í•´ì„: {strength} {direction} ìƒê´€ê´€ê³„")
if p_corr < 0.05:
    print("âœ… í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•¨")
else:
    print("âŒ í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•˜ì§€ ì•ŠìŒ")
````

````python
# 4. ì›” ì§€ì¶œì•¡ ì˜ˆì¸¡ íšŒê·€ ëª¨ë¸
print("\n[4] ì›” ì§€ì¶œì•¡ ì˜ˆì¸¡ íšŒê·€ ëª¨ë¸")
print("-" * 60)

# ì„¸ê·¸ë¨¼íŠ¸ ë”ë¯¸ ë³€ìˆ˜ ìƒì„±
df_reg = customer_segments.copy()
df_reg['seg_ì¼ë°˜'] = (df_reg['segment'] == 'ì¼ë°˜').astype(int)
df_reg['seg_ì¶©ì„±'] = (df_reg['segment'] == 'ì¶©ì„±').astype(int)
df_reg['seg_VIP'] = (df_reg['segment'] == 'VIP').astype(int)

model = smf.ols('monthly_spend ~ age + tenure_months + visit_frequency + satisfaction + seg_ì¼ë°˜ + seg_ì¶©ì„± + seg_VIP', 
                data=df_reg).fit()

print(f"R-squared: {model.rsquared:.4f}")
print(f"Adjusted R-squared: {model.rsquared_adj:.4f}")

print(f"\nìœ ì˜í•œ ë³€ìˆ˜ (p < 0.05):")
sig_vars = model.pvalues[model.pvalues < 0.05]
if len(sig_vars) > 1:  # Intercept ì œì™¸
    for var in sig_vars.index:
        if var != 'Intercept':
            print(f"  - {var}: ê³„ìˆ˜={model.params[var]:.2f}, p={model.pvalues[var]:.4f}")
else:
    print("  ì—†ìŒ (ëœë¤ ë°ì´í„° íŠ¹ì„±)")
````

````python
# 5. ì”ì°¨ ë¶„ì„
print("\n[5] ì”ì°¨ ë¶„ì„")
print("-" * 60)

from scipy.stats import probplot

residuals = model.resid

# ì”ì°¨ ì •ê·œì„± ê²€ì •
stat_resid, p_resid = stats.shapiro(residuals)
print(f"ì”ì°¨ ì •ê·œì„± (Shapiro-Wilk): p-value = {p_resid:.4f}")
if p_resid >= 0.05:
    print("âœ… ì”ì°¨ê°€ ì •ê·œë¶„í¬ë¥¼ ë”°ë¦„")
else:
    print("âš ï¸ ì”ì°¨ê°€ ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ì§€ ì•ŠìŒ")

# ì”ì°¨ ì‹œê°í™”
fig = make_subplots(rows=1, cols=2, subplot_titles=('ì”ì°¨ vs ì˜ˆì¸¡ê°’', 'ì”ì°¨ Q-Q Plot'))

fig.add_trace(
    go.Scatter(x=model.fittedvalues, y=residuals, mode='markers',
               marker=dict(color='royalblue', opacity=0.6)),
    row=1, col=1
)
fig.add_hline(y=0, line_dash="dash", line_color="red", row=1, col=1)

qq_data = probplot(residuals)
fig.add_trace(
    go.Scatter(x=qq_data[0][0], y=qq_data[0][1], mode='markers',
               marker=dict(color='royalblue', opacity=0.6)),
    row=1, col=2
)
x_line = np.linspace(min(qq_data[0][0]), max(qq_data[0][0]), 100)
y_line = qq_data[1][0] * x_line + qq_data[1][1]
fig.add_trace(
    go.Scatter(x=x_line, y=y_line, mode='lines', line=dict(color='red')),
    row=1, col=2
)

fig.update_layout(height=350, width=800, showlegend=False, title_text="ì”ì°¨ ë¶„ì„")
fig.show()
````
