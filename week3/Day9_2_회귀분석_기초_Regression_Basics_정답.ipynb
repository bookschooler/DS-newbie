### Q1. ë‹¨ìˆœ ì„ í˜• íšŒê·€ ëª¨ë¸ êµ¬ì¶• â­

**ë¬¸ì œ**: ê³µë¶€ì‹œê°„ê³¼ ì‹œí—˜ì ìˆ˜ ë°ì´í„°ë¡œ ë‹¨ìˆœ ì„ í˜• íšŒê·€ ëª¨ë¸ì„ ë§Œë“¤ê³ , íšŒê·€ê³„ìˆ˜ë¥¼ ì¶œë ¥í•˜ì„¸ìš”.

```python

test_scores = [55, 60, 65, 70, 75, 80, 85, 88, 92]
```

**ì¶œë ¥**: ì ˆí¸, ê¸°ìš¸ê¸°, íšŒê·€ì‹

````python
# Q1. ì •ë‹µ ì½”ë“œ

study_hours = [2, 3, 4, 5, 6, 7, 8, 9, 10]
test_scores = [55, 60, 65, 70, 75, 80, 85, 88, 92]

# ë°ì´í„° ë³€í™˜ (sklearnì€ 2D ë°°ì—´ í•„ìš”)
X = np.array(study_hours).reshape(-1, 1)
y = np.array(test_scores)

# ëª¨ë¸ ìƒì„± ë° í•™ìŠµ
model_q1 = LinearRegression()
model_q1.fit(X, y)

# íšŒê·€ê³„ìˆ˜ ì¶œë ¥
print("ë‹¨ìˆœ ì„ í˜• íšŒê·€ ê²°ê³¼")
print("=" * 50)
print(f"ì ˆí¸ (Intercept): {model_q1.intercept_:.4f}")
print(f"ê¸°ìš¸ê¸° (Slope): {model_q1.coef_[0]:.4f}")
print(f"\níšŒê·€ì‹: ì‹œí—˜ì ìˆ˜ = {model_q1.intercept_:.2f} + {model_q1.coef_[0]:.2f} x ê³µë¶€ì‹œê°„")
````

````python
# Q1. í…ŒìŠ¤íŠ¸/ê²€ì¦
assert hasattr(model_q1, 'coef_'), "ëª¨ë¸ì´ í•™ìŠµë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤."
assert model_q1.coef_[0] > 0, "ê¸°ìš¸ê¸°ëŠ” ì–‘ìˆ˜ì—¬ì•¼ í•©ë‹ˆë‹¤ (ê³µë¶€í• ìˆ˜ë¡ ì ìˆ˜ ìƒìŠ¹)."
assert model_q1.intercept_ > 0, "ì ˆí¸ì€ ì–‘ìˆ˜ì—¬ì•¼ í•©ë‹ˆë‹¤."
print("í…ŒìŠ¤íŠ¸ í†µê³¼!")
````

### Q3. R-squared ê³„ì‚° â­â­

**ë¬¸ì œ**: Q1 ëª¨ë¸ì˜ R-squaredë¥¼ ê³„ì‚°í•˜ê³  í•´ì„í•˜ì„¸ìš”.

**ì¶œë ¥**: R^2 ê°’, í•´ì„ ë¬¸ì¥

````python
# Q3. ì •ë‹µ ì½”ë“œ

# ë°©ë²• 1: model.score() ì‚¬ìš©
r2_score_method = model_q1.score(X, y)

# ë°©ë²• 2: r2_score í•¨ìˆ˜ ì‚¬ìš©
y_pred = model_q1.predict(X)
r2_func = r2_score(y, y_pred)

# ë°©ë²• 3: ìˆ˜ë™ ê³„ì‚°
ss_total = np.sum((y - y.mean()) ** 2)
ss_residual = np.sum((y - y_pred) ** 2)
r2_manual = 1 - (ss_residual / ss_total)

print("R-squared (ê²°ì •ê³„ìˆ˜) ê³„ì‚°")
print("=" * 50)
print(f"model.score(): {r2_score_method:.4f}")
print(f"r2_score(): {r2_func:.4f}")
print(f"ìˆ˜ë™ ê³„ì‚°: {r2_manual:.4f}")

print(f"\ní•´ì„: ê³µë¶€ì‹œê°„ì´ ì‹œí—˜ì ìˆ˜ ë³€ë™ì˜ {r2_score_method*100:.1f}%ë¥¼ ì„¤ëª…í•©ë‹ˆë‹¤.")
print(f"       ë‚˜ë¨¸ì§€ {(1-r2_score_method)*100:.1f}%ëŠ” ë‹¤ë¥¸ ìš”ì¸ì— ì˜í•´ ê²°ì •ë©ë‹ˆë‹¤.")
````

````python
# Q3. í…ŒìŠ¤íŠ¸/ê²€ì¦
assert 0 <= r2_score_method <= 1, "R-squaredëŠ” 0ê³¼ 1 ì‚¬ì´ì—¬ì•¼ í•©ë‹ˆë‹¤."
assert r2_score_method > 0.9, "ì´ ë°ì´í„°ì—ì„œ R-squaredëŠ” 0.9 ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤."
assert abs(r2_score_method - r2_func) < 0.0001, "ê³„ì‚° ë°©ë²•ì´ ì¼ì¹˜í•´ì•¼ í•©ë‹ˆë‹¤."
print("í…ŒìŠ¤íŠ¸ í†µê³¼!")
````

### ğŸ’¡ Q3 í’€ì´ ì„¤ëª…

**ì ‘ê·¼ ë°©ë²•**:
1. `model.score()` ë˜ëŠ” `r2_score()` í•¨ìˆ˜ ì‚¬ìš©
2. R^2 = 1 - (SSE/SST) ê³µì‹ìœ¼ë¡œ ìˆ˜ë™ ê³„ì‚°

**í•µì‹¬ ê°œë…**:

**í”í•œ ì‹¤ìˆ˜**:

### Q7. statsmodels ìš”ì•½ ë¶„ì„ â­â­â­â­

**ë¬¸ì œ**: Q6 ë°ì´í„°ë¡œ statsmodelsë¥¼ ì‚¬ìš©í•´ íšŒê·€ë¶„ì„ì„ ìˆ˜í–‰í•˜ê³ , p-valueë¡œ ìœ ì˜í•œ ë³€ìˆ˜ë¥¼ íŒë‹¨í•˜ì„¸ìš”.

**ì¶œë ¥**: summary() ê²°ê³¼, ìœ ì˜í•œ ë³€ìˆ˜ ëª©ë¡

````python
# Q7. ì •ë‹µ ì½”ë“œ

# statsmodelsë¡œ íšŒê·€ë¶„ì„ (formula ë°©ì‹)
model_sm = smf.ols('ì›”ë§¤ì¶œ ~ ë©´ì  + ì§ì›ìˆ˜', data=store_data).fit()

# ìš”ì•½ ì¶œë ¥
print(model_sm.summary())
````

````python
# Q7. ìœ ì˜í•œ ë³€ìˆ˜ ë¶„ì„

print("\n" + "=" * 60)
print("ìœ ì˜ì„± ë¶„ì„ (alpha = 0.05)")
print("=" * 60)

# íšŒê·€ê³„ìˆ˜ì™€ p-value ì¶”ì¶œ
coef_table = pd.DataFrame({
    'íšŒê·€ê³„ìˆ˜': model_sm.params,
    'p-value': model_sm.pvalues,
    'ìœ ì˜í•¨': ['O' if p < 0.05 else 'X' for p in model_sm.pvalues]
})

print("\níšŒê·€ê³„ìˆ˜ í…Œì´ë¸”:")
print(coef_table.round(4))

# ìœ ì˜í•œ ë³€ìˆ˜ ì¶œë ¥
print("\nìœ ì˜í•œ ë³€ìˆ˜ (p < 0.05):")
for name, pval in model_sm.pvalues.items():
    if pval < 0.05 and name != 'Intercept':
        coef = model_sm.params[name]
        print(f"  - {name}: ê³„ìˆ˜={coef:.4f}, p-value={pval:.4f}")

# ëª¨ë¸ ì „ì²´ ìœ ì˜ì„±
print(f"\nëª¨ë¸ ì „ì²´ ìœ ì˜ì„±:")
print(f"  F-statistic: {model_sm.fvalue:.4f}")
print(f"  Prob (F-statistic): {model_sm.f_pvalue:.6f}")
if model_sm.f_pvalue < 0.05:
    print("  -> ëª¨ë¸ ì „ì²´ê°€ í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•©ë‹ˆë‹¤.")
````

````python
# Q7. í…ŒìŠ¤íŠ¸/ê²€ì¦
assert model_sm.f_pvalue < 0.05, "ëª¨ë¸ì´ í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•´ì•¼ í•©ë‹ˆë‹¤."
assert model_sm.rsquared > 0.9, "R-squaredê°€ 0.9 ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤."
print("í…ŒìŠ¤íŠ¸ í†µê³¼!")
````

````python
# Q10. ì •ë‹µ ì½”ë“œ - Part 1: ëª¨ë¸ êµ¬ì¶•

shop_data = pd.DataFrame({
    'ê´‘ê³ ë¹„': [100, 150, 200, 120, 180, 220, 160, 190, 140, 210],
    'ë°©ë¬¸ììˆ˜': [1200, 1800, 2500, 1400, 2200, 2800, 1900, 2400, 1600, 2600],
    'í• ì¸ìœ¨': [5, 10, 15, 7, 12, 18, 10, 14, 8, 16],
    'ë§¤ì¶œ': [800, 1200, 1700, 950, 1500, 1900, 1300, 1650, 1100, 1800]
})

print("=" * 70)
print("ì˜¨ë¼ì¸ ì‡¼í•‘ëª° ë§¤ì¶œ ì˜ˆì¸¡ ë¶„ì„ ë¦¬í¬íŠ¸")
print("=" * 70)

# 1. ë‹¤ì¤‘ íšŒê·€ ëª¨ë¸ êµ¬ì¶•
print("\n[1. ë‹¤ì¤‘ íšŒê·€ ëª¨ë¸ êµ¬ì¶•]")
print("-" * 70)

X_shop = shop_data[['ê´‘ê³ ë¹„', 'ë°©ë¬¸ììˆ˜', 'í• ì¸ìœ¨']]
y_shop = shop_data['ë§¤ì¶œ']

# sklearn ëª¨ë¸
model_shop = LinearRegression()
model_shop.fit(X_shop, y_shop)

# statsmodels ëª¨ë¸ (p-value í™•ì¸ìš©)
model_shop_sm = smf.ols('ë§¤ì¶œ ~ ê´‘ê³ ë¹„ + ë°©ë¬¸ììˆ˜ + í• ì¸ìœ¨', data=shop_data).fit()

print(f"\nëª¨ë¸ í•™ìŠµ ì™„ë£Œ")
print(f"  ì¢…ì†ë³€ìˆ˜: ë§¤ì¶œ")
print(f"  ë…ë¦½ë³€ìˆ˜: ê´‘ê³ ë¹„, ë°©ë¬¸ììˆ˜, í• ì¸ìœ¨")
print(f"  ë°ì´í„° ìˆ˜: {len(shop_data)}ê°œ")
````

````python
# Q10. ì •ë‹µ ì½”ë“œ - Part 2: íšŒê·€ê³„ìˆ˜ í•´ì„

print("\n[2. íšŒê·€ê³„ìˆ˜ ë¶„ì„]")
print("-" * 70)

print(f"\nì ˆí¸: {model_shop.intercept_:.2f}ë§Œì›")
print("\níšŒê·€ê³„ìˆ˜:")

coef_interpretation = {
    'ê´‘ê³ ë¹„': ('ë§Œì›', 'ê´‘ê³ ë¹„ 1ë§Œì› ì¦ê°€ ì‹œ'),
    'ë°©ë¬¸ììˆ˜': ('ëª…', 'ë°©ë¬¸ì 1ëª… ì¦ê°€ ì‹œ'),
    'í• ì¸ìœ¨': ('%p', 'í• ì¸ìœ¨ 1%p ì¦ê°€ ì‹œ')
}

for feature, coef in zip(X_shop.columns, model_shop.coef_):
    pval = model_shop_sm.pvalues[feature]
    sig = '*' if pval < 0.05 else ''
    unit, desc = coef_interpretation[feature]
    
    print(f"\n  {feature}: {coef:.4f} (p={pval:.4f}){sig}")
    if coef > 0:
        print(f"    -> {desc} ë§¤ì¶œ ì•½ {abs(coef):.1f}ë§Œì› ì¦ê°€")
    else:
        print(f"    -> {desc} ë§¤ì¶œ ì•½ {abs(coef):.1f}ë§Œì› ê°ì†Œ")

print("\n  (* p < 0.05 í†µê³„ì ìœ¼ë¡œ ìœ ì˜)")
````

````python
# Q10. ì •ë‹µ ì½”ë“œ - Part 3: ëª¨ë¸ í‰ê°€

print("\n[3. ëª¨ë¸ ì„±ëŠ¥ í‰ê°€]")
print("-" * 70)

y_pred_shop = model_shop.predict(X_shop)

mae_shop = mean_absolute_error(y_shop, y_pred_shop)
rmse_shop = np.sqrt(mean_squared_error(y_shop, y_pred_shop))
r2_shop = r2_score(y_shop, y_pred_shop)

print(f"\n  R-squared: {r2_shop:.4f} ({r2_shop*100:.1f}%)")
print(f"  MAE: {mae_shop:.2f}ë§Œì›")
print(f"  RMSE: {rmse_shop:.2f}ë§Œì›")

avg_sales = y_shop.mean()
print(f"\n  í‰ê·  ë§¤ì¶œ: {avg_sales:.0f}ë§Œì›")
print(f"  MAE/í‰ê· : {mae_shop/avg_sales*100:.1f}%")

print("\n  í‰ê°€:")
if r2_shop > 0.9:
    print("    -> R^2 > 0.9: ë§¤ìš° ë†’ì€ ì„¤ëª…ë ¥")
if mae_shop/avg_sales < 0.1:
    print("    -> MAE/í‰ê·  < 10%: ìš°ìˆ˜í•œ ì˜ˆì¸¡ ì •í™•ë„")
````

````python
# Q10. ì •ë‹µ ì½”ë“œ - Part 4: ì‹œê°í™”

print("\n[4. ì‹¤ì œê°’ vs ì˜ˆì¸¡ê°’ ì‹œê°í™”]")
print("-" * 70)

# ì‹¤ì œê°’ vs ì˜ˆì¸¡ê°’ ì‚°ì ë„
fig = go.Figure()

# ì˜ˆì¸¡ vs ì‹¤ì œ
fig.add_trace(go.Scatter(
    x=y_shop,
    y=y_pred_shop,
    mode='markers',
    name='ë°ì´í„°',
    marker=dict(size=12, color='royalblue', line=dict(width=1, color='white')),
    text=[f'ê´‘ê³ ë¹„:{a}ë§Œì›<br>ë°©ë¬¸ì:{v}ëª…<br>í• ì¸:{d}%' 
          for a, v, d in zip(shop_data['ê´‘ê³ ë¹„'], shop_data['ë°©ë¬¸ììˆ˜'], shop_data['í• ì¸ìœ¨'])],
    hovertemplate='ì‹¤ì œ: %{x}ë§Œì›<br>ì˜ˆì¸¡: %{y:.0f}ë§Œì›<br>%{text}<extra></extra>'
))

# ì™„ë²½í•œ ì˜ˆì¸¡ì„ 
min_val = min(y_shop.min(), y_pred_shop.min())
max_val = max(y_shop.max(), y_pred_shop.max())
fig.add_trace(go.Scatter(
    x=[min_val, max_val],
    y=[min_val, max_val],
    mode='lines',
    name='ì™„ë²½í•œ ì˜ˆì¸¡',
    line=dict(color='red', width=2, dash='dash')
))

fig.update_layout(
    title=f'ì‹¤ì œ ë§¤ì¶œ vs ì˜ˆì¸¡ ë§¤ì¶œ (R^2 = {r2_shop:.4f})',
    xaxis_title='ì‹¤ì œ ë§¤ì¶œ (ë§Œì›)',
    yaxis_title='ì˜ˆì¸¡ ë§¤ì¶œ (ë§Œì›)',
    width=700,
    height=500
)

fig.show()
````

````python
# Q10. ì •ë‹µ ì½”ë“œ - Part 5: ë¹„ì¦ˆë‹ˆìŠ¤ ì¸ì‚¬ì´íŠ¸

print("\n[5. ë¹„ì¦ˆë‹ˆìŠ¤ ì¸ì‚¬ì´íŠ¸]")
print("=" * 70)

print("\n[ì£¼ìš” ë°œê²¬ì‚¬í•­]")
print(f"  1. ëª¨ë¸ì´ ë§¤ì¶œ ë³€ë™ì˜ {r2_shop*100:.1f}%ë¥¼ ì„¤ëª… (ë§¤ìš° ë†’ì€ ì„¤ëª…ë ¥)")

# ê°€ì¥ ì˜í–¥ë ¥ í° ë³€ìˆ˜ ì°¾ê¸°
# í‘œì¤€í™” ê³„ìˆ˜ë¡œ ë¹„êµ (ê³µì •í•œ ë¹„êµ)
X_std = (X_shop - X_shop.mean()) / X_shop.std()
model_std = LinearRegression().fit(X_std, y_shop)
importance = dict(zip(X_shop.columns, np.abs(model_std.coef_)))
sorted_importance = sorted(importance.items(), key=lambda x: -x[1])

print(f"  2. ë³€ìˆ˜ ì¤‘ìš”ë„ (í‘œì¤€í™” ê³„ìˆ˜ ê¸°ì¤€):")
for rank, (var, imp) in enumerate(sorted_importance, 1):
    print(f"     {rank}ìœ„: {var} (ì˜í–¥ë ¥: {imp:.2f})")

print(f"  3. ë°©ë¬¸ììˆ˜ê°€ ë§¤ì¶œì— ê°€ì¥ í° ì˜í–¥")

print("\n[ì „ëµì  ê¶Œì¥ì‚¬í•­]")
print("  1. ë°©ë¬¸ì ìœ ì¹˜ì— ë§ˆì¼€íŒ… ì˜ˆì‚° ì§‘ì¤‘")
print(f"     -> ë°©ë¬¸ì 1,000ëª… ì¦ê°€ ì‹œ ë§¤ì¶œ ì•½ {model_shop.coef_[1]*1000:.0f}ë§Œì› ì¦ê°€")

print(f"\n  2. í• ì¸ìœ¨ ì „ëµ ìµœì í™”")
if model_shop.coef_[2] > 0:
    print(f"     -> í• ì¸ìœ¨ ì¦ê°€ê°€ ë§¤ì¶œì— ê¸ì •ì  (í• ì¸ í™•ëŒ€ ê³ ë ¤)")
else:
    print(f"     -> í• ì¸ìœ¨ ì¦ê°€ê°€ ë§¤ì¶œì— ë¶€ì •ì  (í• ì¸ ì¶•ì†Œ ê³ ë ¤)")

print(f"\n  3. ê´‘ê³ ë¹„ ROI ë¶„ì„")
ad_roi = model_shop.coef_[0] - 1  # 1ë§Œì› íˆ¬ì ëŒ€ë¹„ ìˆœìˆ˜ìµ
print(f"     -> ê´‘ê³ ë¹„ 1ë§Œì› íˆ¬ì… ì‹œ ì¶”ê°€ ë§¤ì¶œ {model_shop.coef_[0]:.1f}ë§Œì›")
print(f"     -> ìˆœìˆ˜ìµ: {ad_roi:.1f}ë§Œì› (ROI: {ad_roi*100:.0f}%)")

print("\n" + "=" * 70)
print("ë¶„ì„ ì™„ë£Œ")
print("=" * 70)
````
