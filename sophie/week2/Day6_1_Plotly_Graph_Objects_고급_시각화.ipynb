## 1.1 go.Figure() ÏÉùÏÑ±ÌïòÍ∏∞

Plotly ExpressÎäî Ìïú Ï§ÑÎ°ú Ï∞®Ìä∏Î•º ÎßåÎì§ÏßÄÎßå, Graph ObjectsÎäî Îπà Ï∫îÎ≤ÑÏä§ÏóêÏÑú ÏãúÏûëÌï©ÎãàÎã§.

````python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Îπà Figure ÏÉùÏÑ±
fig = go.Figure()
fig.show()
````

### Ï¥àÍ∏∞ Îç∞Ïù¥ÌÑ∞ÏôÄ Ìï®Íªò ÏÉùÏÑ±

````python
# Îß§Ï∂ú Îç∞Ïù¥ÌÑ∞
months = ['1Ïõî', '2Ïõî', '3Ïõî', '4Ïõî', '5Ïõî', '6Ïõî']
sales = [120, 150, 140, 180, 200, 220]

# Scatter Ï∞®Ìä∏Î°ú ÏÉùÏÑ±
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines+markers'))
fig.show()
````

## 1.2 add_trace()Î°ú Îç∞Ïù¥ÌÑ∞ ÏãúÎ¶¨Ï¶à Ï∂îÍ∞Ä

Ïó¨Îü¨ Îç∞Ïù¥ÌÑ∞ ÏãúÎ¶¨Ï¶àÎ•º ÌïòÎÇòÏùò Ï∞®Ìä∏Ïóê Ï∂îÍ∞ÄÌï† Ïàò ÏûàÏäµÎãàÎã§.

````python
# Ïò®ÎùºÏù∏/Ïò§ÌîÑÎùºÏù∏ Îß§Ï∂ú ÎπÑÍµê
months = ['1Ïõî', '2Ïõî', '3Ïõî', '4Ïõî', '5Ïõî', '6Ïõî']
online_sales = [120, 150, 140, 180, 200, 220]
offline_sales = [200, 190, 180, 170, 160, 150]

# Îπà Figure ÏÉùÏÑ±
fig = go.Figure()

# Ïò®ÎùºÏù∏ Îß§Ï∂ú Ï∂îÍ∞Ä
fig.add_trace(go.Scatter(       #figureÎ•º Î®ºÏ†Ä ÎßåÎì§Í≥† ÎÇòÏÑú Í∑∏ÎûòÌîÑÎ°ú ÎßåÎì§ Îïå
    x=months,                   #items() = [(ÌÇ§, Í∞í),(  ),(    )]  
    y=online_sales,
    mode='lines+markers',
    name='Ïò®ÎùºÏù∏ Îß§Ï∂ú'
))

# Ïò§ÌîÑÎùºÏù∏ Îß§Ï∂ú Ï∂îÍ∞Ä
fig.add_trace(go.Scatter(       
    x=months,
    y=offline_sales,
    mode='lines+markers',
    name='Ïò§ÌîÑÎùºÏù∏ Îß§Ï∂ú'
))

fig.show()
````

### üí° Ïã§Î¨¥ ÏòàÏãú: Ïó¨Îü¨ Ï†úÌíà Îß§Ï∂ú ÎπÑÍµê

Ï†úÌíàÎ≥Ñ ÏõîÍ∞Ñ Îß§Ï∂úÏùÑ Ìïú Ï∞®Ìä∏ÏóêÏÑú ÎπÑÍµêÌï† Ïàò ÏûàÏäµÎãàÎã§.

````python
# Ï†úÌíàÎ≥Ñ Îß§Ï∂ú Îç∞Ïù¥ÌÑ∞
products = {
    'ÎÖ∏Ìä∏Î∂Å': [500, 550, 520, 600, 650, 700],
    'ÌÉúÎ∏îÎ¶ø': [200, 220, 210, 250, 280, 300],
    'Ïä§ÎßàÌä∏Ìè∞': [800, 850, 820, 900, 950, 1000]
}

fig = go.Figure()

# Í∞Å Ï†úÌíàÏùÑ traceÎ°ú Ï∂îÍ∞Ä
for product, sales in products.items():
    fig.add_trace(go.Scatter(
        x=months,
        y=sales,
        mode='lines+markers',
        name=product
    ))

fig.update_layout(title='Ï†úÌíàÎ≥Ñ ÏõîÍ∞Ñ Îß§Ï∂ú Ï∂îÏù¥')        #ÎÇòÏ§ëÏóê Í∞ÅÏ¢Ö Ï∞®Ìä∏ ÏòµÏÖò Ï°∞Ï†ïÌïòÎäî Í≤É 
fig.show()
````

---

## 1.3 update_layout()ÏúºÎ°ú Î†àÏù¥ÏïÑÏõÉ Ïª§Ïä§ÌÑ∞ÎßàÏù¥Ïßï

Ï†úÎ™©, Ï∂ï Î†àÏù¥Î∏î, ÌÅ¨Í∏∞, Î∞∞Í≤ΩÏÉâ Îì± Î™®Îì† Î†àÏù¥ÏïÑÏõÉÏùÑ ÏÑ∏Î∞ÄÌïòÍ≤å Ï°∞Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.

````python
# Í∏∞Î≥∏ Ï∞®Ìä∏
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines+markers'))

# Î†àÏù¥ÏïÑÏõÉ Ïª§Ïä§ÌÑ∞ÎßàÏù¥Ïßï
fig.update_layout(
    title={
        'text': '2024ÎÖÑ ÏÉÅÎ∞òÍ∏∞ Îß§Ï∂ú Ï∂îÏù¥',
        'font': {'size': 24, 'color': 'darkblue'},
        'x': 0.5,  # Ï§ëÏïô Ï†ïÎ†¨
        'xanchor': 'center'
    },
    xaxis_title='Ïõî',
    yaxis_title='Îß§Ï∂ú (ÎßåÏõê)',
    width=800,
    height=500,
    plot_bgcolor='lightgray',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12)
)

fig.show()
````

### üí° Ïã§Î¨¥ ÏòàÏãú: Î≥¥Í≥†ÏÑúÏö© ÍπîÎÅîÌïú Ïä§ÌÉÄÏùº

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x=months, y=online_sales, mode='lines', name='Ïò®ÎùºÏù∏'))
fig.add_trace(go.Scatter(x=months, y=offline_sales, mode='lines', name='Ïò§ÌîÑÎùºÏù∏'))

# Ï†ÑÎ¨∏Ï†ÅÏù∏ Î†àÏù¥ÏïÑÏõÉ
fig.update_layout(
    title='Ï±ÑÎÑêÎ≥Ñ Îß§Ï∂ú ÎπÑÍµê',
    xaxis=dict(
        title='Ïõî',
        showgrid=True,
        gridcolor='lightgray',
        zeroline=False
    ),
    yaxis=dict(
        title='Îß§Ï∂ú (ÎßåÏõê)',
        showgrid=True,
        gridcolor='lightgray'
    ),
    legend=dict(
        x=0.01,
        y=0.99,
        bgcolor='rgba(255, 255, 255, 0.8)',
        bordercolor='gray',
        borderwidth=1
    ),
    hovermode='x unified'  # Ìò∏Î≤Ñ Ïãú Î™®Îì† trace ÌëúÏãú
)

fig.show()
````

---

## 1.4 update_xaxes(), update_yaxes()Î°ú Ï∂ï ÏÑ∏Î∞ÄÌïòÍ≤å ÏÑ§Ï†ï

````python
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines+markers'))

# XÏ∂ï ÏÑ§Ï†ï
fig.update_xaxes(
    title='Ïõî',
    showgrid=True,
    gridcolor='lightgray',
    tickangle=-45  # Î†àÏù¥Î∏î Í∞ÅÎèÑ
)

# YÏ∂ï ÏÑ§Ï†ï
fig.update_yaxes(
    title='Îß§Ï∂ú (ÎßåÏõê)',
    range=[0, 250],  # Ï∂ï Î≤îÏúÑ Í≥†Ï†ï
    dtick=50,        # ÎààÍ∏à Í∞ÑÍ≤©
    showgrid=True,
    gridcolor='lightgray'
)

fig.show()
````

---

## 1.5 ÏÉâÏÉÅ, ÏÑ† Ïä§ÌÉÄÏùº, ÎßàÏª§ Ïª§Ïä§ÌÑ∞ÎßàÏù¥Ïßï

````python
fig = go.Figure()

# ÏÑ† Ïä§ÌÉÄÏùº Ïª§Ïä§ÌÑ∞ÎßàÏù¥Ïßï
fig.add_trace(go.Scatter(
    x=months,
    y=online_sales,
    mode='lines+markers',
    name='Ïò®ÎùºÏù∏',
    line=dict(color='blue', width=3, dash='solid'),
    marker=dict(size=10, color='blue', symbol='circle')
))

fig.add_trace(go.Scatter(
    x=months,
    y=offline_sales,
    mode='lines+markers',
    name='Ïò§ÌîÑÎùºÏù∏',
    line=dict(color='red', width=2, dash='dash'),
    marker=dict(size=8, color='red', symbol='square')
))

fig.show()
````

### üí° Ïã§Î¨¥ ÏòàÏãú: Î™©Ìëú Îã¨ÏÑ±Î•† ÏãúÍ∞ÅÌôî

````python
# Ïã§Ï†ú Îß§Ï∂ú vs Î™©Ìëú
target = [150, 160, 170, 180, 190, 200]

fig = go.Figure()

# Ïã§Ï†ú Îß§Ï∂ú
fig.add_trace(go.Scatter(
    x=months,
    y=sales,
    mode='lines+markers',
    name='Ïã§Ï†ú Îß§Ï∂ú',
    line=dict(color='green', width=3),
    marker=dict(size=10, color='green')
))

# Î™©Ìëú Îß§Ï∂ú (Ï†êÏÑ†)
fig.add_trace(go.Scatter(
    x=months,
    y=target,
    mode='lines',
    name='Î™©Ìëú Îß§Ï∂ú',
    line=dict(color='orange', width=2, dash='dash')
))

fig.update_layout(title='Îß§Ï∂ú vs Î™©Ìëú Îã¨ÏÑ±Î•†')
fig.show()
````

## 2.1 make_subplots()Î°ú Î≥µÌï© Ï∞®Ìä∏ ÎßåÎì§Í∏∞

Ïó¨Îü¨ Ï∞®Ìä∏Î•º Í≤©Ïûê ÌòïÌÉúÎ°ú Î∞∞ÏπòÌï† Ïàò ÏûàÏäµÎãàÎã§.
Ïù¥Ï§ëÏ∂ïÏùÑ Ïç®ÏïºÎê† Îïå?

````python
from plotly.subplots import make_subplots

# 1Ìñâ 2Ïó¥ Subplot ÏÉùÏÑ±
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Ïò®ÎùºÏù∏ Îß§Ï∂ú', 'Ïò§ÌîÑÎùºÏù∏ Îß§Ï∂ú')
)
fig.show( )
````

````python

# Ï≤´ Î≤àÏß∏ Subplot (1Ìñâ 1Ïó¥)
fig.add_trace(
    go.Scatter(x=months, y=online_sales, mode='lines+markers', name='Ïò®ÎùºÏù∏'),
    row=1, col=1
)

# Îëê Î≤àÏß∏ Subplot (1Ìñâ 2Ïó¥)
fig.add_trace(
    go.Scatter(x=months, y=offline_sales, mode='lines+markers', name='Ïò§ÌîÑÎùºÏù∏'),
    row=1, col=2
)

fig.update_layout(height=400, width=900, title_text='Ï±ÑÎÑêÎ≥Ñ Îß§Ï∂ú ÎπÑÍµê')
fig.show()
````

### 2Ìñâ 2Ïó¥ Î≥µÌï© Ï∞®Ìä∏

````python
# Îã§ÏñëÌïú ÏßÄÌëúÎ•º Ìïú ÌôîÎ©¥Ïóê
revenue = [120, 150, 140, 180, 200, 220]
cost = [80, 90, 85, 100, 110, 120]
profit = [r - c for r, c in zip(revenue, cost)]
conversion_rate = [2.5, 3.0, 2.8, 3.5, 4.0, 4.2]

# revenue =np.array  ([120, 150, 140, 180, 200, 220])
# cost = np.array([80, 90, 85, 100, 110, 120])


# 2x2 Subplot
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Îß§Ï∂ú', 'ÎπÑÏö©', 'Ïù¥Ïùµ', 'Ï†ÑÌôòÏú®'),
    specs=[[{}, {}], [{}, {}]]  # Í∞Å ÏÖÄÏùò ÌÉÄÏûÖ ÏßÄÏ†ï
)

# Îß§Ï∂ú (1,1)
fig.add_trace(go.Scatter(x=months, y=revenue, mode='lines', name='Îß§Ï∂ú'), row=1, col=1)

# ÎπÑÏö© (1,2)
fig.add_trace(go.Scatter(x=months, y=cost, mode='lines', name='ÎπÑÏö©'), row=1, col=2)

# Ïù¥Ïùµ (2,1)
fig.add_trace(go.Bar(x=months, y=profit, name='Ïù¥Ïùµ'), row=2, col=1)

# Ï†ÑÌôòÏú® (2,2)
fig.add_trace(go.Scatter(x=months, y=conversion_rate, mode='lines+markers', name='Ï†ÑÌôòÏú®'), row=2, col=2)

fig.update_layout(height=600, width=900, title_text='Í≤ΩÏòÅ ÎåÄÏãúÎ≥¥Îìú', showlegend=False)
fig.show()
````

### üí° Ïã§Î¨¥ ÏòàÏãú: Ï†úÌíàÎ≥Ñ KPI ÎåÄÏãúÎ≥¥Îìú

````python
# Ï†úÌíàÎ≥Ñ ÌåêÎß§Îüâ, ÌèâÍ∑† Í∞ÄÍ≤©, Î¶¨Î∑∞ Ï†êÏàò
products_list = ['ÎÖ∏Ìä∏Î∂Å', 'ÌÉúÎ∏îÎ¶ø', 'Ïä§ÎßàÌä∏Ìè∞', 'Ïù¥Ïñ¥Ìè∞']
quantity_sold = [150, 200, 350, 500]
avg_price = [1200, 600, 900, 150]
avg_rating = [4.5, 4.2, 4.7, 4.3]

fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=('ÌåêÎß§Îüâ', 'ÌèâÍ∑† Í∞ÄÍ≤©', 'ÌèâÍ∑† Î≥ÑÏ†ê')
)

fig.add_trace(go.Bar(x=products_list, y=quantity_sold, name='ÌåêÎß§Îüâ'), row=1, col=1)
fig.add_trace(go.Bar(x=products_list, y=avg_price, name='Í∞ÄÍ≤©'), row=1, col=2)
fig.add_trace(go.Bar(x=products_list, y=avg_rating, name='Î≥ÑÏ†ê'), row=1, col=3)

fig.update_layout(height=400, width=1200, title_text='Ï†úÌíàÎ≥Ñ KPI ÏöîÏïΩ', showlegend=False)
fig.show()
````

## 2.2 Ïù¥Ï§ë YÏ∂ïÏúºÎ°ú ÏÑúÎ°ú Îã§Î•∏ Ïä§ÏºÄÏùº Îç∞Ïù¥ÌÑ∞ ÎπÑÍµê

Îß§Ï∂ú(ÏñµÏõê)Í≥º Ï†ÑÌôòÏú®(%)Ï≤òÎüº Ïä§ÏºÄÏùºÏù¥ Îã§Î•∏ Îç∞Ïù¥ÌÑ∞Î•º Ìïú Ï∞®Ìä∏Ïóê ÌëúÏãúÌï† Îïå Ïú†Ïö©Ìï©ÎãàÎã§.

````python
# Îß§Ï∂ú(ÎßåÏõê)Í≥º Ï†ÑÌôòÏú®(%) Îç∞Ïù¥ÌÑ∞
revenue_data = [120, 150, 140, 180, 200, 220]
conversion_data = [2.5, 3.0, 2.8, 3.5, 4.0, 4.2]

# Secondary YÏ∂ïÏùÑ Í∞ÄÏßÑ Subplot
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Îß§Ï∂ú (Primary YÏ∂ï)
fig.add_trace(
    go.Scatter(x=months, y=revenue_data, name='Îß§Ï∂ú', mode='lines+markers'),
    secondary_y=False
)

# Ï†ÑÌôòÏú® (Secondary YÏ∂ï)
fig.add_trace(
    go.Scatter(x=months, y=conversion_data, name='Ï†ÑÌôòÏú®', mode='lines+markers', line=dict(color='red')),
    secondary_y=True
)

# Ï∂ï Ï†úÎ™© ÏÑ§Ï†ï
fig.update_xaxes(title_text='Ïõî')
fig.update_yaxes(title_text='Îß§Ï∂ú (ÎßåÏõê)', secondary_y=False)
fig.update_yaxes(title_text='Ï†ÑÌôòÏú® (%)', secondary_y=True)

fig.update_layout(title_text='Îß§Ï∂ú vs Ï†ÑÌôòÏú® ÎπÑÍµê', height=500)
fig.show()
````

### üí° Ïã§Î¨¥ ÏòàÏãú: Ïõπ Ìä∏ÎûòÌîΩ vs Ï†ÑÌôòÏú®

````python
# Î∞©Î¨∏Ïûê Ïàò(Î™Ö)ÏôÄ Ï†ÑÌôòÏú®(%)
visitors = [10000, 12000, 11500, 15000, 18000, 20000]
conversion_rate_pct = [2.5, 3.0, 2.8, 3.5, 4.0, 4.2]

fig = make_subplots(specs=[[{"secondary_y": True}]])

# Î∞©Î¨∏Ïûê (Primary)
fig.add_trace(
    go.Bar(x=months, y=visitors, name='Î∞©Î¨∏Ïûê', marker_color='lightblue'),
    secondary_y=False
)

# Ï†ÑÌôòÏú® (Secondary)
fig.add_trace(
    go.Scatter(x=months, y=conversion_rate_pct, name='Ï†ÑÌôòÏú®', mode='lines+markers', line=dict(color='orange', width=3)),
    secondary_y=True
)

fig.update_xaxes(title_text='Ïõî')
fig.update_yaxes(title_text='Î∞©Î¨∏Ïûê Ïàò (Î™Ö)', secondary_y=False)
fig.update_yaxes(title_text='Ï†ÑÌôòÏú® (%)', secondary_y=True, range=[0, 5])

fig.update_layout(title_text='Ïõπ Ìä∏ÎûòÌîΩ vs Ï†ÑÌôòÏú®', hovermode='x unified')
fig.show()
````

## 2.3 AnnotationsÎ°ú Ï∞®Ìä∏Ïóê Ï£ºÏÑù Îã¨Í∏∞

Ï§ëÏöîÌïú Ìè¨Ïù∏Ìä∏, Ïù¥Î≤§Ìä∏, Ïù¥ÏÉÅÏπò Îì±ÏùÑ Í∞ïÏ°∞Ìï† Ïàò ÏûàÏäµÎãàÎã§.
