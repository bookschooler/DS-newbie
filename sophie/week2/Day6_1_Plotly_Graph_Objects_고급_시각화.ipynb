## 1.1 go.Figure() ìƒì„±í•˜ê¸°

Plotly ExpressëŠ” í•œ ì¤„ë¡œ ì°¨íŠ¸ë¥¼ ë§Œë“¤ì§€ë§Œ, Graph ObjectsëŠ” ë¹ˆ ìº”ë²„ìŠ¤ì—ì„œ ì‹œì‘í•©ë‹ˆë‹¤.

````python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# ë¹ˆ Figure ìƒì„±
fig = go.Figure()
fig.show()
````

### ì´ˆê¸° ë°ì´í„°ì™€ í•¨ê»˜ ìƒì„±

````python
# ë§¤ì¶œ ë°ì´í„°
months = ['1ì›”', '2ì›”', '3ì›”', '4ì›”', '5ì›”', '6ì›”']
sales = [120, 150, 140, 180, 200, 220]

# Scatter ì°¨íŠ¸ë¡œ ìƒì„±
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines+markers'))
fig.show()
````

---

## 1.2 add_trace()ë¡œ ë°ì´í„° ì‹œë¦¬ì¦ˆ ì¶”ê°€

ì—¬ëŸ¬ ë°ì´í„° ì‹œë¦¬ì¦ˆë¥¼ í•˜ë‚˜ì˜ ì°¨íŠ¸ì— ì¶”ê°€í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# ì˜¨ë¼ì¸/ì˜¤í”„ë¼ì¸ ë§¤ì¶œ ë¹„êµ
months = ['1ì›”', '2ì›”', '3ì›”', '4ì›”', '5ì›”', '6ì›”']
online_sales = [120, 150, 140, 180, 200, 220]
offline_sales = [200, 190, 180, 170, 160, 150]

# ë¹ˆ Figure ìƒì„±
fig = go.Figure()

# ì˜¨ë¼ì¸ ë§¤ì¶œ ì¶”ê°€
fig.add_trace(go.Scatter(
    x=months,
    y=online_sales,
    mode='lines+markers',
    name='ì˜¨ë¼ì¸ ë§¤ì¶œ'
))

# ì˜¤í”„ë¼ì¸ ë§¤ì¶œ ì¶”ê°€
fig.add_trace(go.Scatter(
    x=months,
    y=offline_sales,
    mode='lines+markers',
    name='ì˜¤í”„ë¼ì¸ ë§¤ì¶œ'
))

fig.show()
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ì—¬ëŸ¬ ì œí’ˆ ë§¤ì¶œ ë¹„êµ

ì œí’ˆë³„ ì›”ê°„ ë§¤ì¶œì„ í•œ ì°¨íŠ¸ì—ì„œ ë¹„êµí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# ì œí’ˆë³„ ë§¤ì¶œ ë°ì´í„°
products = {
    'ë…¸íŠ¸ë¶': [500, 550, 520, 600, 650, 700],
    'íƒœë¸”ë¦¿': [200, 220, 210, 250, 280, 300],
    'ìŠ¤ë§ˆíŠ¸í°': [800, 850, 820, 900, 950, 1000]
}

fig = go.Figure()

# ê° ì œí’ˆì„ traceë¡œ ì¶”ê°€
for product, sales in products.items():
    fig.add_trace(go.Scatter(
        x=months,
        y=sales,
        mode='lines+markers',
        name=product
    ))

fig.update_layout(title='ì œí’ˆë³„ ì›”ê°„ ë§¤ì¶œ ì¶”ì´')
fig.show()
````

---

## 1.3 update_layout()ìœ¼ë¡œ ë ˆì´ì•„ì›ƒ ì»¤ìŠ¤í„°ë§ˆì´ì§•

ì œëª©, ì¶• ë ˆì´ë¸”, í¬ê¸°, ë°°ê²½ìƒ‰ ë“± ëª¨ë“  ë ˆì´ì•„ì›ƒì„ ì„¸ë°€í•˜ê²Œ ì¡°ì •í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# ê¸°ë³¸ ì°¨íŠ¸
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines+markers'))

# ë ˆì´ì•„ì›ƒ ì»¤ìŠ¤í„°ë§ˆì´ì§•
fig.update_layout(
    title={
        'text': '2024ë…„ ìƒë°˜ê¸° ë§¤ì¶œ ì¶”ì´',
        'font': {'size': 24, 'color': 'darkblue'},
        'x': 0.5,  # ì¤‘ì•™ ì •ë ¬
        'xanchor': 'center'
    },
    xaxis_title='ì›”',
    yaxis_title='ë§¤ì¶œ (ë§Œì›)',
    width=800,
    height=500,
    plot_bgcolor='lightgray',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12)
)

fig.show()
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ë³´ê³ ì„œìš© ê¹”ë”í•œ ìŠ¤íƒ€ì¼

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x=months, y=online_sales, mode='lines', name='ì˜¨ë¼ì¸'))
fig.add_trace(go.Scatter(x=months, y=offline_sales, mode='lines', name='ì˜¤í”„ë¼ì¸'))

# ì „ë¬¸ì ì¸ ë ˆì´ì•„ì›ƒ
fig.update_layout(
    title='ì±„ë„ë³„ ë§¤ì¶œ ë¹„êµ',
    xaxis=dict(
        title='ì›”',
        showgrid=True,
        gridcolor='lightgray',
        zeroline=False
    ),
    yaxis=dict(
        title='ë§¤ì¶œ (ë§Œì›)',
        showgrid=True,
        gridcolor='lightgray'
    ),
    legend=dict(
        x=0.01,
        y=0.99,
        bgcolor='rgba(255, 255, 255, 0.8)',
        bordercolor='gray',
        borderwidth=1
    ),
    hovermode='x unified'  # í˜¸ë²„ ì‹œ ëª¨ë“  trace í‘œì‹œ
)

fig.show()
````

## 1.4 update_xaxes(), update_yaxes()ë¡œ ì¶• ì„¸ë°€í•˜ê²Œ ì„¤ì •

````python
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines+markers'))

# Xì¶• ì„¤ì •
fig.update_xaxes(
    title='ì›”',
    showgrid=True,
    gridcolor='lightgray',
    tickangle=-45  # ë ˆì´ë¸” ê°ë„
)

# Yì¶• ì„¤ì •
fig.update_yaxes(
    title='ë§¤ì¶œ (ë§Œì›)',
    range=[0, 250],  # ì¶• ë²”ìœ„ ê³ ì •
    dtick=50,        # ëˆˆê¸ˆ ê°„ê²©
    showgrid=True,
    gridcolor='lightgray'
)

fig.show()
````

## 1.5 ìƒ‰ìƒ, ì„  ìŠ¤íƒ€ì¼, ë§ˆì»¤ ì»¤ìŠ¤í„°ë§ˆì´ì§•

````python
fig = go.Figure()

# ì„  ìŠ¤íƒ€ì¼ ì»¤ìŠ¤í„°ë§ˆì´ì§•
fig.add_trace(go.Scatter(
    x=months,
    y=online_sales,
    mode='lines+markers',
    name='ì˜¨ë¼ì¸',
    line=dict(color='blue', width=3, dash='solid'),
    marker=dict(size=10, color='blue', symbol='circle')
))

fig.add_trace(go.Scatter(
    x=months,
    y=offline_sales,
    mode='lines+markers',
    name='ì˜¤í”„ë¼ì¸',
    line=dict(color='red', width=2, dash='dash'),
    marker=dict(size=8, color='red', symbol='square')
))

fig.show()
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ëª©í‘œ ë‹¬ì„±ë¥  ì‹œê°í™”

````python
# ì‹¤ì œ ë§¤ì¶œ vs ëª©í‘œ
target = [150, 160, 170, 180, 190, 200]

fig = go.Figure()

# ì‹¤ì œ ë§¤ì¶œ
fig.add_trace(go.Scatter(
    x=months,
    y=sales,
    mode='lines+markers',
    name='ì‹¤ì œ ë§¤ì¶œ',
    line=dict(color='green', width=3),
    marker=dict(size=10, color='green')
))

# ëª©í‘œ ë§¤ì¶œ (ì ì„ )
fig.add_trace(go.Scatter(
    x=months,
    y=target,
    mode='lines',
    name='ëª©í‘œ ë§¤ì¶œ',
    line=dict(color='orange', width=2, dash='dash')
))

fig.update_layout(title='ë§¤ì¶œ vs ëª©í‘œ ë‹¬ì„±ë¥ ')
fig.show()
````

## 2.1 make_subplots()ë¡œ ë³µí•© ì°¨íŠ¸ ë§Œë“¤ê¸°

ì—¬ëŸ¬ ì°¨íŠ¸ë¥¼ ê²©ì í˜•íƒœë¡œ ë°°ì¹˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
from plotly.subplots import make_subplots

# 1í–‰ 2ì—´ Subplot ìƒì„±
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('ì˜¨ë¼ì¸ ë§¤ì¶œ', 'ì˜¤í”„ë¼ì¸ ë§¤ì¶œ')
)

# ì²« ë²ˆì§¸ Subplot (1í–‰ 1ì—´)
fig.add_trace(
    go.Scatter(x=months, y=online_sales, mode='lines+markers', name='ì˜¨ë¼ì¸'),
    row=1, col=1
)

# ë‘ ë²ˆì§¸ Subplot (1í–‰ 2ì—´)
fig.add_trace(
    go.Scatter(x=months, y=offline_sales, mode='lines+markers', name='ì˜¤í”„ë¼ì¸'),
    row=1, col=2
)

fig.update_layout(height=400, width=900, title_text='ì±„ë„ë³„ ë§¤ì¶œ ë¹„êµ')
fig.show()
````

### 2í–‰ 2ì—´ ë³µí•© ì°¨íŠ¸

````python
# ë‹¤ì–‘í•œ ì§€í‘œë¥¼ í•œ í™”ë©´ì—
revenue = [120, 150, 140, 180, 200, 220]
cost = [80, 90, 85, 100, 110, 120]
profit = [r - c for r, c in zip(revenue, cost)]
conversion_rate = [2.5, 3.0, 2.8, 3.5, 4.0, 4.2]

# 2x2 Subplot
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('ë§¤ì¶œ', 'ë¹„ìš©', 'ì´ìµ', 'ì „í™˜ìœ¨'),
    specs=[[{}, {}], [{}, {}]]  # ê° ì…€ì˜ íƒ€ì… ì§€ì •
)

# ë§¤ì¶œ (1,1)
fig.add_trace(go.Scatter(x=months, y=revenue, mode='lines', name='ë§¤ì¶œ'), row=1, col=1)

# ë¹„ìš© (1,2)
fig.add_trace(go.Scatter(x=months, y=cost, mode='lines', name='ë¹„ìš©'), row=1, col=2)

# ì´ìµ (2,1)
fig.add_trace(go.Bar(x=months, y=profit, name='ì´ìµ'), row=2, col=1)

# ì „í™˜ìœ¨ (2,2)
fig.add_trace(go.Scatter(x=months, y=conversion_rate, mode='lines+markers', name='ì „í™˜ìœ¨'), row=2, col=2)

fig.update_layout(height=600, width=900, title_text='ê²½ì˜ ëŒ€ì‹œë³´ë“œ', showlegend=False)
fig.show()
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ì œí’ˆë³„ KPI ëŒ€ì‹œë³´ë“œ

````python
# ì œí’ˆë³„ íŒë§¤ëŸ‰, í‰ê·  ê°€ê²©, ë¦¬ë·° ì ìˆ˜
products_list = ['ë…¸íŠ¸ë¶', 'íƒœë¸”ë¦¿', 'ìŠ¤ë§ˆíŠ¸í°', 'ì´ì–´í°']
quantity_sold = [150, 200, 350, 500]
avg_price = [1200, 600, 900, 150]
avg_rating = [4.5, 4.2, 4.7, 4.3]

fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=('íŒë§¤ëŸ‰', 'í‰ê·  ê°€ê²©', 'í‰ê·  ë³„ì ')
)

fig.add_trace(go.Bar(x=products_list, y=quantity_sold, name='íŒë§¤ëŸ‰'), row=1, col=1)
fig.add_trace(go.Bar(x=products_list, y=avg_price, name='ê°€ê²©'), row=1, col=2)
fig.add_trace(go.Bar(x=products_list, y=avg_rating, name='ë³„ì '), row=1, col=3)

fig.update_layout(height=400, width=1200, title_text='ì œí’ˆë³„ KPI ìš”ì•½', showlegend=False)
fig.show()
````

## 2.2 ì´ì¤‘ Yì¶•ìœ¼ë¡œ ì„œë¡œ ë‹¤ë¥¸ ìŠ¤ì¼€ì¼ ë°ì´í„° ë¹„êµ

ë§¤ì¶œ(ì–µì›)ê³¼ ì „í™˜ìœ¨(%)ì²˜ëŸ¼ ìŠ¤ì¼€ì¼ì´ ë‹¤ë¥¸ ë°ì´í„°ë¥¼ í•œ ì°¨íŠ¸ì— í‘œì‹œí•  ë•Œ ìœ ìš©í•©ë‹ˆë‹¤.

````python
# ë§¤ì¶œ(ë§Œì›)ê³¼ ì „í™˜ìœ¨(%) ë°ì´í„°
revenue_data = [120, 150, 140, 180, 200, 220]
conversion_data = [2.5, 3.0, 2.8, 3.5, 4.0, 4.2]

# Secondary Yì¶•ì„ ê°€ì§„ Subplot
fig = make_subplots(specs=[[{"secondary_y": True}]])

# ë§¤ì¶œ (Primary Yì¶•)
fig.add_trace(
    go.Scatter(x=months, y=revenue_data, name='ë§¤ì¶œ', mode='lines+markers'),
    secondary_y=False
)

# ì „í™˜ìœ¨ (Secondary Yì¶•)
fig.add_trace(
    go.Scatter(x=months, y=conversion_data, name='ì „í™˜ìœ¨', mode='lines+markers', line=dict(color='red')),
    secondary_y=True
)

# ì¶• ì œëª© ì„¤ì •
fig.update_xaxes(title_text='ì›”')
fig.update_yaxes(title_text='ë§¤ì¶œ (ë§Œì›)', secondary_y=False)
fig.update_yaxes(title_text='ì „í™˜ìœ¨ (%)', secondary_y=True)

fig.update_layout(title_text='ë§¤ì¶œ vs ì „í™˜ìœ¨ ë¹„êµ', height=500)
fig.show()
````

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ì›¹ íŠ¸ë˜í”½ vs ì „í™˜ìœ¨

````python
# ë°©ë¬¸ì ìˆ˜(ëª…)ì™€ ì „í™˜ìœ¨(%)
visitors = [10000, 12000, 11500, 15000, 18000, 20000]
conversion_rate_pct = [2.5, 3.0, 2.8, 3.5, 4.0, 4.2]

fig = make_subplots(specs=[[{"secondary_y": True}]])

# ë°©ë¬¸ì (Primary)
fig.add_trace(
    go.Bar(x=months, y=visitors, name='ë°©ë¬¸ì', marker_color='lightblue'),
    secondary_y=False
)

# ì „í™˜ìœ¨ (Secondary)
fig.add_trace(
    go.Scatter(x=months, y=conversion_rate_pct, name='ì „í™˜ìœ¨', mode='lines+markers', line=dict(color='orange', width=3)),
    secondary_y=True
)

fig.update_xaxes(title_text='ì›”')
fig.update_yaxes(title_text='ë°©ë¬¸ì ìˆ˜ (ëª…)', secondary_y=False)
fig.update_yaxes(title_text='ì „í™˜ìœ¨ (%)', secondary_y=True, range=[0, 5])

fig.update_layout(title_text='ì›¹ íŠ¸ë˜í”½ vs ì „í™˜ìœ¨', hovermode='x unified')
fig.show()
````

## 2.3 Annotationsë¡œ ì°¨íŠ¸ì— ì£¼ì„ ë‹¬ê¸°

ì¤‘ìš”í•œ í¬ì¸íŠ¸, ì´ë²¤íŠ¸, ì´ìƒì¹˜ ë“±ì„ ê°•ì¡°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

## 2.4 Shapesë¡œ ì˜ì—­, ì„ , ë„í˜• ê·¸ë¦¬ê¸°

ëª©í‘œ ë²”ìœ„, ê²½ê³  êµ¬ì—­, ì´ë²¤íŠ¸ ê¸°ê°„ ë“±ì„ ì‹œê°ì ìœ¼ë¡œ í‘œí˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines+markers'))

# ëª©í‘œ ë²”ìœ„ (ìˆ˜í‰ì„ )
fig.add_hline(
    y=180, 
    line_dash='dash', 
    line_color='green',
    annotation_text='ëª©í‘œì„  (180ë§Œì›)',
    annotation_position='right'
)

# ê²½ê³  ìˆ˜ì¤€ (ìˆ˜í‰ì„ )
fig.add_hline(
    y=130, 
    line_dash='dash', 
    line_color='red',
    annotation_text='ìµœì†Œ ëª©í‘œ (130ë§Œì›)'
)

fig.update_layout(title='ëª©í‘œ ëŒ€ë¹„ ë§¤ì¶œ ì¶”ì´')
fig.show()
````
