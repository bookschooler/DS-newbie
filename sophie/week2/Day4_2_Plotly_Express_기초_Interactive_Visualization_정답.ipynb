### Q3. ÏÑ† Í∑∏ÎûòÌîÑÎ°ú Ìä∏Î†åÎìú ÌëúÌòÑ ‚≠ê‚≠ê

### Q5. ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú Î∂ÑÌè¨ ÌååÏïÖ ‚≠ê‚≠ê‚≠ê

**Î¨∏Ï†ú**: ÏïÑÎûò ÎÇòÏù¥ Îç∞Ïù¥ÌÑ∞Î°ú ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î¶¨Îêò, Íµ¨Í∞Ñ(nbins)ÏùÑ 15Í∞úÎ°ú ÏÑ§Ï†ïÌïòÏÑ∏Ïöî.

````python
# Q5 Ï†ïÎãµ
ages = pd.DataFrame({
    'ÎÇòÏù¥': [23, 25, 28, 30, 32, 35, 38, 40, 42, 45, 48, 50, 52, 55, 58,
             22, 26, 29, 31, 34, 36, 39, 41, 44, 47, 49, 51, 54, 56, 59]
})

# ÌûàÏä§ÌÜ†Í∑∏Îû® (Íµ¨Í∞Ñ 15Í∞ú)
fig = px.histogram(
    ages,
    x='ÎÇòÏù¥',
    nbins=15,  # ÌïµÏã¨: Íµ¨Í∞Ñ Í∞úÏàò ÏßÄÏ†ï
    title='ÎÇòÏù¥ Î∂ÑÌè¨ (15Í∞ú Íµ¨Í∞Ñ)'
)

fig.show()
````

````python
# ÌÖåÏä§Ìä∏: nbins ÏÑ§Ï†ï ÌôïÏù∏
# (Ïã§Ï†ú bin Í∞úÏàòÎäî Îç∞Ïù¥ÌÑ∞ Î≤îÏúÑÏóê Îî∞Îùº ÏïΩÍ∞Ñ Îã¨ÎùºÏßà Ïàò ÏûàÏùå)
assert fig.data[0].nbinsx == 15, "nbinsÎäî 15Î°ú ÏÑ§Ï†ïÎêòÏñ¥Ïïº Ìï©ÎãàÎã§."
print("‚úÖ ÌÖåÏä§Ìä∏ ÌÜµÍ≥º!")
````

- `nbins`Î°ú Íµ¨Í∞Ñ Í∞úÏàòÎ•º Ï°∞Ï†à (Í∏∞Î≥∏Í∞í: ÏûêÎèô Í≥ÑÏÇ∞)

**ÌïµÏã¨ Í∞úÎÖê**:
- `nbins`Í∞Ä ÌÅ¥ÏàòÎ°ù ÏÑ∏Î∞ÄÌïú Î∂ÑÌè¨, ÏûëÏùÑÏàòÎ°ù Ï†ÑÏ≤¥ Ìå®ÌÑ¥ ÌååÏïÖ

1. xÏ∂ï: `Í≤ΩÌóòÎÖÑÏàò`, yÏ∂ï: `Ïó∞Î¥â`
2. Ï†ê ÌÅ¨Í∏∞: `ÌîÑÎ°úÏ†ùÌä∏Ïàò`

4. ÌÖúÌîåÎ¶ø: `plotly_white`
5. hover_dataÏóê `Ïù¥Î¶Ñ` Ìè¨Ìï®

````python
# Q10 Ï†ïÎãµ
employee_data = pd.DataFrame({
    'Ïù¥Î¶Ñ': ['ÍπÄÏ≤†Ïàò', 'Ïù¥ÏòÅÌù¨', 'Î∞ïÎØºÏàò', 'Ï†ïÏàòÏßÑ'],
    'Í≤ΩÌóòÎÖÑÏàò': [3, 5, 8, 10],
    'Ïó∞Î¥â': [4000, 5500, 7000, 9000],
    'ÌîÑÎ°úÏ†ùÌä∏Ïàò': [5, 8, 12, 15],
    'ÏßÅÍ∏â': ['ÏÇ¨Ïõê', 'ÎåÄÎ¶¨', 'Í≥ºÏû•', 'Ï∞®Ïû•']
})

# Ï¢ÖÌï© ÏÇ∞Ï†êÎèÑ (5Í∞ÄÏßÄ Ï°∞Í±¥ Î™®Îëê Ï∂©Ï°±)
fig = px.scatter(
    employee_data,
    x='Í≤ΩÌóòÎÖÑÏàò',          # Ï°∞Í±¥1
    y='Ïó∞Î¥â',              # Ï°∞Í±¥1
    size='ÌîÑÎ°úÏ†ùÌä∏Ïàò',     # Ï°∞Í±¥2
    color='ÏßÅÍ∏â',          # Ï°∞Í±¥3
    template='plotly_white',  # Ï°∞Í±¥4
    hover_data=['Ïù¥Î¶Ñ'],   # Ï°∞Í±¥5
    title='ÏßÅÏõê ÏÑ±Í≥º Î∂ÑÏÑù (Í≤ΩÌóò-Ïó∞Î¥â-ÌîÑÎ°úÏ†ùÌä∏)',
    labels={'Í≤ΩÌóòÎÖÑÏàò': 'Í≤ΩÌóò (ÎÖÑ)', 'Ïó∞Î¥â': 'Ïó∞Î¥â (ÎßåÏõê)'},
    size_max=40
)

fig.show()
````

- `template`, `hover_data`, `labels`ÍπåÏßÄ ÌôúÏö©Ìïú Ï¢ÖÌï© Î¨∏Ï†ú

**ÌïµÏã¨ Í∞úÎÖê**:
- **Îã§Ï∞®Ïõê ÌëúÌòÑ**: x(Í≤ΩÌóò) + y(Ïó∞Î¥â) + size(ÌîÑÎ°úÏ†ùÌä∏) + color(ÏßÅÍ∏â) + hover(Ïù¥Î¶Ñ) = 5Ï∞®Ïõê
