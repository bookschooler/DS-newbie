### Q10. ì¢…í•© ë¬¸ì œ - ë¶„ê¸° ì‹¤ì  ë¶„ì„ â­â­â­â­â­

**ë¬¸ì œ**: ë‹¤ìŒ ë¶„ê¸° ì‹¤ì  ë°ì´í„°ë¥¼ ë¶„ì„í•˜ì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
1. ê° ì§€ì ì˜ ë¶„ê¸° ì´ ë§¤ì¶œ ê³„ì‚°
2. ê° ì›”ì˜ ì „ì²´ ì§€ì  í•©ê³„ ê³„ì‚°
3. ì „ì²´ í‰ê·  ë§¤ì¶œë³´ë‹¤ ë†’ì€ ì‹¤ì ë§Œ ì¶”ì¶œ
4. ìµœê³  ì‹¤ì  ì§€ì ê³¼ ì›” ì°¾ê¸°

````python
# Q10 ì •ë‹µ

quarterly_data = np.array([
    [1200, 1350, 1280],  # ê°•ë‚¨ì 
    [980, 1050, 1120],   # ë¶„ë‹¹ì 
    [1100, 1200, 1150]   # íŒêµì 
])

# 1. ê° ì§€ì ì˜ ë¶„ê¸° ì´ ë§¤ì¶œ (í–‰ë³„ í•©ê³„)
branch_totals = quarterly_data.sum(axis=1)
print("1. ì§€ì ë³„ ë¶„ê¸° ì´ë§¤ì¶œ:")
print(f"   ê°•ë‚¨ì : {branch_totals[0]:,}ë§Œì›")
print(f"   ë¶„ë‹¹ì : {branch_totals[1]:,}ë§Œì›")
print(f"   íŒêµì : {branch_totals[2]:,}ë§Œì›")

# 2. ê° ì›”ì˜ ì „ì²´ ì§€ì  í•©ê³„ (ì—´ë³„ í•©ê³„)
monthly_totals = quarterly_data.sum(axis=0)
print("\n2. ì›”ë³„ ì „ì²´ ë§¤ì¶œ:")
for i, total in enumerate(monthly_totals, 1):
    print(f"   {i}ì›”: {total:,}ë§Œì›")

# 3. ì „ì²´ í‰ê· ë³´ë‹¤ ë†’ì€ ì‹¤ì ë§Œ ì¶”ì¶œ
overall_avg = quarterly_data.mean()
above_avg = quarterly_data[quarterly_data > overall_avg]
print(f"\n3. ì „ì²´ í‰ê· : {overall_avg:.1f}ë§Œì›")
print(f"   í‰ê·  ì´ìƒ ì‹¤ì : {above_avg}")

# 4. ìµœê³  ì‹¤ì  ì§€ì ê³¼ ì›” ì°¾ê¸°
max_value = quarterly_data.max()
max_position = np.unravel_index(quarterly_data.argmax(), quarterly_data.shape)
branch_names = ["ê°•ë‚¨ì ", "ë¶„ë‹¹ì ", "íŒêµì "]
print(f"\n4. ìµœê³  ì‹¤ì : {max_value}ë§Œì›")
print(f"   ì§€ì : {branch_names[max_position[0]]}")
print(f"   ì›”: {max_position[1] + 1}ì›”")
````
