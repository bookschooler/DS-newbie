### 1.2 px.scatter() - ÏÇ∞Ï†êÎèÑ ÏãúÍ∞ÅÌôî

**ÏÇ∞Ï†êÎèÑ (Scatter Plot)**: Îëê Ïó∞ÏÜçÌòï Î≥ÄÏàòÏùò Í¥ÄÍ≥ÑÎ•º Ï†êÏúºÎ°ú ÌëúÌòÑ

**Í∏∞Î≥∏ Î¨∏Î≤ï**:
```python
px.scatter(
    data_frame=df,      # pandas DataFrame
    x='Ïª¨ÎüºÎ™Ö',          # xÏ∂ï Î≥ÄÏàò
    y='Ïª¨ÎüºÎ™Ö',          # yÏ∂ï Î≥ÄÏàò
    title='Ï∞®Ìä∏ Ï†úÎ™©'    # Ï†úÎ™©

```

````python
# Í∞ÑÎã®Ìïú Îç∞Ïù¥ÌÑ∞ÏÖã ÏÉùÏÑ±
sales_data = pd.DataFrame({
    'Í¥ëÍ≥†ÎπÑ': [50, 80, 120, 150, 200, 250, 300, 350],  # ÎßåÏõê
    'Îß§Ï∂ú': [200, 280, 350, 420, 500, 580, 650, 720],  # ÎßåÏõê
    'ÏßÄÏó≠': ['ÏÑúÏö∏', 'Î∂ÄÏÇ∞', 'ÏÑúÏö∏', 'ÎåÄÍµ¨', 'Î∂ÄÏÇ∞', 'ÏÑúÏö∏', 'ÎåÄÍµ¨', 'Î∂ÄÏÇ∞']
})
````

üí° **Ïã§Î¨¥ ÏòàÏãú**: ÎßàÏºÄÌåÖ Î∂ÄÏÑúÏóêÏÑú Í¥ëÍ≥†ÎπÑ ÎåÄÎπÑ Îß§Ï∂ú Ìö®Ïú®ÏùÑ ÏßÄÏó≠Î≥ÑÎ°ú ÎπÑÍµêÌï† Îïå, ÏÉâÏÉÅÏúºÎ°ú ÏßÄÏó≠ÏùÑ Íµ¨Î∂ÑÌïòÎ©¥ ÌïúÎààÏóê Ìå®ÌÑ¥ ÌååÏïÖÏù¥ Í∞ÄÎä•Ìï©ÎãàÎã§.

### 1.3 px.line() - ÏÑ† Í∑∏ÎûòÌîÑ (Ìä∏Î†åÎìú)

**ÏÑ† Í∑∏ÎûòÌîÑ**: ÏãúÍ∞ÑÏóê Îî∞Î•∏ Î≥ÄÌôî, Ìä∏Î†åÎìúÎ•º ÏÑ†ÏúºÎ°ú Ïó∞Í≤∞ÌïòÏó¨ ÌëúÌòÑ

**Í∏∞Î≥∏ Î¨∏Î≤ï**:
```python
px.line(
    data_frame=df,
    x='ÎÇ†Ïßú',
    y='Í∞í',
    title='Ìä∏Î†åÎìú Ï∞®Ìä∏'
)
```

````python
# ÏõîÎ≥Ñ Îß§Ï∂ú Îç∞Ïù¥ÌÑ∞
monthly_revenue = pd.DataFrame({
    'Ïõî': ['1Ïõî', '2Ïõî', '3Ïõî', '4Ïõî', '5Ïõî', '6Ïõî'],
    'Îß§Ï∂ú': [1200, 1350, 1280, 1500, 1650, 1800],  # ÎßåÏõê
    'ÎπÑÏö©': [800, 850, 820, 900, 950, 1000]  # ÎßåÏõê
})

# Îã®Ïùº ÏÑ† Í∑∏ÎûòÌîÑ
fig = px.line(
    monthly_revenue,
    x='Ïõî',
    y='Îß§Ï∂ú',
    title='ÏõîÎ≥Ñ Îß§Ï∂ú Ìä∏Î†åÎìú',
    markers=True  # Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ ÌëúÏãú
)

fig.show()
````

````python
# Îã§Ï§ë ÏÑ† Í∑∏ÎûòÌîÑ (ÏôÄÏù¥Îìú Ìè¨Îß∑)
fig = px.line(
    monthly_revenue,
    x='Ïõî',
    y=['Îß§Ï∂ú', 'ÎπÑÏö©'],  # Ïó¨Îü¨ Ïª¨Îüº
    title='ÏõîÎ≥Ñ Îß§Ï∂ú vs ÎπÑÏö©',
    markers=True
)

fig.show()
````

### 1.4 px.bar() - ÎßâÎåÄ Í∑∏ÎûòÌîÑ

**ÎßâÎåÄ Í∑∏ÎûòÌîÑ**: Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Í∞íÏùÑ ÎßâÎåÄ ÎÜíÏù¥Î°ú ÎπÑÍµê


````python
# Ï†úÌíàÎ≥Ñ ÌåêÎß§Îüâ Îç∞Ïù¥ÌÑ∞
product_sales = pd.DataFrame({
    'Ï†úÌíà': ['ÎÖ∏Ìä∏Î∂Å', 'Î™®ÎãàÌÑ∞', 'ÌÇ§Î≥¥Îìú', 'ÎßàÏö∞Ïä§', 'Ìó§ÎìúÏÖã'],
    'ÌåêÎß§Îüâ': [120, 200, 350, 280, 150],
    'Ïπ¥ÌÖåÍ≥†Î¶¨': ['PC', 'PC', 'Ï£ºÎ≥ÄÍ∏∞Í∏∞', 'Ï£ºÎ≥ÄÍ∏∞Í∏∞', 'Ï£ºÎ≥ÄÍ∏∞Í∏∞']
})

# Í∏∞Î≥∏ ÎßâÎåÄ Í∑∏ÎûòÌîÑ
fig = px.bar(
    product_sales,
    x='Ï†úÌíà',
    y='ÌåêÎß§Îüâ',
    title='Ï†úÌíàÎ≥Ñ ÌåêÎß§Îüâ ÎπÑÍµê'
)

fig.show()
````

üí° **Ïã§Î¨¥ ÏòàÏãú**: Ï†úÌíàÎ™ÖÏù¥ Í∏¥ Í≤ΩÏö∞ Í∞ÄÎ°ú ÎßâÎåÄ Í∑∏ÎûòÌîÑ(`orientation='h'`)Î•º ÏÇ¨Ïö©ÌïòÎ©¥ ÎùºÎ≤®Ïù¥ Í≤πÏπòÏßÄ ÏïäÏïÑ Í∞ÄÎèÖÏÑ±Ïù¥ ÎÜíÏïÑÏßëÎãàÎã§.

### 1.5 px.histogram() - ÌûàÏä§ÌÜ†Í∑∏Îû®

**ÌûàÏä§ÌÜ†Í∑∏Îû®**: Ïó∞ÏÜçÌòï Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨Î•º Íµ¨Í∞Ñ(bin)ÏúºÎ°ú ÎÇòÎàÑÏñ¥ ÌëúÌòÑ

**Í∏∞Î≥∏ Î¨∏Î≤ï**:
```python
px.histogram(
    data_frame=df,
    x='Í∞í',
    nbins=10,  # Íµ¨Í∞Ñ Í∞úÏàò
    title='Î∂ÑÌè¨ Ï∞®Ìä∏'
)
```

````python
# Í≥†Í∞ù ÎÇòÏù¥ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Ï†ïÍ∑úÎ∂ÑÌè¨)
np.random.seed(42)
customer_data = pd.DataFrame({
    'ÎÇòÏù¥': np.random.normal(35, 10, 500).astype(int),  # ÌèâÍ∑† 35, ÌëúÏ§ÄÌé∏Ï∞® 10
    'ÏÑ±Î≥Ñ': np.random.choice(['ÎÇ®ÏÑ±', 'Ïó¨ÏÑ±'], 500)
})

# Í∏∞Î≥∏ ÌûàÏä§ÌÜ†Í∑∏Îû®
fig = px.histogram(
    customer_data,
    x='ÎÇòÏù¥',
    nbins=20,
    title='Í≥†Í∞ù ÎÇòÏù¥ Î∂ÑÌè¨'
)

fig.show()
````

````python
# Í∑∏Î£πÎ≥Ñ ÌûàÏä§ÌÜ†Í∑∏Îû® (Í≤πÏ≥ê Í∑∏Î¶¨Í∏∞)
fig = px.histogram(
    customer_data,
    x='ÎÇòÏù¥',
    color='ÏÑ±Î≥Ñ',
    nbins=20,
    title='ÏÑ±Î≥Ñ Í≥†Í∞ù ÎÇòÏù¥ Î∂ÑÌè¨',
    barmode='overlay',  # Í≤πÏ≥êÏÑú ÌëúÏãú
    opacity=0.7  # Ìà¨Î™ÖÎèÑ Ï°∞Ï†à
)

fig.show()
````

### 1.6 px.box() - Î∞ïÏä§ÌîåÎ°Ø

**Î∞ïÏä§ÌîåÎ°Ø**: ÏµúÏÜåÍ∞í, 1ÏÇ¨Î∂ÑÏúÑÏàò, Ï§ëÏïôÍ∞í, 3ÏÇ¨Î∂ÑÏúÑÏàò, ÏµúÎåÄÍ∞íÏùÑ Î∞ïÏä§Î°ú ÌëúÌòÑ

**Í∏∞Î≥∏ Î¨∏Î≤ï**:
```python
px.box(
    data_frame=df,
    x='Ïπ¥ÌÖåÍ≥†Î¶¨',
    y='Í∞í',
    title='Î∂ÑÌè¨ ÎπÑÍµê'
)
```

````python
# Î∂ÄÏÑúÎ≥Ñ Ïó∞Î¥â Îç∞Ïù¥ÌÑ∞
salary_data = pd.DataFrame({
    'Î∂ÄÏÑú': ['ÏòÅÏóÖ']*50 + ['Í∞úÎ∞ú']*50 + ['ÎßàÏºÄÌåÖ']*50,
    'Ïó∞Î¥â': np.concatenate([
        np.random.normal(4500, 800, 50),  # ÏòÅÏóÖ: ÌèâÍ∑† 4500ÎßåÏõê
        np.random.normal(5500, 1000, 50),  # Í∞úÎ∞ú: ÌèâÍ∑† 5500ÎßåÏõê
        np.random.normal(4000, 600, 50)   # ÎßàÏºÄÌåÖ: ÌèâÍ∑† 4000ÎßåÏõê
    ])
})

# Î∞ïÏä§ÌîåÎ°Ø
fig = px.box(
    salary_data,
    x='Î∂ÄÏÑú',
    y='Ïó∞Î¥â',
    title='Î∂ÄÏÑúÎ≥Ñ Ïó∞Î¥â Î∂ÑÌè¨',
    points='all'  # Î™®Îì† Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ ÌëúÏãú
)

fig.show()
````

üí° **Ïã§Î¨¥ ÏòàÏãú**: Î∞ïÏä§ÌîåÎ°ØÏùÄ Ïù¥ÏÉÅÏπò(outlier)Î•º Ï†êÏúºÎ°ú ÌëúÏãúÌïòÎØÄÎ°ú, Í∏âÏó¨ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú ÎπÑÏ†ïÏÉÅÏ†ÅÏúºÎ°ú ÎÜíÍ±∞ÎÇò ÎÇÆÏùÄ Í∞íÏùÑ ÌïúÎààÏóê Î∞úÍ≤¨Ìï† Ïàò ÏûàÏäµÎãàÎã§.

### 2.1 hover_data, color, size - Îã§Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞ ÌëúÌòÑ

**Ïù∏ÌÑ∞ÎûôÌã∞Î∏å ÏöîÏÜå**:
- `hover_data`: ÎßàÏö∞Ïä§ Ïò§Î≤Ñ Ïãú Ï∂îÍ∞Ä Ï†ïÎ≥¥ ÌëúÏãú
- `color`: Ïπ¥ÌÖåÍ≥†Î¶¨ ÎòêÎäî Ïó∞ÏÜçÍ∞íÏúºÎ°ú ÏÉâÏÉÅ Îß§Ìïë
- `size`: Í∞íÏùò ÌÅ¨Í∏∞Î•º Ï†ê ÌÅ¨Í∏∞Î°ú ÌëúÌòÑ

Ïù¥ ÏÑ∏ Í∞ÄÏßÄÎ•º Ï°∞Ìï©ÌïòÎ©¥ **4Ï∞®Ïõê Ïù¥ÏÉÅÏùò Îç∞Ïù¥ÌÑ∞**Î•º 2D Ï∞®Ìä∏Ïóê ÌëúÌòÑ Í∞ÄÎä•!

````python
# Îã§Ï∞®Ïõê Îß§Ïû• Îç∞Ïù¥ÌÑ∞
store_data = pd.DataFrame({
    'Îß§Ïû•Î™Ö': ['Í∞ïÎÇ®Ï†ê', 'ÌôçÎåÄÏ†ê', 'Ïã†Ï¥åÏ†ê', 'Ïû†Ïã§Ï†ê', 'Í±¥ÎåÄÏ†ê', 'ÏàòÏõêÏ†ê'],
    'Ïõî_Îß§Ï∂ú': [5000, 4200, 3800, 4500, 3500, 3000],  # ÎßåÏõê
    'Î∞©Î¨∏Í∞ù': [1200, 980, 850, 1050, 800, 650],
    'ÏßÅÏõêÏàò': [12, 10, 8, 11, 7, 6],
    'ÏßÄÏó≠': ['ÏÑúÏö∏', 'ÏÑúÏö∏', 'ÏÑúÏö∏', 'ÏÑúÏö∏', 'ÏÑúÏö∏', 'Í≤ΩÍ∏∞'],
    'ÌèâÏ†ê': [4.5, 4.3, 4.1, 4.4, 4.0, 3.9]
})

# Îã§Ï∞®Ïõê ÏÇ∞Ï†êÎèÑ
fig = px.scatter(
    store_data,
    x='Î∞©Î¨∏Í∞ù',
    y='Ïõî_Îß§Ï∂ú',
    size='ÏßÅÏõêÏàò',  # Ï†ê ÌÅ¨Í∏∞ = ÏßÅÏõêÏàò
    color='ÌèâÏ†ê',  # ÏÉâÏÉÅ = ÌèâÏ†ê (Ïó∞ÏÜçÍ∞í)
    hover_data=['Îß§Ïû•Î™Ö', 'ÏßÄÏó≠'],  # ÎßàÏö∞Ïä§ Ïò§Î≤Ñ Ïãú Ï∂îÍ∞Ä Ï†ïÎ≥¥
    title='Îß§Ïû• ÏÑ±Í≥º Î∂ÑÏÑù (4Ï∞®Ïõê)',
    labels={'Î∞©Î¨∏Í∞ù': 'ÏõîÍ∞Ñ Î∞©Î¨∏Í∞ù Ïàò', 'Ïõî_Îß§Ï∂ú': 'Ïõî Îß§Ï∂ú (ÎßåÏõê)'},
    size_max=30  # ÏµúÎåÄ Ï†ê ÌÅ¨Í∏∞
)

fig.show()
````

üí° **Ïã§Î¨¥ ÏòàÏãú**: ÏúÑ Ï∞®Ìä∏Îäî xÏ∂ï(Î∞©Î¨∏Í∞ù), yÏ∂ï(Îß§Ï∂ú), Ï†ê ÌÅ¨Í∏∞(ÏßÅÏõêÏàò), ÏÉâÏÉÅ(ÌèâÏ†ê) Ï¥ù 4Í∞ú Î≥ÄÏàòÎ•º ÎèôÏãúÏóê ÌëúÌòÑÌï©ÎãàÎã§. ÎßàÏö∞Ïä§Î•º Ïò¨Î¶¨Î©¥ Îß§Ïû•Î™ÖÍ≥º ÏßÄÏó≠ÎèÑ ÌôïÏù∏ Í∞ÄÎä•!

### 2.2 facet_row, facet_col - ÏÑúÎ∏åÌîåÎ°Ø

**ÏÑúÎ∏åÌîåÎ°Ø (Facets)**: Ïπ¥ÌÖåÍ≥†Î¶¨Î≥ÑÎ°ú Ïó¨Îü¨ Ï∞®Ìä∏Î•º ÏûêÎèô ÏÉùÏÑ±

**Í∏∞Î≥∏ Î¨∏Î≤ï**:
```python

    df,
    x='Í∞í1',
    y='Í∞í2',
    facet_col='Ïπ¥ÌÖåÍ≥†Î¶¨1',  # Ïó¥ Î∞©Ìñ• Î∂ÑÌï†
    facet_row='Ïπ¥ÌÖåÍ≥†Î¶¨2'   # Ìñâ Î∞©Ìñ• Î∂ÑÌï†
```
