### Q2. ì—¬ëŸ¬ ë°˜í™˜ê°’ - í†µê³„ ê³„ì‚° â­â­

**ë¬¸ì œ**: ìˆ«ì ë¦¬ìŠ¤íŠ¸ì˜ í†µê³„ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
- í•¨ìˆ˜ëª…: `calculate_statistics`
- ë§¤ê°œë³€ìˆ˜: `numbers` (ìˆ«ì ë¦¬ìŠ¤íŠ¸)
- ë°˜í™˜ê°’: ìµœì†Œê°’, ìµœëŒ€ê°’, í‰ê·  (íŠœí”Œ)
- í‰ê· ì€ ì†Œìˆ˜ì  2ìë¦¬ê¹Œì§€

````python
# ì •ë‹µ
def calculate_statistics(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    avg = round(sum(numbers) / len(numbers), 2)
    return min_val, max_val, avg

# í…ŒìŠ¤íŠ¸
min_val, max_val, avg = calculate_statistics([10, 20, 30, 40, 50])
print(f"ìµœì†Œ: {min_val}, ìµœëŒ€: {max_val}, í‰ê· : {avg}")
# ìµœì†Œ: 10, ìµœëŒ€: 50, í‰ê· : 30.0
````

### Q3. ê¸°ë³¸ê°’ ë§¤ê°œë³€ìˆ˜ - í• ì¸ ê³„ì‚° â­â­

**ë¬¸ì œ**: í• ì¸ëœ ê°€ê²©ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
- í•¨ìˆ˜ëª…: `apply_discount`
- ë§¤ê°œë³€ìˆ˜:
  - `price` (ì›ê°€)
  - `discount_rate` (í• ì¸ìœ¨, ê¸°ë³¸ê°’ 0.1 = 10%)
  - `min_price` (ìµœì†Œ ê°€ê²©, ê¸°ë³¸ê°’ 0)
- ë°˜í™˜ê°’: í• ì¸ëœ ê°€ê²© (ì •ìˆ˜)
- í• ì¸ í›„ ê°€ê²©ì´ ìµœì†Œ ê°€ê²©ë³´ë‹¤ ë‚®ìœ¼ë©´ ìµœì†Œ ê°€ê²© ë°˜í™˜

````python
# ì •ë‹µ
def apply_discount(price, discount_rate=0.1, min_price=0):
    discounted = price * (1 - discount_rate)
    final_price = max(discounted, min_price)  # ìµœì†Œ ê°€ê²© ì ìš©
    return int(final_price)

# í…ŒìŠ¤íŠ¸
print(apply_discount(10000))                    # 9000 (10% í• ì¸)
print(apply_discount(10000, 0.2))               # 8000 (20% í• ì¸)
print(apply_discount(10000, 0.5, min_price=6000))  # 6000 (ìµœì†Œê°€ ì ìš©)
````

- ê¸°ë³¸ê°’ ë§¤ê°œë³€ìˆ˜ëŠ” **ë§ˆì§€ë§‰**ì— ìœ„ì¹˜ (`price`ëŠ” í•„ìˆ˜, ë‚˜ë¨¸ì§€ëŠ” ì„ íƒ)

### Q4. Docstring - ë¬¸ì„œí™”ëœ í•¨ìˆ˜ â­â­

**ë¬¸ì œ**: BMI(ì²´ì§ˆëŸ‰ì§€ìˆ˜)ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ê³  Docstringì„ ì¶”ê°€í•˜ì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
- í•¨ìˆ˜ëª…: `calculate_bmi`

- ê³µì‹: `BMI = ì²´ì¤‘(kg) / (ì‹ ì¥(m))Â²`

````python
# ì •ë‹µ
def calculate_bmi(weight, height):
    """
    BMI(ì²´ì§ˆëŸ‰ì§€ìˆ˜)ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
    
    Args:
        weight (float): ì²´ì¤‘ (kg)
        height (float): ì‹ ì¥ (cm)
    
    Returns:
        float: BMI ê°’ (ì†Œìˆ˜ì  1ìë¦¬)
    
    Examples:
        >>> calculate_bmi(70, 175)
        22.9
        >>> calculate_bmi(60, 160)
        23.4
    """
    height_m = height / 100  # cm â†’ m ë³€í™˜
    bmi = weight / (height_m ** 2)
    return round(bmi, 1)

# í…ŒìŠ¤íŠ¸
bmi = calculate_bmi(70, 175)
print(f"BMI: {bmi}")  # 22.9

# Docstring í™•ì¸
help(calculate_bmi)
````

### Q6. *args - í‰ê·  ê³„ì‚° â­â­â­

**ìš”êµ¬ì‚¬í•­**:
- í•¨ìˆ˜ëª…: `calculate_average`

- ì¸ìê°€ ì—†ìœ¼ë©´ 0 ë°˜í™˜
- í‰ê· ì€ ì†Œìˆ˜ì  2ìë¦¬

````python
# ì •ë‹µ
def calculate_average(*args):
    if len(args) == 0:
        return 0
    return round(sum(args) / len(args), 2)

# í…ŒìŠ¤íŠ¸
print(calculate_average(10, 20, 30))       # 20.0
print(calculate_average(5, 10, 15, 20))    # 12.5
print(calculate_average())                 # 0
````

### Q8. í‚¤ì›Œë“œ ì „ìš© ë§¤ê°œë³€ìˆ˜ - ë°ì´í„° í•„í„° â­â­â­â­

**ë¬¸ì œ**: ë°ì´í„°ë¥¼ ì •ì œí•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
- í•¨ìˆ˜ëª…: `clean_data`
- ì¼ë°˜ ë§¤ê°œë³€ìˆ˜: `data` (ë¦¬ìŠ¤íŠ¸)
- í‚¤ì›Œë“œ ì „ìš© ë§¤ê°œë³€ìˆ˜ (`*` ì‚¬ìš©):
  - `remove_duplicates` (ê¸°ë³¸ê°’: False)
  - `remove_nulls` (ê¸°ë³¸ê°’: False)
  - `sort_result` (ê¸°ë³¸ê°’: False)
- None ê°’ì€ nullë¡œ ê°„ì£¼

````python
# ì •ë‹µ
def clean_data(data, *, remove_duplicates=False, remove_nulls=False, sort_result=False):
    result = data.copy()  # ì›ë³¸ ë³´ì¡´
    
    # 1. null ì œê±°
    if remove_nulls:
        result = [x for x in result if x is not None]
    
    # 2. ì¤‘ë³µ ì œê±° (ìˆœì„œ ë³´ì¡´)
    if remove_duplicates:
        seen = set()
        unique = []
        for x in result:
            if x not in seen:
                unique.append(x)
                seen.add(x)
        result = unique
    
    # 3. ì •ë ¬
    if sort_result:
        result = sorted(result)
    
    return result

# í…ŒìŠ¤íŠ¸
data = [3, 1, None, 2, 3, None, 1, 4]
print(clean_data(data))  # [3, 1, None, 2, 3, None, 1, 4]
print(clean_data(data, remove_duplicates=True))  # [3, 1, None, 2, 4]
print(clean_data(data, remove_nulls=True))  # [3, 1, 2, 3, 1, 4]
print(clean_data(data, remove_duplicates=True, remove_nulls=True, sort_result=True))  # [1, 2, 3, 4]
````

**ğŸ’¡ í•´ì„¤**:
- `*` ì´í›„ì˜ ë§¤ê°œë³€ìˆ˜ëŠ” **í‚¤ì›Œë“œë¡œë§Œ** ì „ë‹¬ ê°€ëŠ¥
- `data.copy()`: ì›ë³¸ ë°ì´í„° ë³´ì¡´ (ì¤‘ìš”!)
- ì¤‘ë³µ ì œê±° ì‹œ ìˆœì„œ ë³´ì¡´: `set()`ì€ ìˆœì„œ ë³´ì¥ ì•ˆ ë¨ â†’ ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜ ì‚¬ìš©
- ì²˜ë¦¬ ìˆœì„œ: null ì œê±° â†’ ì¤‘ë³µ ì œê±° â†’ ì •ë ¬

**ëŒ€ì•ˆ - ì¤‘ë³µ ì œê±°**:

### Q9. ëŒë‹¤ - ì œí’ˆ ì²˜ë¦¬ â­â­â­â­

**ë¬¸ì œ**: ëŒë‹¤ í•¨ìˆ˜ë¥¼ í™œìš©í•˜ì—¬ ì œí’ˆ ë°ì´í„°ë¥¼ ì²˜ë¦¬í•˜ì„¸ìš”.

**ìš”êµ¬ì‚¬í•­**:
1. `map()`ê³¼ ëŒë‹¤: ëª¨ë“  ì œí’ˆì— 10% í• ì¸ ì ìš©
2. `filter()`ì™€ ëŒë‹¤: 10ë§Œì› ì´ìƒ ì œí’ˆë§Œ í•„í„°ë§
3. `sorted()`ì™€ ëŒë‹¤: ê°€ê²© ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
4. ìœ„ 3ê°€ì§€ë¥¼ ì¡°í•©í•˜ì—¬ ì²˜ë¦¬í•˜ëŠ” í•¨ìˆ˜ `process_products()` ì‘ì„±

````python
products = [
    {'name': 'ë…¸íŠ¸ë¶', 'price': 1200000, 'category': 'ì „ìì œí’ˆ'},
    {'name': 'ë§ˆìš°ìŠ¤', 'price': 30000, 'category': 'ì „ìì œí’ˆ'},
    {'name': 'ì±…ìƒ', 'price': 150000, 'category': 'ê°€êµ¬'},
    {'name': 'ì˜ì', 'price': 200000, 'category': 'ê°€êµ¬'},
    {'name': 'ëª¨ë‹ˆí„°', 'price': 300000, 'category': 'ì „ìì œí’ˆ'},
]

# ì •ë‹µ
def process_products(products):
    # 1. 10% í• ì¸ ì ìš© (map)
    discounted = list(map(
        lambda p: {**p, 'price': int(p['price'] * 0.9)},
        products
    ))
    
    # 2. 10ë§Œì› ì´ìƒë§Œ í•„í„°ë§ (filter)
    filtered = list(filter(
        lambda p: p['price'] >= 100000,
        discounted
    ))
    
    # 3. ê°€ê²© ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬ (sorted)
    sorted_products = sorted(
        filtered,
        key=lambda p: p['price'],
        reverse=True
    )
    
    return sorted_products

# í…ŒìŠ¤íŠ¸
result = process_products(products)
for p in result:
    print(p)

# ê¸°ëŒ€ ê²°ê³¼:
# {'name': 'ë…¸íŠ¸ë¶', 'price': 1080000, 'category': 'ì „ìì œí’ˆ'}
# {'name': 'ëª¨ë‹ˆí„°', 'price': 270000, 'category': 'ì „ìì œí’ˆ'}
# {'name': 'ì˜ì', 'price': 180000, 'category': 'ê°€êµ¬'}
# {'name': 'ì±…ìƒ', 'price': 135000, 'category': 'ê°€êµ¬'}
````

**ğŸ’¡ í•´ì„¤**:

**1. map() - ë”•ì…”ë„ˆë¦¬ ì—…ë°ì´íŠ¸**:
```python
lambda p: {**p, 'price': int(p['price'] * 0.9)}
```
- `{**p, ...}`: ë”•ì…”ë„ˆë¦¬ ì–¸íŒ¨í‚¹ìœ¼ë¡œ ê¸°ì¡´ í‚¤ ìœ ì§€
- `'price': ìƒˆê°’`: price í‚¤ë§Œ ë®ì–´ì“°ê¸°
- `int()`: ì •ìˆ˜ ë³€í™˜

**2. filter() - ì¡°ê±´ í•„í„°ë§**:
```python
lambda p: p['price'] >= 100000
```
- True ë°˜í™˜í•˜ëŠ” í•­ëª©ë§Œ ìœ ì§€

**3. sorted() - ì •ë ¬ í‚¤ ì§€ì •**:
```python

```
- `key`: ì •ë ¬ ê¸°ì¤€ (price ê°’)
- `reverse=True`: ë‚´ë¦¼ì°¨ìˆœ

````python
import time
import functools

# ì •ë‹µ
def timer(func):
    """í•¨ìˆ˜ ì‹¤í–‰ ì‹œê°„ì„ ì¸¡ì •í•˜ëŠ” ë°ì½”ë ˆì´í„°"""
    @functools.wraps(func)  # ì›ë³¸ í•¨ìˆ˜ì˜ ë©”íƒ€ë°ì´í„° ë³´ì¡´
    def wrapper(*args, **kwargs):
        start = time.time()  # ì‹œì‘ ì‹œê°„
        result = func(*args, **kwargs)  # ì›ë³¸ í•¨ìˆ˜ ì‹¤í–‰
        end = time.time()  # ì¢…ë£Œ ì‹œê°„
        elapsed = (end - start) * 1000  # ì´ˆ â†’ ë°€ë¦¬ì´ˆ
        print(f"{func.__name__}() ì‹¤í–‰ ì‹œê°„: {elapsed:.2f} ms")
        return result
    return wrapper

# í…ŒìŠ¤íŠ¸
@timer
def slow_function():
    """ëŠë¦° í•¨ìˆ˜ ì‹œë®¬ë ˆì´ì…˜"""
    time.sleep(0.5)
    return "ì™„ë£Œ"

result = slow_function()
# ì¶œë ¥: slow_function() ì‹¤í–‰ ì‹œê°„: 500.xx ms
print(f"ê²°ê³¼: {result}")

# ë©”íƒ€ë°ì´í„° í™•ì¸
print(f"\ní•¨ìˆ˜ëª…: {slow_function.__name__}")
print(f"Docstring: {slow_function.__doc__}")
````
