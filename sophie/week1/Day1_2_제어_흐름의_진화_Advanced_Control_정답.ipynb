## Q8. zip - ì„±ì í‘œ ìƒì„± â­â­â­â­

## Q9. for-else - ê²€ìƒ‰ ê¸°ëŠ¥ â­â­â­â­

```python
users = [
    {'name': 'ê¹€ì² ìˆ˜', 'email': 'kim@example.com'},
    {'name': 'ì´ì˜í¬', 'email': 'lee@example.com'},
    {'name': 'ë°•ë¯¼ìˆ˜', 'email': 'park@example.com'}
]

def find_user_by_email(users, email):
    for user in users:
        if user['email'] == email:
            return user
    else:
        return None

# í…ŒìŠ¤íŠ¸
result = find_user_by_email(users, 'lee@example.com')
print(result)

result = find_user_by_email(users, 'notfound@example.com')
print(result)
```
