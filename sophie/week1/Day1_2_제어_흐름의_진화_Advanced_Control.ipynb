íŒ¨í„´ ë§¤ì¹­ í›„ ì¶”ê°€ ì¡°ê±´ì„ ê²€ì‚¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# [ì˜ˆì œ 19] ê°€ë“œ ì¡°ê±´
def classify_number(num):
    match num:
        case n if n < 0:
            return "ìŒìˆ˜"
        case 0:
            return "ì˜"
        case n if n > 0 and n < 10:
            return "í•œ ìë¦¬ ì–‘ìˆ˜"
        case n if n >= 10:
            return "ë‘ ìë¦¬ ì´ìƒ ì–‘ìˆ˜"

print(classify_number(-5))   # ìŒìˆ˜
print(classify_number(0))    # ì˜
print(classify_number(7))    # í•œ ìë¦¬ ì–‘ìˆ˜
print(classify_number(42))   # ë‘ ìë¦¬ ì´ìƒ ì–‘ìˆ˜
````

````python
# [ì˜ˆì œ 27] ì†Œìˆ˜ íŒë³„
def is_prime(n):
    """nì´ ì†Œìˆ˜ì¸ì§€ íŒë³„"""
    if n < 2:
        return False
    
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False  # ì•½ìˆ˜ ë°œê²¬, ì†Œìˆ˜ ì•„ë‹˜
    else:
        return True  # ì•½ìˆ˜ ì—†ìŒ, ì†Œìˆ˜

print(f"17ì€ ì†Œìˆ˜? {is_prime(17)}")
print(f"18ì€ ì†Œìˆ˜? {is_prime(18)}")
````
