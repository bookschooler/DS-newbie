## 2.1 ì§‘í•© ì—°ì‚°

### ìˆ˜í•™ ì§‘í•© ì—°ì‚°ì„ Pythonìœ¼ë¡œ!

````python
# ë‘ ê°œì˜ ì§‘í•©
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(f"A = {A}")
print(f"B = {B}")
````

### 1) í•©ì§‘í•© (Union) - A âˆª B

````python
# ë°©ë²• 1: | ì—°ì‚°ì
union1 = A | B
print(f"A | B = {union1}")

# ë°©ë²• 2: union() ë©”ì†Œë“œ
union2 = A.union(B)
print(f"A.union(B) = {union2}")
````

### 2) êµì§‘í•© (Intersection) - A âˆ© B

````python
# ë°©ë²• 1: & ì—°ì‚°ì
inter1 = A & B
print(f"A & B = {inter1}")

# ë°©ë²• 2: intersection() ë©”ì†Œë“œ
inter2 = A.intersection(B)
print(f"A.intersection(B) = {inter2}")
````

### 3) ì°¨ì§‘í•© (Difference) - A - B

````python
# Aì—ëŠ” ìˆì§€ë§Œ Bì—ëŠ” ì—†ëŠ” ìš”ì†Œ
diff1 = A - B
print(f"A - B = {diff1}")

# Bì—ëŠ” ìˆì§€ë§Œ Aì—ëŠ” ì—†ëŠ” ìš”ì†Œ
diff2 = B - A
print(f"B - A = {diff2}")
````

### 4) ëŒ€ì¹­ ì°¨ì§‘í•© (Symmetric Difference) - A â–³ B

````python
# ë‘˜ ì¤‘ í•˜ë‚˜ì—ë§Œ ìˆëŠ” ìš”ì†Œ (êµì§‘í•©ì˜ ë°˜ëŒ€)
sym_diff1 = A ^ B
print(f"A ^ B = {sym_diff1}")

# ë™ì¼í•œ ê²°ê³¼
sym_diff2 = A.symmetric_difference(B)
print(f"A.symmetric_difference(B) = {sym_diff2}")
````

### ì§‘í•© ì—°ì‚° ìš”ì•½í‘œ

| ì—°ì‚° | ê¸°í˜¸ | ë©”ì†Œë“œ | ì˜ë¯¸ |
|------|------|--------|------|
| í•©ì§‘í•© | `A \| B` | `A.union(B)` | A ë˜ëŠ” B |
| êµì§‘í•© | `A & B` | `A.intersection(B)` | A ê·¸ë¦¬ê³  B |
| ì°¨ì§‘í•© | `A - B` | `A.difference(B)` | Aì´ì§€ë§Œ B ì•„ë‹˜ |
| ëŒ€ì¹­ì°¨ì§‘í•© | `A ^ B` | `A.symmetric_difference(B)` | A ë˜ëŠ” Bì´ì§€ë§Œ ë‘˜ ë‹¤ëŠ” ì•„ë‹˜ |

## 2.4 frozenset (ë¶ˆë³€ ì§‘í•©)

### ğŸ’¡ ì‹¤ë¬´ ì˜ˆì‹œ: ê¶Œí•œ ê·¸ë£¹ ê´€ë¦¬

````python
# ê° ì—­í• ë³„ ê¶Œí•œ (ë³€ê²½ë˜ë©´ ì•ˆ ë¨)
permissions = {
    "admin": frozenset(["read", "write", "delete", "manage"]),
    "editor": frozenset(["read", "write"]),
    "viewer": frozenset(["read"])
}

# ì‚¬ìš©ì ì—­í•  í™•ì¸
def check_permission(role, action):
    if action in permissions.get(role, frozenset()):
        return f"âœ… {role}ì€(ëŠ”) {action} ê¶Œí•œì´ ìˆìŠµë‹ˆë‹¤."
    else:
        return f"âŒ {role}ì€(ëŠ”) {action} ê¶Œí•œì´ ì—†ìŠµë‹ˆë‹¤."

print(check_permission("admin", "delete"))
print(check_permission("viewer", "delete"))
print(check_permission("editor", "write"))
````

````python
# A/B í…ŒìŠ¤íŠ¸ ì°¸ì—¬ì
group_a = {"user001", "user002", "user003", "user004", "user005", "user006"}
group_b = {"user007", "user008", "user009", "user010", "user011", "user012"}

# êµ¬ë§¤ ì „í™˜ ì‚¬ìš©ì
converted_a = {"user001", "user003", "user005"}  # Aê·¸ë£¹ ì „í™˜ìœ¨ 50%
converted_b = {"user007", "user008", "user009", "user010"}  # Bê·¸ë£¹ ì „í™˜ìœ¨ 66%

print("=== A/B í…ŒìŠ¤íŠ¸ ê²°ê³¼ ===")

# 1. ì „í™˜ìœ¨ ê³„ì‚°
conv_rate_a = len(converted_a) / len(group_a) * 100
conv_rate_b = len(converted_b) / len(group_b) * 100

print(f"Aê·¸ë£¹ ì „í™˜ìœ¨: {conv_rate_a:.1f}%")
print(f"Bê·¸ë£¹ ì „í™˜ìœ¨: {conv_rate_b:.1f}%")
````

````python

# 2. ë¯¸ì „í™˜ ì‚¬ìš©ì (ë¦¬íƒ€ê²ŸíŒ… ëŒ€ìƒ)
non_converted_a = group_a - converted_a
non_converted_b = group_b - converted_b

print(f"\nAê·¸ë£¹ ë¯¸ì „í™˜: {non_converted_a}")
print(f"Bê·¸ë£¹ ë¯¸ì „í™˜: {non_converted_b}")
````

````python

# 3. ì „ì²´ ì°¸ì—¬ì & ì „í™˜ì
all_users = group_a | group_b
all_converted = converted_a | converted_b

print(f"\nì´ ì°¸ì—¬ì: {len(all_users)}ëª…")
print(f"ì´ ì „í™˜ì: {len(all_converted)}ëª…")
print(f"ì „ì²´ ì „í™˜ìœ¨: {len(all_converted)/len(all_users)*100:.1f}%")
````

````python

# 4. ë°ì´í„° ë¬´ê²°ì„± ê²€ì¦
overlap = group_a & group_b
if overlap:
    print(f"\nâš ï¸ ê²½ê³ : A/B ê·¸ë£¹ ì¤‘ë³µ ì‚¬ìš©ì ë°œê²¬: {overlap}")
else:
    print("\nâœ… ê·¸ë£¹ ë¶„ë¦¬ ì •ìƒ")
````
