````python
class Encoder(nn.Module):
    """
    Encoder: ì…ë ¥ ì‹œí€€ìŠ¤ë¥¼ Context Vectorë¡œ ë³€í™˜
    
    - Embedding: í† í° ì¸ë±ìŠ¤ -> ë°€ì§‘ ë²¡í„°
    - LSTM: ì‹œí€€ìŠ¤ ì²˜ë¦¬ ë° ë¬¸ë§¥ ì¸ì½”ë”©
    """
    def __init__(self, vocab_size, embed_dim, hidden_dim, n_layers=1, dropout=0.1):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(
            embed_dim, 
            hidden_dim, 
            num_layers=n_layers,
            dropout=dropout if n_layers > 1 else 0,
            batch_first=True
        )
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, src):
        """
        Args:
            src: ì…ë ¥ ì‹œí€€ìŠ¤ (batch, seq_len)
        Returns:
            outputs: ëª¨ë“  ì‹œì ì˜ hidden states (batch, seq_len, hidden_dim)
            hidden: ë§ˆì§€ë§‰ hidden state (n_layers, batch, hidden_dim)
            cell: ë§ˆì§€ë§‰ cell state (n_layers, batch, hidden_dim)
        """
        # ì„ë² ë”© ë° ë“œë¡­ì•„ì›ƒ
        embedded = self.dropout(self.embedding(src))  # (batch, seq_len, embed_dim)
        
        # LSTM í†µê³¼
        outputs, (hidden, cell) = self.lstm(embedded)
        # outputs: (batch, seq_len, hidden_dim)
        # hidden: (n_layers, batch, hidden_dim)
        # cell: (n_layers, batch, hidden_dim)
        
        return outputs, hidden, cell

# í…ŒìŠ¤íŠ¸
encoder = Encoder(vocab_size=100, embed_dim=32, hidden_dim=64)
dummy_src = torch.randint(0, 100, (4, 10))  # ë°°ì¹˜ 4, ê¸¸ì´ 10
enc_outputs, enc_hidden, enc_cell = encoder(dummy_src)

print(f"ì…ë ¥ shape: {dummy_src.shape}")
print(f"Encoder outputs shape: {enc_outputs.shape}")
print(f"Encoder hidden shape: {enc_hidden.shape}")
print(f"Encoder cell shape: {enc_cell.shape}")
````

````python
class Decoder(nn.Module):
    """
    Decoder: Context Vectorë¡œë¶€í„° ì¶œë ¥ ì‹œí€€ìŠ¤ ìƒì„±
    
    - í•œ ë²ˆì— í•˜ë‚˜ì˜ í† í°ë§Œ ì²˜ë¦¬
    - ì´ì „ hidden stateë¥¼ ë°›ì•„ ë‹¤ìŒ í† í° ì˜ˆì¸¡
    """
    def __init__(self, vocab_size, embed_dim, hidden_dim, n_layers=1, dropout=0.1):
        super().__init__()
        self.vocab_size = vocab_size
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(
            embed_dim, 
            hidden_dim, 
            num_layers=n_layers,
            dropout=dropout if n_layers > 1 else 0,
            batch_first=True
        )
        self.fc_out = nn.Linear(hidden_dim, vocab_size)  # ì¶œë ¥ ì–´íœ˜ í¬ê¸°ë¡œ ë³€í™˜
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, input_token, hidden, cell):
        """
        Args:
            input_token: í˜„ì¬ ì…ë ¥ í† í° (batch,) ë˜ëŠ” (batch, 1)
            hidden: ì´ì „ hidden state (n_layers, batch, hidden_dim)
            cell: ì´ì „ cell state (n_layers, batch, hidden_dim)
        Returns:
            prediction: ë‹¤ìŒ í† í° ì˜ˆì¸¡ ë¡œì§“ (batch, vocab_size)
            hidden: ìƒˆë¡œìš´ hidden state
            cell: ìƒˆë¡œìš´ cell state
        """
        # ì…ë ¥ì´ 1Dë©´ 2Dë¡œ ë³€í™˜
        if input_token.dim() == 1:
            input_token = input_token.unsqueeze(1)  # (batch, 1)
        
        # ì„ë² ë”©
        embedded = self.dropout(self.embedding(input_token))  # (batch, 1, embed_dim)
        
        # LSTM (í•œ ìŠ¤í…ë§Œ ì²˜ë¦¬)
        output, (hidden, cell) = self.lstm(embedded, (hidden, cell))
        # output: (batch, 1, hidden_dim)
        
        # ì¶œë ¥ ì˜ˆì¸¡
        prediction = self.fc_out(output.squeeze(1))  # (batch, vocab_size)
        
        return prediction, hidden, cell

# í…ŒìŠ¤íŠ¸
decoder = Decoder(vocab_size=100, embed_dim=32, hidden_dim=64)
dummy_input = torch.randint(0, 100, (4,))  # ë°°ì¹˜ 4
dummy_hidden = torch.randn(1, 4, 64)
dummy_cell = torch.randn(1, 4, 64)

pred, new_hidden, new_cell = decoder(dummy_input, dummy_hidden, dummy_cell)
print(f"ì…ë ¥ í† í° shape: {dummy_input.shape}")
print(f"ì˜ˆì¸¡ output shape: {pred.shape}")
print(f"ìƒˆ hidden shape: {new_hidden.shape}")
````


````python
# Teacher Forcing ë¹„ìœ¨ì— ë”°ë¥¸ í•™ìŠµ ì˜ˆì‹œ
def training_step_with_teacher_forcing(decoder, target_seq, hidden, cell, teacher_forcing_ratio=0.5):
    """
    Teacher Forcingì„ ì ìš©í•œ í›ˆë ¨ ìŠ¤í… ì‹œë®¬ë ˆì´ì…˜
    """
    batch_size, seq_len = target_seq.shape
    vocab_size = decoder.vocab_size
    
    # ì¶œë ¥ ì €ì¥ìš©
    outputs = torch.zeros(batch_size, seq_len, vocab_size)
    
    # ì²« ì…ë ¥ì€ <SOS> í† í° (ì—¬ê¸°ì„œëŠ” 0ìœ¼ë¡œ ê°€ì •)
    input_token = torch.zeros(batch_size, dtype=torch.long)
    
    teacher_forcing_used = []
    
    for t in range(seq_len):
        # ë””ì½”ë” í˜¸ì¶œ
        prediction, hidden, cell = decoder(input_token, hidden, cell)
        outputs[:, t] = prediction
        
        # Teacher Forcing ê²°ì •
        use_teacher = random.random() < teacher_forcing_ratio
        teacher_forcing_used.append(use_teacher)
        
        if use_teacher:
            input_token = target_seq[:, t]  # ì •ë‹µ ì‚¬ìš©
        else:
            input_token = prediction.argmax(1)  # ì˜ˆì¸¡ ì‚¬ìš©
    
    return outputs, teacher_forcing_used

# í…ŒìŠ¤íŠ¸
target = torch.randint(0, 100, (4, 5))  # ë°°ì¹˜ 4, ê¸¸ì´ 5
_, tf_used = training_step_with_teacher_forcing(
    decoder, target, dummy_hidden, dummy_cell, teacher_forcing_ratio=0.7
)
print(f"Teacher Forcing ì‚¬ìš© ì—¬ë¶€: {tf_used}")
print(f"Teacher Forcing ë¹„ìœ¨: {sum(tf_used)/len(tf_used):.2%}")
````

## 4.2 Attentionì˜ ì•„ì´ë””ì–´

**í•µì‹¬ ì•„ì´ë””ì–´**: ë””ì½”ë”ê°€ ê° ì¶œë ¥ì„ ìƒì„±í•  ë•Œ, ì¸ì½”ë”ì˜ ëª¨ë“  hidden states ì¤‘ ê´€ë ¨ ìˆëŠ” ë¶€ë¶„ì— 'ì§‘ì¤‘'

### Attentionì˜ 3ìš”ì†Œ: Query, Key, Value

| ìš”ì†Œ | ì—­í•  | Seq2Seqì—ì„œì˜ ëŒ€ì‘ |
|------|------|-------------------|
| Query | "ë¬´ì—‡ì„ ì°¾ê³  ìˆëŠ”ê°€?" | ë””ì½”ë”ì˜ í˜„ì¬ hidden state |
| Key | "ì—¬ê¸°ì— ì´ëŸ° ì •ë³´ê°€ ìˆë‹¤" | ì¸ì½”ë”ì˜ ê° hidden state |
| Value | "ì‹¤ì œ ì •ë³´ ë‚´ìš©" | ì¸ì½”ë”ì˜ ê° hidden state |

````python
class Attention(nn.Module):
    """
    Attention ë©”ì»¤ë‹ˆì¦˜ (Bahdanau-style Additive Attention)
    
    Attention Score = v^T * tanh(W_h * h_enc + W_s * h_dec)
    """
    def __init__(self, hidden_dim):
        super().__init__()
        # Attention ìŠ¤ì½”ì–´ ê³„ì‚°ì„ ìœ„í•œ ë ˆì´ì–´ë“¤
        self.W_enc = nn.Linear(hidden_dim, hidden_dim, bias=False)  # ì¸ì½”ë” hidden
        self.W_dec = nn.Linear(hidden_dim, hidden_dim, bias=False)  # ë””ì½”ë” hidden
        self.v = nn.Linear(hidden_dim, 1, bias=False)  # ìŠ¤ì¹¼ë¼ ìŠ¤ì½”ì–´ ì¶œë ¥
        
    def forward(self, decoder_hidden, encoder_outputs, mask=None):
        """
        Args:
            decoder_hidden: ë””ì½”ë” í˜„ì¬ hidden (batch, hidden_dim)
            encoder_outputs: ì¸ì½”ë” ì „ì²´ ì¶œë ¥ (batch, src_len, hidden_dim)
            mask: íŒ¨ë”© ë§ˆìŠ¤í¬ (batch, src_len)
        Returns:
            context: ê°€ì¤‘í•©ëœ ì»¨í…ìŠ¤íŠ¸ ë²¡í„° (batch, hidden_dim)
            attention_weights: ì–´í…ì…˜ ê°€ì¤‘ì¹˜ (batch, src_len)
        """
        src_len = encoder_outputs.shape[1]
        
        # ë””ì½”ë” hiddenì„ src_len ë§Œí¼ í™•ì¥
        decoder_hidden = decoder_hidden.unsqueeze(1).repeat(1, src_len, 1)
        # (batch, src_len, hidden_dim)
        
        # Attention ìŠ¤ì½”ì–´ ê³„ì‚°
        energy = torch.tanh(
            self.W_enc(encoder_outputs) + self.W_dec(decoder_hidden)
        )  # (batch, src_len, hidden_dim)
        
        attention_scores = self.v(energy).squeeze(-1)  # (batch, src_len)
        
        # ë§ˆìŠ¤í‚¹ (íŒ¨ë”© ìœ„ì¹˜ëŠ” -inf)
        if mask is not None:
            attention_scores = attention_scores.masked_fill(mask == 0, -1e10)
        
        # Softmaxë¡œ ê°€ì¤‘ì¹˜ ì •ê·œí™”
        attention_weights = F.softmax(attention_scores, dim=1)
        
        # ê°€ì¤‘í•©ìœ¼ë¡œ Context Vector ìƒì„±
        # (batch, 1, src_len) @ (batch, src_len, hidden_dim) -> (batch, 1, hidden_dim)
        context = torch.bmm(attention_weights.unsqueeze(1), encoder_outputs)
        context = context.squeeze(1)  # (batch, hidden_dim)
        
        return context, attention_weights

# í…ŒìŠ¤íŠ¸
attention = Attention(hidden_dim=64)
dec_hidden = torch.randn(4, 64)  # ë””ì½”ë” hidden
enc_outputs = torch.randn(4, 10, 64)  # ì¸ì½”ë” ì¶œë ¥ (ê¸¸ì´ 10)

context, weights = attention(dec_hidden, enc_outputs)
print(f"Context Vector shape: {context.shape}")
print(f"Attention Weights shape: {weights.shape}")
print(f"Attention Weights í•©: {weights.sum(dim=1)}")
````

````python
class AttentionDecoder(nn.Module):
    """
    Attentionì´ ì ìš©ëœ Decoder
    
    ë§¤ ìŠ¤í…ë§ˆë‹¤ ì¸ì½”ë” ì¶œë ¥ì— ëŒ€í•œ Attentionì„ ê³„ì‚°í•˜ì—¬
    ë™ì ìœ¼ë¡œ í•„ìš”í•œ ì •ë³´ë¥¼ ê°€ì ¸ì˜´
    """
    def __init__(self, vocab_size, embed_dim, hidden_dim, n_layers=1, dropout=0.1):
        super().__init__()
        self.vocab_size = vocab_size
        self.hidden_dim = hidden_dim
        
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.attention = Attention(hidden_dim)
        
        # LSTM ì…ë ¥: ì„ë² ë”© + Context Vector
        self.lstm = nn.LSTM(
            embed_dim + hidden_dim,  # ì„ë² ë”© + context
            hidden_dim,
            num_layers=n_layers,
            dropout=dropout if n_layers > 1 else 0,
            batch_first=True
        )
        
        # ì¶œë ¥ ë ˆì´ì–´: hidden + context + embedding
        self.fc_out = nn.Linear(hidden_dim * 2 + embed_dim, vocab_size)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, input_token, hidden, cell, encoder_outputs, mask=None):
        """
        Args:
            input_token: í˜„ì¬ ì…ë ¥ í† í° (batch,)
            hidden: ì´ì „ hidden state (n_layers, batch, hidden_dim)
            cell: ì´ì „ cell state
            encoder_outputs: ì¸ì½”ë” ì „ì²´ ì¶œë ¥ (batch, src_len, hidden_dim)
            mask: íŒ¨ë”© ë§ˆìŠ¤í¬
        Returns:
            prediction: í† í° ì˜ˆì¸¡ ë¡œì§“ (batch, vocab_size)
            hidden: ìƒˆ hidden state
            cell: ìƒˆ cell state
            attention_weights: ì–´í…ì…˜ ê°€ì¤‘ì¹˜
        """
        # ì„ë² ë”©
        embedded = self.dropout(self.embedding(input_token.unsqueeze(1)))  # (batch, 1, embed_dim)
        
        # Attention ê³„ì‚° (ë§ˆì§€ë§‰ ë ˆì´ì–´ì˜ hidden ì‚¬ìš©)
        context, attention_weights = self.attention(
            hidden[-1],  # (batch, hidden_dim)
            encoder_outputs,
            mask
        )
        
        # LSTM ì…ë ¥: ì„ë² ë”© + Context Vector
        lstm_input = torch.cat([embedded, context.unsqueeze(1)], dim=2)
        # (batch, 1, embed_dim + hidden_dim)
        
        # LSTM í†µê³¼
        output, (hidden, cell) = self.lstm(lstm_input, (hidden, cell))
        # output: (batch, 1, hidden_dim)
        
        # ì¶œë ¥ ì˜ˆì¸¡: hidden + context + embedding ê²°í•©
        output = output.squeeze(1)  # (batch, hidden_dim)
        embedded = embedded.squeeze(1)  # (batch, embed_dim)
        
        prediction = self.fc_out(torch.cat([output, context, embedded], dim=1))
        # (batch, vocab_size)
        
        return prediction, hidden, cell, attention_weights

# í…ŒìŠ¤íŠ¸
attn_decoder = AttentionDecoder(vocab_size=100, embed_dim=32, hidden_dim=64)
pred, _, _, attn_weights = attn_decoder(
    dummy_input, dummy_hidden, dummy_cell, enc_outputs
)
print(f"ì˜ˆì¸¡ shape: {pred.shape}")
print(f"Attention weights shape: {attn_weights.shape}")
````

# 6. ì™„ì„±ëœ Seq2Seq ëª¨ë¸

---

Encoderì™€ AttentionDecoderë¥¼ ê²°í•©í•œ ì „ì²´ Seq2Seq ëª¨ë¸ì…ë‹ˆë‹¤.

````python
class Seq2Seq(nn.Module):
    """
    ì™„ì„±ëœ Seq2Seq ëª¨ë¸ (Attention í¬í•¨)
    """
    def __init__(self, encoder, decoder, sos_idx, eos_idx, pad_idx):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.sos_idx = sos_idx  # ì‹œì‘ í† í°
        self.eos_idx = eos_idx  # ì¢…ë£Œ í† í°
        self.pad_idx = pad_idx  # íŒ¨ë”© í† í°
        
    def create_mask(self, src):
        """íŒ¨ë”© ìœ„ì¹˜ ë§ˆìŠ¤í¬ ìƒì„±"""
        return (src != self.pad_idx)
    
    def forward(self, src, tgt, teacher_forcing_ratio=0.5):
        """
        í›ˆë ¨ ì‹œ ì‚¬ìš© (Teacher Forcing ì ìš©)
        
        Args:
            src: ì…ë ¥ ì‹œí€€ìŠ¤ (batch, src_len)
            tgt: íƒ€ê²Ÿ ì‹œí€€ìŠ¤ (batch, tgt_len)
            teacher_forcing_ratio: Teacher Forcing ë¹„ìœ¨
        Returns:
            outputs: ì˜ˆì¸¡ ë¡œì§“ (batch, tgt_len, vocab_size)
        """
        batch_size, tgt_len = tgt.shape
        vocab_size = self.decoder.vocab_size
        
        # ì¶œë ¥ ì €ì¥
        outputs = torch.zeros(batch_size, tgt_len, vocab_size, device=src.device)
        
        # ì¸ì½”ë” ì‹¤í–‰
        encoder_outputs, hidden, cell = self.encoder(src)
        mask = self.create_mask(src)
        
        # ì²« ì…ë ¥ì€ <SOS>
        input_token = tgt[:, 0]
        
        # ë””ì½”ë” ìˆœì°¨ ì‹¤í–‰
        for t in range(1, tgt_len):
            prediction, hidden, cell, _ = self.decoder(
                input_token, hidden, cell, encoder_outputs, mask
            )
            outputs[:, t] = prediction
            
            # Teacher Forcing ì ìš©
            teacher_force = random.random() < teacher_forcing_ratio
            top1 = prediction.argmax(1)
            input_token = tgt[:, t] if teacher_force else top1
        
        return outputs
    
    def inference(self, src, max_len=50):
        """
        ì¶”ë¡  ì‹œ ì‚¬ìš© (Greedy Decoding)
        """
        self.eval()
        batch_size = src.shape[0]
        
        with torch.no_grad():
            # ì¸ì½”ë” ì‹¤í–‰
            encoder_outputs, hidden, cell = self.encoder(src)
            mask = self.create_mask(src)
            
            # ì‹œì‘ í† í°
            input_token = torch.full((batch_size,), self.sos_idx, device=src.device)
            
            outputs = []
            attentions = []
            
            for _ in range(max_len):
                prediction, hidden, cell, attn_weights = self.decoder(
                    input_token, hidden, cell, encoder_outputs, mask
                )
                
                top1 = prediction.argmax(1)
                outputs.append(top1)
                attentions.append(attn_weights)
                
                # EOS í† í°ì´ë©´ ì¢…ë£Œ
                if (top1 == self.eos_idx).all():
                    break
                    
                input_token = top1
            
            return torch.stack(outputs, dim=1), torch.stack(attentions, dim=1)
````

## 7.1 ë°ì´í„°ì…‹ ì¤€ë¹„

**ê³¼ì œ**: ë‹¤ì–‘í•œ ì˜ë¬¸ ë‚ ì§œ í˜•ì‹ì„ í‘œì¤€ ISO í˜•ì‹ìœ¼ë¡œ ë³€í™˜

````python
from torch.utils.data import Dataset, DataLoader

# íŠ¹ìˆ˜ í† í°
PAD_IDX = 0
SOS_IDX = 1
EOS_IDX = 2

# ì…ë ¥ ì–´íœ˜ (ë¬¸ì ë‹¨ìœ„)
INPUT_CHARS = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789 ,/"
OUTPUT_CHARS = "0123456789-"

# ì–´íœ˜ ì‚¬ì „ êµ¬ì¶•
input_vocab = {c: i + 3 for i, c in enumerate(INPUT_CHARS)}  # 3ë¶€í„° ì‹œì‘ (0,1,2ëŠ” íŠ¹ìˆ˜í† í°)
input_vocab['<pad>'] = PAD_IDX
input_vocab['<sos>'] = SOS_IDX
input_vocab['<eos>'] = EOS_IDX

output_vocab = {c: i + 3 for i, c in enumerate(OUTPUT_CHARS)}
output_vocab['<pad>'] = PAD_IDX
output_vocab['<sos>'] = SOS_IDX
output_vocab['<eos>'] = EOS_IDX

# ì—­ë°©í–¥ ì‚¬ì „
idx2input = {v: k for k, v in input_vocab.items()}
idx2output = {v: k for k, v in output_vocab.items()}

print(f"ì…ë ¥ ì–´íœ˜ í¬ê¸°: {len(input_vocab)}")
print(f"ì¶œë ¥ ì–´íœ˜ í¬ê¸°: {len(output_vocab)}")
````

````python
# ë‚ ì§œ ë°ì´í„° ìƒì„±
import datetime

MONTHS_FULL = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
MONTHS_SHORT = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

def generate_date_pair():
    """ëœë¤ ë‚ ì§œ ìŒ ìƒì„±"""
    year = random.randint(1900, 2099)
    month = random.randint(1, 12)
    day = random.randint(1, 28)  # ë‹¨ìˆœí™”ë¥¼ ìœ„í•´ 28ì¼ê¹Œì§€
    
    # ë‹¤ì–‘í•œ ì…ë ¥ í˜•ì‹
    formats = [
        f"{MONTHS_FULL[month-1]} {day}, {year}",     # January 15, 2025
        f"{day} {MONTHS_SHORT[month-1]} {year}",    # 15 Jan 2025
        f"{month:02d}/{day:02d}/{year}",            # 01/15/2025
        f"{MONTHS_SHORT[month-1]} {day}, {year}",   # Jan 15, 2025
    ]
    
    input_text = random.choice(formats)
    output_text = f"{year}-{month:02d}-{day:02d}"  # ISO í˜•ì‹
    
    return input_text, output_text

# ìƒ˜í”Œ ìƒì„±
for _ in range(5):
    inp, out = generate_date_pair()
    print(f"ì…ë ¥: {inp:25s} -> ì¶œë ¥: {out}")
````

````python
class DateDataset(Dataset):
    """ë‚ ì§œ ë³€í™˜ ë°ì´í„°ì…‹"""
    def __init__(self, n_samples=10000, max_src_len=25, max_tgt_len=12):
        self.data = []
        self.max_src_len = max_src_len
        self.max_tgt_len = max_tgt_len
        
        for _ in range(n_samples):
            input_text, output_text = generate_date_pair()
            
            # í† í°í™”
            src_tokens = [input_vocab.get(c, PAD_IDX) for c in input_text]
            tgt_tokens = [SOS_IDX] + [output_vocab.get(c, PAD_IDX) for c in output_text] + [EOS_IDX]
            
            # íŒ¨ë”©
            src_tokens = src_tokens[:max_src_len] + [PAD_IDX] * max(0, max_src_len - len(src_tokens))
            tgt_tokens = tgt_tokens[:max_tgt_len] + [PAD_IDX] * max(0, max_tgt_len - len(tgt_tokens))
            
            self.data.append((src_tokens, tgt_tokens))
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        src, tgt = self.data[idx]
        return torch.tensor(src), torch.tensor(tgt)

# ë°ì´í„°ì…‹ ë° ë°ì´í„°ë¡œë”
train_dataset = DateDataset(n_samples=10000)
val_dataset = DateDataset(n_samples=1000)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)

print(f"í›ˆë ¨ ìƒ˜í”Œ ìˆ˜: {len(train_dataset)}")
print(f"ê²€ì¦ ìƒ˜í”Œ ìˆ˜: {len(val_dataset)}")
````

## 7.2 ëª¨ë¸ ìƒì„± ë° í›ˆë ¨

````python
# í•˜ì´í¼íŒŒë¼ë¯¸í„°
EMBED_DIM = 64
HIDDEN_DIM = 128
N_LAYERS = 1
DROPOUT = 0.1

# ëª¨ë¸ ìƒì„±
encoder = Encoder(
    vocab_size=len(input_vocab),
    embed_dim=EMBED_DIM,
    hidden_dim=HIDDEN_DIM,
    n_layers=N_LAYERS,
    dropout=DROPOUT
)

decoder = AttentionDecoder(
    vocab_size=len(output_vocab),
    embed_dim=EMBED_DIM,
    hidden_dim=HIDDEN_DIM,
    n_layers=N_LAYERS,
    dropout=DROPOUT
)

model = Seq2Seq(
    encoder=encoder,
    decoder=decoder,
    sos_idx=SOS_IDX,
    eos_idx=EOS_IDX,
    pad_idx=PAD_IDX
).to(device)

# íŒŒë¼ë¯¸í„° ìˆ˜
total_params = sum(p.numel() for p in model.parameters())
print(f"ì´ íŒŒë¼ë¯¸í„° ìˆ˜: {total_params:,}")
````

````python
# í›ˆë ¨ ì„¤ì •
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

def train_epoch(model, loader, criterion, optimizer, teacher_forcing_ratio=0.5):
    model.train()
    total_loss = 0
    
    for src, tgt in loader:
        src, tgt = src.to(device), tgt.to(device)
        
        optimizer.zero_grad()
        
        # ìˆœì „íŒŒ
        outputs = model(src, tgt, teacher_forcing_ratio)
        
        # ì†ì‹¤ ê³„ì‚° (ì²« ë²ˆì§¸ í† í°(<SOS>) ì œì™¸)
        output_dim = outputs.shape[-1]
        outputs = outputs[:, 1:].reshape(-1, output_dim)
        tgt = tgt[:, 1:].reshape(-1)
        
        loss = criterion(outputs, tgt)
        
        # ì—­ì „íŒŒ
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        
        total_loss += loss.item()
    
    return total_loss / len(loader)

def evaluate(model, loader, criterion):
    model.eval()
    total_loss = 0
    
    with torch.no_grad():
        for src, tgt in loader:
            src, tgt = src.to(device), tgt.to(device)
            
            outputs = model(src, tgt, teacher_forcing_ratio=0)  # ì¶”ë¡  ëª¨ë“œ
            
            output_dim = outputs.shape[-1]
            outputs = outputs[:, 1:].reshape(-1, output_dim)
            tgt = tgt[:, 1:].reshape(-1)
            
            loss = criterion(outputs, tgt)
            total_loss += loss.item()
    
    return total_loss / len(loader)
````

````python
# í›ˆë ¨ ì‹¤í–‰
N_EPOCHS = 20
history = {'train_loss': [], 'val_loss': []}

print("í›ˆë ¨ ì‹œì‘!")
print("=" * 50)

for epoch in range(1, N_EPOCHS + 1):
    # Teacher Forcing ë¹„ìœ¨ ì ì§„ì  ê°ì†Œ
    tf_ratio = max(0.5 - epoch * 0.02, 0.1)
    
    train_loss = train_epoch(model, train_loader, criterion, optimizer, tf_ratio)
    val_loss = evaluate(model, val_loader, criterion)
    
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    
    if epoch % 5 == 0 or epoch == 1:
        print(f"Epoch {epoch:2d}/{N_EPOCHS} | TF ratio: {tf_ratio:.2f} | "
              f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

print("=" * 50)
print("í›ˆë ¨ ì™„ë£Œ!")
````

````python
def decode_sequence(tokens, idx2char):
    """í† í°ì„ ë¬¸ìì—´ë¡œ ë³€í™˜"""
    chars = []
    for t in tokens:
        t = t.item() if hasattr(t, 'item') else t
        if t == EOS_IDX:
            break
        if t not in [PAD_IDX, SOS_IDX, EOS_IDX]:
            chars.append(idx2char.get(t, '?'))
    return ''.join(chars)

def predict(model, input_text):
    """ë‹¨ì¼ ì…ë ¥ ì˜ˆì¸¡"""
    model.eval()
    
    # í† í°í™”
    src_tokens = [input_vocab.get(c, PAD_IDX) for c in input_text]
    src_tokens = src_tokens[:25] + [PAD_IDX] * max(0, 25 - len(src_tokens))
    src = torch.tensor([src_tokens], device=device)
    
    # ì¶”ë¡ 
    outputs, attentions = model.inference(src, max_len=12)
    
    # ë””ì½”ë”©
    prediction = decode_sequence(outputs[0], idx2output)
    
    return prediction, attentions[0]

# í…ŒìŠ¤íŠ¸
test_inputs = [
    "January 15, 2025",
    "15 Jan 2025",
    "01/15/2025",
    "March 7, 1999",
    "Dec 25, 2000",
    "07/04/1776",
]

print("ë‚ ì§œ ë³€í™˜ ê²°ê³¼")
print("=" * 60)
for inp in test_inputs:
    pred, _ = predict(model, inp)
    print(f"ì…ë ¥: {inp:25s} -> ì˜ˆì¸¡: {pred}")
````
